## –¶–µ–ª—å —Ä–∞–±–æ—Ç—ã
–û–≤–ª–∞–¥–µ—Ç—å –Ω–∞–≤—ã–∫–∞–º–∏ —Ä–∞–±–æ—Ç—ã —Å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º–∏ –ø–∞–∫–µ—Ç–∞–º–∏ –¥–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –ø–æ–¥–≥—Ä—É–ø–ø –≤ —Å–æ—Ü–∏–∞–ª—å–Ω—ã—Ö —Å–µ—Ç—è—Ö, –∞ —Ç–∞–∫–∂–µ –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏—è, –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
–∏ –∞–Ω–∞–ª–∏–∑–∞ —Å–µ—Ç–µ–π –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏.

## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è

 - ***–ì—Ä—É–ø–ø–∞*** - —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å –≤—Å–µ—Ö –∞–∫—Ç–æ—Ä–æ–≤, —Å–≤—è–∑–∏ –º–µ–∂–¥—É –∫–æ—Ç–æ—Ä—ã–º–∏ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –∏–∑—É—á–µ–Ω—ã.

- ***–ü–æ–¥–≥—Ä—É–ø–ø–∞*** - –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–æ —É–∑–ª–æ–≤ $N_s$

 - ***–ü–æ–¥–≥—Ä–∞—Ñ***: –î–∞–Ω –≥—Ä–∞—Ñ $Y(N, U)$. –ü–æ–¥–≥—Ä–∞—Ñ–æ–º $Y_s$ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –≥—Ä–∞—Ñ $Y_s = Y_s(N_s, U_s)$, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–∞ —É–∑–ª–æ–≤ $N_s \subseteq N$ –∏ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–∞ –ª–∏–Ω–∏–π (—Ä–µ–±–µ—Ä) $U_s \subseteq U$, —Å–≤—è–∑—ã–≤–∞—é—â–∏–µ —É–∑–ª—ã –≤ $Y_s$

 - ***–ö–ª–∏–∫–∞*** - –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–ª–Ω—ã–π –ø–æ–¥–≥—Ä–∞—Ñ –∏–∑ —Ç—Ä–µ—Ö –∏–ª–∏ –±–æ–ª–µ–µ —É–∑–ª–æ–≤ 

 - **n-–∫–ª–∏–∫–∞** - –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–∞–∏–±–æ–ª—å—à–µ–µ –≥–µ–æ–¥–µ–∑–∏—á–µ—Å–∫–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –ª—é–±—ã–º–∏ –¥–≤—É–º—è —É–∑–ª–∞–º–∏ –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç n. (–ü—É—Ç—å –º–æ–∂–µ—Ç –ø—Ä–æ—Ö–æ–¥–∏—Ç—å –ø–æ —É–∑–ª–∞–º –≤–Ω–µ –ø–æ–¥–≥—Ä–∞—Ñ–∞).

 - **n-–∫–ª–∞–Ω** - n-–∫–ª–∏–∫–∞, –≤ –∫–æ—Ç–æ—Ä–æ–π –≥–µ–æ–¥–µ–∑–∏—á–µ—Å–∫–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É –ª—é–±—ã–º–∏ –¥–≤—É–º—è —É–∑–ª–∞–º–∏ –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç n –¥–ª—è –ø—É—Ç–µ–π –≤–Ω—É—Ç—Ä–∏ –ø–æ–¥–≥—Ä–∞—Ñ–∞. (–ü—É—Ç—å –Ω–µ –º–æ–∂–µ—Ç –ø—Ä–æ—Ö–æ–¥–∏—Ç—å –ø–æ —É–∑–ª–∞–º –≤–Ω–µ –ø–æ–¥–≥—Ä–∞—Ñ–∞).

 - ***k-—è–¥—Ä–æ*** - —ç—Ç–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–∞–∂–¥–∞—è –≤–µ—Ä—à–∏–Ω–∞ —Å–≤—è–∑–∞–Ω–∞ –º–∏–Ω–∏–º—É–º —Å $k$ –¥—Ä—É–≥–∏–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏ —ç—Ç–æ–≥–æ –∂–µ –ø–æ–¥–≥—Ä–∞—Ñ–∞

 - ***k-–ø–ª–µ–∫—Å*** - —ç—Ç–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–¥–≥—Ä–∞—Ñ, —Å–æ–¥–µ—Ä–∂–∞—â–∏–π $g_s$ —É–∑–ª–æ–≤, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–∞–∂–¥—ã–π —É–∑–µ–ª —Å–º–µ–∂–µ–Ω –Ω–µ –º–µ–Ω–µ–µ —Å —á–µ–º —Å $(g_s - k)$ —É–∑–ª–∞–º–∏ –ø–æ–¥–≥—Ä–∞—Ñ–∞:

$$d_s(i) \geq g_s - k, \forall n \in N_s$$
    
–≥–¥–µ $d_s(i)$ - —Å—Ç–µ–ø–µ–Ω—å —É–∑–ª–∞ $n_i$ –≤ –ø–æ–¥–≥—Ä–∞—Ñ–µ $Y_s$

  - ***–ú–∞—Ç—Ä–∏—Ü–∞ –∏–Ω—Ü–µ–¥–µ–Ω—Ç–Ω–æ—Å—Ç–∏*** - –º–∞—Ç—Ä–∏—Ü–∞ A = ${a_{ij}}$ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ $g*h$ —Å —ç–ª–µ–º–µ–Ω—Ç–∞–º–∏:
\begin{equation*}
a_{ij} = 
 \begin{cases}
   1, &\text{–∞–∫—Ç–æ—Ä i —Å–≤—è–∑–∞–Ω —Å —Å–æ–±—ã—Ç–∏–µ–º j}\\
   0, &\text{–∏–Ω–∞—á–µ}
 \end{cases}
\end{equation*}

  - ***–ì–∏–ø–µ—Ä–≥—Ä–∞—Ñ*** - –æ–±–æ–±—â–µ–Ω–∏–µ –ø–æ–Ω—è—Ç–∏—è –≥—Ä–∞—Ñ–∞; –≥–∏–ø–µ—Ä–≥—Ä–∞—Ñ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –æ–±—ä–µ–∫—Ç–æ–≤ –∏ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤ –æ–±—ä–µ–∫—Ç–æ–≤, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–∞–∂–¥—ã–π –æ–±—ä–µ–∫—Ç –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ–º—É –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤—É, –ø—Ä–∏—á–µ–º –Ω–∏ –æ–¥–Ω–æ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–æ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –ø—É—Å—Ç—ã–º:
$$H = (A, B)$$
–≥–¥–µ $A = \{a_1,a_2,...,a_g\}$ - –º–Ω–æ–∂–µ—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤

$B = \{b_1, b_2, ..., b_n\}$ - –º–Ω–æ–∂–µ—Å—Ç–≤–æ —Ä–µ–±–µ—Ä


  - ***–°–µ—Ç—å –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏*** - —Å–µ—Ç—å, –≤ –∫–æ—Ç–æ—Ä–æ–π —É—á–∞—Å—Ç–Ω–∏–∫–∏ –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω—ã –¥—Ä—É–≥ —Å –¥—Ä—É–≥–æ–º –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å–æ—Ç—Ä—É–¥–Ω–∏—á–µ—Å—Ç–≤–∞ –≤ —Ä–∞–º–∫–∞—Ö –∫–∞–∫–æ–π-—Ç–æ –≥—Ä—É–ø–ø—ã –∏–ª–∏ —Å–æ–≤–º–µ—Å—Ç–Ω–æ–≥–æ —É—á–∞—Å—Ç–∏—è –≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º —Å–æ–±—ã—Ç–∏–∏. –°–µ—Ç—å —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ 2 –∏–∑–º–µ—Ä–µ–Ω–∏–π: –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –∞–∫—Ç–æ—Ä–æ–≤ $N = \{n_1, n_2, ..., n_g\}$ –∏ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ —Å–æ–±—ã—Ç–∏–π $M = \{m_1, m_2, ... , m_n\}$

  - ***–ú–∞—Ç—Ä–∏—Ü–∞ —Å–æ—É—á–∞—Å—Ç–∏—è –∞–∫—Ç–æ—Ä–æ–≤***:
$$X^N = \{x_{ij}^N\}$$

$$x_{ij}^N = \sum_{k=1}^ha_{ik}a_{jk}$$

$$X^N = AA^T$$

  - ***–ú–∞—Ç—Ä–∏—Ü–∞ –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è —Å–æ–±—ã—Ç–∏–π***:
$$X^M = \{x_{kl}^M\}$$

$$x_{kl}^M = \sum_{i=1}^ga_{ik}a_{il}$$

$$X^M = A^TA$$

  - ***–°—Ç–µ–ø–µ–Ω—å —É—á–∞—Å—Ç–∏—è*** - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π, —Å –∫–æ—Ç–æ—Ä—ã–º–∏ —Å–≤—è–∑–∞–Ω –∫–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä:
$$a_{i+} = \sum_{j=1}^ha_{ij} = x_{ii}^N$$

  - ***–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π, —Å –∫–æ—Ç–æ—Ä—ã–º –∞–∫—Ç–æ—Ä—ã –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω—ã***:

$$\overline{a_{i+}} = \frac{1}{g}\sum_{i=1}^g\sum_{j=1}^ha_{ij} = \frac{a_{++}}{g} = \frac{1}{g}\sum_{i=1}^gx_{ii}^N$$

  - ***–†–∞–∑–º–µ—Ä —Å–æ–±—ã—Ç–∏–π***:
$$a_{+j} = \sum_{i=1}^ga_{ij} = x_{jj}^M$$

  - ***–°—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–º–µ—Ä —Å–æ–±—ã—Ç–∏–π*** - —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–æ—Ä–æ–≤ –≤ –∫–∞–∂–¥–æ–º —Å–æ–±—ã—Ç–∏–∏ –∏–ª–∏ —Å—Ä–µ–¥–Ω—è—è —Å—Ç–µ–ø–µ–Ω—å —É–∑–ª–æ–≤, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∏—Ö —Å–æ–±—ã—Ç–∏—è –≤ –¥–≤—É–¥–æ–ª—å–Ω–æ–º –≥—Ä–∞—Ñ–µ:
$$\overline{a_{+j}} = \frac{1}{n}\sum_{i=1}^g\sum_{j=1}^ha_{ij} = \frac{a_{++}}{h} = \frac{1}{h}\sum_{j=1}^hx_{jj}^M$$

  - ***–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏—è —Å–æ—É—á–∞—Å—Ç—è $Œî_{(N)}$***:
$$Œî_{(N)} = \frac{1}{g(g-1)}\sum_{i=1}^g\sum_{j=1}^gx_{ij}^N, j \neq i$$

  - ***–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏—è –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è $Œî_{(M)}$***:
$$Œî_{(M)} = \frac{1}{h(h-1)}\sum_{k=1}^h\sum_{l=1}^hx_{kl}^M, l \neq k$$

 - **–ö–ª–∏–∫–∞ —É—Ä–æ–≤–Ω—è c** –≤ –≤–∑–≤–µ—à–µ–Ω–Ω–æ–º –≥—Ä–∞—Ñ–µ - –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–ª–Ω—ã–π –ø–æ–¥–≥—Ä–∞—Ñ –∏–∑ —å—Ä—ë—Ö –∏ –±–æ–ª–µ–µ —É–∑–ª–æ–≤, –≤—Å–µ –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö —è–≤–ª—è—é—Ç—Å—è —Å–º–µ–∂–Ω—ã–º–∏ –Ω–∞ —É—Ä–æ–≤–Ω–µ c (–º–µ–¥–ª—É –≤—Å–µ–º–∏ –ø–∞—Ä–∞–º–∏ —É–∑–ª–æ–≤ –µ—Å—Ç—å —Ä—ë–±—Ä–∞ —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ $\ge c$).

 - **–ö–ª–∏–∫–∞ —É—Ä–æ–≤–Ω—è c** –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏–π —Å–æ—É—á–∞—Å—Ç–∏—è - –ø–æ–¥–≥—Ä–∞—Ñ, –≥–¥–µ –≤—Å–µ –ø–∞—Ä—ã –∞–∫—Ç–æ—Ä–æ–≤ —Å–æ–≤–º–µ—Ç—Å–Ω–æ —É—á–∞—Å—Ç–≤—É—é—Ç –Ω–µ –º–µ–Ω–µ–µ —á–µ–º –≤ c —Å–æ–±—ã—Ç–∏—è—Ö.

 - **–ö–ª–∏–∫–∞ —É—Ä–æ–≤–Ω—è c** –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏–π –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è - –ø–æ–¥–≥—Ä–∞—Ñ, –≥–¥–µ –≤—Å–µ –ø–∞—Ä—ã —Å–æ–±—ã—Ç–∏–π –∏–º–µ–µ—Ç –Ω–µ –º–µ–Ω–µ–µ c –æ–±—â–∏—Ö —á–ª–µ–Ω–æ–≤.

## –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã
1. –ò–∑ —Ñ–∞–π–ª–∞ florentine_marital.csv (–¥–ª—è —á—ë—Ç–Ω—ã—Ö –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ (62)) –±—ã–ª–∞ –∏–∑–≤–ª–µ—á–µ–Ω–∞ —Å–æ—Ü–∏–æ–º–∞—Ç—Ä–∏—Ü–∞ —Å –ø–æ–º–æ—â—å—é —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π. 
–ó–∞—Ç–µ–º —Å–µ—Ç—å –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∞ —Å –ø–æ–º–æ—â—å—é –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –§—Ä—é—Ö—Ç–µ—Ä–º–∞–Ω–∞-–†–µ–π–Ω–≥–æ–ª—å–¥–∞.

```r
library(igraph)

data <- read.csv('./florentine_marital.csv', comment.char='#')
rownames(data) <- colnames(data)
so—Åmat <- as.matrix(data)

graph <- graph.adjacency(so—Åmat, mode="undirected")
plot(graph, layout=layout_with_fr(graph), main="A–ª–≥–æ—Ä–∏—Ç–º —É–∫–ª–∞–¥–∫–∏ –§—Ä—é—Ö—Ç–µ—Ä–º–∞–Ω–∞‚Äì–†–µ–π–Ω–≥–æ–ª—å–¥–∞")
```

2. –ë—ã–ª–∞ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∞ –¥–∞–Ω–Ω–∞—è —Å–µ—Ç—å –Ω–∞ –ø—Ä–µ–¥–º–µ—Ç –ø–æ–¥–≥—Ä—É–ø–ø —Å –ø–æ–º–æ—â—å—é:

 - ùëõ-–∫–ª–∏–∫;

 - ùëõ-–∫–ª–∞–Ω–æ–≤;

 - ùëõ-–∫–ª—É–±–æ–≤;

 -  ùëò-—è–¥–µ—Ä;

 - ùëò-–ø–ª–µ–∫—Å–æ–≤.

–£–∫–∞–∑–∞—Ç—å –∫–∞–∫–∏–µ –ø–æ–¥–≥—Ä—É–ø–ø—ã –ø–ª—É—á–∏–ª–∏—Å—å. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

–£—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ–æ—Ö–æ–¥–∏–º—ã—Ö –ø–∞–∫–µ—Ç–æ–≤.

```r
if (!require("BiocManager", quietly = TRUE))
    install.packages("BiocManager")

BiocManager::install("RBGL")
```

```r
library(RBGL)

res <- kCliques(igraph.to.graphNEL(graph))
res

ncliques <- res
res <- c()
for (n in names(ncliques)){
    cand <- ncliques[[n]]
    clan <- list()
    i <- 1
    for (cand.i in 1:length(cand)){
        g.n.cand <- induced_subgraph(graph,cand[[cand.i]])
        if (diameter(g.n.cand) <= n){ 
          clan[[i]] <- cand[[cand.i]]
          i <- i+1
        }
      }
    res[[n]] <- clan
}
names(res) <- paste(1:length(names(ncliques)), 'clan', sep='-')
res


info <- coreness(graph, mode= "all")
res <- lapply(0:max(as.integer(info)), function(x) names(info)[which(as.integer(info) >= x)])
names(res) <- paste(1:length(res)-1, 'core', sep='-')
res
```

–í—ã–≤–æ–¥: –î–ª—è –ø–∞—Ä 3 and 2, 4 and 1, 5 and 3, 6 and 3, 10 and 4, 11 and 3, 14 and 3 –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ —à–∞–Ω—Å–æ–≤ —Ä–∞–≤–Ω–æ Inf , —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç —á—Ç–æ –µ—Å–ª–∏ –∞–∫—Ç–æ—Ä –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –æ–¥–Ω–æ–º—É —Å–æ–±—ã—Ç–∏—é –ø–∞—Ä—ã —Ç–æ –æ–Ω —Ç–æ—á–Ω–æ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –∏ –¥—Ä—É–≥–æ–º—É —Å–æ–±—ã—Ç–∏—é –ø–∞—Ä—ã.
–¢–∞–∫–∂–µ –º–Ω–æ–≥–æ –ø–∞—Ä –∏–º–µ—é—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ —à–∞–Ω—Å–æ–≤ 0, —á—Ç–æ –≥–æ–≤–æ—Ä–∏—Ç –æ –≤–∑–∞–∏–º–Ω–æ–º –∏—Å–∫–ª—é—á–µ–Ω–∏–∏ –∞–∫—Ç–æ—Ä–æ–≤.

```r
countZero <- 0
countInf <- 0
for(k in 1:(h-1)){
    for(l in (k+1):h){
        a <- X[, c(k, l)]
        res <- sum(colSums(t(a) == c(1, 1)) == ncol(a))*
        sum(colSums(t(a) == c(0, 0)) == ncol(a))/
        sum(colSums(t(a) == c(0, 1)) == ncol(a))/
        sum(colSums(t(a) == c(1, 0)) == ncol(a))   
        cat('Theta(E', k, ', E', l, ') = ', res, '\n', sep="")
        countZero <- countZero + (res!=0)
        countInf <- countInf + (res==Inf)
    }
}
cat('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ–Ω—É–ª–µ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π:', countZero, '\n')
cat('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ–Ω—É–ª–µ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π (%):', countZero/(h*(h-1)/2)*100, '\n')
cat('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π:', countInf, '\n')
cat('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π (%):', countInf/(h*(h-1)/2)*100, '\n')
```
