## Основные теоретические положения
### Определения
* Социоматрица - матрица инцидентности.
* Размер сети - количество акторов в сети.
* Плотность сети - отношение количества ребер в сети к теоретически возможному при данном количестве акторов.
* Диаметр сети - величина максимального кратчайшего пути между двумя акторами.
* Коэффициент кластеризации - отношение числа закрытых триад к общему количеству триад.
* Дисперсия
$\frac{1}{g}\sum_{i=1}^{g}(d(n_i) - \bar{d})^2$, где <br>
  $g$ - количество акторов <br>
  $d(n_i)$ - степень актора $n_i$ <br>
  $\bar{d}$ - средняя степень акторов <br>
* Мера центральности - показатель центральности или близости к центру в теории графов и анализе сетей; определяет наиболее важные вершины графа. 
* Центральность по степени
$C_D^{\prime}(n_i)=\frac{d(n_i)}{g - 1}$
* Центральность по близости
$C_C^{\prime}(n_i)=\frac{g - 1}{\sum_{j=1}^{g}d(n_i, n_j)}$, где <br>
  $d(n_i, n_j)$ - расстояние между акторами $n_i$ и $n_j$
* Центральность по посредничеству
$C_B^{\prime}(n_i)=\frac{2\sum_{j < k}\frac{g_{jk}(n_i)}{g_{jk}}}{(g-1)(g-2)}$, где <br>
  $g_{jk}(n_i)$ - количество кратчайших путей между акторами с индексами $j$ и $k$, содержащих актор $n_i$ <br>
  $g_{jk}$ - количество кратчайших путей между акторами с индексами $j$ и $k$
* Центральность по информации <br>
$C_I^{\prime}(n_i)=\frac{C_I(n_i)}{\sum_{i}C_I(n_i)}$ <br>
$C_I(n_i)=\frac{1}{c_{ii} + \frac{(\textbf{T} - \textbf{2R})}{g}}$ <br>
 Переменные $c_{ii}, T, R$ вычисляются следующим образом: <br>
 Создается матрица $A$. 
 $a_{ii} = 1 + $ сумма линий, инцедентных актору $n_i$
$
  a_{ij} =
    \begin{cases}
      1 & \text{if nodes $n_i$ and $n_j$ are not adjacent}\\
      {1 - x_{ij}} & \text{if nodes $n_i$ and $n_j$ are adjacent. }\\
    \end{cases}       
$ <br>
$C = A^{-1}$ <br>
$T = \sum_{i=1}^{g}c_{ii}$ <br>
$R = \sum_{j=1}^{g}c_{ij}$ <br>
*Централизация по степени <br>
$C_D = \frac{\sum_{i=1}^{g}[C_D(n^{*}) - C_D(n_i)]}{(g-1)(g-2)}$ <br>
$C_D(n^{*})$ - максимум среди всех значений центральности
*Централизация по близости
$C_C=\frac{(2g-3)\sum_{i=1}{g}[C_C^{\prime}(n^{*}) - C_C^{\prime}(n_i)]}{(g-1)(g-2)}$
*Централизация по посредничеству
$C_B = \frac{\sum_{i=1}^{g}[C_B^{\prime}(n^{*}) - C_B^{\prime}(n_i)]}{(g-1)}$
*Централизация по информации - среднее вектора, представляющего значения центральности по информации для каждого актора.
*Престиж актора - позволяет оценить, насколько часто актор встречается в формировании кратчайших путей.
*Престиж по степени <br>
$P_D^{\prime}(n_i) = \frac{x_{+i}}{g-1}$ <br>
$x_{+i}$ - количество входящих ребер
*Престиж по близости <br>
$P_P(n_i) = \frac{I_i/(g-1)}{\sum d(n_j, n_i) / I_i}$ <br>
$I_i$ - количество акторов, из которых актор $i$ достижим
*Престиж по рангу
$(I - X^{T})p = 0$ <br>
$I$ - единичная матрица <br>
$X$ - социоматрица <br>
$p$ - вектор значений престижа

9) Заполнить таблицу для различных мер центральности. Примечание: меры центральности вычислить для неизолированных вершин:

| $$n_i$$ | $$C'_D(n_i)$$ | $$C'_C(n_i)$$ | $$C'_B(n_i)$$ |  $$C'_I(n_i)$$ |
|:-----:|:-----:|---|---|---|
|       |       |   |   |   |


```r
membership <- igraph::components(df_graph, mode="strong")$membership
counts <- igraph::components(df_graph, mode="strong")$csize
isolated <- membership[counts[membership] == 1]

drop_isolated <- delete_vertices(df_graph, names(isolated))

g <- gorder(drop_isolated)

actor_degree_centrality <- igraph::degree(drop_isolated) / (g - 1) 

dist <- distances(drop_isolated)

actor_closeness_centrality <- igraph::closeness(drop_isolated)

actor_betweenness_centrality <- igraph::betweenness(drop_isolated, directed=FALSE, normalized=TRUE)

actor_inf_centrality <- infocent(as.matrix(as_adjacency_matrix(drop_isolated)))

measures <- data.frame(degree_centrality=actor_degree_centrality,
                        betweenness_centrality=actor_betweenness_centrality,
                        closeness_centrality=actor_closeness_centrality,
                        inf_centrality=actor_inf_centrality)
measures
```

13) Заполнить таблицу для различных мер престижа:

| $$n_i$$ | $$P'_D(n_i)$$ | $$P'_P(n_i)$$ |
|:-----:|:-----:|---|
|       |       |   |

```r
g <- gorder(graph_data)
degree_prestige <- igraph::degree(graph_data, mode="in") / (g - 1)
dists <- distances(graph_data, mode="out")
count_proximity_prestige <- function(x){
  x <- x[is.finite(x)]
  I <- length(x)
  sum_dists <- sum(x)
  return(I ** 2 / (sum_dists * (g - 1)))
}
proximity_prestige = apply(dists, FUN=count_proximity_prestige, MARGIN=2)

table = data.frame(degree_prestige, proximity_prestige)
table
```
