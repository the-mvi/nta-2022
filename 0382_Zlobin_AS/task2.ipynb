## Основные теоретические положения
  
Группа (group) - это совокупность всех акторов, связи между которыми должны быть изучены. (Рассматривается одно отношение)  
Подгруппа(subgroup) - любое подмножество акторов и все связи между ними.
  
Клика(clique) - максимальный полный подграф из трёх или более узлов  
n-клика(n-clique) - максимальный подграф, в котором наибольшее геодезическое - расстояние между любыми двумя узлами не превышает n:  
$ n: d(i, j) <= n\: \forall n_i, n_j \in N_s $  
n-кланы(n-clans) - это  n-клика, в которой геодезическое расстояние d(i, j) между всеми узлами в подграфе не превышает n  для путей внутри подграфа  
n-клубы(n-clubs) - максимальный подграф диаметра ≤ n  

  
k-плекс (k-plex) - это максимальный подграф, содержащий g_s узлов, в котором каждый узел смежен не менее чем $ (g_s - k) $ узлами подграфа:
    $ d_s(i) >= g_s-k,  \forall n_i \in N_s $,  
    где $d_s(i)$ - степень узла $n_i$ в подграфе $G$

k-ядро(k-core) - это максимальный подграф, в котором каждая вершина связана минимум с k другими вершинами этого же подграфа  
  * вложенные
  * не перекрываются
  * легко опрделеить
  
(Seidman, 1983) Дан подграф $\mathscr G_s$  с множеством узлов $\mathscr N_s$. Множество узлом $\mathfrak N_s$ будем называть LS-множеством, если любое собственное подмножество $\mathscr{Q}$ ( $\mathscr{N}_s $имеет больше линий к узлам $ \mathscr{N}_S$ \ L, чем к $\mathscr{N} $ \ 
- (Borgatti, Everett and Shirey, 1990) Рассматриваются пары узлов множества $\mathfrak N_s$ подграфа $\mathscr Q_s.$ Множество узлов  $\mathfrak N_S$ называется $\lambda$ - множество $(\lambda-set)$ если люба пара узлов в $\mathscr N_s$ имеет большее линейное соединение, чеб любая пара узлов, состоящая из одного узла из лямбда-набора и второго узла извне:
    
    где $ \lambda(i, j) $ - минимальное количество линий, которое необходимо удалить из графа, чтобы не осталось путей между двумя узлами
Мера Бока–Хусейна, мера Альбы.  
(Bock and Husain, 1950) Дана сеть в которой $\mathscr N$ = g и $\mathscr N_S = g_s$. Мера сплоченности подгрупп по Боке Хусейну имеет вид:  
$ M_{BH} = (\dfrac{1}{g_s(g_s-1)}\sum_{i \in N_s} \sum_{j \in N_s} x_{ij})/(\dfrac{1}{g_s(g-g_s)}\sum_{i \in N_s} \sum_{j \notin N_s} x_{ij}) $  
(Alba (1973)) Дана сеть с параметрами $\mathscr N = g $ и $\mathscr L = L$. В сети есть подгруппа с праметрами $N_S = g_s$ и $L_s = L_s$. В качестве меры Альбы используется гипергеометрическая вероятность того, что наблюдаемое количество строк в подграфе равно q:  
$ Pr(L_S = q) = \binom{L}{Q}\binom{\binom{g}{2}-L}{\binom{g_s}{2}-q}/\binom{\binom{g}{2}}{\binom{g_s}{2}} $
  
* Слабосвязанная n-клика (weakly connected n-clique) - подграф, в котором все узлы слабо n-связаны,  и нет дополнительных узлом, также слабо n-связаны, и нет дополнительных узлов, также слабо n-связных со всеми узлами подграфа
* Односторонне связанная n-клика(unilaterally connected n-clique) - подграф, в коотором все узлы односторонне n-связаны, и нет дополнительных узлов, которые также односторонне n-связаны со всеми узлами подграsфа
* Сильно связная n-клика (strongly connected n-clique) - подграф, в котором все узлы сильно n-связаны, и нет дополнительных узлов, также сильно n-связных со всеми узлами подграфа  
  
Модулярность(modularity) - характеристика структуры сети, в частности степень кластеризации узлов, когда внутри кластера наблюдается много узлов (высокая плотность), а между кластерами мало узлов(низкая плотность)
    - Показатель модулярности определяется как разница между доле связей внутри группы и ожидаемой долей связей, если бы связи были размещены случайно
    - Показатель модулярности может принимать значения в диапазоне от -1/2 до +1
    - Чем выше показатель, тем больше уровень кластеризации сети в соответствии с данной группировкой узлов  
  
Сеть аффилированности (affilation network) - сеть, в которой участники аффилированы друг с другом на основе сотрудничества в рамках какой-то группы или совместного участия в определенном событии.
    - Сеть состоит из двух измерений (two-modes):  
    множество акторов $\mathscr N = \{n_1, n_2, ..., n_g\}$  
    и множества событий (events) $\mathscr M = \{m_1, m_2, ..., m_h\}$
  
- Матрица соучастия акторов (actor co-membership matrix):
    
    $X^\mathscr N = \{x_{ij}^\mathscr N\}:$  
    $ x_{ij}^\mathscr N = \sum_{k=1}^h a_{ik}a_{jk} X^\mathscr N = XX^T $

- Матрица перекрытия событий (event overlap matrix): $X^\mathscr M = \{x_{kl}^\mathscr M\}:$  
$ x_{kl}^\mathscr M = \sum_{k=1}^h a_{ik}a_{il} X^\mathscr M = X^TX $  
  
_Степень участия_ - количество событий, с которым связан каждый актор.

$a_{i+} = \sum_{j = 1}^{h} a_{ij} = x_{ii}^{\cal{N}}$

_Размер событий:_

$a_{+j} = \sum_{i = 1}^{g} a_{ij} = x_{ii}^{\cal{M}}$

### n-кланы 

```r
nclan <- function(g,n){
  g <- as.undirected(g)
  E(g)$weight <- 1
  ncliques <- kCliques(ugraph(igraph.to.graphNEL(g)))
  n.cand <- ncliques[[n]]
  n.clan <- list()
  n.clan.i <- 1
  for (n.cand.i in 1:length(n.cand)){
    g.n.cand <- induced_subgraph(g,n.cand[[n.cand.i]])
    if (diameter(g.n.cand)<=n){
      n.clan[[n.clan.i]] <- n.cand[[n.cand.i]]
      n.clan.i <- n.clan.i+1
    }
  }
  return(n.clan)
}
lapply(c(1:length(ncliques)), function(x) {nclan(dat_graph, x)})
```

**Edge-betweenness**
Мера центральности в графе, основанная на кратчайших путях. Для любой пары вершин в связном графе существует по меньшей мере один (кратчайший) путь между вершинами, для которого минимально либо число рёбер, по которым путь проходит, (для невзвешенных графов), либо сумма весов этих рёбер (для взвешенных графов). Степень посредничества для каждой вершины равна числу этих кратчайших путей через вершину.  
**Leading eigenvector** 
Суть метода заключается в определении матрицы модульности , Bкоторая является матрицей смежности (ненаправленной) сети и содержит вероятность наличия определенных ребер в соответствии с «моделью конфигурации». Другими словами, элемент есть вероятность того, что существует ребро между вершинами и в случайной сети, в которой степени всех вершин такие же, как и во входном графе.  
**Fast-greedy**
Каждой вершине графа инициализируется отдельное сообщество, а затем объединяются пары сообществ, приводящие к максимальному увеличению модулярности. При этом объединяются только инцидентные пары вершин, так как, в противном случае, модулярность не может увеличиться. Результат - ярус дендрограммы, на котором модулярность максимальна.

```r
osh <- function(A){
  x_k_l <- t(A) %*% A
  x_not_kl <- (1-t(A))%*%(1-A)
  x_not_k_l <- (1-t(A))%*%A
  x_k_not_l <- t(A)%*%(1-A)
  
  res <- c()
 for(i in 2:nrow(x_k_l)){
   for(j in 1:(i-1)){
     name <- paste(i,"and",j,"events",sep=" ")
     chance <- (x_k_l[i,j]*x_not_kl[i,j])/(x_not_k_l[i,j]*x_k_not_l[i,j])
     res[name] <-chance
   }
 }
 return(res)
}

print(osh(affil_matrix))
```
