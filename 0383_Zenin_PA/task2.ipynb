## –°–µ—Ç–∏ –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏
<br><br>
**–°–µ—Ç—å –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏** - —Å–µ—Ç—å, –≤ –∫–æ—Ç–æ—Ä–æ–π —É—á–∞—Å—Ç–Ω–∏–∫–∏ –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω—ã –¥—Ä—É–≥ —Å –¥—Ä—É–≥–æ–º –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å–æ—Ç—Ä—É–¥–Ω–∏—á–µ—Å—Ç–≤–∞ –≤ —Ä–∞–º–∫–∞—Ö –∫–∞–∫–æ–π-—Ç–æ –≥—Ä—É–ø–ø—ã –∏–ª–∏ —Å–æ–≤–º–µ—Å—Ç–Ω–æ–≥–æ —É—á–∞—Å—Ç–∏—è –≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º —Å–æ–±—ã—Ç–∏–∏. –°–µ—Ç—å —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ 2 –∏–∑–º–µ—Ä–µ–Ω–∏–π: –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –∞–∫—Ç–æ—Ä–æ–≤ $N = \{n_1, n_2, ..., n_g\}$ –∏ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ —Å–æ–±—ã—Ç–∏–π $M = \{m_1, m_2, ... , m_n\}$

**–ú–∞—Ç—Ä–∏—Ü–∞ —Å–æ—É—á–∞—Å—Ç–∏—è –∞–∫—Ç–æ—Ä–æ–≤**:<br><br>
$$X^N = \{x_{ij}^N\}$$

$$x_{ij}^N = \sum_{k=1}^ha_{ik}a_{jk}$$

$$X^N = AA^T$$<br><br>

**–ú–∞—Ç—Ä–∏—Ü–∞ –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è —Å–æ–±—ã—Ç–∏–π**:<br><br>
$$X^M = \{x_{kl}^M\}$$

$$x_{kl}^M = \sum_{i=1}^ga_{ik}a_{il}$$

$$X^M = A^TA$$<br><br>

**–°—Ç–µ–ø–µ–Ω—å —É—á–∞—Å—Ç–∏—è** - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π, —Å –∫–æ—Ç–æ—Ä—ã–º–∏ —Å–≤—è–∑–∞–Ω –∫–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä:<br><br>
$$a_{i+} = \sum_{j=1}^ha_{ij} = x_{ii}^N$$<br><br>

**–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π, —Å –∫–æ—Ç–æ—Ä—ã–º –∞–∫—Ç–æ—Ä—ã –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω—ã**:<br><br>

$$\overline{a_{i+}} = \frac{1}{g}\sum_{i=1}^g\sum_{j=1}^ha_{ij} = \frac{a_{++}}{g} = \frac{1}{g}\sum_{i=1}^gx_{ii}^N$$<br><br>

**–†–∞–∑–º–µ—Ä —Å–æ–±—ã—Ç–∏–π**:<br><br>
$$a_{+j} = \sum_{i=1}^ga_{ij} = x_{jj}^M$$<br><br>

**–°—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–º–µ—Ä —Å–æ–±—ã—Ç–∏–π** - —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–æ—Ä–æ–≤ –≤ –∫–∞–∂–¥–æ–º —Å–æ–±—ã—Ç–∏–∏ –∏–ª–∏ —Å—Ä–µ–¥–Ω—è—è —Å—Ç–µ–ø–µ–Ω—å —É–∑–ª–æ–≤, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∏—Ö —Å–æ–±—ã—Ç–∏—è –≤ –¥–≤—É–¥–æ–ª—å–Ω–æ–º –≥—Ä–∞—Ñ–µ:<br><br>
$$\overline{a_{+j}} = \frac{1}{n}\sum_{i=1}^g\sum_{j=1}^ha_{ij} = \frac{a_{++}}{h} = \frac{1}{h}\sum_{j=1}^hx_{jj}^M$$<br><br>

**–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏—è —Å–æ—É—á–∞—Å—Ç—è $Œî_{(N)}$**:<br><br>
$$Œî_{(N)} = \frac{1}{g(g-1)}\sum_{i=1}^g\sum_{j=1}^gx_{ij}^N, j \neq i$$<br><br>

**–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏—è –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è $Œî_{(M)}$**:<br><br>
$$Œî_{(M)} = \frac{1}{h(h-1)}\sum_{k=1}^h\sum_{l=1}^hx_{kl}^M, l \neq k$$<br><br>

**–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è**:



* **–î–∏–∞–º–µ—Ç—Ä –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —Å–µ—Ç–∏** - –¥–ª–∏–Ω–∞ —Å–∞–º–æ–≥–æ –¥–ª–∏–Ω–Ω–æ–≥–æ –ø—É—Ç–∏ –º–µ–∂–¥—É –ª—é–±–æ–π –ø–∞—Ä–æ–π –∞–∫—Ç–æ—Ä–æ–≤ –∏/–∏–ª–∏ —Å–æ–±—ã—Ç–∏–π

* **–ö–ª–∏–∫–∞ —É—Ä–æ–≤–Ω—è $c$ –≤ –≤–∑–≤–µ—à–µ–Ω–Ω–æ–º –≥—Ä–∞—Ñ–µ** - –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–ª–Ω—ã–π –ø–æ–¥–≥—Ä–∞—Ñ –∏–∑ 3 –∏–ª–∏ –±–æ–ª–µ–µ —É–∑–ª–æ–≤, –≤—Å–µ –∏–∑ –∫–æ—Ç–æ—Ä—ã–∑ —è–≤–ª—è—é—Ç—Å—è —Å–º–µ–∂–Ω—ã–º–∏ –Ω–∞ —É—Ä–æ–≤–Ω–µ $c$, —Ç.–µ. –º–µ–∂–¥—É –≤—Å–µ–º–∏ –ø–∞—Ä–∞–º–∏ —É–∑–ª–æ–≤ –µ—Å—Ç—å —Ä–µ–±—Ä–∞ —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –±–æ–ª—å—à–∏–º–∏ –∏–ª–∏ —Ä–∞–≤–Ω—ã–º–∏ $c$

* **–ö–ª–∏–∫–∞ —É—Ä–æ–≤–Ω—è $c$ –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏—è —Å–æ—É—á–∞—Å—Ç–∏—è** - –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –≤—Å–µ –ø–∞—Ä—ã –∞–∫—Ç–æ—Ä–æ–≤ —Å–æ–≤–º–µ—Å—Ç–Ω–æ —É—á–∞–≤—Å—Ç–≤—É—é—Ç –Ω–µ –º–µ–Ω–µ–µ, —á–µ–º –≤ $c$ —Å–æ–±—ã—Ç–∏—è—Ö

* **–ö–ª–∏–∫–∞ —É—Ä–æ–≤–Ω—è $c$ –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏—è –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è** - –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –≤—Å–µ –ø–∞—Ä—ã —Å–æ–±—ã—Ç–∏–π –∏–º–µ—é—Ç –Ω–µ –º–µ–Ω–µ–µ $c$ –æ–±—â–∏—Ö —á–ª–µ–Ω–æ–≤

* $x_{kl}^M$ - —á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω—ã —Å —Å–æ–±—ã—Ç–∏—è–º–∏ $k$ –∏ $l$

* $x_{\overline{kl}}^M$ - —á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—Ç –Ω–∏ –æ–¥–Ω–æ–º—É —Å–æ–±—ã—Ç–∏—é 

* $x_{k\overline{l}}^M$ - —á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã —Å —Å–æ–±—ã—Ç–∏–µ–º $k$, –Ω–æ –Ω–µ —Å —Å–æ–±—ã—Ç–∏–µ–º $l$

* $x_{\overline{k}l}^M$ - —á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω—ã —Å —Å–æ–±—ã—Ç–∏–µ–º $l$, –Ω–æ –Ω–µ —Å —Å–æ–±—ã—Ç–∏–µ–º $k$<br><br>

–°–ø—Ä–∞–≤–µ–¥–ª–∏–≤–æ —Å–ª–µ–¥—É—â–µ–µ:<br><br>
$$x_{kl}^M + x_{k\overline{l}}^M + x_{k\overline{l}}^M + x_{\overline{k}l}^M$$<br><br>

**–û—Ç–Ω–æ—à–µ–Ω–∏–µ —à–∞–Ω—Å–æ–≤** - –º–µ—Ä–∞ –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è –º–µ–∂–¥—É —Å–æ–±—ã—Ç–∏—è–º–∏, –∫–æ—Ç–æ—Ä–∞—è –Ω–µ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —Ä–∞–∑–º–µ—Ä–∞ —Å–æ–±—ã—Ç–∏–π:<br><br>
  $$\theta_{kl} = \frac{x_{kl}^M/x_{\overline{k}l}^M}{x_{k\overline{l}}^M/x_{\overline{kl}}^M} = \frac{x_{kl}^Mx_{\overline{kl}}^M}{x_{k\overline{l}}^Mx_{\overline{k}l}^M}$$<br><br>

  **–õ–æ–≥–∞—Ä–∏—Ñ–º –æ—Ç–Ω–æ—à–µ–Ω–∏—è —à–∞–Ω—Å–æ–≤**:<br><br>
  $$ln\theta_{kl} = ln(\frac{x_{kl}^Mx_{\overline{kl}}^M}{x_{k\overline{l}}^Mx_{\overline{k}l}^M})$$

N-cliques:

```r
n_cliques <- function(g, n){
  gr <- connect(g, order=n)
  return(max_cliques(gr,min=3))
}

get_n_cliques <- function(g) {
    N <- 1
    SIZE <- 2
    RESULT <- list()

        while(SIZE > 1) {
        cliques <- n_cliques(g, N)

        cliques <- lapply(cliques, function(x) names(x))
        names(cliques) <- rep(N, times = length(cliques))
        RESULT <- c(RESULT,cliques)

        N <- N + 1
        SIZE <- length(cliques)
        }

  return (RESULT)
}

print("N-cliques:")
print(get_n_cliques(gr))
```

N-clans

```r
n_clan <-function(g, clique, N) {
    diameter <- diameter(subgraph(g,clique), directed = FALSE)
    if(diameter > N){
      return(FALSE)
    }
    return (TRUE)
}

    get_n_clans <- function(g) {
    cliques <- get_n_cliques(g)

    tmpres <- 1:length(cliques)
    tmpres <-sapply(tmpres, function(x) n_clan(g,cliques[[x]], names(cliques)[x]))

    return(cliques[tmpres])
}

print("N-clans:")
print(get_n_clans(gr))
```

K-kernels

```r
get_k_kernels <- function(g) {
  info <- coreness(g, mode= "all")
  RESULT <-lapply(0:max(as.integer(info)), function(x) names(info)[which(as.integer(info) >= x)])

  names(RESULT) <- sapply(1:length(RESULT), function(x) x-1)

  return(RESULT)
}

print("K-kernels:")
print(get_k_kernels(gr))
```

4. –ò–∑ —Ñ–∞–π–ª–∞ ceo.csv –∏–∑–≤–ª–µ—á—å –º–∞—Ç—Ä–∏—Ü—É –∏–Ω—Ü–∏–¥–µ–Ω—Ç–Ω–æ—Å—Ç–∏ —Å –ø–æ–º–æ—â—å—é —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π. –û—Ç–æ–±—Ä–∞–∑–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É

—Å–µ—Ç–µ–π: —Å–µ—Ç—å —Å–æ—É—á–∞—Å—Ç–∏—è –∏ —Å–µ—Ç—å –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è —Å–æ–±—ã—Ç–∏–π. –ü—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Å–µ—Ç–∏ –≤ –≤–∏–¥–µ —Ü–µ–Ω–Ω–æ—Å—Ç–Ω—ã—Ö –º–∞—Ç—Ä–∏—Ü –∏ –≤–∑–≤–µ—à–µ–Ω–Ω—ã—Ö –≥—Ä–∞—Ñ–æ–≤.

```r

partmatr <- data2_matr %*% t(data2_matr)
partnet  <- graph.incidence(partmatr, weighted = TRUE)

event    <- t(data2_matr) %*% data2_matr
eventnet <- graph.incidence(event, weighted = TRUE)

partmatr
plot(partnet, edge.label = E(partnet)$weight, main = "–°–µ—Ç—å —Å–æ—É—á–∞—Å—Ç–∏—è")

event
plot(eventnet, edge.label = E(eventnet)$weight, main = "–°–µ—Ç—å –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è —Å–æ–±—ã—Ç–∏–π")
```

```r
print("–°–æ—É—á–∞—Å—Ç–∏–µ:")

for(i in 3:4){
    ng <- delete_edges(partnet, E(partnet)[E(partnet)$weight < i])
    print(max_cliques(ng, min=3))
  } 

print("\n\n\n")
print("–ü–µ—Ä–µ–∫—Ä—ã—Ç–∏–µ:")

for(i in 2:6){
    ng <- delete_edges(eventnet, E(eventnet)[E(eventnet)$weight < i])
    print(max_cliques(ng, min=3))
  } 
```

13. –î–ª—è –∫–∞–∂–¥—ã—Ö –ø–∞—Ä —Å–æ–±—ã—Ç–∏–π –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –æ—Ç–Ω–æ—à–µ–Ω–∏–µ —à–∞–Ω—Å–æ–≤. –°–¥–µ–ª–∞—Ç—å

```r
dt <- matrix(0, nrow = 15, ncol = 15)


for(k in 1:(length(event_size)-1)){
    for(l in (k+1):length(event_size)){
        a <- data2_matr[, c(k, l)]
        res <- sum(colSums(t(a) == c(1, 1)) == ncol(a))*
        sum(colSums(t(a) == c(0, 0)) == ncol(a))/
        sum(colSums(t(a) == c(0, 1)) == ncol(a))/
        sum(colSums(t(a) == c(1, 0)) == ncol(a))
        dt[k,l] <- res
        #cat('Theta(M', k, ', M', l, ') = ', res, '\n', sep="")
    }
}

tbl <- as.table(dt)
rownames(tbl) <- c(1:15)
colnames(tbl) <- c(1:15)
tbl
```
