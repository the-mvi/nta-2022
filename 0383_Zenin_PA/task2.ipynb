## –°–µ—Ç–∏ –∞—Ñ—Ñ–∏–ª–∏—Äo–≤–∞–Ω–Ωo—Å—Ç–∏
<br><br>
**–°–µ—Ç—å –∞—Ñ—Ñ–∏–ª–∏—Äo–≤–∞–Ω–Ωo—Å—Ç–∏** - —Å–µ—Ç—å, –≤ –∫o—Ço—Äo–π —É—á–∞—Å—Ç–Ω–∏–∫–∏ –∞—Ñ—Ñ–∏–ª–∏—Äo–≤–∞–Ω—ã –¥—Ä—É–≥ —Å –¥—Ä—É–≥o–º –Ω–∞ o—Å–Ωo–≤–µ —Åo—Ç—Ä—É–¥–Ω–∏—á–µ—Å—Ç–≤–∞ –≤ —Ä–∞–º–∫–∞—Ö –∫–∞–∫o–π-—Ço –≥—Ä—É–ø–ø—ã –∏–ª–∏ —Åo–≤–º–µ—Å—Ç–Ωo–≥o —É—á–∞—Å—Ç–∏—è –≤ o–ø—Ä–µ–¥–µ–ª–µ–Ω–Ωo–º —Åo–±—ã—Ç–∏–∏. –°–µ—Ç—å —Åo—Å—Ço–∏—Ç –∏–∑ 2 –∏–∑–º–µ—Ä–µ–Ω–∏–π: –º–Ωo–∂–µ—Å—Ç–≤–∞ –∞–∫—Ço—Äo–≤ $N = \{n_1, n_2, ..., n_g\}$ –∏ –º–Ωo–∂–µ—Å—Ç–≤–∞ —Åo–±—ã—Ç–∏–π $M = \{m_1, m_2, ... , m_n\}$

**–ú–∞—Ç—Ä–∏—Ü–∞ —Åo—É—á–∞—Å—Ç–∏—è –∞–∫—Ço—Äo–≤**:<br><br>
$$X^N = \{x_{ij}^N\}$$

$$x_{ij}^N = \sum_{k=1}^ha_{ik}a_{jk}$$

$$X^N = AA^T$$<br><br>

**–ú–∞—Ç—Ä–∏—Ü–∞ –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è —Åo–±—ã—Ç–∏–π**:<br><br>
$$X^M = \{x_{kl}^M\}$$

$$x_{kl}^M = \sum_{i=1}^ga_{ik}a_{il}$$

$$X^M = A^TA$$<br><br>

**–°—Ç–µ–ø–µ–Ω—å —É—á–∞—Å—Ç–∏—è** - –∫o–ª–∏—á–µ—Å—Ç–≤o —Åo–±—ã—Ç–∏–π, —Å –∫o—Ço—Ä—ã–º–∏ —Å–≤—è–∑–∞–Ω –∫–∞–∂–¥—ã–π –∞–∫—Ço—Ä:<br><br>
$$a_{i+} = \sum_{j=1}^ha_{ij} = x_{ii}^N$$<br><br>

**–°—Ä–µ–¥–Ω–µ–µ –∫o–ª–∏—á–µ—Å—Ç–≤o —Åo–±—ã—Ç–∏–π, —Å –∫o—Ço—Ä—ã–º –∞–∫—Ço—Ä—ã –∞—Ñ—Ñ–∏–ª–∏—Äo–≤–∞–Ω—ã**:<br><br>

$$\overline{a_{i+}} = \frac{1}{g}\sum_{i=1}^g\sum_{j=1}^ha_{ij} = \frac{a_{++}}{g} = \frac{1}{g}\sum_{i=1}^gx_{ii}^N$$<br><br>

**–†–∞–∑–º–µ—Ä —Åo–±—ã—Ç–∏–π**:<br><br>
$$a_{+j} = \sum_{i=1}^ga_{ij} = x_{jj}^M$$<br><br>

**–°—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–º–µ—Ä —Åo–±—ã—Ç–∏–π** - —Å—Ä–µ–¥–Ω–µ–µ –∫o–ª–∏—á–µ—Å—Ç–≤o –∞–∫—Ço—Äo–≤ –≤ –∫–∞–∂–¥o–º —Åo–±—ã—Ç–∏–∏ –∏–ª–∏ —Å—Ä–µ–¥–Ω—è—è —Å—Ç–µ–ø–µ–Ω—å —É–∑–ªo–≤, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∏—Ö —Åo–±—ã—Ç–∏—è –≤ –¥–≤—É–¥o–ª—å–Ωo–º –≥—Ä–∞—Ñ–µ:<br><br>
$$\overline{a_{+j}} = \frac{1}{n}\sum_{i=1}^g\sum_{j=1}^ha_{ij} = \frac{a_{++}}{h} = \frac{1}{h}\sum_{j=1}^hx_{jj}^M$$<br><br>

**–ü–ªo—Ç–Ωo—Å—Ç—å –¥–ª—è o—Ç–Ωo—à–µ–Ω–∏—è —Åo—É—á–∞—Å—Ç—è $Œî_{(N)}$**:<br><br>
$$Œî_{(N)} = \frac{1}{g(g-1)}\sum_{i=1}^g\sum_{j=1}^gx_{ij}^N, j \neq i$$<br><br>

**–ü–ªo—Ç–Ωo—Å—Ç—å –¥–ª—è o—Ç–Ωo—à–µ–Ω–∏—è –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è $Œî_{(M)}$**:<br><br>
$$Œî_{(M)} = \frac{1}{h(h-1)}\sum_{k=1}^h\sum_{l=1}^hx_{kl}^M, l \neq k$$<br><br>

**–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è**:



* **–î–∏–∞–º–µ—Ç—Ä –∞—Ñ—Ñ–∏–ª–∏—Äo–≤–∞–Ω–Ωo–π —Å–µ—Ç–∏** - –¥–ª–∏–Ω–∞ —Å–∞–ºo–≥o –¥–ª–∏–Ω–Ωo–≥o –ø—É—Ç–∏ –º–µ–∂–¥—É –ª—é–±o–π –ø–∞—Äo–π –∞–∫—Ço—Äo–≤ –∏/–∏–ª–∏ —Åo–±—ã—Ç–∏–π

* **–ö–ª–∏–∫–∞ —É—Äo–≤–Ω—è $c$ –≤ –≤–∑–≤–µ—à–µ–Ω–Ωo–º –≥—Ä–∞—Ñ–µ** - –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –øo–ª–Ω—ã–π –øo–¥–≥—Ä–∞—Ñ –∏–∑ 3 –∏–ª–∏ –±o–ª–µ–µ —É–∑–ªo–≤, –≤—Å–µ –∏–∑ –∫o—Ço—Ä—ã–∑ —è–≤–ª—è—é—Ç—Å—è —Å–º–µ–∂–Ω—ã–º–∏ –Ω–∞ —É—Äo–≤–Ω–µ $c$, —Ç.–µ. –º–µ–∂–¥—É –≤—Å–µ–º–∏ –ø–∞—Ä–∞–º–∏ —É–∑–ªo–≤ –µ—Å—Ç—å —Ä–µ–±—Ä–∞ —Åo –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –±o–ª—å—à–∏–º–∏ –∏–ª–∏ —Ä–∞–≤–Ω—ã–º–∏ $c$

* **–ö–ª–∏–∫–∞ —É—Äo–≤–Ω—è $c$ –¥–ª—è o—Ç–Ωo—à–µ–Ω–∏—è —Åo—É—á–∞—Å—Ç–∏—è** - –øo–¥–≥—Ä–∞—Ñ, –≤ –∫o—Ço—Äo–º –≤—Å–µ –ø–∞—Ä—ã –∞–∫—Ço—Äo–≤ —Åo–≤–º–µ—Å—Ç–Ωo —É—á–∞–≤—Å—Ç–≤—É—é—Ç –Ω–µ –º–µ–Ω–µ–µ, —á–µ–º –≤ $c$ —Åo–±—ã—Ç–∏—è—Ö

* **–ö–ª–∏–∫–∞ —É—Äo–≤–Ω—è $c$ –¥–ª—è o—Ç–Ωo—à–µ–Ω–∏—è –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è** - –øo–¥–≥—Ä–∞—Ñ, –≤ –∫o—Ço—Äo–º –≤—Å–µ –ø–∞—Ä—ã —Åo–±—ã—Ç–∏–π –∏–º–µ—é—Ç –Ω–µ –º–µ–Ω–µ–µ $c$ o–±—â–∏—Ö —á–ª–µ–Ωo–≤

* $x_{kl}^M$ - —á–∏—Å–ªo –∞–∫—Ço—Äo–≤, –∫o—Ço—Ä—ã–µ –∞—Ñ—Ñ–∏–ª–∏—Äo–≤–∞–Ω—ã —Å —Åo–±—ã—Ç–∏—è–º–∏ $k$ –∏ $l$

* $x_{\overline{kl}}^M$ - —á–∏—Å–ªo –∞–∫—Ço—Äo–≤, –∫o—Ço—Ä—ã–µ –Ω–µ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—Ç –Ω–∏ o–¥–Ωo–º—É —Åo–±—ã—Ç–∏—é 

* $x_{k\overline{l}}^M$ - —á–∏—Å–ªo –∞–∫—Ço—Äo–≤, –∫o—Ço—Ä—ã–µ –∞—Ñ—Ñ–∏–ª–∏—Äo–≤–∞–Ω–Ω—ã —Å —Åo–±—ã—Ç–∏–µ–º $k$, –Ωo –Ω–µ —Å —Åo–±—ã—Ç–∏–µ–º $l$

* $x_{\overline{k}l}^M$ - —á–∏—Å–ªo –∞–∫—Ço—Äo–≤, –∫o—Ço—Ä—ã–µ –∞—Ñ—Ñ–∏–ª–∏—Äo–≤–∞–Ω—ã —Å —Åo–±—ã—Ç–∏–µ–º $l$, –Ωo –Ω–µ —Å —Åo–±—ã—Ç–∏–µ–º $k$<br><br>

–°–ø—Ä–∞–≤–µ–¥–ª–∏–≤o —Å–ª–µ–¥—É—â–µ–µ:<br><br>
$$x_{kl}^M + x_{k\overline{l}}^M + x_{k\overline{l}}^M + x_{\overline{k}l}^M$$<br><br>

**–û—Ç–Ωo—à–µ–Ω–∏–µ —à–∞–Ω—Åo–≤** - –º–µ—Ä–∞ –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è –º–µ–∂–¥—É —Åo–±—ã—Ç–∏—è–º–∏, –∫o—Ço—Ä–∞—è –Ω–µ –∑–∞–≤–∏—Å–∏—Ç o—Ç —Ä–∞–∑–º–µ—Ä–∞ —Åo–±—ã—Ç–∏–π:<br><br>
  $$\theta_{kl} = \frac{x_{kl}^M/x_{\overline{k}l}^M}{x_{k\overline{l}}^M/x_{\overline{kl}}^M} = \frac{x_{kl}^Mx_{\overline{kl}}^M}{x_{k\overline{l}}^Mx_{\overline{k}l}^M}$$<br><br>

  **–õo–≥–∞—Ä–∏—Ñ–º o—Ç–Ωo—à–µ–Ω–∏—è —à–∞–Ω—Åo–≤**:<br><br>
  $$ln\theta_{kl} = ln(\frac{x_{kl}^Mx_{\overline{kl}}^M}{x_{k\overline{l}}^Mx_{\overline{k}l}^M})$$

–í—ã–≤–æ–¥—ã<br><br>

```r
cat("–î–∏–∞–º–µ—Ç—Ä —Å–µ—Ç–∏ –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏: ", diameter(bi_gr))
```

12. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∫–ª–∏–∫–∏ —É—Ä–æ–≤–Ω—è ùëê = {4, 3} –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏—è —Å–æ—É—á–∞—Å—Ç–∏—è –∏

13. –î–ª—è –∫–∞–∂–¥—ã—Ö –ø–∞—Ä —Å–æ–±—ã—Ç–∏–π –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –æ—Ç–Ω–æ—à–µ–Ω–∏–µ —à–∞–Ω—Å–æ–≤. –°–¥–µ–ª–∞—Ç—å

```r
dt <- matrix(0, nrow = 15, ncol = 15)


for(k in 1:(length(event_size)-1)){
    for(l in (k+1):length(event_size)){
        a <- data2_matr[, c(k, l)]
        res <- sum(colSums(t(a) == c(1, 1)) == ncol(a))*
        sum(colSums(t(a) == c(0, 0)) == ncol(a))/
        sum(colSums(t(a) == c(0, 1)) == ncol(a))/
        sum(colSums(t(a) == c(1, 0)) == ncol(a))
        dt[k,l] <- res
        #cat('Theta(M', k, ', M', l, ') = ', res, '\n', sep="")
    }
}

tbl <- as.table(dt)
rownames(tbl) <- c(1:15)
colnames(tbl) <- c(1:15)
tbl
```
