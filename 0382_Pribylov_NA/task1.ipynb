–í—ã–≤–æ–¥—ã –∏ —Ä–∞—Å—Å—á–µ—Ç—ã –≤—Ä—É—á–Ω—É—é:
* –ü–ª–æ—Ç–Ω–æ—Å—Ç—å —Å–µ—Ç–∏ (–Ω–µ–æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π): $Œî = \frac {ùêø} {{g\choose 2}} = \frac {2 * ùêø} {g * (g - 1)} = \frac {2*20} {16*15} = 0.1666667$ , –≥–¥–µ $ùêø$ - —á–∏—Å–ª–æ –ª–∏–Ω–∏–π, $g$ - —á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤ –≤ —Å–µ—Ç–∏.
> –ß–µ–º –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –±–ª–∏–∂–µ –∫ 1, —Ç–µ–º —Å–µ—Ç—å —Å—á–∏—Ç–∞–µ—Ç—Å—è –±–æ–ª–µ–µ –ø–ª–æ—Ç–Ω–æ–π. –î–∞–Ω–Ω–∞—è —Å–µ—Ç—å –æ–±–ª–∞–¥–∞–µ—Ç –Ω–µ–±–æ–ª—å—à–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç—å—é - —Å–≤—è–∑–µ–π –≤ —Å–µ—Ç–∏ –Ω–µ –º–Ω–æ–≥–æ.

* –ß–∏—Å–ª–æ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç —Å–≤—è–∑–Ω–æ—Å—Ç–∏, –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–µ –ø–æ –≥—Ä–∞—Ñ–∏–∫—É, —Ä–∞–≤–Ω–æ 2 (–∏–∑–æ–ª–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –≤–µ—Ä—à–∏–Ω–∞ –∏ –æ—Å—Ç–∞–ª—å–Ω–æ–π –≥—Ä–∞—Ñ). –ß–∏—Å–ª–æ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç —Å–≤—è–∑–Ω–æ—Å—Ç–∏ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, —Å–∫–æ–ª—å–∫–æ –∏–º–µ–µ—Ç—Å—è —Å–≤—è–∑–Ω—ã—Ö –ø–æ–¥–≥—Ä—É–ø–ø –≤ —Å–µ—Ç–∏, —Ç–æ –µ—Å—Ç—å –ø–æ–¥–≥—Ä—É–ø–ø, –≥–¥–µ –∫–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä –ø—Ä—è–º–æ –∏–ª–∏ –∫–æ—Å–≤–µ–Ω–Ω–æ —Å–≤—è–∑–∞–Ω —Å–æ –≤—Å–µ–º –¥—Ä—É–≥–∏–º–∏ –≤ –ø–æ–¥–≥—Ä—É–ø–ø–µ.
> –í –¥–∞–Ω–Ω–æ–π —Å–µ—Ç–∏, –≤—Å–µ —Ñ–ª–æ—Ä–µ–Ω—Ç–∏–π—Å–∫–∏–µ —Å–µ–º—å–∏, –∫—Ä–æ–º–µ Pucci, –∏–º–µ—é—Ç –æ–±—â–∏—Ö —Ä–æ–¥—Å—Ç–≤–µ–Ω–Ω–∏–∫–æ–≤.

* –î–∏–∞–º–µ—Ç—Ä —Å–µ—Ç–∏ —Ä–∞–≤–µ–Ω 5.
> –°–∞–º–∞—è –¥–∞–ª—å–Ω—è—è –∏–∑ –∫—Ä–∞—Ç—á–∞–π—à–∏—Ö —Å–≤—è–∑—å –º–µ–∂–¥—É —Å–µ–º—å—è–º–∏ Pazzi –∏ Lamberteschi.

* –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ —Å–µ—Ç–∏ —Ä–∞–≤–µ–Ω 0.1914894
> –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Ç–µ–Ω–¥–µ–Ω—Ü–∏—é –∫ —Å–æ–∑–¥–∞–Ω–∏—é –∑–∞–∫—Ä—ã—Ç—ã—Ö —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–æ–≤. –†–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω–∞—è —Å–µ—Ç—å –Ω–µ –∏–º–µ–µ—Ç —Å–∏–ª—å–Ω–æ–π —Ç–µ–Ω–¥–µ–Ω—Ü–∏–∏ –∫ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏. –¢. –µ. —Ç—Ä–∏ —Å–µ–º—å–∏ —Å–æ–∑–¥–∞—é—Ç "—Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫" (—Å–µ–º—å—è $A$ —Å–≤—è–∑–∞–Ω–∞ —Å —Å–µ–º—å–µ–π $B$, $B$ —Å–≤—è–∑–∞–Ω–∞ —Å $C$, $C$ —Å $A$) –∏–∑ —Å—É–ø—Ä—É–∂–µ—Å–∫–∏—Ö –Ω–µ—á–∞—Å—Ç–æ.

8. –ë—ã–ª–∞ –≤—ã—á–∏—Å–ª–µ–Ω–∞ —Å—Ä–µ–¥–Ω—è—è —Å—Ç–µ–ø–µ–Ω—å —É–∑–ª–æ–≤ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—è.

```r
fm_degrees <- degree(fm_network, gmode='graph')
cat("–°—Ä–µ–¥–Ω—è—è —Å—Ç–µ–ø–µ–Ω—å —É–∑–ª–æ–≤ = ", mean(fm_degrees))
cat("\n–î–∏—Å–ø–µ—Ä—Å–∏—è = ", var(fm_degrees))
```

>–°—Ä–µ–¥–Ω—è—è —Å—Ç–µ–ø–µ–Ω—å –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Å –∫–∞–∫–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∏–º–µ–µ—Ç –ø—Ä—è–º—É—é —Å–≤—è–∑—å –∫–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä –≤ —Å—Ä–µ–¥–Ω–µ–º. –í —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω–æ–π —Å–µ—Ç–∏ –∫–∞–∂–¥–∞—è —Å–µ–º—å—è —Å–≤—è–∑–∞–Ω–∞ –≤ —Å—Ä–µ–¥–Ω–µ–º —Å 2-3 –¥—Ä—É–≥–∏–º–∏ —Å–µ–º—å—è–º–∏.

>–î–∏—Å–ø–µ—Ä—Å–∏—è –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Ä–∞–∑–±—Ä–æ—Å –≤–µ–ª–∏—á–∏–Ω—ã –æ—Ç –µ–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è. –í —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω–æ–π —Å–µ—Ç–∏ —Å–µ–º—å–∏ –∏–º–µ—é—Ç –æ—Ç 0 –¥–æ 5 —Å–≤—è–∑–µ–π.

9. –ë—ã–ª–∞ –∑–∞–ø–æ–ª–Ω–µ–Ω–∞ —Ç–∞–±–ª–∏—Ü–∞ –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –º–µ—Ä —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç–∏ –¥–ª—è –Ω–µ–∏–∑–æ–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –≤–µ—Ä—à–∏–Ω.

```r
# —Ä–∞–±–æ—Ç–∞–µ–º —Å —Å–µ—Ç—å—é –±–µ–∑ –∏–∑–æ–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –≤–µ—Ä—à–∏–Ω
fm_sociomatrix_no_strays <- fm_sociomatrix[rowSums(fm_sociomatrix)!=0, colSums(fm_sociomatrix)!=0]
fm_graph_no_strays <- graph.adjacency(fm_sociomatrix_no_strays, mode = 'undirected')
fm_network_no_strays <- network(fm_sociomatrix_no_strays, directed = FALSE, matrix.type = "adjacency")

centralities <- data.frame(names = colnames(fm_sociomatrix_no_strays))
amount_of_actors <- ncol(fm_sociomatrix_no_strays)

# —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ —Å—Ç–µ–ø–µ–Ω–∏
degree_centrality <- centr_degree(fm_graph_no_strays, mode = "all", normalized = TRUE)

# —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ –±–ª–∏–∑–æ—Å—Ç–∏
closeness_centrality <- centr_clo(fm_graph_no_strays, normalized = TRUE)

# —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ –ø–æ—Å—Ä–µ–¥–Ω–∏—á–µ—Å—Ç–≤—É
betweenness_centrality <- centr_betw(fm_graph_no_strays, normalized = TRUE)

# —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏
information_centrality <- infocent(fm_sociomatrix_no_strays, gmode = "graph")

# –Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç–∏
centralities <- cbind(centralities, 
                      degree = degree_centrality$res / (amount_of_actors - 1), 
                      closeness = closeness_centrality$res, 
                      betweenness = betweenness_centrality$res * 2 / ((amount_of_actors - 1) * (amount_of_actors - 2)), 
                      information = information_centrality / sum(information_centrality))
centralities
```

–í—ã–≤–æ–¥—ã:
> –¶–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É–µ—Ç –≤–∞–∂–Ω–æ—Å—Ç—å –∞–∫—Ç–æ—Ä–∞. Medichi —è–≤–ª—è–µ—Ç—Å—è –Ω–∞–∏–±–æ–ª–µ–µ —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–π –ø–æ —Ç–µ–º –∏–ª–∏ –∏–Ω—ã–º –≤—Å–µ–º –º–µ—Ä–∞–º —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç–∏.

10. –ë—ã–ª–æ –≤—ã—á–∏—Å–ª–µ–Ω–æ –∑–Ω–∞—á–µ–Ω–∏–µ —Ü–µ–Ω—Ç—Ä–∞–ª–∏–∑–∞—Ü–∏–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –º–µ—Ä—ã.

```r
centralizations <- data.frame(type = c("degree", "closeness", "betweenness", "information"), 
                              centralization = c(degree_centrality$centralization, 
                                                 closeness_centrality$centralization, 
                                                 betweenness_centrality$centralization, 
                                                 sum((information_centrality / sum(information_centrality) - sum(information_centrality) / amount_of_actors)^2) / amount_of_actors))
centralizations
```

–í—ã–≤–æ–¥—ã:
>–°–µ—Ç—å –º–µ–Ω—å—à–µ –≤—Å–µ–≥–æ —Ü–µ–Ω—Ç—Ä–∞–ª–∏–∑–æ–≤–∞–Ω–∞ –ø–æ —Å—Ç–µ–ø–µ–Ω–∏, –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ - –ø–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏.

11. –ë—ã–ª–∏ –≤—ã—è–≤–ª–µ–Ω—ã —Ç–æ—á–∫–∏ —Å–æ—á–ª–µ–Ω–µ–Ω–∏—è –∏ –º–æ—Å—Ç—ã —Å–µ—Ç–∏. –î–æ–≥–∞–¥–∫–∏ –±—ã–ª–∏ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω—ã —Å –ø–æ–º–æ—â—å—é —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π. –ë—ã–ª –ø–æ—Å—Ç—Ä–æ–µ–Ω –≥—Ä–∞—Ñ–∏–∫ —Å–µ—Ç–∏, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º –≤—ã–¥–µ–ª–µ–Ω—ã
—Ç–æ—á–∫–∏ —Å–æ—á–ª–µ–Ω–µ–Ω–∏—è –∏ –º–æ—Å—Ç—ã.

–¢–æ—á–∫–∏ —Å–æ—á–ª–µ–Ω–µ–Ω–∏—è:

–ú–æ—Å—Ç—ã:

13. –ë—ã–ª–∞ –∑–∞–ø–æ–ª–Ω–µ–Ω–∞ —Ç–∞–±–ª–∏—Ü–∞ –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –º–µ—Ä –ø—Ä–µ—Å—Ç–∏–∂–∞.

```r
prestige_by_degree <- prestige(kf_network, gmode="digraph", cmode="indegree")
gd <- geodist(kf_network)
gd <- gd$gdist[which(rowSums(gd$gdist) != Inf), ]
inp <- nrow(gd)
prestige_by_proximity <- (inp^2)/((network.size(kf_network)-1)*colSums(gd))
prestiges <- data.frame(n=network.vertex.names(kf_network), 
                        P_d=prestige_by_degree/(network.size(kf_network)-1), 
                        P_p=prestige_by_proximity)
prestiges
```
