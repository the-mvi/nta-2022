#### 8. –í—ã—á–∏—Å–ª–∏—Ç—å —Å—Ä–µ–¥–Ω—é—é —Å—Ç–µ–ø–µ–Ω—å —É–∑–ª–æ–≤ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—é. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

–í —Å—Ä–µ–¥–Ω–µ–º —Å–µ–º—å—è –∏–º–µ–µ—Ç 2-3 –±—Ä–∞—á–Ω—ã–µ —Å–≤—è–∑–∏ —Å –¥—Ä—É–≥–∏–º–∏ —Å–µ–º—å—è–º–∏.
–û–¥–Ω–∞–∫–æ –¥–∏—Å–ø–µ—Ä—Å–∏—è –æ—á–µ–Ω—å –±–æ–ª—å—à–∞—è (–ø–æ—á—Ç–∏ —Ä–∞–≤–Ω–∞ —Å—Ä–µ–¥–Ω–µ–π —Å—Ç–µ–ø–µ–Ω–∏ —É–∑–ª–∞) - —ç—Ç–æ –∑–Ω–∞—á–∏—Ç, —á—Ç–æ –µ—Å—Ç—å —Å–µ–º—å–∏, —É –∫–æ—Ç–æ—Ä—ã—Ö –∫–æ–ª-–≤–æ –±—Ä–∞—á–Ω—ã—Ö —Å–≤—è–∑–µ–π –±–ª–∏–∑–∫–æ –∫ –Ω—É–ª—é –∏–ª–∏ –Ω–∞–æ–±–æ—Ä–æ—Ç, –±–æ–ª—å—à–µ –≤ 2 —Ä–∞–∑–∞ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è.

#### 9. –ó–∞–ø–æ–ª–Ω–∏—Ç—å —Ç–∞–±–ª–∏—Ü—É –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –º–µ—Ä —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç–∏. –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ: –º–µ—Ä—ã —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç–∏ –≤—ã—á–∏—Å–ª–∏—Ç—å –¥–ª—è –Ω–µ–∏–∑–æ–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –≤–µ—Ä—à–∏–Ω.

```python
c = {name: [] for name in names}
component = nx.subgraph(network, next(nx.connected_components(network)))
degree_centrality = nx.degree_centrality(component)
for name in degree_centrality:
    c[name].append(degree_centrality[name])
    
closeness_centrality = nx.closeness_centrality(component)
for name in closeness_centrality:
    c[name].append(closeness_centrality[name])
    
betweenness_centrality = nx.betweenness_centrality(component)
for name in betweenness_centrality:
    c[name].append(betweenness_centrality[name])
    
information_centrality = nx.information_centrality(component)
for name in information_centrality:
    c[name].append(information_centrality[name])
    
data = {name: lst for name, lst in c.items() if lst}
df = pd.DataFrame(data=data).T
df.columns = ["c_d", "c_c", "c_b", "c_i"]
print(df)
```

–ß–µ–º –±–ª–∏–∂–µ –∞–∫—Ç–æ—Ä –∫ —Ü–µ–Ω—Ç—Ä—É, —Ç–µ–º –±–æ–ª—å—à—É—é —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –æ–Ω –∏–º–µ–µ—Ç. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –º—ã –º–æ–∂–µ–º —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥ –æ —Ç–æ–º, —á—Ç–æ Medici –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –±–ª–∏–∂–µ –≤—Å–µ—Ö –∫ —Ü–µ–Ω—Ç—Ä—É –ø–æ –≤—Å–µ–º —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—è–º.

#### 10. –í—ã—á–∏—Å–ª–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ —Ü–µ–Ω—Ç—Ä–∞–ª–∏–∑–∞—Ü–∏–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –º–µ—Ä—ã. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```python
g = len(names)
C_d = sum(max(df["c_d"]) - df["c_d"]) / (g - 1) / (g - 2)
C_c = sum(max(df["c_c"]) - df["c_c"]) / (g - 1) / (g - 2) * (2 * g - 3)
C_b = sum(max(df["c_b"]) - df["c_b"]) / (g - 1)
C_i = sum((df["c_i"] - sum(df["c_i"])) ** 2) / g
C_d, C_c, C_b, C_i
```

#### 11. –í—ã—è–≤–∏—Ç—å —Ç–æ—á–∫–∏ —Å–æ—á–ª–µ–Ω–µ–Ω–∏—è –∏ –º–æ—Å—Ç—ã —Å–µ—Ç–∏. –ü–æ–¥—Ç–≤–µ—Ä–¥–∏—Ç—å —Å–≤–æ–∏ –¥–æ–≥–∞–¥–∫–∏ —Å –ø–æ–º–æ—â—å—é —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π (–µ—Å–ª–∏ –∏—Ö –Ω–µ—Ç ‚Äì —Ä–µ–∞–ª–∏–∑—É–π—Ç–µ –∏—Ö —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ). –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ —Å–µ—Ç–∏, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º –≤—ã–¥–µ–ª–µ–Ω—ã —Ç–æ—á–∫–∏ —Å–æ—á–ª–µ–Ω–µ–Ω–∏—è –∏ –º–æ—Å—Ç—ã.

```python
art_points = [x for x in nx.articulation_points(network)]
bridges = [x for x in nx.bridges(network)]
pos = nx.kamada_kawai_layout(network)

# –†–∏—Å—É–µ–º –≤—Å–µ –≤–µ—Ä—à–∏–Ω—ã –∏ –≤—ã–¥–µ–ª—è–µ–º –∫—Ä–∞—Å–Ω—ã–º —Ç–æ—á–∫–∏ —Å–æ—á–ª–µ–Ω–µ–Ω–∏—è
nx.draw_networkx_nodes(network, pos)
nx.draw_networkx_nodes(network, pos, nodelist=art_points, node_color="tab:red")

# –†–∏—Å—É–µ–º —Ä–µ–±—Ä–∞ –∏ –≤—ã–¥–µ–ª—è–µ–º –∑–µ–ª–µ–Ω—ã–º –º–æ—Å—Ç—ã
nx.draw_networkx_edges(network, pos)
nx.draw_networkx_edges(network, pos, edgelist=bridges, edge_color="tab:green", width=5)
# –†–∏—Å—É–µ–º –∏–º–µ–Ω–∞ –∞–∫—Ç–æ—Ä–æ–≤
nx.draw_networkx_labels(network, pos, {name: name for name in names})
```

#### 13. –ó–∞–ø–æ–ª–Ω–∏—Ç—å —Ç–∞–±–ª–∏—Ü—É –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –º–µ—Ä –ø—Ä–µ—Å—Ç–∏–∂–∞.

```python
def I_i(n_i, network):
    return [node for node in network.nodes if nx.has_path(network, node, n_i)]

g = network.number_of_nodes()
c = {node: [] for node in network.nodes}

for name in network.nodes:
    c[name].append(network.in_degree(name) / (g - 1))

for n_i in network.nodes:
    I = I_i(n_i, network)
    c[n_i].append(len(I) ** 2 / (g - 1) / sum(nx.shortest_path_length(network, n_j, n_i) for n_j in I))

data = {name: lst for name, lst in c.items() if lst}
df = pd.DataFrame(data=data).T
df.columns = ['P_d', 'P_p']
df
```
