## Основные теоретические положения
**Акторы** - это дискретные,индивидуальные,корпоративные или коллективные социальные единицы.<br><br>
**Социоматрица** - матрица $X_{g×g}$,описывающая смежные узлы в сети: X={$x_{ij}$}, i=1,...,g (g-количество акторов).<br><br>

**Алгоритм Фрюхтермана-Рейнгольдса** - силовой алгоритм размещения графа.Его суть заключается в том,что между каждой парой вершин действует сила отталкивания.Если две вершины связаны ребром,то между ними действует сила притяжения.Силы притяжения и отталкивания вычисляются по следующим формулам:<br><br>
$F_{rep} (p_u,p_v) = l^2/x $ -сила отталкивания.<br>
$F_{attr} (p_u,p_v) = x^2/l $ -сила притяжения.<br>
Здесь $l= \sqrt{area/numberOfVertices}$ - идеальная длина,которая зависит от площади и количества вершин,$x$- расстояние между двумя вершинами.<br>

Сила пружины вычисляется как сумма силы отталкивания и сил притяжения: $F_{spring}(p_u,p_v) = F_{rep} (p_u,p_v) + F_{attr} (p_u,p_v)$<br><br>
**Алгоритм Камада-Каваи** - применяется для неориентированных взешенных графов.Граф представляется в виде динамической системы, в которой каждые две вершины соединены «пружиной» желаемой длины. Затем рассчитывается оптимальное представление графа – состояние, в котором суммарная энергия системы будет минимальной. Модель:<br> $E = \sum_{i=1}^{n-1}\sum_{j=i+1}^{n}{1\over2}k_{ij}{(|p_i-p_j|-l_{ij})}^2$,где <br> $n$-количество вершин, $p_i$ - положение вершины i, $k_{ij}$- сила пружины между $p_i$ и $p_j$, $l_{ij} = L× d_{ij}$,$L$-желаемая длина ребра в графе, $d_{ij}$ - кратчайший путь между вершинами $i$ и $j$,$k_{ij} = {K\over{d^2}},$K$-константа$ <br><br>

**Пятичисловая сводка сети:**

</ul>

 </li>

</ol>

<h5><strong> Средняя степень узлов:</strong></h5>
$\bar d = {1\over g} \sum_{i=1}^{g}d(n_i)$

<h5><strong>Дисперсия степеней узлов:</strong></h5>
${S_d}^2 = {1\over g} \sum_{i=1}^{g}{(d(n_i)-\bar d)}^2$


<h5><strong>Меры центральности:</strong></h5>
<ul>
<li>нормированная центральность по степени:${C_d}^\prime(n_i) = {d(n_i) \over g-1}$  </li>
<li>Нормированная центральность по близости:${C_c}^\prime(n_i) = {{g-1} \over {\sum_{j=1}^g d(n_i,n_j)}} = (g-1)C_c(n_i)$ </li>
<li>Нормированная центральность по посредничеству: $C'_B(n_i) = \frac{C_B(n_i)}{(g-1)(g-2)/2}$, где ${C_B(n_i)} = \sum_{j<k}\frac{g_{jk}(n_i)}{g_{jk}}$ </li>
<li>Нормированная центральность по информации: $C'_I(n_i) = \frac{C_I(n_i)}{\sum_iC_I(n_i)}$ </li>
</ul>

<h5><strong>Централизация:</strong></h5>
<ul>

<li>Централизация по посредничеству:  $C_B = \frac{2\sum_{i=1}^g(C'_B(n^*)-C'_B(n_i))}{(g-1)^2(g-2)}$</li>
</ul>

```r
five_digit_summary <-function(net){
  orientation <-get.network.attribute(net,"directed")
  graph <-intergraph::asIgraph(net)
  res = c()
  
  vertex_count <- network.size(net)
  
  density <- network.density(net)
  
  mode <- "weak"
  if(orientation){
    mode <- "strong"
  }
  components_count <- count_components(graph, mode = mode)
  
  net_diameter <- diameter(graph,directed = orientation)

  trans_mode <-"graph"
  if(orientation){
    trans_mode <- "digraph"
  }
  clustering_coefficient <- gtrans(net,mode = trans_mode)

  return (c(size = vertex_count,density = density,components_count = components_count,
            net_diameter = net_diameter, clustering_coefficient = clustering_coefficient))
}
print(five_digit_summary(net_1))
```

8) Вычислить среднюю степень узлов и дисперсию. Сделать выводы.

```r
calcualte_degree_variance <- function(degrees, mean_degree){
  temp <- (degrees - mean_degree)**2
  return (mean (temp))
}

directed_version <-function(graph){
  vertex_count <-vcount(graph)
  edge_count <- ecount(graph)
  mean_degree <- edge_count/vertex_count
  indegree_variance <-calcualte_degree_variance(degree(graph,mode= "in"),mean_degree)
  outdegree_variance <-calcualte_degree_variance(degree(graph,mode="out"),mean_degree)
  return(c(mean_degree = mean_degree, indegree_variance = indegree_variance,
           outdegree_variance = outdegree_variance))
}



average_degree_and_variance <- function(net){
  graph <-intergraph::asIgraph(net)
  if(get.network.attribute(net,"directed")){
    return(directed_version(graph))
  }
  
  vertex_count <- network.size(net)
  edge_count <-network.edgecount(net)
  average_degree <- (2*edge_count)/vertex_count
  
  
  dispersion <- calcualte_degree_variance(degree(graph),average_degree)
  return (c(dispersion = dispersion, average_degree = average_degree))
}

print(average_degree_and_variance(net_1))
```

Выводы: Каждая семья в среднем имеет по 2-3 связи, при этом средний разброс относительно этого значения составляет около 2.

9) Заполнить таблицу для различных мер центральности.$$C_D^\prime(n_i)\ , C_C^\prime(n_i)\ , C_B^\prime(n_i)\ , C_I^\prime(n_i)\ $$  Меры центральности вычислить для неизолированных вершин.

```r
isolated <-function(graph){
  all_degrees <- degree(graph)
  names(all_degrees) <-1:length(all_degrees)
  isolated_vertices <- all_degrees[all_degrees < 1]
  return (as.numeric(names(isolated_vertices)))
}

task_nine <- function(net){
  graph <-intergraph::asIgraph(net)
  graph <-set_vertex_attr(graph,"name",value = rownames(as.matrix(net)))
  to_delete <- isolated(graph)
  graph <- delete_vertices(graph,to_delete)
  vertex_count <-vcount(graph)
  all_degrees <- degree(graph)
  res1 <- all_degrees/(vertex_count -1)
 
  res2 <- closeness(graph)*(vertex_count-1)

  res3 <- betweenness(graph)/((vertex_count-1)*(vertex_count-2)/2)
 
  net <- network::delete.vertices(net,to_delete)
  info_centrality <- infocent(net, gmode = "graph")
  res4 <-info_centrality/sum(info_centrality)
  names(res4) <- rownames(as.matrix(net))
  
  return (data.frame( degree_centrality = res1, closeness_centrality = res2,
                between_centrality = res3, info_centrality = res4))
  
}
task_nine(net_1)


```

13) Заполнить таблицу для различных мер престижа:
$$P_D^\prime(n_i)\ , P_P^\prime(n_i)\  $$

```r
task_thirteen <- function(net){
  graph <- intergraph::asIgraph(net)
  
  degree_prestige <- degree(graph,mode= "in")/(vcount(graph)-1)
  paths <-shortest.paths(graph,mode = "out")
  domains <- apply(paths,2, function(x) length(which(is.finite(x)& x!=0)))
  domains_sum <- apply(paths,2,function(x) sum(x[is.finite(x)]))
  proximity_prestige <- (domains/(vcount(graph)-1))/(domains_sum/domains)
  return(data.frame(degree_prestige = degree_prestige, proximity_prestige = proximity_prestige))
  
}

task_thirteen(net_2)
```
