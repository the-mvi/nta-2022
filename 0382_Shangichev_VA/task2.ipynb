* n-–∫–ª—É–± - –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–¥–≥—Ä–∞—Ñ –¥–∏–∞–º–µ—Ç—Ä–∞ ‚â§ n  

* ùëò-–ø–ª–µ–∫—Å—ã.  
k-–ø–ª–µ–∫—Å - —ç—Ç–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–¥–≥—Ä–∞—Ñ, —Å–æ–¥–µ—Ä–∂–∞—â–∏–π $g_s$ —É–∑–ª–æ–≤, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–∞–∂–¥—ã–π —É–∑–µ–ª —Å–º–µ–∂–µ–Ω –Ω–µ –º–µ–Ω–µ–µ —á–µ–º $ (g_s - k) $ —É–∑–ª–∞–º–∏ –ø–æ–¥–≥—Ä–∞—Ñ–∞:
    
    $ d_s(i) >= g_s-k,  \forall n_i \in N_s $,  

    –≥–¥–µ $d_s(i)$ - —Å—Ç–µ–ø–µ–Ω—å —É–∑–ª–∞ $n_i$ –≤ –ø–æ–¥–≥—Ä–∞—Ñ–µ $G$

* ùëò-—è–¥—Ä–∞.
k-—è–¥—Ä–æ(k-core) - —ç—Ç–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–∞–∂–¥–∞—è –≤–µ—Ä—à–∏–Ω–∞ —Å–≤—è–∑–∞–Ω–∞ –º–∏–Ω–∏–º—É–º —Å k –¥—Ä—É–≥–∏–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏ —ç—Ç–æ–≥–æ –∂–µ –ø–æ–¥–≥—Ä–∞—Ñ–∞  
  * –≤–ª–æ–∂–µ–Ω–Ω—ã–µ
  * –Ω–µ –ø–µ—Ä–µ–∫—Ä—ã–≤–∞—é—Ç—Å—è
  * –ª–µ–≥–∫–æ –æ–ø—Ä–¥–µ–ª–µ–∏—Ç—å
* ùêøùëÜ-–º–Ω–æ–∂–µ—Å—Ç–≤–æ. ùúÜ-–º–Ω–æ–∂–µ—Å—Ç–≤–æ.  
(Seidman, 1983) –î–∞–Ω –ø–æ–¥–≥—Ä–∞—Ñ $\mathscr G_s$  —Å –º–Ω–æ–∂–µ—Å—Ç–≤–æ–º —É–∑–ª–æ–≤ $\mathscr N_s$. –ú–Ω–æ–∂–µ—Å—Ç–≤–æ —É–∑–ª–æ–º $\mathfrak N_s$ –±—É–¥–µ–º –Ω–∞–∑—ã–≤–∞—Ç—å LS-–º–Ω–æ–∂–µ—Å—Ç–≤–æ–º, –µ—Å–ª–∏ –ª—é–±–æ–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–µ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–æ $\mathscr{Q}$ ( $\mathscr{N}_s $–∏–º–µ–µ—Ç –±–æ–ª—å—à–µ –ª–∏–Ω–∏–π –∫ —É–∑–ª–∞–º $ \mathscr{N}_S$ \ L, —á–µ–º –∫ $\mathscr{N} $ \ 

* –ú–µ—Ä—ã —Å–ø–ª–æ—á—ë–Ω–Ω–æ—Å—Ç–∏ –ø–æ–¥–≥—Ä—É–ø–ø. –ú–µ—Ä–∞ –ë–æ–∫–∞‚Äì–•—É—Å–µ–π–Ω–∞, –º–µ—Ä–∞ –ê–ª—å–±—ã.  
(Bock and Husain, 1950) –î–∞–Ω–∞ —Å–µ—Ç—å –≤ –∫–æ—Ç–æ—Ä–æ–π #  $\mathscr N$ = g –∏ # $\mathscr N_S = g_s$. –ú–µ—Ä–∞ —Å–ø–ª–æ—á–µ–Ω–Ω–æ—Å—Ç–∏ –ø–æ–¥–≥—Ä—É–ø–ø –ø–æ –ë–æ–∫–µ –•—É—Å–µ–π–Ω—É –∏–º–µ–µ—Ç –≤–∏–¥:  
$ M_{BH} = (\dfrac{1}{g_s(g_s-1)}\sum_{i \in N_s} \sum_{j \in N_s} x_{ij})/(\dfrac{1}{g_s(g-g_s)}\sum_{i \in N_s} \sum_{j \notin N_s} x_{ij}) $  
(Alba (1973)) –î–∞–Ω–∞ —Å–µ—Ç—å —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ $\mathscr N = g $ –∏ $\mathscr L = L$. –í —Å–µ—Ç–∏ –µ—Å—Ç—å –ø–æ–¥–≥—Ä—É–ø–ø–∞ —Å –ø—Ä–∞–º–µ—Ç—Ä–∞–º–∏ $N_S = g_s$ –∏ $L_s = L_s$. –í –∫–∞—á–µ—Å—Ç–≤–µ –º–µ—Ä—ã –ê–ª—å–±—ã –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –≥–∏–ø–µ—Ä–≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –Ω–∞–±–ª—é–¥–∞–µ–º–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –≤ –ø–æ–¥–≥—Ä–∞—Ñ–µ —Ä–∞–≤–Ω–æ q:  
$ Pr(L_S = q) = \binom{L}{Q}\binom{\binom{g}{2}-L}{\binom{g_s}{2}-q}/\binom{\binom{g}{2}}{\binom{g_s}{2}} $
 
* –°–ª–∞–±–æ—Å–≤—è–∑–∞–Ω–Ω–∞—è n-–∫–ª–∏–∫–∞  - –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –≤—Å–µ —É–∑–ª—ã —Å–ª–∞–±–æ n-—Å–≤—è–∑–∞–Ω—ã,  –∏ –Ω–µ—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —É–∑–ª–æ–º, —Ç–∞–∫–∂–µ —Å–ª–∞–±–æ n-—Å–≤—è–∑–∞–Ω—ã, –∏ –Ω–µ—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —É–∑–ª–æ–≤, —Ç–∞–∫–∂–µ —Å–ª–∞–±–æ n-—Å–≤—è–∑–Ω—ã—Ö —Å–æ –≤—Å–µ–º–∏ —É–∑–ª–∞–º–∏ –ø–æ–¥–≥—Ä–∞—Ñ–∞.

* –û–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ —Å–≤—è–∑–∞–Ω–Ω–∞—è n-–∫–ª–∏–∫–∞(unilaterally connected n-clique) - –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ–æ—Ç–æ—Ä–æ–º –≤—Å–µ —É–∑–ª—ã –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ n-—Å–≤—è–∑–∞–Ω—ã, –∏ –Ω–µ—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —É–∑–ª–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ —Ç–∞–∫–∂–µ –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ n-—Å–≤—è–∑–∞–Ω—ã —Å–æ –≤—Å–µ–º–∏ —É–∑–ª–∞–º–∏ –ø–æ–¥–≥—Ä–∞s—Ñ–∞

* –°–∏–ª—å–Ω–æ —Å–≤—è–∑–Ω–∞—è n-–∫–ª–∏–∫–∞ (strongly connected n-clique) - –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –≤—Å–µ —É–∑–ª—ã —Å–∏–ª—å–Ω–æ n-—Å–≤—è–∑–∞–Ω—ã, –∏ –Ω–µ—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —É–∑–ª–æ–≤, —Ç–∞–∫–∂–µ —Å–∏–ª—å–Ω–æ n-—Å–≤—è–∑–Ω—ã—Ö —Å–æ –≤—Å–µ–º–∏ —É–∑–ª–∞–º–∏ –ø–æ–¥–≥—Ä–∞—Ñ–∞  

* –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –º–æ–¥—É–ª—è—Ä–Ω–æ—Å—Ç–∏.  
–ú–æ–¥—É–ª—è—Ä–Ω–æ—Å—Ç—å(modularity) - —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã —Å–µ—Ç–∏, –≤ —á–∞—Å—Ç–Ω–æ—Å—Ç–∏ —Å—Ç–µ–ø–µ–Ω—å –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ —É–∑–ª–æ–≤, –∫–æ–≥–¥–∞ –≤–Ω—É—Ç—Ä–∏ –∫–ª–∞—Å—Ç–µ—Ä–∞ –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è –º–Ω–æ–≥–æ —É–∑–ª–æ–≤ (–≤—ã—Å–æ–∫–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å), –∞ –º–µ–∂–¥—É –∫–ª–∞—Å—Ç–µ—Ä–∞–º–∏ –º–∞–ª–æ —É–∑–ª–æ–≤(–Ω–∏–∑–∫–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å)
    - –ü–æ–∫–∞–∑–∞—Ç–µ–ª—å –º–æ–¥—É–ª—è—Ä–Ω–æ—Å—Ç–∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∫–∞–∫ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –¥–æ–ª–µ —Å–≤—è–∑–µ–π –≤–Ω—É—Ç—Ä–∏ –≥—Ä—É–ø–ø—ã –∏ –æ–∂–∏–¥–∞–µ–º–æ–π –¥–æ–ª–µ–π —Å–≤—è–∑–µ–π, –µ—Å–ª–∏ –±—ã —Å–≤—è–∑–∏ –±—ã–ª–∏ —Ä–∞–∑–º–µ—â–µ–Ω—ã —Å–ª—É—á–∞–π–Ω–æ
    - –ü–æ–∫–∞–∑–∞—Ç–µ–ª—å –º–æ–¥—É–ª—è—Ä–Ω–æ—Å—Ç–∏ –º–æ–∂–µ—Ç –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ –æ—Ç -1/2 –¥–æ +1
    - –ß–µ–º –≤—ã—à–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å, —Ç–µ–º –±–æ–ª—å—à–µ —É—Ä–æ–≤–µ–Ω—å –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ —Å–µ—Ç–∏ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –¥–∞–Ω–Ω–æ–π –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–æ–π —É–∑–ª–æ–≤  

* –°–µ—Ç—å –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ.  
–°–µ—Ç—å –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏ (affilation network) - —Å–µ—Ç—å, –≤ –∫–æ—Ç–æ—Ä–æ–π —É—á–∞—Å—Ç–Ω–∏–∫–∏ –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω—ã –¥—Ä—É–≥ —Å –¥—Ä—É–≥–æ–º –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å–æ—Ç—Ä—É–¥–Ω–∏—á–µ—Å—Ç–≤–∞ –≤ —Ä–∞–º–∫–∞—Ö –∫–∞–∫–æ–π-—Ç–æ –≥—Ä—É–ø–ø—ã –∏–ª–∏ —Å–æ–≤–º–µ—Å—Ç–Ω–æ–≥–æ —É—á–∞—Å—Ç–∏—è –≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º —Å–æ–±—ã—Ç–∏–∏.
    - –°–µ—Ç—å —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –¥–≤—É—Ö –∏–∑–º–µ—Ä–µ–Ω–∏–π (two-modes):  
    –º–Ω–æ–∂–µ—Å—Ç–≤–æ –∞–∫—Ç–æ—Ä–æ–≤ $\mathscr N = \{n_1, n_2, ..., n_g\}$  
    –∏ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ —Å–æ–±—ã—Ç–∏–π (events) $\mathscr M = \{m_1, m_2, ..., m_h\}$

```r
ncliques <- kCliques(igraph.to.graphNEL(graph_data))
ncliques
```

```r
result = c()
for (n in names(ncliques)){
    cand <- ncliques[[n]]
    clan <- list()
    i <- 1
    for (cand.i in 1:length(cand)){
        g.n.cand <- induced_subgraph(graph_data, cand[[cand.i]])
        if (diameter(g.n.cand) <= n){ 
          clan[[i]] <- cand[[cand.i]]
          i <- i+1
        }
      }
    result[[n]] <- clan
}



```
