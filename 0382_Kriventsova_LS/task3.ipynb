–í–∞—Ä–∏–∞–Ω—Ç ‚Ññ37 –í—ã–ø–æ–ª–Ω–∏–ª–∞ –ö—Ä–∏–≤–µ–Ω—Ü–æ–≤–∞ –õ.–°., —Å—Ç—É–¥–µ–Ω—Ç–∫–∞ –≥—Ä. 0382 –ü–ú–ò –õ–≠–¢–ò.

## –¶–µ–ª—å —Ä–∞–±–æ—Ç—ã
–û–≤–ª–∞–¥–µ—Ç—å –Ω–∞–≤—ã–∫–∞–º–∏ —Ä–∞–±–æ—Ç—ã —Å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º–∏ –ø–∞–∫–µ—Ç–∞–º–∏ –¥–ª—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ —Å–æ—Ü–∏–∞–ª—å–Ω—ã—Ö —Å–µ—Ç–µ–π.

## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è
#### $\textbf{–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥—É–≥ –¥–ª—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.}$
* $\textbf{–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–≥.}$ 
$$L \sim B(g(g - 1), \frac{1}{2})$$

* $\textbf{–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–≥ —Ä–∞–≤–Ω–æ}$ $l$:
$$P(L = l) = \binom {g(g - 1)}{l} \cdot  2^{g(1 - g)}$$

* $\textbf{–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ.}$
$$E(L) = \binom {g}{2}$$

* $\textbf{–î–∏—Å–ø–µ—Ä—Å–∏—è.}$
$$D(L) = \frac{1}{2} \binom {g}{2}$$

#### $\textbf{–°—Ç–∞—Ç–∏—á–µ—Å–∫–∏–µ –≥–∏–ø–æ—Ç–µ–∑—ã –¥–ª—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:}$
* $\textbf{–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞}$ $H_0$. 
$$L \sim B(g(g - 1), \frac{1}{2})$$

* $\textbf{–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞.}$
$$z_l = \frac{l - E(L)}{\sqrt {D(L)}} = \frac {l-\frac {g(g-1)}{2}}{\sqrt {\frac {g(g-1)}{4}}},$$

–ì–¥–µ $l = \sum_{i = 1}^g\sum_{j = 1}^gx_{ij}$

* –¢–∞–∫–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —è–≤–ª—è–µ—Ç—Å—è –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏.
#### $\textbf{–í–∞–∂–Ω—ã–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:}$

$\textbf{–î–∏–∞–¥–∞}$ - —ç–ª–µ–º–µ–Ω—Ç —Å–µ—Ç–∏, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∏–π —Å–æ–±–æ–π –ø–∞—Ä—É –∞–∫—Ç–æ—Ä–æ–≤ –∏ –≤–æ–∑–º–æ–∂–Ω–Ω—É—é —Å–≤—è–∑—å –º–µ–∂–¥—É –Ω–∏–º–∏.

$\textbf{–í–∏–¥—ã –¥–∏–∞–¥.}$
* –í–∑–∞–∏–º–Ω–∞—è –¥–∏–∞–¥–∞: $D_{i,j} = (1, 1)$

* –ê—Å—Å–∏–º–µ—Ç—Ä–∏—á–Ω–∞—è –¥–∏–∞–¥–∞: $D_{i,j} = (1, 0)$, $D_{i,j} = (0, 1)$

* –ù—É–ª–µ–≤–∞—è –¥–∏–∞–¥–∞:  $D_{i,j} = (0, 0)$

$\textbf{–¶–µ–Ω–∑—É—Å –¥–∏–∞–¥}$ <br> - —Ç—Ä–æ–π–∫–∞ —á–∏—Å–µ–ª $<M, A, N>$, <br>–ì–¥–µ:
* $M = \sum_{i < j}(x_{ij}x_{ji})$

* $A = x_{++} - 2M$

* $N = C^2_g - A - M$

$\textbf{–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ö–∞—Ü–∞-–ü–∞—É—ç–ª–ª–∞}$ <br>- —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞ –∏–∑–º–µ—Ä–µ–Ω–∏—è —Ç–µ–Ω–¥–µ–Ω—Ü–∏–∏ –∞–∫—Ç–æ—Ä–æ–≤ –≤ –≥—Ä—É–ø–ø–µ.

$$-\infty < \rho_{kp} \leq 1$$

$\textbf{–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –æ—Å–Ω–æ–≤–∞–Ω}$ –Ω–∞ –æ–∂–∏–¥–∞–Ω–∏–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥, –ø—Ä–µ–¥–ø–æ–ª–≥–∞—è, —á—Ç–æ –≤—ã–±–æ—Ä –¥–µ–ª–∞–µ—Ç—Å—è –∞–∫—Ç–æ—Ä–∞–º–∏ —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º.

$\textbf{–ü—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ:}$

* –û—Ü–µ–Ω–∫–∞ $p_{kp}$ –ø–æ –º–µ—Ç–æ–¥—É –º–æ–º–µ–Ω—Ç–æ–≤.
$${\rho{'}_{kp} = \frac {2(g-1)^2M-L^2+L_2}{L(g-1)^2-L^2+L_2}},$$
–≥–¥–µ $L = \sum_i x_{i+}$, $L_2 = \sum_i x^2_{i+}$

* –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –∞–∫—Ç–æ—Ä—ã –∏–º–µ—é—Ç –≤–∑–∞–∏–º–Ω—ã–µ –æ—Ç–Ω–æ—à–µ–Ω–∏—è.
$$\frac {x_{i+}x_{j+}}{(g-1)^2}$$

$\textbf{–ü—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –≤—ã–±–æ—Ä–µ:}$

* –ö–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä –¥–µ–ª–∞–µ—Ç $d$ –≤—ã–±–æ—Ä–æ–≤ –∏–∑ $(g-1)$ –∞–∫—Ç–æ—Ä–æ–≤, –¥–æ—Å—Ç—É–ø–Ω—ã—Ö –¥–ª—è –≤—ã–±–æ—Ä–∞.

* –û—Ü–µ–Ω–∫–∞ $\rho_{kp}$ –ø–æ –º–µ—Ç–æ–¥—É –º–æ–º–µ–Ω—Ç–æ–≤.
$$\hat \rho_{kp} = \frac {2(g-1)M-gd^2}{gd(g-1-d)}$$

* –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤–∑–∞–∏–º–Ω–æ–≥–æ –≤—ã–±–æ—Ä–∞ –º–µ–∂–¥—É –ª—é–±–æ–π –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–π –ø–∞—Ä–æ–π –∞–∫—Ç–æ—Ä–æ–≤.
$$\frac {d^2}{(g-1)^2}$$

$\textbf{–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è}$ <br>- –¥–∞–Ω–Ω—ã–π –∏–Ω–¥–µ–∫—Å –æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–º –∑–Ω–∞—á–µ–Ω–∏—è—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥.
$$\rho_{B} = \frac{M - M_{min}}{M_{max} - M_{min}},$$
–ì–¥–µ:
* $M_{min} = max_{0 \leq t \leq g}f(t)$

* $M_{max} = [\frac{1}{2}(\sum_{i=1}^g - max_{0 \leq t \leq g}g(t))]$

* $f(t) = \sum_{i=1}^tx_{i+} - t(g-1) - C^2_t$ 

* $g(t) = \sum_{i=1}^tx_{i+} - t(t-1) - \sum_{i=t+1}^gmin(t, x_{i+})$

#### $\textbf{–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –≥—Ä–∞—Ñ–∞—Ö.}$
* $\#G_d(\cal {N}) = 2^{g(g - 1)}$, –≥–¥–µ $G_d(\cal {N})$ - –º–Ω–æ–∂–µ—Å—Ç–≤–æ –≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö –æ—Ä–≥—Ä–∞—Ñ–æ–≤ –±–µ–∑ –ø–µ—Ç–µ–ª—å —Å g —É–∑–ª–∞–º–∏

* $\textbf{–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å —Å–æ—Ü–∏–æ–º–∞—Ç—Ä–∏—Ü—É}$ $x$.
$$P(X = x) = 2^{g(g - 1)}$$

* $\textbf{–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—è–≤–ª–µ–Ω–∏—è –¥—É–≥–∏ –º–µ–∂–¥—É –∞–∫—Ç–æ—Ä–∞–º–∏ –¥–ª—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.}$
$$P(X_{ij} = 1) = \begin{cases}
    \frac{1}{2}, i \ne j\\
    0, i = j
\end{cases}$$

* $\textbf{–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—è–≤–ª–µ–Ω–∏—è –¥—É–≥–∏ –º–µ–∂–¥—É –∞–∫—Ç–æ—Ä–∞–º–∏ –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ë–µ—Ä–Ω—É–ª–ª–∏.}$
$$P(X_{ij} = 1) = \begin{cases}
    p_{ij}, i \ne j\\
    0, i = j
\end{cases}$$
####  $\textbf{–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –≥–∏–ø–æ—Ç–µ–∑—ã –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ë–µ—Ä–Ω—É–ª–ª–∏}$:
*  $\textbf{–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞}$ $H_0$.
$$L \sim B(g(g - 1), P_0)$$

*  $\textbf{–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞.}$
$$z_l = \frac{l - E(L)}{\sqrt {D(L)}} = \frac {l-P_0g(g-1)}{\sqrt {P_0(1-P_0)g(g-1)}},$$
–≥–¥–µ $l = \sum_{i = 1}^g\sum_{j = 1}^gx_{ij}$

* –¢–∞–∫–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —è–≤–ª—è–µ—Ç—Å—è –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏.

####  $\textbf{–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –æ—Ü–µ–Ω–∫–∏ $P$:}$
* $\textbf{–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $P$.}$
$$\hat {P} = \frac{l}{g(g - 1)}$$

* $\textbf{–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª.}$
$$\hat {P}_{lower} \leq P \leq  \hat {P}_{upper}$$

* $\hat {P}_{lower} = \hat {P} - z_{\frac {\alpha}{2}} \sqrt {\frac {\hat {P}(1 - \hat{P})}{g(g - 1)}}$

* $\hat {P}_{upper} = \hat {P} + z_{\frac {\alpha}{2}} \sqrt {\frac {\hat {P}(1 - \hat{P})}{g(g - 1)}}$

* $z_{\frac{\alpha}{2}}$: $\Phi(z_{\frac{\alpha}{2}}) = 0.5 - \frac{\alpha}{2}$

####  $\textbf{–û—Ü–µ–Ω–∫–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ $P$ –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ë–µ—Ä–Ω—É–ª–ª–∏:}$
*  $\textbf{–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–≥.}$
$$L \sim B(g(g - 1), P_0)$$

*  $\textbf{–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–≥ —Ä–∞–≤–Ω–æ}$ $l$.
$$P(L = l) = \binom {g(g - 1)}{l} = P_0^l(1 - P_0)^{g(g - 1) - l}$$

*  $\textbf{–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ.}$
$$E(L) = P_0g(g - 1)$$

*  $\textbf{–î–∏—Å–ø–µ—Ä—Å–∏—è.}$
$$D(L) = P_0(1 - P_0)g(g - 1)$$




## –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏
–° –ø–æ–º–æ—â—å—é –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö –ø–∞–∫–µ—Ç–æ–≤ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫ –ø—Ä–æ–≤–µ—Å—Ç–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ —Å–ª—É—á–∞–π–Ω–æ —Å–º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∞. –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ–¥–µ—Ä–∂–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å.

##### –ü—É–Ω–∫—Ç 1.
–° –ø–æ–º–æ—â—å—é –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ igraph —Å–º–æ–¥–µ–ª–∏—Ä–æ–≤–∞—Ç—å —Å–ª—É—á–∞–π–Ω—ã–π –æ—Ä–≥—Ä–∞—Ñ $\cal G_d$ –Ω–∞ –æ—Å–Ω–æ–≤–µ –º–æ–¥–µ–ª–∏ –≠—Ä–¥–µ—à–∞‚Äî–†–µ–Ω—å–∏: $\cal G_ùëë = \cal {G}(ùëî, ùëù)$, –≥–¥–µ ùëî ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–æ—Ä–æ–≤, ùëù ‚Äì –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—è–≤–ª–µ–Ω–∏—è –¥—É–≥–∏ –º–µ–∂–¥—É –∞–∫—Ç–æ—Ä–∞–º–∏. –í R –¥–ª—è –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∞ –≠—Ä–¥–µ—à–∞‚Äî–†–µ–Ω—å–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ñ—É–Ω–∫—Ü–∏—è erdos.renyi.game(). –û–∑–Ω–∞–∫–æ–º—å—Ç–µ—Å—å —Å –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª—å—é –∏ –æ–ø–∏—à–∏—Ç–µ –µ—ë.

–í –º–æ–¥–µ–ª–∏ $G(n,p)$ –≥—Ä–∞—Ñ —Å—Ç—Ä–æ–∏—Ç—Å—è –ø—É—Ç—ë–º —Å–ª—É—á–∞–π–Ω–æ–≥–æ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è —Ä—ë–±–µ—Ä. –ö–∞–∂–¥–æ–µ —Ä–µ–±—Ä–æ –≤–∫–ª—é—á–∞–µ—Ç—Å—è –≤ –≥—Ä–∞—Ñ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é p –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Ä—ë–±–µ—Ä. –≠–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ, –≤—Å–µ –≥—Ä–∞—Ñ—ã —Å n —É–∑–ª–∞–º–∏ –∏ M —Ä—ë–±—Ä–∞–º–∏ –∏–º–µ—é—Ç –æ–¥–∏–Ω–∞–∫–æ–≤—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å:

${p^{M}(1-p)^{{n \choose 2}-M}}$

```r
g <- 62
p <- 0.78
gr <-erdos.renyi.game(g, p, type ='gnp', directed = TRUE)
gr
```

##### –ü—É–Ω–∫—Ç 2.
–û—Ç–æ–±—Ä–∞–∑–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–π –æ—Ä–≥—Ä–∞—Ñ.

```r
plot(gr, main = '–°–ª—É—á–∞–π–Ω—ã–π –æ—Ä–≥—Ä–∞—Ñ –Ω–∞ –æ—Å–Ω–æ–≤–µ –º–æ–¥–µ–ª–∏ –≠—Ä–¥–µ—à–∞‚Äî–†–µ–Ω—å–∏')
```

##### –ü—É–Ω–∫—Ç 3.
 –í—ã—á–∏—Å–ª–∏—Ç—å ¬´–ø—è—Ç–∏—á–∏—Å–ª–æ–≤—É—é —Å–≤–æ–¥–∫—É¬ª –¥–ª—è –≥—Ä–∞—Ñ–∞ $\cal {G}_d$.

```r
grAsNetwork <- asNetwork(gr)

cat("–†–∞–∑–º–µ—Ä:\n", network.size(grAsNetwork))

cat("\n–ü–ª–æ—Ç–Ω–æ—Å—Ç—å:\n", gden(grAsNetwork))

cat("\n–ö–æ–º–ø–æ–Ω–µ–Ω—Ç—ã:\n", igraph::components(gr)$no)

components <- component.largest(grAsNetwork,result="graph")
gd <- geodist(components)
cat("\n–î–∏–∞–º–µ—Ç—Ä:\n", max(gd$gdist))

cat("\n–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏:", gtrans(grAsNetwork))
```

##### –ü—É–Ω–∫—Ç 5.
–í—ã—á–∏—Å–ª–∏—Ç—å –æ—Ü–µ–Ω–∫—É –∏–Ω–¥–µ–∫—Å–∞ –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ö–∞—Ü–∞‚Äî–ü–∞—É—ç–ª–∞ –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–æ–º –≤—ã–±–æ—Ä–µ –∞–∫—Ç–æ—Ä–æ–≤ ùëë –∏ –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

–í—ã–≤–æ–¥—ã:
–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –æ—Ü–µ–Ω–∫—É –∏–Ω–¥–µ–∫—Å–∞ –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ö–∞—Ü–∞-–ü–∞—É—ç–ª–∞ –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–æ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –≤—ã–±–æ—Ä–æ–≤ $d$.
–í—ã–±–µ—Ä–µ–º –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ $d = 26$. –ö–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä –≤—ã–±–∏—Ä–∞–µ—Ç –∏–∑ $g - 1 = 62 - 1 = 61$ –∞–∫—Ç–æ—Ä–æ–≤.
–ü—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –≤—ã–±–æ—Ä–µ –∞–∫—Ç–æ—Ä—ã –ø—Ä–µ–¥–ø–æ—á–∏—Ç–∞—é—Ç –æ–±—Ä–∞–∑–æ–≤—ã–≤–∞—Ç—å –≤–∑–∞–∏–º–Ω—ã–µ —Å–≤—è–∑–∏ (1.7352).
–í—ã—á–∏—Å–ª–∏–º –æ—Ü–µ–Ω–∫—É –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ –∏ –ø–æ–ª—É—á–∏–º, —á—Ç–æ –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ –∞–∫—Ç–æ—Ä—ã —Å–∫–ª–æ–Ω–Ω—ã –æ–±—Ä–∞–∑–æ–≤—ã–≤–∞—Ç—å –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–µ —Å–≤—è–∑–∏ (–∏–ª–∏ –Ω–µ –æ–±—Ä–∞–∑–æ–≤—ã–≤–∞—Ç—å –≤–æ–≤—Å–µ) (0.02673543).

```r
g <- length(V(gr))
d <- 26 # –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ
M <- census$mut
P_fix <- (2*M*(g-1) - g*(d**2))/(g*d*(g-1-d))
L <- length(E(gr))
L2 <- 0
matrix1 <- as.matrix(net)
for(i in 1:nrow(matrix1)){
    sum1 <- 0
    for (j in 1:ncol(matrix1)){
        sum1 <- sum1 + matrix1[i, j]
    }
    L2 <- L2 + (sum1**2)
}
P_free <- (2*M*((g-1)**2)-L**2 + L2)/(L*((g-1)**2)-L**2+L2)
cat(P_fix, " ",P_free)
```

##### –ü—É–Ω–∫—Ç 7.
–í—ã—á–∏—Å–ª–∏—Ç—å –∏–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```r
vec <- c()
vec2 <- c()
for(t in 0:g){
    s <- 0
    if (t>0){
        for(i in 1:t){
            s <- s + sum(matr[i, ])
        }
    }
    f_t <- s - t*(g-1) - choose(t, 2)
    vec <- append(vec, f_t)
    s2 <- 0
    if (t+1 <= g){
        for(i in (t+1):g){
            s2 <- s2 + min(t, sum(matr[i, ]))
        }
    }
    g_t <- s - t*(t-1)-s2
    vec2 <- append(vec2, g_t)
}
M_min <- max(vec)
M_max <- floor(0.5*(sum(matr)-max(vec2)))
P_B <- (M - M_min)/(M_max - M_min)
P_B
```

–í—ã–≤–æ–¥—ã:
–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è —Ä–∞–≤–µ–Ω –ø—Ä–∏–º–µ—Ä–Ω–æ $p = 0.78$.

##### –ü—É–Ω–∫—Ç 8.
–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —á–∏—Å–ª–æ –¥—É–≥ $ùêø$ –≤ –æ—Ä–≥—Ä–∞—Ñ–µ $\cal G_ùëë$ –∏–º–µ–µ—Ç –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ $L$~$Bin(g(g-1), 1/2)$  –ø—Ä–∏
—É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $\alpha = 0.05$. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è.

```r
cat("alpha = ", 0.05)
binom.test(L, g*(g-1), p = 0.5)
```

–¢–∞–∫ –∫–∞–∫ p-value –æ—á–µ–Ω—å –º–∞–ª–æ (–ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å $\alpha$), —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è.
–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è —Ä–µ–±—Ä–∞ –º–µ–∂–¥—É –ª—é–±—ã–º–∏ –¥–≤—É–º—è –∞–∫—Ç–æ—Ä–∞–º–∏ –≤ —Ç–∞–∫–æ–º —Å–ª—É—á–∞–µ –Ω–µ —Ä–∞–≤–Ω–∞ 0.5. –¢–æ–≥–¥–∞ p-value –∏ —è–≤–ª—è–µ—Ç—Å—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º —É—Ä–æ–≤–Ω–µ–º –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è $p-value < 2.2e-16.$


##### –ü—É–Ω–∫—Ç 9.
–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —á–∏—Å–ª–æ –¥—É–≥ $L$ –≤ –æ—Ä–≥—Ä–∞—Ñ–µ $\cal G_ùëë$ –∏–º–µ–µ—Ç –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ $L$~$Bin(g(g-1), P_0)$  –ø—Ä–∏
—É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $\alpha$. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è.

```r
binom.test(L, g*(g-1), p = 0.57)
```

–¢–∞–∫ –∫–∞–∫ $p-value$ –æ—á–µ–Ω—å –º–∞–ª–æ (–ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å $\alpha$), —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è.
–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è —Ä–µ–±—Ä–∞ –º–µ–∂–¥—É –ª—é–±—ã–º–∏ –¥–≤—É–º—è –∞–∫—Ç–æ—Ä–∞–º–∏ –≤ —Ç–∞–∫–æ–º —Å–ª—É—á–∞–µ –Ω–µ —Ä–∞–≤–Ω–∞ p0 = 0.57. –¢–æ–≥–¥–∞ p-value –∏ —è–≤–ª—è–µ—Ç—Å—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º —É—Ä–æ–≤–Ω–µ–º –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è $p-value < 2.2e-16.$


##### –ü—É–Ω–∫—Ç 10.
–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $p$ —Å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–π –Ω–∞–¥—ë–∂–Ω–æ—Å—Ç—å—é $(1 - \alpha)$. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```r
P_hat <- L/(g*(g-1))
f_z <- 0.5 - 0.05/2
z_a2 <- 2.58 # –¢–∞–±–ª–∏—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —Ñ—É–Ω–∫—Ü–∏–∏ –õ–∞–ø–ª–∞—Å–∞ f_z = 0.475
P_lower <- P_hat-z_a2*sqrt((P_hat*(1-P_hat))/(g*(g-1)))
P_upper <- P_hat+z_a2*sqrt((P_hat*(1-P_hat))/(g*(g-1)))
cat('[', P_lower, ',', P_upper, ']')
```

–¢–∞–∫ –∫–∞–∫ $p = 0.78$, —Ç–æ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –≤—ã—á–∏—Å–ª–µ–Ω –≤–µ—Ä–Ω–æ (–ø—Ä–∏ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–π –Ω–∞–¥—ë–∂–Ω–æ—Å—Ç–∏ $1 - \alpha$ .
