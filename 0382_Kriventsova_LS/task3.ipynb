## Основные теоретические положения
#### $\textbf{Статистический анализ количества дуг для равномерного распределения.}$
* $\textbf{Количество дуг.}$ 
$$L \sim B(g(g - 1), \frac{1}{2})$$

* $\textbf{Вероятность, что количество дуг равно}$ $l$:
$$P(L = l) = \binom {g(g - 1)}{l} \cdot  2^{g(1 - g)}$$

* $\textbf{Математическое ожидание.}$
$$E(L) = \binom {g}{2}$$

* $\textbf{Дисперсия.}$
$$D(L) = \frac{1}{2} \binom {g}{2}$$

#### $\textbf{Статические гипотезы для равномерного распределения:}$
* $\textbf{Нулевая гипотеза}$ $H_0$. 
$$L \sim B(g(g - 1), \frac{1}{2})$$

* $\textbf{Статистика.}$
$$z_l = \frac{l - E(L)}{\sqrt {D(L)}} = \frac {l-\frac {g(g-1)}{2}}{\sqrt {\frac {g(g-1)}{4}}},$$

Где $l = \sum_{i = 1}^g\sum_{j = 1}^gx_{ij}$

* Такая статистика является приблизительно стандартным нормальным распределением при заданном уровне значимости.
#### $\textbf{Важные определения:}$

$\textbf{Диада}$ - элемент сети, представляющий собой пару акторов и возможнную связь между ними.

$\textbf{Виды диад.}$
* Взаимная диада: $D_{i,j} = (1, 1)$

* Ассиметричная диада: $D_{i,j} = (1, 0)$, $D_{i,j} = (0, 1)$

* Нулевая диада:  $D_{i,j} = (0, 0)$

$\textbf{Цензус диад}$ <br> - тройка чисел $<M, A, N>$, <br>Где:
* $M = \sum_{i < j}(x_{ij}x_{ji})$

* $A = x_{++} - 2M$

* $N = C^2_g - A - M$

$\textbf{Индекс взаимности Каца-Пауэлла}$ <br>- характеристика измерения тенденции акторов в группе.

$$-\infty < \rho_{kp} \leq 1$$

$\textbf{Индекс взаимности основан}$ на ожидании количества взаимных диад, предполгая, что выбор делается акторами случайным образом.

$\textbf{При свободном выборе:}$

* Оценка $p_{kp}$ по методу моментов.
$${\rho{'}_{kp} = \frac {2(g-1)^2M-L^2+L_2}{L(g-1)^2-L^2+L_2}},$$
где $L = \sum_i x_{i+}$, $L_2 = \sum_i x^2_{i+}$

* Вероятность того, что акторы имеют взаимные отношения.
$$\frac {x_{i+}x_{j+}}{(g-1)^2}$$

$\textbf{При фиксированном выборе:}$

* Каждый актор делает $d$ выборов из $(g-1)$ акторов, доступных для выбора.

* Оценка $\rho_{kp}$ по методу моментов.
$$\hat \rho_{kp} = \frac {2(g-1)M-gd^2}{gd(g-1-d)}$$

* Вероятность взаимного выбора между любой произвольной парой акторов.
$$\frac {d^2}{(g-1)^2}$$

$\textbf{Индекс взаимности Бандиопадхьяйя}$ <br>- данный индекс основан на минимальном и максимальном значениях количества взаимных диад.
$$\rho_{B} = \frac{M - M_{min}}{M_{max} - M_{min}},$$
Где:
* $M_{min} = max_{0 \leq t \leq g}f(t)$

* $M_{max} = [\frac{1}{2}(\sum_{i=1}^g - max_{0 \leq t \leq g}g(t))]$

* $f(t) = \sum_{i=1}^tx_{i+} - t(g-1) - C^2_t$ 

* $g(t) = \sum_{i=1}^tx_{i+} - t(t-1) - \sum_{i=t+1}^gmin(t, x_{i+})$

#### $\textbf{Распределения на графах.}$
* $\#G_d(\cal {N}) = 2^{g(g - 1)}$, где $G_d(\cal {N})$ - множество всех возможных орграфов без петель с g узлами

* $\textbf{Вероятность получить социоматрицу}$ $x$.
$$P(X = x) = 2^{g(g - 1)}$$

* $\textbf{Вероятность появления дуги между акторами для равномерного распределения.}$
$$P(X_{ij} = 1) = \begin{cases}
    \frac{1}{2}, i \ne j\\
    0, i = j
\end{cases}$$

* $\textbf{Вероятность появления дуги между акторами для распределения Бернулли.}$
$$P(X_{ij} = 1) = \begin{cases}
    p_{ij}, i \ne j\\
    0, i = j
\end{cases}$$
####  $\textbf{Статистические гипотезы для распределения Бернулли}$:
*  $\textbf{Нулевая гипотеза}$ $H_0$.
$$L \sim B(g(g - 1), P_0)$$

*  $\textbf{Статистика.}$
$$z_l = \frac{l - E(L)}{\sqrt {D(L)}} = \frac {l-P_0g(g-1)}{\sqrt {P_0(1-P_0)g(g-1)}},$$
где $l = \sum_{i = 1}^g\sum_{j = 1}^gx_{ij}$

* Такая статистика является приблизительно стандартным нормальным распределением при заданном уровне значимости.

####  $\textbf{Доверительный интервал для оценки $P$:}$
* $\textbf{Оценка параметра $P$.}$
$$\hat {P} = \frac{l}{g(g - 1)}$$

* $\textbf{Доверительный интервал.}$
$$\hat {P}_{lower} \leq P \leq  \hat {P}_{upper}$$

* $\hat {P}_{lower} = \hat {P} - z_{\frac {\alpha}{2}} \sqrt {\frac {\hat {P}(1 - \hat{P})}{g(g - 1)}}$

* $\hat {P}_{upper} = \hat {P} + z_{\frac {\alpha}{2}} \sqrt {\frac {\hat {P}(1 - \hat{P})}{g(g - 1)}}$

* $z_{\frac{\alpha}{2}}$: $\Phi(z_{\frac{\alpha}{2}}) = 0.5 - \frac{\alpha}{2}$

####  $\textbf{Оценка вероятности $P$ для распределения Бернулли:}$
*  $\textbf{Количество дуг.}$
$$L \sim B(g(g - 1), P_0)$$

*  $\textbf{Вероятность, что количество дуг равно}$ $l$.
$$P(L = l) = \binom {g(g - 1)}{l} = P_0^l(1 - P_0)^{g(g - 1) - l}$$

*  $\textbf{Математическое ожидание.}$
$$E(L) = P_0g(g - 1)$$

*  $\textbf{Дисперсия.}$
$$D(L) = P_0(1 - P_0)g(g - 1)$$

##### Пункт 1.
С помощью библиотеки igraph смоделировать случайный орграф $\cal G_d$ на основе модели Эрдеша—Реньи: $\cal G_𝑑 = \cal {G}(𝑔, 𝑝)$, где 𝑔 – количество акторов, 𝑝 – вероятность появления дуги между акторами. В R для модели случайного графа Эрдеша—Реньи существует функция erdos.renyi.game(). Ознакомьтесь с данной моделью и опишите её.

В модели $G(n,p)$ граф строится путём случайного добавления рёбер. Каждое ребро включается в граф с вероятностью p независимо от остальных рёбер. Эквивалентно, все графы с n узлами и M рёбрами имеют одинаковую вероятность:

${p^{M}(1-p)^{{n \choose 2}-M}}$

```r
g <- 62
p <- 0.78
gr <-erdos.renyi.game(g, p, type ='gnp', directed = TRUE)
gr
```

Отобразить полученный орграф.

##### Пункт 3.
 Вычислить «пятичисловую сводку» для графа $\cal {G}_d$.

```r
grAsNetwork <- asNetwork(gr)

cat("Размер:\n", network.size(grAsNetwork))

cat("\nПлотность:\n", gden(grAsNetwork))

cat("\nКомпоненты:\n", igraph::components(gr)$no)

components <- component.largest(grAsNetwork,result="graph")
gd <- geodist(components)
cat("\nДиаметр:\n", max(gd$gdist))

cat("\nКоэффициент кластеризации:", gtrans(grAsNetwork))
```

##### Пункт 7.
