# Основные теоретические положения

* $\textbf{Степень участия}$ - количество событий, с которым связан каждый актор.

$$a_{i+} = \sum_{j = 1}^{h} a_{ij} = x_{ii}^{\cal{N}}$$

* $\textbf{Размер событий:}$

$$a_{+j} = \sum_{i = 1}^{g} a_{ij} = x_{ii}^{\cal{M}}$$

* $\textbf{n-клика}$ - максимальный подграф, в котором наибольшее геодезическое расстояние между любыми двумя узлами не превышает n.

* $\textbf{n-клан}$ - n-клика, в которой геодезическое расстояние между всеми узлами в подграфе не превышает n для путей внутри подграфа.

* $\textbf{n-клуб}$ - максимальный подграф диаметра меньше либо равно n.

* $\textbf{k-ядро}$ - максимальный подграф, в котором каждая вершина связана минимум с _k_ другими вершинами этого же подграфа.

* $\textbf{k-плекс}$ - максимальный подграф, содержащий $g_s$ узлов, в котором каждый узел смежен с не менее чем с $g_s - k$ узлами подграфа.

* $\textbf{Полупуть}$ от узла _i_ до узла _j_ - последовательность различных узлов, где все последовательные пары узлов соединены дугой от первого ко второму или наоборот.

* $\textbf{Клика уровня с в взвешенном графе}$ - максимальный полный подграф из трех или более узлов, все из которых являются смежными на уровне _c_, то есть между всеми парами узлов есть ребра со значениями, большими или равными _c_.

* $\textbf{Клика уровня с для отношения соучастия}$ - подграф, в котором все пары акторов совместно участвуют не менее чем в _c_ событиях.

* $\textbf{Клика уровня с для отношения перекрытия}$ - подграф, в котором все пары событий имеют не менее _с_ общих членов.

* $\textbf{Модулярность}$ - характеристика структуры сети, в частности степень кластеризации узлов, когда внутри кластера наблюдается много узлов, а между кластерами мало узлов. показатель модулярности определяется как разница между долей связей внутри группы и ожидаемой доли связей, если бы связи были размещены случайно.

* $\textbf{Гиперграф}$ - обобщение понятия графа: гиперграф состоит из множества объектов и множества подмножеств объектов, в котором каждый объект принадлежит хотя бы одному подмножеству, причем ни одно подмножество не является пустым.

$$\cal {H = H(A, B)}$$, где $$\cal{A} = \{a_1, ..., a_n\}$$ - множество объектов; $$\cal{B} = \{b_1, ..., b_n\}$$ - множество рёбер. 

* $\textbf{Двойственный гипеграф для гипперграфа}$ 
$\cal {H = H(A, B)}:$

$$\cal {H* = H(B, A)}$$, где $$\cal{A} = \{a_1, ..., a_n\}$$ - множество рёбер; $$\cal{B} = \{b_1, ..., b_n\}$$ - множество объектов.

* $\textbf{Матрица соучастия акторов:}$

$$X^{\cal{N}} = \{x_{ij}^{\cal{N}} \}$$

$$x_{ij}^{\cal{N}} = \sum_{k = 1}^{h} a_{ik}a_{jk}$$

* $\textbf{Матрица перекрытия событий:}$

$$X^{\cal{M}} = \{x_{kl}^{\cal{M}} \}$$

$$x_{kl}^{\cal{M}} = \sum_{i = 1}^{g} a_{ik}a_{jl}$$


* $\textbf{Средний размер событий}$ - среднее количество акторов в каждом событии или средняя степень узлов, прeдставляющих события в двудольном графе.

$$\overline {a}_{ij} = \frac{1}{h} \sum_{j = 1}^{h} x_{jj}^{\cal{M}}$$
## Сети аффилированности:
* $\textbf{Сеть аффилированности}$ - сеть, в которой участники аффилированы друг с другом на основе сотрудничества в рамках какой-то группы или совместного участия в определенном событии. Сеть состоит из двух измерений.

* $\textbf{Связная сеть аффилированности}$ - сеть, в которой все пары узлов достижимы.

* $\textbf{Диаметр аффилированной сети}$ -длина самого длинного пути между любой парой акторов и/или событий.

### Их свойства:
* Количество перекрывающихся связей между событиями частично зависит от количества событий, к которым принадлежат акторы.

* Количество связей соучастия между акторами частично зависит от размера событий.

* Актор, принадлежащий $a_{i+}$ событиям, создает $C_{a_{i+}}^2$ парных связей между событиями.

* Событие с участиеми $a_{+j}$ участников создает $C_{a_{+о}}^2$ связей между парами акторов.

* Плотность для дихотомического отношения - доля присутствующих связей.

* Плоность для ценностных отношений - среднее значение связей.

```r
plex_k <- function(g,k,m){
    g.sym <- as.undirected(g)
    k.plex <- list()
    g.sym.degmk <- induced_subgraph(g.sym,igraph::degree(g.sym)>=(m-k))
    if(length(V(g.sym.degmk)) < m) {
        return(k.plex)
    } 
    k.cand <- combn(V(g.sym.degmk)$name,m)
    k.plex <- list() 
    k.plex.i <- 1 
    for (k.cand.i in 1:dim(k.cand)[2]){ 
        g.k.cand <- induced_subgraph(g.sym.degmk,k.cand[,k.cand.i]) 
        if (min(igraph::degree(g.k.cand))>=(m-k)){
            k.plex[[k.plex.i]] <- k.cand[,k.cand.i] 
            k.plex.i <- k.plex.i+1 
        }
    }
    return(k.plex) #return the entire list
}

n_clan <- function(g,n,ncliques){
    E(g)$weight <- 1 
    n.cand <- ncliques[[n]] 
    n.clan <- list() 
    n.clan.i <- 1 
    for (n.cand.i in 1:length(n.cand)){ 
        g.n.cand <- induced_subgraph(g,n.cand[[n.cand.i]]) 
        if (diameter(g.n.cand)<=n){ 
          n.clan[[n.clan.i]] <- n.cand[[n.cand.i]] 
          n.clan.i <- n.clan.i+1 
        }
      }
    return(n.clan) 
}
n_clans <- function(x){
    x <- as.undirected(x)
    ncliques <- kCliques(ugraph(igraph.to.graphNEL(x)))
    result <- c()
    for (i in names(cl)){
        result[[i]] <- nclan(x, i, ncliques)
    }
    return(result)
}

plexes_k <- function(x) {
    result <- c()
    for (i in 2:(length(V(x))-6)) {
        for (j in 1:(i-1)) {
            plex <- plex_k(x, j, i)
            if(length(plex) > 0) {
                result[[paste(toString(i), toString(j), "plex", sep="_")]] <- plex
            }
        }
    }
    return(result)
}



kernel_k <- function(g, k) {
    g.sym <- as.undirected(g)
    k.kernel <- list()
    g.sym.degk <- induced_subgraph(g.sym,igraph::degree(g.sym)>=k)
    if(length(V(g.sym.degk)) < k) {
        return(k.kernel)
    } 
    flag <- 0
    for (j in length(V(g.sym.degk)):1){
        if(flag) {
            break
        }
        k.cand <- combn(V(g.sym.degk)$name,j) 
        k.kernel <- list() 
        k.kernel.i <- 1 
        for (k.cand.i in 1:dim(k.cand)[2]){ 
            g.k.cand <- induced_subgraph(g.sym.degk,k.cand[,k.cand.i]) 
            if (min(igraph::degree(g.k.cand))>=k){ #if minimum degree of sugraph is > k, k-kernel!
                k.kernel[[k.kernel.i]] <- k.cand[,k.cand.i] 
                k.kernel.i <- k.kernel.i+1 
                flag <- 1
            }
        }
    }
    return(k.kernel) #return the entire list
}

kernels_k <- function(x) {
    result <- c()
    for (k in 1:length(V(x))) {
        kernel <- kernel_k(x, k)
        if(length(kernel) > 0) {
            result[[paste(toString(k), "kernel", sep="_")]] <- kernel
        }
    }
    return(result)
}

#k плексы
cat("i_j_plex, Где i - это мощность множества вершин подграфа, а j - k")
plexesk <- plexes_k(socio_graph_not_isolated)
plexesk

#n клики
cat("n cliques")
kClqs <- kCliques(igraph.to.graphNEL(undirected_graph))
kClqs

#n кланы
cat("n cliques => n clans")
nclans <- n_clans(undirected_graph)
nclans
       
#k ядра
cat("k kernels")
kernelsk <- kernels_k(socio_graph_not_isolated)
kernelsk
```

## Алгоритмы. Краткое описание
* InfoMAP: <br>
    В алгоритме используется подход на случайных блужданиях и кодах Хаффмана.<br>
    Каждая вершина имеет своё значение - вероятность посещения. С этими значениями применяются коды Хаффмана - кодируется "код блуждателя". Такая последовательность имеет свою длину, при этом, пользуясь иерархическим кодированием* возможно сократить эту длину. 
    *Иерархическое кодирование  - кодирование сообщества, после чего кодируем и вершины, попавшие в него. Коды вершин из разных групп могут совпадать. На выходе из сообщества записывается уникальный для него "код выхода".<br>
    Следовательно, в данном способе минимизируется длина кода прогулки блуждателя жадным способом.<br>
    
* Louvain:<br>
    В алгоритме применяется стратегия соияния (с коэффициентом модульности). В начале, каждый узел в отдельном сообществе подвергается инициализации, а также считается "выигрыш модульности Q" - от добавления этого узла к сообществу-соседу. Так, выбирается сосед, где такой коэффициент наибольший (для объединения).
    Как новые узлы до двух сообществ рассматриваются объединённые сообщества. Если полученный КМ оказывается меньше нуля, то такая комбинация останавливается.<br>

* Optimal:<br> 
Алгормтм применяет максимизацию модульности чтобы найти сообщества. Преобразованием максимизации модульности производится расчёт. Сложность алгоритма => экспоненциальная.
Алгоритм имеет экспоненциальную временную сложность.
