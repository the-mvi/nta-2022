–†–∞–∑–¥–µ–ª–∏–º —Å—É–º–º—É —Å—Ç–µ–ø–µ–Ω–µ–π –Ω–∞ –∫–æ–ª-–≤–æ –≤–µ—Ä—à–∏–Ω
$ \overline{d} = \frac {1} {g} * \sum_{i=1}^{g} {d(n_i)}$ ,–≥–¥–µ $g$ - —á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤ –≤ —Å–µ—Ç–∏.
* –î–∏—Å–ø–µ—Ä—Å–∏—è –±—ã–ª–∞ —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–∞ –ø–æ —Ñ–æ—Ä–º—É–ª–µ: $S_{D}^{2} = \frac {1} {g} * \sum_{i=1}^{g} {(d(n_i) - \overline{d})^2}$ (–º–µ—Ç–æ–¥ *var()*)
>–ü–æ —Å—Ä–µ–¥–Ω–µ–π —Å—Ç–µ–ø–µ–Ω–∏ –º–æ–∂–Ω–æ —Å—É–¥–∏—Ç—å —Å–æ —Å–∫–æ–ª—å–∫–∏–º–∏ –∞–∫—Ç–æ—Ä–∞–º–∏ –∫–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä –≤ —Å—Ä–µ–¥–Ω–µ–º –∏–º–µ–µ—Ç –ø—Ä—è–º—É—é —Å–≤—è–∑—å. –í —ç—Ç–æ–π —Å–µ—Ç–∏ –∫–∞–∂–¥–∞—è —Å–µ–º—å—è —Å–≤—è–∑–∞–Ω–∞ –≤ —Å—Ä–µ–¥–Ω–µ–º —Å $2-3$ –¥—Ä—É–≥–∏–º–∏ —Å–µ–º—å—è–º–∏.
>–î–∏—Å–ø–µ—Ä—Å–∏—è –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Ä–∞–∑–±—Ä–æ—Å –≤–µ–ª–∏—á–∏–Ω—ã –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è. –ó–¥–µ—Å—å –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ —Å–µ–º—å–∏ –º–æ–≥—É—Ç –Ω–µ –∏–º–µ—Ç—å —Å–≤—è–∑–µ–π, –∞ –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –¥–∞–∂–µ –∏–º–µ—Ç—å 5.

9. –ó–∞–ø–æ–ª–Ω–µ–Ω–∞ —Ç–∞–±–ª–∏—Ü–∞ –¥–ª—è –º–µ—Ä —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç–∏ –¥–ª—è –Ω–µ–∏–∑–æ–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –≤–µ—Ä—à–∏–Ω.

````r
# –£–¥–∞–ª–µ–Ω–∏–µ –∏–∑–æ–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –≤–µ—Ä—à–∏–Ω.
non_isolated_vertices <- sociomatrix[which(rowSums(sociomatrix) > 0), which(colSums(sociomatrix) > 0)]
socio_network_not_isolated_actors <- network(non_isolated_vertices, matrix.type = "adjacency", directed = FALSE)
socio_graph_not_isolated_actors <- graph.adjacency(non_isolated_vertices, mode="undirected")

result <- data.frame(names = colnames(non_isolated_vertices))
number_actor <- ncol(non_isolated_vertices)
# –¶–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç–∏:
# –ü–æ —Å—Ç–µ–ø–µ–Ω–∏.
degree_centr <- centr_degree(socio_graph_not_isolated_actors, mode = "all", normalized = TRUE)

# –ü–æ –ø–æ—Å—Ä–µ–¥–Ω–∏—á–µ—Å—Ç–≤—É.
betweenness_centr <- centr_betw(socio_graph_not_isolated_actors, normalized = TRUE)

# –ü–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏.
information_centr <- infocent(socio_network_not_isolated_actors, gmode = "graph")

# –ü–æ –±–ª–∏–∑–æ—Å—Ç–∏.
closeness_centr <- centr_clo(socio_graph_not_isolated_actors, normalized = TRUE)

# –ù–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ.
result <- cbind(result, deg = degree_centr$res / (number_actor - 1), 
closen = closeness_centr$res, betw = betweenness_centr$res * 2 / ((number_actor - 1) * (number_actor - 2)), 
inform = information_centr / sum(information_centr))
result
````
