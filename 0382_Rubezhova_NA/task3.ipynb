## Основные теоретические положения



---
####*Диады. Цензус диад.*
&emsp;*Определение.*<br>
&emsp;&emsp;**Диада** - элемент сети, представляющий собой пару акторов и возможнную связь между ними.

>&emsp;&emsp;*Виды диад*:
* Взаимная диада: $D_{i,j} = (1, 1)$
* Ассиметричная диада: $D_{i,j} = (1, 0)$ или $D_{i,j} = (0, 1)$
* Нулевая диада:  $D_{i,j} = (0, 0)$

&emsp;*Определение.*<br>
&emsp;&emsp;**Цензус диад** - тройка чисел $<M, \ A, \ N>$, где:<br>
&emsp;&emsp;&emsp;&emsp; $M = \sum_{i < j}(x_{ij}x_{ji})$ - число взаимных диад<br>
&emsp;&emsp;&emsp;&emsp; $A = x_{++} - 2M$ - число ассиметричных диад<br>
&emsp;&emsp;&emsp;&emsp; $N = C^2_g - A - M$ - число нулевых диад<br><br>

---
####*Индекс взаимности Каца-Пауэлла.*
&emsp;*Определение.*<br>
&emsp;&emsp;**Индекс взаимности Каца-Пауэлла** - характеристика измерения тенденции акторов в группе:

$$-\infty < \rho_{KP} \leq 1$$

>Индекс взаимности основан на ожидании количества взаимных диад, предполгая, что выбор делается акторами случайным образом.

>&emsp;*При **фиксированном** выборе*:
* Каждый актор делает $d$ выборов из $(g-1)$ акторов, доступных для выбора
* Оценка $\rho_{KP}$ по методу моментов:
$$\hat \rho_{KP} = \frac {2(g-1)M-gd^2}{gd(g-1-d)}$$
* Вероятность взаимного выбора между любой произвольной парой акторов:
$$\frac {d^2}{(g-1)^2}$$

>&emsp;*При **свободном** выборе*:
* Оценка $p_{KP}$ по методу моментов:
$${\rho{'}_{KP} = \frac {2(g-1)^2M-L^2+L_2}{L(g-1)^2-L^2+L_2}},$$
где $L = \sum_i x_{i+}$, $L_2 = \sum_i x^2_{i+}$
* Вероятность того, что акторы имеют взаимные отношения:
$$\frac {x_{i+}x_{j+}}{(g-1)^2}$$

---
####*Индекс взаимности Бандиопадхьяйя.*
&emsp;*Определение.*<br>
&emsp;&emsp;**Индекс взаимности Бандиопадхьяйя** - данный индекс основан на минимальном и максимальном значениях количества взаимных диад. Рассчитывается по формуле:
$$\rho_{B} = \frac{M - M_{min}}{M_{max} - M_{min}}, \ \ где$$
* &emsp;$M_{min} = max_{0 \leq t \leq g}f(t)$

* &emsp;$M_{max} = [\frac{1}{2}(\sum_{i=1}^g - max_{0 \leq t \leq g}g(t))]$

* &emsp;$f(t) = \sum_{i=1}^tx_{i+} - t(g-1) - C^2_t$ 

* &emsp;$g(t) = \sum_{i=1}^tx_{i+} - t(t-1) - \sum_{i=t+1}^gmin(t, x_{i+})$
<br>
---
#### *Распределения на графах.*
&emsp;$\#G_d(\cal {N}) = 2^{g(g - 1)}$, где $G_d(\cal {N})$ - множество всех возможных орграфов без петель с g узлами

>* Вероятность получить социоматрицу $x$:
$$P(X = x) = 2^{g(g - 1)}$$
* Вероятность появления дуги между акторами для равномерного распределения:
$$P(X_{ij} = 1) = \begin{cases}
    \frac{1}{2}, i \ne j\\
    0, i = j
\end{cases}$$
* Вероятность появления дуги между акторами для распределения Бернулли:
$$P(X_{ij} = 1) = \begin{cases}
    p_{ij}, i \ne j\\
    0, i = j
\end{cases}$$

---

#### *Статистический анализ количества дуг для **равномерного распределения**.*
>* Число дуг: 
$$L \sim B(g(g - 1), \frac{1}{2})$$
* Вероятность, что число дуг равно $l$:
$$P(L = l) = \binom {g(g - 1)}{l} \cdot  2^{g(1 - g)}$$
* Математическое ожидание:
$$E(L) = \binom {g}{2}$$
* Дисперсия:
$$D(L) = \frac{1}{2} \binom {g}{2}$$

---

#### *Статиcтические **гипотезы для равномерного распределения**.*
>* Нулевая гипотеза $H_0$: 
$$L \sim B(g(g - 1), \frac{1}{2})$$
* Статистика:
$$z_l = \frac{l - E(L)}{\sqrt {D(L)}} = \frac {l-\frac {g(g-1)}{2}}{\sqrt {\frac {g(g-1)}{4}}},$$
где $l = \sum_{i = 1}^g\sum_{j = 1}^gx_{ij}$

* Данная статистика является приблизительно стандартным нормальным распределением при заданном уровне значимости.

---

#### *Оценка вероятности **$P$ для распределения Бернулли**.*
>* Число дуг:
$$L \sim B(g(g - 1), P_0)$$
* Вероятность, что количество дуг равно $l$:
$$P(L = l) = \binom {g(g - 1)}{l} = P_0^l(1 - P_0)^{g(g - 1) - l}$$
* Математическое ожидание:
$$E(L) = P_0g(g - 1)$$
* Дисперсия:
$$D(L) = P_0(1 - P_0)g(g - 1)$$

---

#### *Статистические **гипотезы для распределения Бернулли**.*
>* Нулевая гипотеза $H_0$:
$$L \sim B(g(g - 1), P_0)$$
* Статистика:
$$z_l = \frac{l - E(L)}{\sqrt {D(L)}} = \frac {l-P_0g(g-1)}{\sqrt {P_0(1-P_0)g(g-1)}},$$
где $l = \sum_{i = 1}^g\sum_{j = 1}^gx_{ij}$

* Данная статистика является приблизительно стандартным нормальным распределением при заданном уровне значимости.

---

#### *Д**оверительный интервал** для оценки $P$.*
>* Оценка параметра $P$:
$$\hat {P} = \frac{l}{g(g - 1)},$$  где $l = \sum_{i = 1}^g\sum_{j = 1}^gx_{ij}$
* Доверительный интервал:
$$\hat {P}_{lower} \leq P \leq  \hat {P}_{upper}, \ где$$
* &emsp;&emsp;$\hat {P}_{lower} = \hat {P} - z_{\frac {\alpha}{2}} \sqrt {\frac {\hat {P}(1 - \hat{P})}{g(g - 1)}}$
* &emsp;&emsp;$\hat {P}_{upper} = \hat {P} + z_{\frac {\alpha}{2}} \sqrt {\frac {\hat {P}(1 - \hat{P})}{g(g - 1)}}$
&emsp;&emsp;При $z_{\frac{\alpha}{2}}$: $\Phi(z_{\frac{\alpha}{2}}) = 0.5 - \frac{\alpha}{2}$

##**Задание 1.**


Случайный орграф $G_𝑑$ **был смоделирован** с помощью функции *erdos.renyi.game()* из библиотеки *igraph* на основе **модели Эрдеша—Реньи**: <br> &emsp;$G_𝑑 = G(𝑔,𝑝)$, где $𝑔$– количество акторов, $𝑝$ – вероятность появления дуги между акторами. <br><br>

**Описание модели.** В модели Эрдеша-Реньи при построении графа из $n$ вершин рёбра добавляются *случайным образом*: каждое ребро добавляется независимо от других рёбер с вероятностью $p$. 

```r
library(igraph)
g <- 91 # количество акторов
p <- 0.62 # вероятность появления дуги между акторами

graph <- erdos.renyi.game(g, p, type ='gnp', directed = TRUE)
edgesGraph <- as_edgelist(graph, names = TRUE)
```

##**Задание 5.**
Была вычислена оценка индекса взаимности Каца—Пауэла при фиксированом выборе акторов $𝑑 \in \mathbb{N}$ и при свободном выборе. Сделаны выводы.<br>


При фиксированном выборе предполагается, что актор делает $d=60$ выборов из $(g-1)$ акторов, доступных для выбора.

```r
# при фиксированном d=41
d <- 60
# оценка индекса взаимности по методу моментов при фикс. d=60
ro1 <- (2*(g-1)*M - g*(d**2))/(g*d*(g-1-d))
cat("Индекс взаимности Каца-Пауэла при фикс. выборе акторов d=60:\t", ro1, "\n")

library(intergraph)
netGraph <- asNetwork(graph)
matrGraph <- as.matrix(netGraph)

# при свободном выборе
L <- sum(rowSums(matrGraph)) # число ребер, сумма элементов вектора(сумм строк матрицы), т.е. суммирует суммы строк матрицы
L_2 <- sum(rowSums(matrGraph)^2) # сумма квадратов сумм строк матрицы, т.е. суммирует квадраты сумм строк матрицы

# оценка индекса взаимности по методу моментов
ro2 <- (2*M*((g-1)**2)-L**2 + L_2)/(L*((g-1)**2)-L**2+L_2)
cat("Индекс взаимности Каца-Пауэла при свободном выборе акторов:\t", ro2)
```

##**Задание 7.**
Был вычислен индекс взаимности Бандиопадхьяйя. А также сделаны выводы.

&emsp;&emsp;**Индекс взаимности Бандиопадхьяйя** основан на минимальном и максимальном значениях количества **взаимных диад**:<br> 
&emsp;$M_{min} = max_{0 \leq t \leq g}f(t)$,&emsp;где $$f(t) = \sum_{i=1}^tx_{i+} - t(g-1) - C^2_t$$<br>

&emsp;$M_{max} = [\frac{1}{2}(\sum_{i=1}^g - max_{0 \leq t \leq g}g(t))]$,&emsp;где $$g(t) = \sum_{i=1}^tx_{i+} - t(t-1) - \sum_{i=t+1}^gmin(t, x_{i+})$$

Индекс взаимности Бандиопадхьяйя рассчитывается по формуле:
$$\rho_{B} = \frac{M - M_{min}}{M_{max} - M_{min}}, $$

```r
f_t <- c() # будет хранить значения f(t)
g_t <- c() # будет хранить значения g(t)

# перебираем от 0 до g (0<=t<=g)
for(t in 0:g){
    sum <- 0
    if (t>0){
        # перебираем i-ые строки и суммируем по строкам
        for(i in 1:t) {
            sum <- sum+sum(matrGraph[i, ])
        }
    }
    #вычисляем текущее f(t)
    f_cur <- sum-t*(g-1)-choose(t, 2)
    # запоминаем значение f(t), чтобы потом найти максимум
    f_t <- append(f_t, f_cur)
    sum2 <- 0
    if (t + 1 <= g){
        for(i in (t+1):g){
            sum2 <-sum2+min(t,sum(matrGraph[i, ]))
        }
    }
    #вычисляем текущее g(t)
    g_cur <-sum-t*(t-1)-sum2
    # запоминаем значение g(t), чтобы потом найти максимум
    g_t<-append(g_t, g_cur)
}
m_min <- max(f_t)
m_max <- floor( 0.5*(sum(matrGraph)-max(g_t)) )
# вычисляем индекс взаимности по формуле
index <- ( M - m_min ) / ( m_max - m_min )
cat("Индекс взаимности Бандиопадхьяйя:", index)
```

##**Задание 10.**
Построить доверительный интервал для оценки параметра $𝑝$ с доверительной надёжностью $(1 − 𝛼)$. Сделать выводы.

>Итак, оценка параметра $p$:
$$\hat {p} = \frac{L}{g(g - 1)},$$
где $L = \sum_{i = 1}^g\sum_{j = 1}^gx_{ij}$<br>

>А доверительный интервал:
$$\hat {P}_{lower} \leq P \leq  \hat {P}_{upper}, \ где$$
* &emsp;&emsp;$\hat {P}_{lower} = \hat {P} - z_{\frac {\alpha}{2}} \sqrt {\frac {\hat {P}(1 - \hat{P})}{g(g - 1)}}$
* &emsp;&emsp;$\hat {P}_{upper} = \hat {P} + z_{\frac {\alpha}{2}} \sqrt {\frac {\hat {P}(1 - \hat{P})}{g(g - 1)}}$
&emsp;&emsp;При $z_{\frac{\alpha}{2}}$: $\Phi(z_{\frac{\alpha}{2}}) = 0.5 - \frac{\alpha}{2}$

*Вычислим:*

```r
estimP<- L/(g*(g-1))
cat("Оценка параметра p =", estimP, '\n')
alpha <- 0.05
z <- abs(qnorm(alpha/2))
low <- estimP - z * sqrt( estimP *( 1-estimP ) / ( g*(g-1) ) )
up <- estimP + z * sqrt( estimP * ( 1-estimP ) / ( g*(g-1) ) )
cat("Доверительный интервал = [", low,", ", up,"]")
```
