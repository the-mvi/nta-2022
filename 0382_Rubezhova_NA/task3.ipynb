>&emsp;*–ü—Ä–∏ **—Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º** –≤—ã–±–æ—Ä–µ*:
* –ö–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä –¥–µ–ª–∞–µ—Ç $d$ –≤—ã–±–æ—Ä–æ–≤ –∏–∑ $(g-1)$ –∞–∫—Ç–æ—Ä–æ–≤, –¥–æ—Å—Ç—É–ø–Ω—ã—Ö –¥–ª—è –≤—ã–±–æ—Ä–∞
* –û—Ü–µ–Ω–∫–∞ $\rho_{KP}$ –ø–æ –º–µ—Ç–æ–¥—É –º–æ–º–µ–Ω—Ç–æ–≤:
$$\hat \rho_{KP} = \frac {2(g-1)M-gd^2}{gd(g-1-d)}$$
* –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤–∑–∞–∏–º–Ω–æ–≥–æ –≤—ã–±–æ—Ä–∞ –º–µ–∂–¥—É –ª—é–±–æ–π –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–π –ø–∞—Ä–æ–π –∞–∫—Ç–æ—Ä–æ–≤:

>&emsp;*–ü—Ä–∏ **—Å–≤–æ–±–æ–¥–Ω–æ–º** –≤—ã–±–æ—Ä–µ*:
* –û—Ü–µ–Ω–∫–∞ $p_{KP}$ –ø–æ –º–µ—Ç–æ–¥—É –º–æ–º–µ–Ω—Ç–æ–≤:
$${\rho{'}_{KP} = \frac {2(g-1)^2M-L^2+L_2}{L(g-1)^2-L^2+L_2}},$$
–≥–¥–µ $L = \sum_i x_{i+}$, $L_2 = \sum_i x^2_{i+}$
* –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –∞–∫—Ç–æ—Ä—ã –∏–º–µ—é—Ç –≤–∑–∞–∏–º–Ω—ã–µ –æ—Ç–Ω–æ—à–µ–Ω–∏—è:

&emsp;*–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ.*<br>
&emsp;&emsp;**–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è** - –¥–∞–Ω–Ω—ã–π –∏–Ω–¥–µ–∫—Å –æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–º –∑–Ω–∞—á–µ–Ω–∏—è—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥. –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ:

* &emsp;$M_{min} = max_{0 \leq t \leq g}f(t)$

* &emsp;$f(t) = \sum_{i=1}^tx_{i+} - t(g-1) - C^2_t$ 

* &emsp;$g(t) = \sum_{i=1}^tx_{i+} - t(t-1) - \sum_{i=t+1}^gmin(t, x_{i+})$

#### *–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –≥—Ä–∞—Ñ–∞—Ö.*

#### *–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥—É–≥ –¥–ª—è **—Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è**.*

#### *–°—Ç–∞—Ç–∏c—Ç–∏—á–µ—Å–∫–∏–µ **–≥–∏–ø–æ—Ç–µ–∑—ã –¥–ª—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è**.*
>* –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ $H_0$:

#### *–û—Ü–µ–Ω–∫–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ **$P$ –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ë–µ—Ä–Ω—É–ª–ª–∏**.*
>* –ß–∏—Å–ª–æ –¥—É–≥:
$$L \sim B(g(g - 1), P_0)$$
* –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–≥ —Ä–∞–≤–Ω–æ $l$:
$$P(L = l) = \binom {g(g - 1)}{l} = P_0^l(1 - P_0)^{g(g - 1) - l}$$
* –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ:
$$E(L) = P_0g(g - 1)$$
* –î–∏—Å–ø–µ—Ä—Å–∏—è:
$$D(L) = P_0(1 - P_0)g(g - 1)$$

---

#### *–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ **–≥–∏–ø–æ—Ç–µ–∑—ã –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ë–µ—Ä–Ω—É–ª–ª–∏**.*

$$L \sim B(g(g - 1), P_0)$$

$$z_l = \frac{l - E(L)}{\sqrt {D(L)}} = \frac {l-P_0g(g-1)}{\sqrt {P_0(1-P_0)g(g-1)}},$$
–≥–¥–µ $l = \sum_{i = 1}^g\sum_{j = 1}^gx_{ij}$

* –î–∞–Ω–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —è–≤–ª—è–µ—Ç—Å—è –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏.

---

#### *–î**–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª** –¥–ª—è –æ—Ü–µ–Ω–∫–∏ $P$.*
>* –û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $P$:
$$\hat {P} = \frac{l}{g(g - 1)},$$  –≥–¥–µ $l = \sum_{i = 1}^g\sum_{j = 1}^gx_{ij}$
* –î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª:
$$\hat {P}_{lower} \leq P \leq  \hat {P}_{upper}, \ –≥–¥–µ$$
* &emsp;&emsp;$\hat {P}_{lower} = \hat {P} - z_{\frac {\alpha}{2}} \sqrt {\frac {\hat {P}(1 - \hat{P})}{g(g - 1)}}$
* &emsp;&emsp;$\hat {P}_{upper} = \hat {P} + z_{\frac {\alpha}{2}} \sqrt {\frac {\hat {P}(1 - \hat{P})}{g(g - 1)}}$
&emsp;&emsp;–ü—Ä–∏ $z_{\frac{\alpha}{2}}$: $\Phi(z_{\frac{\alpha}{2}}) = 0.5 - \frac{\alpha}{2}$

**–û–ø–∏—Å–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏.** –í –º–æ–¥–µ–ª–∏ –≠—Ä–¥–µ—à–∞-–†–µ–Ω—å–∏ –ø—Ä–∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ –≥—Ä–∞—Ñ–∞ –∏–∑ $n$ –≤–µ—Ä—à–∏–Ω —Ä—ë–±—Ä–∞ –¥–æ–±–∞–≤–ª—è—é—Ç—Å—è *—Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º*: –∫–∞–∂–¥–æ–µ —Ä–µ–±—Ä–æ –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –¥—Ä—É–≥–∏—Ö —Ä—ë–±–µ—Ä —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é $p$. 

```r
library(igraph)
g <- 91 # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–æ—Ä–æ–≤
p <- 0.62 # –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—è–≤–ª–µ–Ω–∏—è –¥—É–≥–∏ –º–µ–∂–¥—É –∞–∫—Ç–æ—Ä–∞–º–∏

graph <- erdos.renyi.game(g, p, type ='gnp', directed = TRUE)
edgesGraph <- as_edgelist(graph, names = TRUE)
```

##**–ó–∞–¥–∞–Ω–∏–µ 3.**
**¬´–ü—è—Ç–∏—á–∏—Å–ª–æ–≤–∞—è —Å–≤–æ–¥–∫–∞¬ª** –¥–ª—è –≥—Ä–∞—Ñ–∞ $G_ùëë$. –î–ª—è –µ—ë –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ñ—É–Ω–∫—Ü–∏—è–º–∏ –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫ *igraph, intergraph*. –¢–∞–∫–∂–µ —Å–¥–µ–ª–∞–µ–º **–≤—ã–≤–æ–¥—ã**.<br>

&emsp;&emsp;&emsp; *(1)* **–†–∞–∑–º–µ—Ä –≥—Ä–∞—Ñ–∞:**&emsp;**91** (–¥–∞–Ω–æ –≤ –∑–∞–¥–∞–Ω–∏–∏, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–æ—Ä–æ–≤);</br> 

&emsp;&emsp;&emsp;&nbsp;*(2)* **–ü–ª–æ—Ç–Ω–æ—Å—Ç—å** –¥–ª—è –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∞ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è *–ø–æ —Ñ–æ—Ä–º—É–ª–µ:*<br/>&emsp;&emsp;&emsp;&nbsp;$‚ñ≥ = \dfrac{L}{A_2^g} = \dfrac{L}{g(g-1)}$ <br/>&emsp;&emsp;&emsp;&nbsp;–í—ã—á–∏—Å–ª–∏–ª–∏ —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ *gden()*, –ø–æ–ª—É—á–∏–ª–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç—å: **0.03552051**. <br/>&emsp;&emsp;&emsp;&nbsp;**–í—ã–≤–æ–¥**. –ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≥—Ä–∞—Ñ–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –æ—Ç–Ω–æ—à–µ–Ω–∏–µ —á–∏—Å–ª–∞ —Ä–µ–±–µ—Ä –∫ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–º—É, –∞ –ø–æ—Å–∫–æ–ª—å–∫—É *—á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤ –≤–µ–ª–∏–∫–æ –∏ –≥—Ä–∞—Ñ –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π*, –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–µ —á–∏—Å–ª–æ —Ä–µ–±–µ—Ä –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–ª—å—à–æ–µ, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ, *–ø–ª–æ—Ç–Ω–æ—Å—Ç—å*, –≤—ã—á–∏—Å–ª–µ–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏–µ–π *gden()* *–æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –Ω–µ–±–æ–ª—å—à–∞—è*.</br> 

&emsp;&emsp;&emsp;&nbsp;*(3)* ***–ß–∏—Å–ª–æ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç —Å–≤—è–∑–Ω–æ—Å—Ç–∏*** &emsp;<br/>
&emsp;&emsp;&emsp;&nbsp;–í—ã—á–∏—Å–ª–µ–Ω–Ω–æ–µ —Ñ—É–Ω–∫—Ü–∏–µ–π *components()* –∑–Ω–∞—á–µ–Ω–∏–µ: &emsp;**1**.<br/> &emsp;&emsp;&emsp;&nbsp;**–í—ã–≤–æ–¥**. –í –¥–∞–Ω–Ω–æ–º –≥—Ä–∞—Ñ–µ –≤—Å–µ–≥–æ 1 –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞ —Å–≤—è–∑–Ω–æ—Å—Ç–∏, —ç—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –≥—Ä–∞—Ñ **—Å–≤—è–∑–Ω—ã–π**. **–ß–∏—Å–ª–æ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç - 1**.</br> 

&emsp;&emsp;&emsp;&nbsp;*(4)* ***–î–∏–∞–º–µ—Ç—Ä —Å–µ—Ç–∏*** . <br/>&emsp;&emsp;&emsp;&nbsp;–í—ã—á–∏—Å–ª–∏–º –º–∞—Ç—Ä–∏—Ü—É –≥–µ–æ–¥–µ–∑–∏—á–µ—Å–∫–∏—Ö —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ *geodist()* –∏ –Ω–∞–π–¥–µ–º —Å—Ä–µ–¥–∏ –µ–µ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π, –æ—Ç–ª–∏—á–Ω—ã–π –æ—Ç –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç–∏(–∫–æ–≥–¥–∞ –ø—É—Ç–∏ –Ω–µ—Ç). –ü–æ–ª—É—á–∏–ª–∏ **diameter = 2**.<br/>&emsp;&emsp;&emsp;&nbsp;**–í—ã–≤–æ–¥**. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∏–∑ –∫—Ä–∞—Ç—á–∞–π—à–∏—Ö —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –º–µ–∂–¥—É –≤—Å–µ–º–∏ –ø–∞—Ä–∞–º–∏ —É–∑–ª–æ–≤ —Ä–∞–≤–Ω–æ 2. </br> 

&emsp;&emsp;&emsp;&nbsp;*(5)* ***–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏***. <br/>&emsp;&emsp;&emsp;&nbsp;–ó–Ω–∞—á–µ–Ω–∏–µ, –≤—ã—á–∏—Å–ª–µ–Ω–Ω–æ–µ —Ñ—É–Ω–∫—Ü–∏–µ–π *gtrans()*: &emsp;0<br/>&emsp;&emsp;&emsp;&nbsp;**–í—ã–≤–æ–¥.**&nbsp; –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –ø–æ—Å–∫–æ–ª—å–∫—É –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ - —ç—Ç–æ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∑–∞–∫—Ä—ã—Ç—ã—Ö —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–æ–≤ –ø–æ –æ—Ç–Ω–æ—à–µ–Ω–∏—é –∫ –æ–±—â–µ–º—É  –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –æ—Ç–∫—Ä—ã—Ç—ã—Ö –∏ –∑–∞–∫—Ä—ã—Ç—ã—Ö —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–æ–≤, —Ç–æ –º–æ–∂–µ–º —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ –≤ –¥–∞–Ω–Ω–æ–º –≥—Ä–∞—Ñ–µ **–æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—Ç –∑–∞–∫—Ä—ã—Ç—ã–µ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∏**, —Ç.–µ. –Ω–∏–≥–¥–µ –Ω–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —Ç—Ä–∞–Ω–∑–∏—Ç–∏–≤–Ω–æ—Å—Ç—å .

```r
library(statnet)
library(igraph)
library(intergraph)

cat("–†–∞–∑–º–µ—Ä: ", g, '\n')
cat("–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≥—Ä–∞—Ñ–∞: ", gden(edgesGraph), '\n')

components <- igraph::components(graph)$no
cat("–ß–∏—Å–ª–æ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç —Å–≤—è–∑–Ω–æ—Å—Ç–∏: ", components, '\n')

gd <- geodist(edgesGraph)
cat("–î–∏–∞–º–µ—Ç—Ä: ", max(gd$gdist[gd$gdist != Inf]), '\n')

cat("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏: ", gtrans(edgesGraph), '\n')
```

##**–ó–∞–¥–∞–Ω–∏–µ 4.**
–ë—ã–ª –≤—ã—á–∏—Å–ª–µ–Ω **—Ü–µ–Ω–∑—É—Å –¥–∏–∞–¥**, –∞ —Ç–∞–∫–∂–µ —Å–¥–µ–ª–∞–Ω—ã –≤—ã–≤–æ–¥—ã.<br>
&emsp;–° –≤—ã—á–∏—Å–ª–µ–Ω–Ω—ã–º —Ü–µ–Ω–∑—É—Å–æ–º –¥–∏–∞–¥ –º–æ–∂–Ω–æ –æ–∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è –Ω–∏–∂–µ, –ø–æ—Å–ª–µ –∫–æ–¥–∞.<br>
&emsp;**–í—ã–≤–æ–¥.** –ü–æ –≤—ã—á–∏—Å–ª–µ–Ω–Ω–æ–º—É —Ñ—É–Ω–∫—Ü–∏–µ–π *dyad_census* –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ *igraph* **—Ü–µ–Ω–∑—É—Å—É –¥–∏–∞–¥** –º–æ–∂–µ–º –∑–∞–∫–ª—é—á–∏—Ç—å, —á—Ç–æ –≤ –≥—Ä–∞—Ñ–µ –ø—Ä–µ–æ–±–ª–∞–¥–∞—é—Ç- **–∞—Å—Å–∏–º–µ—Ç—Ä–∏—á–Ω—ã–µ –¥–∏–∞–¥—ã** (–ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å–æ –≤–∑–∞–∏–º–Ω—ã–º–∏ –∏ –Ω—É–ª–µ–≤—ã–º–∏).<br>
&emsp; –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –≤ –≥—Ä–∞—Ñ–µ –º–µ–∂–¥—É –¥–∏–∞–¥–∞–º–∏ **—á–∞—â–µ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –∞—Å—Å–∏–º–µ—Ç—Ä–∏—á–Ω–∞—è —Å–≤—è–∑—å**, –≤–∏–¥–∞: $$  D_{ij}=(1, 0) \ –∏–ª–∏ \ D_{ij}=(0, 1),$$ –≤ –æ—Ç–ª–∏—á–∏–µ –æ—Ç –≤–∑–∞–∏–º–Ω–æ–π –∏–ª–∏ –Ω—É–ª–µ–≤–æ–π —Å–≤—è–∑–∏ $$D_{ij}=(1, 1) \ –∏–ª–∏ \ D_{ij}=(0, 0)$$

```r
library(igraph)
census <- dyad_census(graph)
M <- census$mut
cat("M=", M,"- —á–∏—Å–ª–æ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥", '\n')
A <- census$asym
cat("A=", A, "- —á–∏—Å–ª–æ –∞—Å—Å–∏–º–µ—Ç—Ä–∏—á–Ω—ã—Ö –¥–∏–∞–¥",'\n')
N <- census$null
cat("N=", N, "- —á–∏—Å–ª–æ –Ω—É–ª–µ–≤—ã—Ö –¥–∏–∞–¥")
```

##**–ó–∞–¥–∞–Ω–∏–µ 5.**
–ë—ã–ª–∞ –≤—ã—á–∏—Å–ª–µ–Ω–∞ –æ—Ü–µ–Ω–∫–∞ –∏–Ω–¥–µ–∫—Å–∞ –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ö–∞—Ü–∞‚Äî–ü–∞—É—ç–ª–∞ –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–æ–º –≤—ã–±–æ—Ä–µ –∞–∫—Ç–æ—Ä–æ–≤ $ùëë \in \mathbb{N}$ –∏ –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ. –°–¥–µ–ª–∞–Ω—ã –≤—ã–≤–æ–¥—ã.<br>


–ü—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –≤—ã–±–æ—Ä–µ –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è, —á—Ç–æ –∞–∫—Ç–æ—Ä –¥–µ–ª–∞–µ—Ç $d=60$ –≤—ã–±–æ—Ä–æ–≤ –∏–∑ $(g-1)$ –∞–∫—Ç–æ—Ä–æ–≤, –¥–æ—Å—Ç—É–ø–Ω—ã—Ö –¥–ª—è –≤—ã–±–æ—Ä–∞.

```r
# –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º d=41
d <- 60
# –æ—Ü–µ–Ω–∫–∞ –∏–Ω–¥–µ–∫—Å–∞ –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ø–æ –º–µ—Ç–æ–¥—É –º–æ–º–µ–Ω—Ç–æ–≤ –ø—Ä–∏ —Ñ–∏–∫—Å. d=60
ro1 <- (2*(g-1)*M - g*(d**2))/(g*d*(g-1-d))
cat("–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ö–∞—Ü–∞-–ü–∞—É—ç–ª–∞ –ø—Ä–∏ —Ñ–∏–∫—Å. –≤—ã–±–æ—Ä–µ –∞–∫—Ç–æ—Ä–æ–≤ d=60:\t", ro1, "\n")

library(intergraph)
netGraph <- asNetwork(graph)
matrGraph <- as.matrix(netGraph)

# –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ
L <- sum(rowSums(matrGraph)) # —á–∏—Å–ª–æ —Ä–µ–±–µ—Ä, —Å—É–º–º–∞ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤–µ–∫—Ç–æ—Ä–∞(—Å—É–º–º —Å—Ç—Ä–æ–∫ –º–∞—Ç—Ä–∏—Ü—ã), —Ç.–µ. —Å—É–º–º–∏—Ä—É–µ—Ç —Å—É–º–º—ã —Å—Ç—Ä–æ–∫ –º–∞—Ç—Ä–∏—Ü—ã
L_2 <- sum(rowSums(matrGraph)^2) # —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ —Å—É–º–º —Å—Ç—Ä–æ–∫ –º–∞—Ç—Ä–∏—Ü—ã, —Ç.–µ. —Å—É–º–º–∏—Ä—É–µ—Ç –∫–≤–∞–¥—Ä–∞—Ç—ã —Å—É–º–º —Å—Ç—Ä–æ–∫ –º–∞—Ç—Ä–∏—Ü—ã

# –æ—Ü–µ–Ω–∫–∞ –∏–Ω–¥–µ–∫—Å–∞ –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ø–æ –º–µ—Ç–æ–¥—É –º–æ–º–µ–Ω—Ç–æ–≤
ro2 <- (2*M*((g-1)**2)-L**2 + L_2)/(L*((g-1)**2)-L**2+L_2)
cat("–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ö–∞—Ü–∞-–ü–∞—É—ç–ª–∞ –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ –∞–∫—Ç–æ—Ä–æ–≤:\t", ro2)
```

##**–ó–∞–¥–∞–Ω–∏–µ 6.**
–ë—ã–ª–∏ –≤—ã—á–∏—Å–ª–µ–Ω—ã –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Ç–æ–≥–æ, —á—Ç–æ —Å–≤—è–∑—å –±—É–¥–µ—Ç –≤–∑–∞–∏–º–Ω–æ–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Å–ª—É—á–∞—è.

**–ü—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –≤—ã–±–æ—Ä–µ.**<br>
&emsp;–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ  –æ–±—â–∞—è –¥–∏–∞–¥–∞, –≤ –∫–æ—Ç–æ—Ä–æ–π —É—á–∞—Å—Ç–≤—É—é—Ç –∞–∫—Ç–æ—Ä—ã, —è–≤–ª—è–µ—Ç—Å—è –≤–∑–∞–∏–º–Ω–æ–π –¥–∏–∞–¥–æ–π **–ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –≤—ã–±–æ—Ä–µ**, —Ä–∞–≤–Ω–∞:
$$P(i‚Üíj‚àßj‚Üíi)=\frac{d}{g-1}‚ãÖ(\frac{d}{g-1}+œÅ_{KP}‚ãÖ\frac{g-1-d}{g-1})$$

```r
# –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –≤—ã–±–æ—Ä–µ
p1 <- (d/(g-1))*((d/(g-1))+ro1*(g-1-d)/(g-1))
cat("–ü—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –≤—ã–±–æ—Ä–µ:\t", p1)
```

**–ü—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ.**<br>
&emsp;–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –∞–∫—Ç–æ—Ä—ã –∏–º–µ—é—Ç –≤–∑–∞–∏–º–Ω—ã–µ –æ—Ç–Ω–æ—à–µ–Ω–∏—è **–ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ**, —Ä–∞–≤–Ω–∞:<br>
$$P(i‚Üíj‚àßj‚Üíi)=\frac{X_{i+}\cdot X_{j+}}{(g-1)^2}$$


```r
# –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ
p2 <- (rowSums(matrGraph)*colSums(matrGraph))/(g - 1)^2
p2  <- sum(p2)/g
cat("–ü—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ:\t", p2)
```

##**–ó–∞–¥–∞–Ω–∏–µ 7.**
–ë—ã–ª –≤—ã—á–∏—Å–ª–µ–Ω –∏–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è. –ê —Ç–∞–∫–∂–µ —Å–¥–µ–ª–∞–Ω—ã –≤—ã–≤–æ–¥—ã.

&emsp;&emsp;**–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è** –æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–º –∑–Ω–∞—á–µ–Ω–∏—è—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ **–≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥**:<br> 
&emsp;$M_{min} = max_{0 \leq t \leq g}f(t)$,&emsp;–≥–¥–µ $$f(t) = \sum_{i=1}^tx_{i+} - t(g-1) - C^2_t$$<br>

&emsp;$M_{max} = [\frac{1}{2}(\sum_{i=1}^g - max_{0 \leq t \leq g}g(t))]$,&emsp;–≥–¥–µ $$g(t) = \sum_{i=1}^tx_{i+} - t(t-1) - \sum_{i=t+1}^gmin(t, x_{i+})$$

–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ:
$$\rho_{B} = \frac{M - M_{min}}{M_{max} - M_{min}}, $$

```r
f_t <- c() # –±—É–¥–µ—Ç —Ö—Ä–∞–Ω–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è f(t)
g_t <- c() # –±—É–¥–µ—Ç —Ö—Ä–∞–Ω–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è g(t)

# –ø–µ—Ä–µ–±–∏—Ä–∞–µ–º –æ—Ç 0 –¥–æ g (0<=t<=g)
for(t in 0:g){
    sum <- 0
    if (t>0){
        # –ø–µ—Ä–µ–±–∏—Ä–∞–µ–º i-—ã–µ —Å—Ç—Ä–æ–∫–∏ –∏ —Å—É–º–º–∏—Ä—É–µ–º –ø–æ —Å—Ç—Ä–æ–∫–∞–º
        for(i in 1:t) {
            sum <- sum+sum(matrGraph[i, ])
        }
    }
    #–≤—ã—á–∏—Å–ª—è–µ–º —Ç–µ–∫—É—â–µ–µ f(t)
    f_cur <- sum-t*(g-1)-choose(t, 2)
    # –∑–∞–ø–æ–º–∏–Ω–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ f(t), —á—Ç–æ–±—ã –ø–æ—Ç–æ–º –Ω–∞–π—Ç–∏ –º–∞–∫—Å–∏–º—É–º
    f_t <- append(f_t, f_cur)
    sum2 <- 0
    if (t + 1 <= g){
        for(i in (t+1):g){
            sum2 <-sum2+min(t,sum(matrGraph[i, ]))
        }
    }
    #–≤—ã—á–∏—Å–ª—è–µ–º —Ç–µ–∫—É—â–µ–µ g(t)
    g_cur <-sum-t*(t-1)-sum2
    # –∑–∞–ø–æ–º–∏–Ω–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ g(t), —á—Ç–æ–±—ã –ø–æ—Ç–æ–º –Ω–∞–π—Ç–∏ –º–∞–∫—Å–∏–º—É–º
    g_t<-append(g_t, g_cur)
}
m_min <- max(f_t)
m_max <- floor( 0.5*(sum(matrGraph)-max(g_t)) )
# –≤—ã—á–∏—Å–ª—è–µ–º –∏–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ø–æ —Ñ–æ—Ä–º—É–ª–µ
index <- ( M - m_min ) / ( m_max - m_min )
cat("–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è:", index)
```

##**–ó–∞–¥–∞–Ω–∏–µ 8.**
–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —á–∏—Å–ª–æ –¥—É–≥ $L$ –≤ –æ—Ä–≥—Ä–∞—Ñ–µ
$G_ùëë$ –∏–º–µ–µ—Ç –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ $L$ ‚àº $Bin( ùëî(ùëî ‚àí 1)$, $\frac {1} {2}$) –ø—Ä–∏
—É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $ùõº$. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è.

–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ $H_0: L ‚àº Bin(g(g-1), \frac{1}{2})$, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å "—É—Å–ø–µ—Ö–∞" –≤ –∫–∞–∂–¥–æ–º –∏—Å–ø—ã—Ç–∞–Ω–∏–∏ —Ä–∞–≤–Ω–∞ 0.5<br>
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ $H_A: L\not ‚àº Bin(g(g-1), \frac{1}{2})$ (–ª—é–±–æ–µ –¥—Ä—É–≥–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ)<br><br>
–í—ã–ø–æ–ª–Ω–∏–º —Ç–æ—á–Ω—É—é –ø—Ä–æ–≤–µ—Ä–∫—É –ø—Ä–æ—Å—Ç–æ–π –Ω—É–ª–µ–≤–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ *binom.test()*

##**–ó–∞–¥–∞–Ω–∏–µ 9.**
–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —á–∏—Å–ª–æ –¥—É–≥ $ùêø$ –≤ –æ—Ä–≥—Ä–∞—Ñ–µ $G_ùëë$
–∏–º–µ–µ—Ç –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ $ùêø$ ‚àº $Bin(ùëî(ùëî‚àí1), ùëÉ_0)$ –ø—Ä–∏ —É—Ä–æ–≤–Ω–µ
–∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ $ùõº$. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –ø—Ä–∏
–∫–æ—Ç–æ—Ä–æ–º –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è.

–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ $H_0: L ‚àº Bin(g(g-1), P_0)$, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å "—É—Å–ø–µ—Ö–∞" –≤ –∫–∞–∂–¥–æ–º –∏—Å–ø—ã—Ç–∞–Ω–∏–∏ —Ä–∞–≤–Ω–∞ $P_0$<br>
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ $H_A: L\not ‚àº Bin(g(g-1), P_0)$ (–ª—é–±–æ–µ –¥—Ä—É–≥–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ)<br><br>

–í 44 –≤–∞—Ä–∏–∞–Ω—Ç–µ $P_0=0.66, \ \alpha=0.05$

–í—ã–ø–æ–ª–Ω–∏–º —Ç–æ—á–Ω—É—é –ø—Ä–æ–≤–µ—Ä–∫—É –ø—Ä–æ—Å—Ç–æ–π –Ω—É–ª–µ–≤–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ *binom.test()*

##**–ó–∞–¥–∞–Ω–∏–µ 10.**
–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $ùëù$ —Å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–π –Ω–∞–¥—ë–∂–Ω–æ—Å—Ç—å—é $(1 ‚àí ùõº)$. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

>–ò—Ç–∞–∫, –æ—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $p$:
$$\hat {p} = \frac{L}{g(g - 1)},$$
–≥–¥–µ $L = \sum_{i = 1}^g\sum_{j = 1}^gx_{ij}$<br>

>–ê –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª:
$$\hat {P}_{lower} \leq P \leq  \hat {P}_{upper}, \ –≥–¥–µ$$
* &emsp;&emsp;$\hat {P}_{lower} = \hat {P} - z_{\frac {\alpha}{2}} \sqrt {\frac {\hat {P}(1 - \hat{P})}{g(g - 1)}}$
* &emsp;&emsp;$\hat {P}_{upper} = \hat {P} + z_{\frac {\alpha}{2}} \sqrt {\frac {\hat {P}(1 - \hat{P})}{g(g - 1)}}$
&emsp;&emsp;–ü—Ä–∏ $z_{\frac{\alpha}{2}}$: $\Phi(z_{\frac{\alpha}{2}}) = 0.5 - \frac{\alpha}{2}$

*–í—ã—á–∏—Å–ª–∏–º:*

```r
estimP<- L/(g*(g-1))
cat("–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ p =", estimP, '\n')
alpha <- 0.05
z <- abs(qnorm(alpha/2))
low <- estimP - z * sqrt( estimP *( 1-estimP ) / ( g*(g-1) ) )
up <- estimP + z * sqrt( estimP * ( 1-estimP ) / ( g*(g-1) ) )
cat("–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª = [", low,", ", up,"]")
```
