## Основные теоретические положения
---
* **_n-клика_** - максимальный подграф, в котором наибольшее геодезическое расстояние между любыми двумя узлами не превышает n.

* **_n-клан_** - n-клика, в которой геодезическое расстояние между всеми узлами в подграфе не превышает n для путей внутри подграфа.

* **_n-клуб_** - максимальный подграф диаметра меньше либо равно n.

---
* **_k-плекс_** - максимальный подграф, содержащий $g_s$ узлов, в котором каждый узел смежен с не менее чем с $g_s - k$ узлами подграфа.

* **_k-ядро_** - максимальный подграф, в котором каждая вершина связана минимум с _k_ другими вершинами этого же подграфа.

---
&emsp;**_Модулярность_** - характеристика структуры сети, в частности степень кластеризации узлов, когда внутри кластера наблюдается много узлов, а между кластерами мало узлов(низкая плотность). <br>
- Показатель модулярности определяется как разница между долей связей внутри группы и ожидаемой доли связей, если бы связи были размещены случайно.<br>
- Показатель модулярности принимает значения в диапазоне $[-\frac{1}{2}; 1]$.<br>
- Чем выше показатель, тем выше уровень кластеризации сети в соответствии с данной группировкой узлов.<br>
- Есть функция в библиотеке *igraph* *modularity()*

---
&emsp;**_Сеть аффилированности (affilation network)_** - сеть, в которой участники аффилированы друг с другом на основе сотрудничества в рамках какой-то группы или совместного участия в определенном событии.<br>
&emsp;Сеть состоит из двух измерений: <br>
- Множества акторов   $\mathscr N = \{n_1, n_2, ..., n_g\}$
- Множества событий $\mathscr M = \{m_1, m_2, ..., m_h\}$

&emsp;**_Математическое представление сетей аффилированности:_** <br>  
- **_Двудольный граф(bipartite graph)_** - это граф, узлы которого можно разделить на два подмножества, а все рёбра соединяют узлы, принадлежащие разным подмножествам.<br>
&emsp;&emsp;Представление биграфа в виде социоматрицы размера $(g+h) \ \times \ (g+h)$:

$$
\begin{pmatrix}
0_g & A \\
A^T & 0_h 
\end{pmatrix}
$$
    
- **_Гиперграф (hypergraph)_** - обобщение понятия графа; гиперграф состоит из множества объектов и множества и множества подмножеств объектов, в котором каждый объект принадлежит хотя бы одному подмножеству, причём ни одно подмножество не является пустым:
    
    $$
    \mathscr H = \mathscr{H(A, B),}
    $$
    
    где 
    
 &emsp;&emsp;   $\mathscr A = \{a_1, a_2, ..., a_g\} -$  множество объектов
    
 &emsp;&emsp;   $\mathscr B = \{B_1, B_2, …, B_h\}$
    
- **_Двойственный гиперграф (dual hypergraph)_** для гиперграфа $\mathscr H = \mathscr {H(A, B)}$:
    
    $$
    \mathscr {H^* = H(B, A)},
    $$
    
    где
    
  &emsp;&emsp; $\mathscr{B = \{b_1, b_2, ..., b_h\}}$<br>
  &emsp;&emsp; $\mathscr{A} = \{A_1, A_2, ..., A_h\}$

&emsp;**_Социометрическое представление сети аффилированности_:**<br>
- **_Матрица инцидентности (affilation network)_** - матрица А = $a_{ij}$ размерности  $g \ \times \ h$ c элементами
    
$$
a_{ij} =  \begin{equation*}  \left\{
    \begin{aligned}
      & 1, \:если\: актор\:i\: связан\: с \: событием \ j\\
      & 0, иначе
    \end{aligned}
  \right.
\end{equation*}
$$

---
**_Одномодальные проекции:_**<br>
* **_Матрица соучастия акторов:_**

&emsp;&emsp;&emsp;$X^{\cal{N}} = \{x_{ij}^{\cal{N}} \}$

&emsp;&emsp;&emsp;$x_{ij}^{\cal{N}} = \sum_{k = 1}^{h} a_{ik}a_{jk}$

* **_Матрица перекрытия событий:_**

&emsp;&emsp;&emsp;$X^{\cal{M}} = \{x_{kl}^{\cal{M}} \}$

&emsp;&emsp;&emsp;$x_{kl}^{\cal{M}} = \sum_{i = 1}^{g} a_{ik}a_{jl}$

---
**_Свойства акторов и событий:_**<br>
* **_Степень участия_** - количество событий, с которым связан каждый актор.

&emsp;&emsp;&emsp;$a_{i+} = \sum_{j = 1}^{h} a_{ij} = x_{ii}^{\cal{N}}$

* **_Среднее количество событий, с к-ми акторы аффилированы_** - 

&emsp;&emsp;&emsp;$\bar a_{i+} = 1/g \sum_{i=1}^g \sum_{j=1}^h a_{ij} = a_{++} / g = 1/g \sum ^g _{i=1} x_{ii}^\mathscr N$ 

* **_Размер событий:_**

&emsp;&emsp;&emsp;$a_{+j} = \sum_{i = 1}^{g} a_{ij} = x_{ii}^{\cal{M}}$

* **_Средний размер событий_** - среднее количество акторов в каждом событии или средняя степень узлов, прндставляющих события в двудольном графе.

&emsp;&emsp;&emsp;$\overline {a}_{ij} = \frac{1}{h} \sum_{j = 1}^{h} x_{jj}^{\cal{M}}$

---
**_Сплоченные подмножества акторов или событий:_**<br>
&emsp;**_Клика уровня C в взвешенном графе_** - максимальный полный подграф из трех или более узлов, все из которых являются смежными на уровне *C*, то есть между всеми парами узлов есть ребра со значениями, большими или равными *C*.

&emsp;**_Клика уровня C для отношения соучастия_** - подграф, в котором все пары акторов совместно участвуют не менее чем в _C_ событиях.

&emsp;**_Клика уровня C для отношения перекрытия_** - подграф, в котором все пары событий имеют не менее _C_ общих членов.

---
**_Диаметр сети аффилированности_** 

* _Диаметр аффилированной сети_ -длина самого длинного пути между любой парой акторов и/или событий.

* _Связная сеть аффилированности_ - сеть, в которой все пары узлов достижимы.


---
**_Свойства сетей аффилированности:_**
1. Количество перекрывающихся связей между событиями частично зависит от количества событий, к которым принадлежат акторы.

2. Количество связей соучастия между акторами частично зависит от размера событий.

3. Актор, принадлежащий $a_{i+}$ событиям, создает $C_{a_{i+}}^2$ парных связей между событиями.

4. Событие с участиеми $a_{+j}$ участников создает $C_{a_{+о}}^2$ связей между парами акторов.

5. Плотность для дихотомического отношения - доля присутствующих связей.

6. Плотность для ценностных отношений - среднее значение связей.

## **Задание 2.**
Данная сеть была исследована на предмет подгрупп с помощью:<br>&emsp;𝑛-клик;<br>&emsp;𝑛-кланов; <br>&emsp;𝑘-ядер;<br> Указать какие подгруппы получились. Сделать выводы.

```r
# установка библиотеки RBGL
if (!require("BiocManager", quietly = TRUE))
    install.packages("BiocManager")

BiocManager::install("RBGL")
```

```r
library(RBGL)
library(igraph)
# Исследование на предмет подгрупп
# с помощью n-клик
nCliques <- kCliques(igraph.to.graphNEL(gr))
#nCliques
# с помощью n-кланов
nclan <- function(g, n, ncliques){
    E(g)$weight <- 1 #just in case g has weights - does not modify original graph
    n.cand <- ncliques[[n]] #n-clique candidates to be an n-clan
    n.clan <- list() #initializes a list to store the n-clans
    n.clan.i <- 1 #initializes a list pointer
    for (n.cand.i in 1:length(n.cand)){ #loop over all of the candidates
        g.n.cand <- induced_subgraph(g,n.cand[[n.cand.i]]) #get the subgraph
        if (diameter(g.n.cand)<=n){ #check diameter of the subgraph
          n.clan[[n.clan.i]] <- n.cand[[n.cand.i]] #add n-clan to the list
          n.clan.i <- n.clan.i+1 #increment list pointer
        }
      }
    return(n.clan) #return the entire list
}

nclans <- function(g, ncliques){
    result <- c()
    for (i in names(ncliques)){
        result[[i]] <- nclan(g, i, ncliques)
    }
    return(result)
}

nClans <- nclans(gr, nCliques)
nClans
# с помощью k-ядер
inf <- coreness(gr, mode= "all")
kKern <- lapply(0:max(as.integer(inf)), function(x) names(inf)[which(as.integer(inf) >= x)])
cat('k-ядра\n')
kKern
```

Также были найдены n-кланы (это n-клики, в которых геодезическое расстояние не превышает n для путей внутри подграфа). И k-ядра, которые являются максимльными подграфами, в которых каждая вершина связана минимум с k другими вершинами из этого же подграфа.<br>

&emsp;Чем выше показатель модулярности, тем выше уровень кластеризации в соответствии с данной подгруппой.

Алгоритм *Spinglass*: Алгоритм медленный и недетерменированный. Задается количество состояний поворота N - это будет верхняя граница для количества обнаруженных сообществ, и количество итераций. Каждый актор сети находится в одном из этих N состояний. На каждой итерации на основе состояний смежных акторов определяется, куда будет совершен поворот актора. Акторы, находящиеся в одном состоянии поворота, относятся к одному сообществу.

Алгоритм *Walktrap*: Считается матрица вероятностей перехода, на основе которой строится метрика, позволяющая разбить акторов на кластеры - сообщества. Идея алгоритма - если случайно блуждать по графу, то вероятность "хождения" между акторами из одного сообщества выше, чем вероятность перехода в другое. 

Алгоритм *Label propagation*: Входные данные - сеть, используется список ее связей. Данный алгоритм основан на идее распространения ярлыков. Изначально каждый актор сети имеет свой собственный ярлык, затем акторам на каждой итерации присваивается ярлык «соседей», причем выбирается ярлык, который более представлен среди соседей данного актора. Критерием для прекращения итераций является условие "каждый актор имеет соседей со "своим" ярлыком не меньше, чем с другим". Алгоритм является простым и быстрым в работе.
