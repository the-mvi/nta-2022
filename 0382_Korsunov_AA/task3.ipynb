## Основные теоретические положения

**Диада** - элемент сети, представляющий собой пару акторов и (возможную) связь между ними: $D_{ij} = (X_{ij}, X{ji}), i \neq j$
****
**Взаимная диада**: $D_{ij}=(1,1)$
****
**Асимметричная диада**: $D_{ij} = (1,0) или D_{ij} = (0,1)$
****
**Нулевая диада**: $D_{ij} = (0,0)$
****
**Цензус диад** - тройка чисел $(M,A,N)$, где

$$M = \frac{1}{2}trace(XX)$$
$$A = trace(XX^T) - trace(XX)$$
$$N = C_g^2 - trace(XX^T) + \frac{1}{2}trace(XX)$$
****
**Индекс взаимности Каца-Пауэлла** - характеристика изменения тенденции акторов в группе
$$-\infty < p_{кр} \leq 1$$

*   Индекс взаимности основан на ожидании количества взаимных диад, предполагая, что выбор делается акторами случайным образом

а) Фиксированны выбор:

Предполагается, что каждый актор делает d выборов из (g-1) акторов, доступных для выбора. Вероятность взаимного выбора между любой произвольной парой акторов равна $d^2/(g-1)^2$

Математическое ожидание взаимных диад:
$$E_M = \frac{g(g-1)}{2}\frac{d^2}{(g-1)^2} = \frac{gd^2}{2(g-1)}$$

Вероятность того, что общая диада, в которой учавствуют акторы, является взаимной диадой:
$$p(i\rightarrow j \wedge j \rightarrow i) = \frac{d}{g-1}(\frac{d}{g-1}+p_{кр}\frac{g-1-d}{g-1})$$

Условная вероятность:
$$p(j\rightarrow i | i \rightarrow j) = p(j\rightarrow i)+p_{кр}p(j\nrightarrow i)$$

$$E(M|p) = \frac{gd^2}{2(g-1)}(1-p_{кр})+\frac{gd}{2}p_{кр}$$

Оценка $p_{кр}$ по методу моментов:
$$\hat p_{кр} = \frac{2(g-1)M-gd^2}{gd(g-1-d)}$$

б) Свободный выбор:

Пусть $i$-ый актор выбирает $d_0(n_i) = x_{i+}$  других акторов

Пусть $L = \sum_ix_{i+}$, $L_2 = \sum_ix_{i+}^2$

Вероятнось того, что акторы имеют взаимные отношения равна $x_{i+}x_{j+}/(g-1)^2$

Математическое ожидание взаимных диад:

$$E_M = \frac{L^2-L_2}{2(g-1)^2}$$

Математическое ожидание взаимных диад с учетом вероятности:
$$E(M|p') = \frac{L^2-L_2}{2(g-1)}(1-p'_{кр})+\frac{L}{2}p_{кр}$$

Оценка $p_{кр}$ по методу моментов:
$$p'_{кр} = \frac{2(g-1)^2M-L^2+L_2}{L(g-1)^2-L^2+L_2}$$
****

**Индекс взаимности Бандиопадхьяйя**:

Данный индекс основан на минимальном и максимальном значениях количества взаимных диад

Значение индекса:
$$p_B = \frac{M-M_{min}}{M_{max}-M_{min}}$$

$$M_{min} = maxf(t)_{0 \leq t \leq g}$$

$$M_{max}=[1/2(\sum_{i=1}^gx_{i+})-maxg(t)_{0 \leq t \leq g}]$$

$$f(t) = \sum_{i=1}^tx_{i+}-t(g-1)-C_t^2$$

$$g(t) - \sum_{i=1}^tx_{i+}-t(t-1)-\sum_{i=t+1}^gmin(t,x_{i+})$$
****
**Равномерное распределение и распределение Бернулли**:
$\#G_d(N) = 2^{g(g-1)}$

Вероятность получить социоматрицу x: $$p(X=x)= 2^{g(1-g)}$$

Вероятность появления дуги между акторами:
\begin{equation*}
p(x_{ij}=1) = 
 \begin{cases}
   \frac{1}{2} &i \neq g\\
   0 &i = g
 \end{cases}
\end{equation*}

\begin{equation*}
p(x_{ij}=1) = 
 \begin{cases}
   p_{ij} &i \neq j\\
   0 &i = j
 \end{cases}
\end{equation*}
****

**Нулева гипотеза**: $H_0: L$~$B(g(g-1,P_0)$

Статистика:
$$z_l = \frac{l-E(L)}{\sqrt{D(L)}} = \frac{l-g(g-1)/2}{\sqrt{g(g-1)/4}}$$ где $l = \sum_{i=1}^g\sum_{j=1}^gx_{ij}$

Данная статистика является приблизительно стандартным нормальным распределением при заданном уровне значимости
****

**Оценка параметра p**:
$$\hat p = \frac{\sum_{i=1}^g\sum_{j=1}^gx_{ij}}{g(g-1)} = \frac{1}{g(g-1)}$$
****
**Доверительный интервал**:
$$\hat p_{lower} \leq p \leq \hat p_{upper}$$

$\hat p_{lower} = \hat p - z_{a/2}\sqrt(\hat p (1-\hat p)/g(g-1))$

$\hat p_{upper} = \hat p + z_{a/2}\sqrt(\hat p (1-\hat p)/g(g-1))$

Пункт 1

С помощью библиотеки igraph смоделировать случайный орграф
$G_𝑑$ на основе модели Эрдеша—Реньи: $G_𝑑 = G (𝑔, 𝑝)$, где 𝑔 – количество акторов, 𝑝 – вероятность появления дуги между акторами. В R
для модели случайного графа Эрдеша—Реньи существует функция
erdos.renyi.game(). Ознакомьтесь с данной моделью и опишите
её.

В модели ${\displaystyle G(g,p)}$ граф строится путём случайного добавления рёбер. Каждое ребро включается в граф с вероятностью p независимо от остальных рёбер. Эквивалентно, все графы с $n$ узлами и $M$ рёбрами имеют одинаковую вероятность.
$${\displaystyle p^{M}(1-p)^{{n \choose 2}-M}.}$$
Параметр $p$ в этой модели можно рассматривать как функцию веса. По мере роста $p$ от 0 к 1 модель включает с большей вероятностью графы с бо́льшим числом рёбер. В частности, случай $p$=0,5 соответствует случаю, когда все ${\displaystyle 2^{\binom {n}{2}}}$ графы с $n$ вершинами будут выбраны с равной вероятностью.

```r
graph <- erdos.renyi.game(g, p, directed = T)
```

 Пункт 3
 
 Вычислить «пятичисловую сводку» для графа $G_𝑑$

```r
net <- network(as.matrix(graph), directed = T)

print("Размер:")
size_graph <- network.size(net)#размер
print(size_graph)
print("------------------------------")

print("Плотность:")
S_net <- network.density(net)#плотность
print(S_net)
print("------------------------------")

print("Количество компонент")
count_components_net <- count_components(graph) #количество компонент
print(count_components_net)
print("------------------------------")

print("Диаметр")
diametr_net <- diameter(graph) #диаметр
print(diametr_net)
print("------------------------------")

print("Коэффициент кластеризации")
klast <- gtrans(net) #коэффициент кластеризации
print(klast)
```

Пункт 5

Вычислить оценку индекса взаимности Каца—Пауэла при фиксированом выборе акторов $𝑑$ и при свободном выборе. Сделать выводы.

```r
d <- 32 #пусть d будет равен 32
M <- dyad_cen$mut

L <- sum(degree(graph, mode= "out"))
L_2 <- sum(degree(graph, mode= "out")^2)

print("Оценка индекса при фиксированном выборе:")
p_fix <- (2*(size_graph-1)*M-size_graph*d**2)/(size_graph*d*(size_graph-1-d))
print(p_fix)

print("Оценка индекса при свободном выборе:")
p_free <- (2*(size_graph-1)^2 * M - L^2 + L_2)/(L*(size_graph)^2 - L^2 + L_2)
print(p_free)
```

Выводы:

В обоих случаях оценка близка к 0, можно сделать вывод, что акторы больше склонны образовывать односторонние связи, чем взаимные

Пункт 6

Вычислить вероятность того, что связь будет взаимной для каждого случая.

```r
print("При фиксированном выборе:")
p_together_fix <- (d/(size_graph-1)) * (d/(size_graph-1)+p_fix*(size_graph-1-d)/(size_graph-1))
print(p_together_fix)

print("----------------------------")
print("При свободном выборе:")

p_together_free <- degree(graph, mode="out")*degree(graph, mode="in")/(size_graph-1)^2
print("Максимальная:")
print(max(p_together_free))
print("Минимальная:")
print(min(p_together_free))
print("Средняя:")
print(mean(p_together_free))
```

Пункт 7

Вычислить индекс взаимности Бандиопадхьяйя. Сделать выводы

```r
temp <- degree(graph, mode= "out")

f_f <- function(t)
{
  sum(temp[1:t]) - t*(size_graph-1) - choose(t,2)
} 

g_f <- function(t)
{
  min_vec <- c()
  for(i in (t+1):size_graph)
  {
      min_vec <- c(min_vec, min(t, temp[i]))
  }
  na.omit(sum(temp[1:t]) - t*(t-1) - sum(min_vec))
}

M_min <- max(sapply(0:size_graph, f_f))
M_max <- floor((1/2)*(sum(temp) - max(unlist(sapply(0:size_graph, g_f)))))
p_B <- (M - M_min)/(M_max - M_min)

print(p_B)
```

Выводы:

Индекс взаимности Бандиопадхьяйя примерно равен p (p = 0.51), т.е. примерно равен вероятности наличия связи между двумя акторами
