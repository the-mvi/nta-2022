## –¶–µ–ª—å —Ä–∞–±–æ—Ç—ã
–û–≤–ª–∞–¥–µ—Ç—å –Ω–∞–≤—ã–∫–∞–º–∏ —Ä–∞–±–æ—Ç—ã —Å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º–∏ –ø–∞–∫–µ—Ç–∞–º–∏ –¥–ª—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ —Å–æ—Ü–∏–∞–ª—å–Ω—ã—Ö —Å–µ—Ç–µ–π.

## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è

**–î–∏–∞–¥–∞** - —ç–ª–µ–º–µ–Ω—Ç —Å–µ—Ç–∏, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∏–π —Å–æ–±–æ–π –ø–∞—Ä—É –∞–∫—Ç–æ—Ä–æ–≤ –∏ (–≤–æ–∑–º–æ–∂–Ω—É—é) —Å–≤—è–∑—å –º–µ–∂–¥—É –Ω–∏–º–∏: $D_{ij} = (X_{ij}, X{ji}), i \neq j$
****
**–í–∑–∞–∏–º–Ω–∞—è –¥–∏–∞–¥–∞**: $D_{ij}=(1,1)$
****
**–ê—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞—è –¥–∏–∞–¥–∞**: $D_{ij} = (1,0) –∏–ª–∏ D_{ij} = (0,1)$
****
**–ù—É–ª–µ–≤–∞—è –¥–∏–∞–¥–∞**: $D_{ij} = (0,0)$
****
**–¶–µ–Ω–∑—É—Å –¥–∏–∞–¥** - —Ç—Ä–æ–π–∫–∞ —á–∏—Å–µ–ª $(M,A,N)$, –≥–¥–µ

$$M = \frac{1}{2}trace(XX)$$
$$A = trace(XX^T) - trace(XX)$$
$$N = C_g^2 - trace(XX^T) + \frac{1}{2}trace(XX)$$
****
**–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ö–∞—Ü–∞-–ü–∞—É—ç–ª–ª–∞** - —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞ –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–Ω–¥–µ–Ω—Ü–∏–∏ –∞–∫—Ç–æ—Ä–æ–≤ –≤ –≥—Ä—É–ø–ø–µ
$$-\infty < p_{–∫—Ä} \leq 1$$

*   –ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ –æ–∂–∏–¥–∞–Ω–∏–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥, –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞—è, —á—Ç–æ –≤—ã–±–æ—Ä –¥–µ–ª–∞–µ—Ç—Å—è –∞–∫—Ç–æ—Ä–∞–º–∏ —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º

–∞) –§–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã –≤—ã–±–æ—Ä:

–ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è, —á—Ç–æ –∫–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä –¥–µ–ª–∞–µ—Ç d –≤—ã–±–æ—Ä–æ–≤ –∏–∑ (g-1) –∞–∫—Ç–æ—Ä–æ–≤, –¥–æ—Å—Ç—É–ø–Ω—ã—Ö –¥–ª—è –≤—ã–±–æ—Ä–∞. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤–∑–∞–∏–º–Ω–æ–≥–æ –≤—ã–±–æ—Ä–∞ –º–µ–∂–¥—É –ª—é–±–æ–π –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–π –ø–∞—Ä–æ–π –∞–∫—Ç–æ—Ä–æ–≤ —Ä–∞–≤–Ω–∞ $d^2/(g-1)^2$

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥:
$$E_M = \frac{g(g-1)}{2}\frac{d^2}{(g-1)^2} = \frac{gd^2}{2(g-1)}$$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –æ–±—â–∞—è –¥–∏–∞–¥–∞, –≤ –∫–æ—Ç–æ—Ä–æ–π —É—á–∞–≤—Å—Ç–≤—É—é—Ç –∞–∫—Ç–æ—Ä—ã, —è–≤–ª—è–µ—Ç—Å—è –≤–∑–∞–∏–º–Ω–æ–π –¥–∏–∞–¥–æ–π:
$$p(i\rightarrow j \wedge j \rightarrow i) = \frac{d}{g-1}(\frac{d}{g-1}+p_{–∫—Ä}\frac{g-1-d}{g-1})$$

–£—Å–ª–æ–≤–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å:
$$p(j\rightarrow i | i \rightarrow j) = p(j\rightarrow i)+p_{–∫—Ä}p(j\nrightarrow i)$$

$$E(M|p) = \frac{gd^2}{2(g-1)}(1-p_{–∫—Ä})+\frac{gd}{2}p_{–∫—Ä}$$

–û—Ü–µ–Ω–∫–∞ $p_{–∫—Ä}$ –ø–æ –º–µ—Ç–æ–¥—É –º–æ–º–µ–Ω—Ç–æ–≤:
$$\hat p_{–∫—Ä} = \frac{2(g-1)M-gd^2}{gd(g-1-d)}$$

–±) –°–≤–æ–±–æ–¥–Ω—ã–π –≤—ã–±–æ—Ä:

–ü—É—Å—Ç—å $i$-—ã–π –∞–∫—Ç–æ—Ä –≤—ã–±–∏—Ä–∞–µ—Ç $d_0(n_i) = x_{i+}$  –¥—Ä—É–≥–∏—Ö –∞–∫—Ç–æ—Ä–æ–≤

–ü—É—Å—Ç—å $L = \sum_ix_{i+}$, $L_2 = \sum_ix_{i+}^2$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—å —Ç–æ–≥–æ, —á—Ç–æ –∞–∫—Ç–æ—Ä—ã –∏–º–µ—é—Ç –≤–∑–∞–∏–º–Ω—ã–µ –æ—Ç–Ω–æ—à–µ–Ω–∏—è —Ä–∞–≤–Ω–∞ $x_{i+}x_{j+}/(g-1)^2$

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥:

$$E_M = \frac{L^2-L_2}{2(g-1)^2}$$

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥ —Å —É—á–µ—Ç–æ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏:
$$E(M|p') = \frac{L^2-L_2}{2(g-1)}(1-p'_{–∫—Ä})+\frac{L}{2}p_{–∫—Ä}$$

–û—Ü–µ–Ω–∫–∞ $p_{–∫—Ä}$ –ø–æ –º–µ—Ç–æ–¥—É –º–æ–º–µ–Ω—Ç–æ–≤:
$$p'_{–∫—Ä} = \frac{2(g-1)^2M-L^2+L_2}{L(g-1)^2-L^2+L_2}$$
****

**–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è**:

–î–∞–Ω–Ω—ã–π –∏–Ω–¥–µ–∫—Å –æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–º –∑–Ω–∞—á–µ–Ω–∏—è—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥

–ó–Ω–∞—á–µ–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–∞:
$$p_B = \frac{M-M_{min}}{M_{max}-M_{min}}$$

$$M_{min} = maxf(t)_{0 \leq t \leq g}$$

$$M_{max}=[1/2(\sum_{i=1}^gx_{i+})-maxg(t)_{0 \leq t \leq g}]$$

$$f(t) = \sum_{i=1}^tx_{i+}-t(g-1)-C_t^2$$

$$g(t) - \sum_{i=1}^tx_{i+}-t(t-1)-\sum_{i=t+1}^gmin(t,x_{i+})$$
****
**–†–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ë–µ—Ä–Ω—É–ª–ª–∏**:
$\#G_d(N) = 2^{g(g-1)}$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å —Å–æ—Ü–∏–æ–º–∞—Ç—Ä–∏—Ü—É x: $$p(X=x)= 2^{g(1-g)}$$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—è–≤–ª–µ–Ω–∏—è –¥—É–≥–∏ –º–µ–∂–¥—É –∞–∫—Ç–æ—Ä–∞–º–∏:
\begin{equation*}
p(x_{ij}=1) = 
 \begin{cases}
   \frac{1}{2} &i \neq g\\
   0 &i = g
 \end{cases}
\end{equation*}

\begin{equation*}
p(x_{ij}=1) = 
 \begin{cases}
   p_{ij} &i \neq j\\
   0 &i = j
 \end{cases}
\end{equation*}
****

**–ù—É–ª–µ–≤–∞ –≥–∏–ø–æ—Ç–µ–∑–∞**: $H_0: L$~$B(g(g-1,P_0)$

–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:
$$z_l = \frac{l-E(L)}{\sqrt{D(L)}} = \frac{l-g(g-1)/2}{\sqrt{g(g-1)/4}}$$ –≥–¥–µ $l = \sum_{i=1}^g\sum_{j=1}^gx_{ij}$

–î–∞–Ω–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —è–≤–ª—è–µ—Ç—Å—è –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
****

**–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ p**:
$$\hat p = \frac{\sum_{i=1}^g\sum_{j=1}^gx_{ij}}{g(g-1)} = \frac{1}{g(g-1)}$$
****
**–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª**:
$$\hat p_{lower} \leq p \leq \hat p_{upper}$$

$\hat p_{lower} = \hat p - z_{a/2}\sqrt(\hat p (1-\hat p)/g(g-1))$

$\hat p_{upper} = \hat p + z_{a/2}\sqrt(\hat p (1-\hat p)/g(g-1))$

$z_{a/2}: –§(z_a/2) = 0/5 - a/2$



## –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏
–° –ø–æ–º–æ—â—å—é –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö –ø–∞–∫–µ—Ç–æ–≤ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫ –ø—Ä–æ–≤–µ—Å—Ç–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ —Å–ª—É—á–∞–π–Ω–æ —Å–º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∞. –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ–¥–µ—Ä–∂–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å.

–í –º–æ–¥–µ–ª–∏ ${\displaystyle G(g,p)}$ –≥—Ä–∞—Ñ —Å—Ç—Ä–æ–∏—Ç—Å—è –ø—É—Ç—ë–º —Å–ª—É—á–∞–π–Ω–æ–≥–æ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è —Ä—ë–±–µ—Ä. –ö–∞–∂–¥–æ–µ —Ä–µ–±—Ä–æ –≤–∫–ª—é—á–∞–µ—Ç—Å—è –≤ –≥—Ä–∞—Ñ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é p –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Ä—ë–±–µ—Ä. –≠–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ, –≤—Å–µ –≥—Ä–∞—Ñ—ã —Å $n$ —É–∑–ª–∞–º–∏ –∏ $M$ —Ä—ë–±—Ä–∞–º–∏ –∏–º–µ—é—Ç –æ–¥–∏–Ω–∞–∫–æ–≤—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å.

–ü–∞—Ä–∞–º–µ—Ç—Ä $p$ –≤ —ç—Ç–æ–π –º–æ–¥–µ–ª–∏ –º–æ–∂–Ω–æ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å –∫–∞–∫ —Ñ—É–Ω–∫—Ü–∏—é –≤–µ—Å–∞. –ü–æ –º–µ—Ä–µ —Ä–æ—Å—Ç–∞ $p$ –æ—Ç 0 –∫ 1 –º–æ–¥–µ–ª—å –≤–∫–ª—é—á–∞–µ—Ç —Å –±–æ–ª—å—à–µ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –≥—Ä–∞—Ñ—ã —Å –±–æÃÅ–ª—å—à–∏–º —á–∏—Å–ª–æ–º —Ä—ë–±–µ—Ä. –í —á–∞—Å—Ç–Ω–æ—Å—Ç–∏, —Å–ª—É—á–∞–π $p$=0,5 —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Å–ª—É—á–∞—é, –∫–æ–≥–¥–∞ –≤—Å–µ ${\displaystyle 2^{\binom {n}{2}}}$ –≥—Ä–∞—Ñ—ã —Å $n$ –≤–µ—Ä—à–∏–Ω–∞–º–∏ –±—É–¥—É—Ç –≤—ã–±—Ä–∞–Ω—ã —Å —Ä–∞–≤–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é.

–ü—Ä–µ–æ–±–ª–∞–¥–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞—Å—Å–∏–º–µ—Ç—Ä–∏—á–Ω—ã—Ö –¥–∏–∞–¥

```r
d <- 32 #–ø—É—Å—Ç—å d –±—É–¥–µ—Ç —Ä–∞–≤–µ–Ω 32
M <- dyad_cen$mut

L <- sum(degree(graph, mode= "out"))
L_2 <- sum(degree(graph, mode= "out")^2)

print("–û—Ü–µ–Ω–∫–∞ –∏–Ω–¥–µ–∫—Å–∞ –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –≤—ã–±–æ—Ä–µ:")
p_fix <- (2*(size_graph-1)*M-size_graph*d**2)/(size_graph*d*(size_graph-1-d))
print(p_fix)

print("–û—Ü–µ–Ω–∫–∞ –∏–Ω–¥–µ–∫—Å–∞ –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ:")
p_free <- (2*(size_graph-1)^2 * M - L^2 + L_2)/(L*(size_graph)^2 - L^2 + L_2)
print(p_free)
```

–í—ã–≤–æ–¥—ã:

–í –æ–±–æ–∏—Ö —Å–ª—É—á–∞—è—Ö –æ—Ü–µ–Ω–∫–∞ –±–ª–∏–∑–∫–∞ –∫ 0, –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ –∞–∫—Ç–æ—Ä—ã –±–æ–ª—å—à–µ —Å–∫–ª–æ–Ω–Ω—ã –æ–±—Ä–∞–∑–æ–≤—ã–≤–∞—Ç—å –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–µ —Å–≤—è–∑–∏, —á–µ–º –≤–∑–∞–∏–º–Ω—ã–µ

```r
print("–ü—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –≤—ã–±–æ—Ä–µ:")
p_together_fix <- (d/(size_graph-1)) * (d/(size_graph-1)+p_fix*(size_graph-1-d)/(size_graph-1))
print(p_together_fix)

print("----------------------------")
print("–ü—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ:")

p_together_free <- degree(graph, mode="out")*degree(graph, mode="in")/(size_graph-1)^2
print("–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è:")
print(max(p_together_free))
print("–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è:")
print(min(p_together_free))
print("–°—Ä–µ–¥–Ω—è—è:")
print(mean(p_together_free))
```

–ü—É–Ω–∫—Ç 7

–í—ã—á–∏—Å–ª–∏—Ç—å –∏–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã

```r
temp <- degree(graph, mode= "out")

f_f <- function(t)
{
  sum(temp[1:t]) - t*(size_graph-1) - choose(t,2)
} 

g_f <- function(t)
{
  min_vec <- c()
  for(i in (t+1):size_graph)
  {
      min_vec <- c(min_vec, min(t, temp[i]))
  }
  na.omit(sum(temp[1:t]) - t*(t-1) - sum(min_vec))
}

M_min <- max(sapply(0:size_graph, f_f))
M_max <- floor((1/2)*(sum(temp) - max(unlist(sapply(0:size_graph, g_f)))))
p_B <- (M - M_min)/(M_max - M_min)

print(p_B)
```

–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è –ø—Ä–∏–º–µ—Ä–Ω–æ —Ä–∞–≤–µ–Ω p (p = 0.51), —Ç.–µ. –ø—Ä–∏–º–µ—Ä–Ω–æ —Ä–∞–≤–µ–Ω –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –Ω–∞–ª–∏—á–∏—è —Å–≤—è–∑–∏ –º–µ–∂–¥—É –¥–≤—É–º—è –∞–∫—Ç–æ—Ä–∞–º–∏

–ü—Ä–æ–≤–µ—Ä–∏—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —á–∏—Å–ª–æ –¥—É–≥ ùêø –≤ –æ—Ä–≥—Ä–∞—Ñ–µ
$G_ùëë$ –∏–º–µ–µ—Ç –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ ùêø ‚àº Bin(ùëî(ùëî ‚àí 1), 1/2) –ø—Ä–∏
—É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ ùõº. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è.

```r
alpha <- 0.05

hip <- binom.test(length(E(graph)), g*(g-1), p = 0.5)

print("–í–µ—Ä–Ω–∞ –ª–∏ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞:")
print(hip$p.value >= alpha)
print("–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏:")
print(hip$p.value)
```
