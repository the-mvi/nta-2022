## Основные теоретические положения
***Акторы*** - это дискретные индивидуальные, корпоративные или коллективные социальные единиц.

***Социоматрица*** - матрица X размерности gxg, описывающая смежные узлы в сети: $X = {x_{ij}}, i, j = 1, ..., g, g$ - количество акторов.

***Алгоритм Фрюхтермана-Рейнгольдса*** - Алгоритм Фрюхтермана-Рейнгольда
является силовым алгоритмом размещения графа. Суть этого алгоритма заключается в том, что между каждой парой вершин действует сила отталкивания. Если две вершины связаны ребром, то между ними действует сила притяжения. Силы притяжения и отталкивания вычисляются по следующим формулам:

$F_{rep}(p_u,p_v)=l^2 / x$ - сила отталкивания 

$F_{attr}(p_u,p_v)=x^2 / l$ - сила притяжения

где $l = \sqrt{(area/numberOfVertices)}$ – идеальная длина, зависящая от площади и количества вершин, $x$ – расстояние между двумя вершинами.

Сила пружины вычисляется как сумма силы отталкивания и сил притяжения:
$F_{spring(p_u,p_v)} = F{attr(p_u,p_v)}+F{rep(p_u,p_v)}$

***Алгоритм Камада-Каваи*** - Алгоритм Камада-Каваи применяется для неориентированных взвешенных графов. Граф представляется в виде динамической системы, в которой каждые две вершины соединены «пружиной» желаемой длины. Затем рассчитывается оптимальное представление графа – состояние, в котором суммарная энергия системы будет минимальной.
Модель:

$E = ∑_{i=1}^{n-1}\sum_{j=i+1}^n\frac{1}{2}k_{ij}(|p_i-p_j|-l_{ij})^2$

$n$ - кол-во вершин,

$p_i$ - положение вершины $i$,

$k_{ij}$ – сила пружины между $p_i$ и $p_j$,
$l_{ij} = L\times d{ij}$

$L$ – желаемая длина ребра в графе (на плоскости),

$d_{ij}$ – кратчайший путь между вершинами $i$ и $j$

$k_{ij} = \frac{K}{d^2}$

$K$ – константа

***Пятичисленная сводка сети:***

$(a)$ *Размер* - количество акторов сети

$(b)$ *Плотность* - доля имеющихся связей по отношению к максимально возможному количеству связей
*   Для ориентированных сетей: $\Delta = \frac{L}{A_g^2}$ 
*   Для неориентированных сетейЖ $\Delta = \frac{L}{C_g^2}$

$(c)$ *Компонента* - подгруппа, в которой все акторы связаны друг с другом прямо или косвенно

$(d)$ *Диаметр* - наибольшая длина любого из кратчайших путей между всеми парами узлов

$(e)$ *Коэффициент кластеризации* - доля закрытых треугольников по отношению к общему количеству открытых и закрытых треугольников

***Средняя степень узлов***: $\bar{d} = \frac{1}{g}\sum_{i=1}^gd(n_i)$, где $d(n_i)$ - степень узла - число прямых, инцидентных узлу n_i

***Дисперсия степеней узлов***: $S_D^2 = \frac{1}{g}\sum_{i=1}^g(d(n_i) - \bar{d})^2$ 

***Меры центральности***:

$(a)$ *Нормированная центральность по степени*: $C'_D(n_i) = \frac{d(n_i)}{g-1}$

$(b)$ *Нормированная центральность по близости*: $C'_С(n_i) = \frac{g-1}{\sum_{j=1}^gd(n_i,n_j)} = (g-1)C_C(n_i)$, где $C_C(n_i) = (\sum_{j=1}^gd(n_i,n_j))^{-1}$

$(c)$ *Нормированная центральность по посредничеству*: $C'_B(n_i) = \frac{C_B(n_i)}{(g-1)(g-2)/2}$, где ${C_B(n_i)} = \sum_{j<k}\frac{g_{jk}(n_i)}{g_{jk}}$

$(d)$ *Нормированная центральность по информации*: $C'_I(n_i) = \frac{C_I(n_i)}{\sum_iC_I(n_i)}$

***Централизация***:

$(a)$ *Централизация по степени*: $C_D = \frac{\sum_{i=1}^g(C_D(n^*)-C_D(n_i))}{(g-1)(g-2)}$

$(b)$ *Централизация по близости*: $C_С = \frac{\sum_{i=1}^g(C'_C(n^*)-C'_C(n_i)}{(g-2)(g-1)/(2g-3)}$

$(c)$ *Централизация по посредничеству*: $C_B = \frac{2\sum_{i=1}^g(C'_B(n^*)-C'_B(n_i))}{(g-1)^2(g-2)}$

$(d)$ *Централизация по информации*: $C_I = \frac{1}{g}\sum_{i=1}^g(C'_I(n_i)-\bar{C}_I)^2$, где $\bar{C}_I = \sum_iC_I(n_i)$

***Точки сочленения*** - узлы, удаление которых увеличит число компонент связности

***Мосты*** - ребра, удаление которых разделит одну компоненту связности на две.

***Престиж*** - мера, характеризующая актор с большим количеством входных связей.

***Меры престижа***:

$(a)$ *Нормированный престиж по степени*: $P'_D(n_i) = \frac{x_{+i}}{g-1}$, где $x_{+i}$ - входящая степень каждого узла  
$(b)$ *Нормированный престиж по близости*: $P_P(n_i) = \frac{1}{g}\sum_{i=1}^g(P_P(n_i)-\bar{P}_P)^2$, где $\bar{P}_p = \frac{1}{g}\sum_{i=1}^gP_P(n_i)$
