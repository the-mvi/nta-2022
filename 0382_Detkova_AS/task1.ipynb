* –°—Ç–µ–ø–µ–Ω–∏ —É–∑–ª–æ–≤ –±—ã–ª–∏ –Ω–∞–π–¥–µ–Ω—ã —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ *degree()*, –∫—É–¥–∞ –±—ã–ª–∞ –ø–µ—Ä–µ–¥–∞–Ω–∞ —Å–µ—Ç—å maritals_network –∏ –∞—Ä–≥—É–º–µ–Ω—Ç *gmode = "graph"*, –∫–æ—Ç–æ—Ä—ã–π —Å–æ–æ–±—â–∞–µ—Ç —Ñ—É–Ω–∫—Ü–∏–∏, —á—Ç–æ –≥—Ä–∞—Ñ –Ω–µ–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–π ("digraph" - –≥—Ä–∞—Ñ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–π). –î–∞–ª–µ–µ –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–π —Å—Ç–µ–ø–µ–Ω–∏ —Å—É–º–º–∞ —Å—Ç–µ–ø–µ–Ω–µ–π –±—ã–ª–∞ —Ä–∞–∑–¥–µ–ª–µ–Ω–∞ –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–µ—Ä—à–∏–Ω –≤ –≥—Ä–∞—Ñ–µ: $ \overline{d} = \frac {1} {g} * \sum_{i=1}^{g} {d(n_i)}$ , $g$ - —á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤ –≤ —Å–µ—Ç–∏.

````r
degrees <- degree(friendship_network, gmode = "digraph", cmode = "indegree")
mid_degree <- sum(degrees) / length(degrees)
cat("Average input degree of nodes =", mid_degree)
disp <- var(degrees)
cat("\nInput variance =", disp)

degrees <- degree(friendship_network, gmode = "digraph", cmode = "outdegree")
mid_degree <- sum(degrees) / length(degrees)
cat("\nAverage output degree of nodes =", mid_degree)

disp <- var(degrees)
cat("\nOutput variance =", disp)
````

````r
prest_deg <- prestige(friendship_network, gmode="digraph", cmode="indegree")

# –Ω–∞—Ö–æ–¥–∏–º –≥–µ–æ–¥–µ–∑–∏—á–µ—Å–∫–∏–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è
gd <- geodist(friendship_network)
# —É–±–∏—Ä–∞–µ–º —Å—Ç—Ä–æ–∫–∏ –≤–µ—Ä—à–∏–Ω, –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ—Ç –∏—Å—Ö–æ–¥—è—â–∏—Ö —Å–≤—è–∑–µ–π
# —Å—É–º–º–∞ –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤ - —Å—É–º–º–∞ –≥–µ–æ–¥–µ–∑–∏—á–µ—Å–∫–∏—Ö 
# —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –æ—Ç –∞–∫—Ç–æ—Ä–æ–≤, –≤—Ö–æ–¥—è—â–∏—Ö –≤ –æ–±–ª–∞—Å—Ç—å –≤–ª–∏—è–Ω–∏—è i-–≥–æ –∞–∫—Ç–æ—Ä–∞,
# –¥–æ —Å–∞–º–æ–≥–æ i-–≥–æ –∞–∫—Ç–æ—Ä–∞
gd <- gd$gdist[which(rowSums(gd$gdist) != Inf), ]
inp <- nrow(gd)

#–ø–æ—Å—á–∏—Ç–∞–ª–∏ –ø–æ —Ñ–æ—Ä–º—É–ª–∫–µ
prest_prox <- (inp^2)/((network.size(friendship_network)-1)*colSums(gd))

prest <- data.frame(n=network.vertex.names(friendship_network), Pd=prest_deg/(network.size(friendship_network)-1), Pp=prest_prox)
prest
````
