## Основные теоретические положения
*Акторы* — дискретные индивидуальные, корпоративные или коллективные социальные единицы.  
*Триада* — элемент сети, представляющий собой тройку акторов и возможно связь между ними.
*Группа* — совокупность всех акторов, связи между которыми должны быть изучены.
*Подгруппа* — любое подмножество акторов и все связи между ними.  
*Отношение* — совокупность связей определенного вида между членами группы.  
*Социальная сеть* — объект, представляющий собой совокупность конечно набора (наборов/акторов и отношения или отношений), определенного на них.  

*Граф* — модель социальной сети с дихотомическим отношением между каждой парой акторов.
$$G = \mathcal{G}(\mathcal{N}, \mathcal{L})$$ 
где $\mathcal{N}$ — набор узлов, $\mathcal{L}$ — набор линий.  

*Социоматрица* — матрица $X$ размерности $g \times g$, описывающая смежные узлы в сети.  
$X=\{x_{ij_r}\}$, где $i,j=1..g$, $r$—дополнительный индекс отношения, если отношений несколько

Числовые характеристика графа (первые пять - пятичисловая сводка Тьюки):  
1. Размер сети — количество акторов в сети
2. Плотность — доля имеющихся связей по отношению к максимально возможному количеству связей в сети. 
    * Для ориентированных сетей считается как $\Delta = L / A_{g}^{2}$
    * Для неориентированных сетей считается как $\Delta = L / {g \choose 2}$
3. Компонента — подгруппа, в которой все акторы связаны друг с другом прямо или косвенно.  
4. Диаметр полной сети — наибольшая длина любого из кратчайших путей между всеми парами узлов.
5. Коэффициент кластеризации — доля закрытых треугольников по отношению к общему количеству открытых и закрытых треугольников. Закрытый треугольник — триада с 3 связями, открытый треугольник — триада с 2 или 3 связями.
6. Степень узла
    * Для неориентированных сетей — число прямых, инцидентных узлу $n_i$
    * Для ориентированных сетей рассматривают понятия степени входных связей (число прямых, входящих в узел $n_i$) и степени выходных связей (число прямых, исходящих из узла $n_i$).  

Алгоритм Фрюхтермана-Рейнгольда — силовой алгоритм укладки сетей, основан на идее о том, что вершины являются атомными частицами или небесными телами, оказывающими силы притяжения и отталкивания друг на друга. Сила притяжения определяется как $f_a(d)=\frac{d^2}{k}$, а сила отталкивания как  $f_r(d)=\frac{k^2}{d}$, d - расстояние между вершинами, $K = C \cdot \sqrt{area/{\#V}}$, $C$ - const, $area$ - площадь графического окна отрисовки. Уменьшая температуру линейно до 0, можно получить оптимальную конфигурацию.

Алгоритм Камада-Каваи — силовой алгоритм укладки сетей, основан на идее о настройке пружинной системы таким образом, чтобы минимизация энергии системы соответствовала минимизации разницы между геометрическим и графовым расстоянием. Сила пружины определяется через $k_{ij}=K/d_{ij}^2$. Энергия определяется как $\sum_{i=1}^{n-1}\sum_{j=i+1}^{n} \frac{1}{2}k_{ij}(|p_i-p_j|-l_{ij})^2$

Центральность - мера, характеризующая актора с большим количеством связей (прямых или косвенных).
$$C_A(n_i), i=1..g$$

Престиж - мера,  характеризующая актора с большим количеством входных связей
$$P_A(n_i), i=1..g$$

Индекс централизации:
$$C_A = \frac {\sum_{i=1}^{g}[C_A(n*)-C_A(n_i)]} {max \sum_{i=1}^{g}[C_A(n*)-C_A(n_i)]}$$

Центральность по степени — степень каждого актора. $C_D(n_i)=d(n_i)$ 

Нормаированная центральность по степени: $C_D(n_i)'= \frac{d(n_i)}{g-1}$ 

Централизация по степени: $C_D=\frac {\sum_{i=1}^{g}[C_D(n*)-C_D(n_i)]} {(g-1)(g-2)}$ 

Центральность по близости - более важными считаются узлы, наиболее близкие к остальным узлам сети:  
$$C_C(n_i)=(\sum_{j=1}^{g} d(n_i, n_j))^{-1}$$

Нормированная центральность по близости: 
$$C_C(n_i)=(g-1) \cdot C_C(n_i)$$

Центральность по посредничеству:
$$C_B(n_i) = \sum_{j<k} g_{jk}(n_i)/g_{jk}$$

Нормированная центральность по посредничеству:
$$C_B(n_i)'=C_B(n_i)/ {g-1 \choose 2}$$

Вычисление центральности по информации:

1. Создание матрицы A размера $g \times g$ с элементами $\{a_{ij}\}$, где $a_{ii}=1+$ сумма значений всех линий, инцидентных с $n_i$
\begin{equation*}
a_{ij} = 
 \begin{cases}
   1 &\text{$n_i$, $n_j$ не смежны}\\
   {1-x_{ij}} &\text{$n_i$, $n_j$ смежны}
 \end{cases}
\end{equation*}

2. Создание матрицы $C = A^{-1}$ c элементами $\{c_{ij}\}$
3. Центральность $C_I(n_i) = \frac {1} {c_{ii} + (T - 2R)/g}$, где $T=\sum_{i=1}^{g}c_{ii}=tr(C)$

Престиж по степени — входная степень каждого узла. $P_D(n_i)=d_I(n_i)=x_{+i}$

Нормированный престиж по степени $P_D(n_i)'=\frac {x_{+i}} {g-1}$

Область влияния автора $I_i$ — множество акторов, прямо или косвенно связанных с указанными актором.

Престиж по близости — определяет, насколько все акторы близки к заданному по расстоянию до него.

$$P_p(n_i) = \frac {\#I_i / (g-1)} {\sum_{j: n_j \in I_i} d(n_j, n_i) / \#I_i}$$

Алгоритм Фрюхтермана-Рейнгольда представляет вершины как систему атомных частиц или небесных тел, на которые действуют силы притяжения и отталкивания друг на друга. Суть алгоритма заключается в уменьшении температуры системы. Сила притяжения и сила отталкивания зависят от расстояния между вершинами и параметра $K=\sqrt{\frac{area}{\#V}}$. Данные зависимости носят квадратический характер (сила притяжения ~ $d^2$, сила отталкивания ~$K^2$), поэтому вершины располагаются достаточно удалённо друг от друга.

Алгоритм Камада-Каваи основывается на следующей идее: настройка пружинной системы таким образом, чтобы минимизация энергии соответствовала минимизации разницы между геометрическим и графовым расстоянием. При использовании этого алгоритма вершины расположились друг к другу ближе, чем при алгоритме Фрюхтермана-Рейнгольда. При этом вершина Pucci находится заметно удаленнеё других и занимает много места вокруг себя.

```r
# выделим только неизолированные вершины
isolated_vertices <- is.isolate(socialnet, 1:ncol(sociomatrix))
non_isolated_socialnet <- as.network(socialnet[!isolated_vertices, !isolated_vertices], directed=FALSE)

rownames(non_isolated_socialnet) <- rownames(socialnet)[!isolated_vertices, !isolated_vertices]
colnames(non_isolated_socialnet) <- colnames(socialnet)[!isolated_vertices, !isolated_vertices]
cat('Размер сети былы:', network.size(socialnet), '; стал:', network.size(non_isolated_socialnet))

non_isolated_socialnet.graph <- 
    graph_from_adjacency_matrix(sociomatrix[!isolated_vertices, !isolated_vertices], mode="undirected")

non_isolated_sociomatrix <- sociomatrix[!isolated_vertices, !isolated_vertices]

plot(non_isolated_socialnet.graph)
```

```r
g = network.size(non_isolated_socialnet)

centr_by_degrees <- centr_degree(non_isolated_socialnet.graph)$res
centr_by_degrees_norm <- centr_by_degrees / (g - 1)

centr_by_closeness <- centr_clo(non_isolated_socialnet.graph)$res
centr_by_closeness_norm <- centr_by_closeness

centr_by_betw <- centr_betw(non_isolated_socialnet.graph, directed = FALSE)$res
centr_by_betw_norm <- centr_by_betw * 2 / ((g-1) * (g-2))

centr_by_info <- infocent(non_isolated_socialnet, gmode='graph') 
centr_by_info_norm <- centr_by_info / sum(centr_by_info)

centr_table <- data.frame(
    "n_i"=rownames(non_isolated_sociomatrix),
    "by_degrees_norm"=centr_by_degrees_norm,
    "by_closeness_norm"=centr_by_closeness_norm,
    "by_betweenness"=centr_by_betw_norm,
    "by_info_norm"=centr_by_info_norm
)
centr_table
```

По всем мерам центральности самым влиятельным оказался род Medici. То есть, у данного рода наибольшее количество связей (центральность по степени), они наиболее близки к остальным родам (центральность по близости), чаще других связывают другие рода с остальными (центральность по посредничеству), и через них идет большее количество потока "информации" в графе (центральность по информации)

Самым невлиятельным (если рассматривать все централизации) оказался род Pazzi. Они связаны только с Salviati, которые также связаны только с Medici.

## 12.7) Пятичисловая сводка сети

Вычислим плотность сети вручную. Сеть ориентированная, значит используем формулу $\Delta = L/ A_{g}^{2}$, где L - количество связей, а g - количество акторов сети.
В нашей сети L=102, g=21, что можно узнать из сводки в пункте 12.6).  
$\Delta=102/A_{21}^{2}=\frac{102 \cdot 19!}{21!} = \frac{102}{20 \cdot 21} 
= \frac{51}{10*21}=\frac{17}{10 \cdot 7} \approx 0.242857$

По графику определить число компонент очень трудно.

```r
cat('Размер сети:', network.size(socialnet))

cat('\nПлотность сети:', network.density(socialnet))

cat('\nЧисло компонент:', components(socialnet))

cat('\nДиаметр сети:', max(geodist(socialnet, inf.replace = -1)$gdist))

cat('\nКоэффициент кластеризации:', gtrans(socialnet, mode='digraph'))
```

Размер сети - 21, то есть мы рассматриваем 21 сотрудника.

Значения плотности сети, посчитанные вручную и программно, совпали. Плотность сети оказалась равной 0.2428571, то есть доля имеющихся ребёр от максимально возможного количества рёбер равна 0.2428571, что говорит о достаточно немалом количестве связей.

Количество компонент в сети оказалось равно трём, причем одна компонента содержит большинство акторов, вторая компонента состоит из актора V7, третья - из актора V11. То есть почти между всеми сотрудниками есть какой-то путь по друзьям, а V7 и V11 "изолированы".

Диаметр сети оказался равным 5, то есть между одним сотрудником и другим есть связь не более чем через 5 сотрудников.

Коэффициент кластеризации, отражающий долю закрытых треугольников к открытым и закрытым, равен примерно 0.46, что означает умеренную  тенденцию к транзитивности. То есть если сотрудник А дружит с сотрудником B, а  B с  С, то, нестрого говоря, есть 46% вероятность того, что А дружит с С

## 12.8) Степени узлов

```r
in_degrees <- degree(socialnet, gmode='digraph', cmode='indegree')
out_degrees <- degree(socialnet, gmode='digraph', cmode='outdegree')
cat('Степень каждого узла, входящие:', in_degrees)
cat('\nСтепень каждого узла, исходящие:', out_degrees)
cat('\nСреднее входящих:', mean(in_degrees))
cat('\nСреднее исходящих:', mean(out_degrees))

var_in <- sum(sapply(in_degrees, function(x) {(x-mean(in_degrees))^2})) / network.size(socialnet) 
var_out <- sum(sapply(out_degrees, function(x) {(x-mean(out_degrees))^2})) / network.size(socialnet)
cat('\nДисперсия входящих:', var_in)
cat('\nДисперсия исходящих:', var_out)
```

Средние значения входящих и исходящих степеней акторов оказались равны примерно 4.85. 

Дисперсия входящих степеней относительно небольшая, то есть входящие степени многих акторов действительно равны 4 или 5, либо близки к этим значениям.

Дисперсия исходящих степеней получилась гораздо большей, то есть исходящие степени акторов сильно отклоняются от 4 или 5.
