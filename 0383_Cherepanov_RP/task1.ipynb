## Основные теоретические положения
**Акторы** — дискретные социальные единицы.

**Реляционная связь** — свойство, описывающее взаимосвязь между акторами.

**Социальная сеть** — конечное множество наборов акторов и отношения, определенные на них.

**Граф** — модель социальной сети с ненаправленным дихотомическим отношением между парами узлов.

**Социоматрица** — матрица размерности $g \times g$, описывающая смежные узлы в сети: $X = \{x_{i, j; r}\}, i, j = 1, ..., g, r - $ номер отношения.

**Размер сети** — количество акторов в сети.

**Плотность сети** — отношение числа ребер к максимально возможному.

**Диаметр сети** — максимальная длина минимального пути между двумя 
нами графа.

**Коэффициент кластеризации** — отношение количества закрытых 
треугольников к  открытым и закрытым треугольникам.

**Силовые алгоритмы** — итеративные алгоритмы, основанные на уменьшении общей энергии сети, оцениваемой разными способами.

**Центральность** - мера, характеризующая акторы с большим количеством связей (прямых или косвенных).

Центральность по степени:  $C_D^\prime(n_i) =\frac{d(n_i)}{g-1}$.

Центральность по близости: $C_C^\prime(n_i)=\frac{g-1}{\sum_{i=1}^gd(n_i,n_j)}$.

Центральность по посредничеству:  $C_B(n_i)=\sum_{i<k}g_{ik}(n_i)/g_{jk}$.

Центральность по информации: $C_I^{'}=\frac{C_I(n_i)}{\sum_{i}C_I(n_i)}$.

**Точка сочленения** — верщина, удаление которой увеличивает количество компонент связности.

**Мост** — ребро, удаление которого увеличивает количество компонент связности.

**Централизация** - мера, характеризующая равномерную распределенность центральности акторов в сети.

Централизация по степени: $C_D = \frac{\sum_{i=1}^g(C_D(n^{*})-C_D(n_i))}{(g-1)(g-2)}$.

Централизация по близости: $C_C = \frac{\sum_{i=1}^g(C_C^{'}(n^{*})-C_C^{'}(n_i))}{(g-1)(g-2)/(2g-3)}$.

Централизация по посредничеству: $C_B = \frac{\sum_{i=1}^g(C_B^{'}(n^{*})-C_B^{'}(n_i))}{g-1}$.

Централизация по информации: $C_I =\frac{1}{g}\sum_{i=1}^g(C_I^{'}(n^{*})-\bar{C_I}(n_i))^2$.

### Пункт 5
Отобразим сеть с помощью алгоритмов Фрюхтермана-Рейнгольда и Камада-Каваи. Это силовые алгоритмы укладки графов. Общая идея состоит в том, что ребра графа — это "пружины", соединяющие вершины, расположенные случайно. На каждой итерации высчитывается сила, действующая на вершины, и эти вершины перемещаются.

Разница между алгоритмами заключается в том, что итерация алгоритма Фрюхтермана-Рейнгольда подразумевает подсчет сил, действующих на все вершины, а затем — их перемещение вершин. При этом на каждой итерации уменьшается коэффициент, на который умножаются действующие силы — "температура".

В алгоритме Камада-Каваи минимизуется разница между геометрическим и графовым расстояниями между вершинами. При этом на очередной итерации выбирается точка с максимальной энергией и перемещается, пока остальные точки остаются на месте.

```r
plot(f_graph, layout=layout_with_fr(f_graph))
```

```r
plot(f_graph, layout=layout_with_kk(f_graph))
```

### Пункт 7
Вычисляем пятичисловую сводку сети:

1.   Размер сети — количество вершин.
2.   Плотность сети — отношение числа ребер к максимально возможному. Количество ребер в графе — 20, вершин — 16. Максимальное число ребер — $\frac{16*15}{2} = 120$.

$\frac{20}{120} = \frac{1}{6} = 0.1\bar{6}$
3.   Число компонент. На графике видно, что одна вершина не связана с остальными, поэтому компонент связности две.
4.   Диаметр сети — максимальная длина минимального пути между двумя вершинами графа.
5.   Коэффициент кластеризации — отношение количества закрытых треугольников к открытым.




```r
df <- as.data.frame(gorder(f_graph))
df <- cbind(df, network.density(f_net))
df <- cbind(df, count_components(f_graph))
df <- cbind(df, diameter(f_graph))
df <- cbind(df, transitivity(f_graph))
colnames(df) <- c('Size', 'Density', 'Comps_num', 'Diameter', 'Transitivity')
df
```

Плотность графа, посчитанная с помощью функции, совпала с вычисленным вручную значением, как и количество компонент.

Диаметр графа равен 5, то есть длина пути между двумя самыми удаленными ребрами составляет 5 ребер.

Коэффициент кластеризации графа оказался чуть менее 0.2, что значит, что точки имеют достаточно слабую тенденцию к объединению в закрытые треугольники.

### Пункт 8

Степени узлов вычислим с помощью функции "*degree*". Средрее значение и дисперсию этого набора величин — функциями "*mean*" и "*var*" соответственно.


```r
f_degrees <- degree(f_graph)
df <- as.data.frame(mean(f_degrees))
df <- cbind(df, var(f_degrees))
colnames(df) <- c('mean', 'var')
df
```

### Пункт 9
Для подсчета мер центральности уберем из графа вершину, не входяющую в компоненту связности. Сделать это можно с помощью "вычитания" вершин из графа и функции "*network::delete.vertices*" в зависимости от используемой структуры.



```r
f_graph <- f_graph - vertex(which(degree(f_graph) == 0))
network::delete.vertices(f_net, isolates(f_net))
df <- as.data.frame(degree(f_graph, normalized = TRUE))
df <- cbind(df, closeness(f_graph, normalized = TRUE))
df <- cbind(df, betweenness(f_graph, normalized = TRUE))
df <- cbind(df, infocent(f_net) / sum(infocent(f_net)))
colnames(df) <- c('degree', 'closeness', 'betweenness', 'infocent')
df
```

На графе хорошо выделяется вершина "Medici", так как она превосходит все остальные вершины по всем мерам центральности.

### Пункт 11
Мосты и точки сочленения можно найти с помощью функций "*bridges*" и "*articulation.points*" соответственно.

```r
f_bridges <- bridges(f_graph)
f_art_points <- articulation.points(f_graph)
vert_colors <- rep('blue', times = gorder(f_graph))
edge_colors <- rep('black', times = gsize(f_graph))
vert_colors[f_art_points] <- 'red'
edge_colors[f_bridges] <- 'green'

plot(f_graph, edge.color = edge_colors, vertex.color=vert_colors)
```
