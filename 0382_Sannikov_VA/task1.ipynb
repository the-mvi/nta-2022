## Основные теоретические положения
**Сеть** - все, что окружает каждого человека, личные взаимосвязи.

**Акторы** - дискретные индивидуальные, корпоративные или коллективные социальные единицы.

**Реляционная связь** - свойство, описывающее взаимодействие между двумя акторами.


**Группа** - совокупность всех акторов, связи между которыми должны быть изучены.

**Отношение** - совокупность связей определенного вида между членами группы.


"**Социальная сеть** - объект, представляющий собой совокупность конечного набора (или наборов) акторов и отношения (или отношений), определенных на нем.

**Числовые характеристики сети:**

*   Размер сети - количество акторов в сети;
*   Плотность - доля имеющихся связей к общему возможному числу связей;
*   Компонента - подгруппа, в которой все акторы связаны друг с другом прямо или косвенно;
*   Диаметр сети - наибольшая длина любого из кратчайших путей между парами узлов;
*   Степень узла - число связей инцидентных узлу $n_i$;
*   Коэффициент кластеризации - характеристика тенденции к созданию закрытых треугольников (триад, т. е. таких структур, где между акторами есть все три связи). Этот коэффициент измеряется с помощью транзитивности.

**Силовые алгоритмы укладки сетей** - итеративные алгоритмы, основанные на уменьшении общей энергии сети, где энергия может быть определена различными способами:

* Алгоритм Фрюхтермана-Рейндгольда.
Вершины - атомные частицы или небесные тела, которые отталкиваются или притягиваются (силы). Действуют силы притяжения и отталкиваниния: $F_a(d) = \frac {d^2} {k}$ ; $F_r(d) = \frac {-k^2} {d}$ , $d$ - расстояние между вершинами (евклидово); $k = C \sqrt {\frac {area} {\#V}}$ , $area$ - площадь окна, где укладывается граф, $C$ - константа, определяется опытным путем, $\#V$ - количество вершин в сети.
* Алгоритм Камада-Каваи.
Идея: узлы соединены пружинами, нужно настроить пружинную сеть таким образом, чтобы минимальная энергия системы соответствовала минимальной разнице между геометрическим расстоянием и графовым расстоянием между узлами (т. е. найти минимум энергии системы). Сила пружины рассчитывается как: $k_{ij} = \frac {const} {d_{ij}^2}$ , где $d_{ij}$ - кратчайшее расстояние между вершинами $i$ и $j$.

**Центральность** - мера, характеризующая актор с наибольшим количеством связей (прямых или косвеннных). Обозн.: $C_A(n_i)$ , $i=1,..,g$.

**Централизация** - мера, характеризующая группу акторов. Обозн.: $C_A$ . $C_A = \frac {\sum_{i=1}^g [C_A(n^*)-C_A(n_i)]} {max \sum_{i=1}^g [C_A(n^*)-C_A(n_i)]}$ , $C_A(n^*) = max_i C_A(n_i)$ - максимальное значение центральности.

* Центральность по степени - степень каждого актора: $C_D(n_i) = deg(n_i)$.
Нормированная центральность по степени: $C_D'(n_i) = \frac {deg(n_i)} {g-1}$<br>
Централизация по степени: $C_D = \frac {\sum_{i=1}^g [C_D(n^*)-C_D(n_i)]} {max \\sum_{i=1}^g [C_D(n^*)-C_D(n_i)]} = \frac {\sum_{i=1}^g [C_D(n^*)-C_D(n_i)]} {(g-1)(g-2)}$
* Центральность по близости - более важными считаются узлы наиболее близкие к остальным узлам сети: $C_C(n_i) = (\sum_{i=1}^g d(n_i, n_j))^-1$ , $d(n_i, n_j)$ - расстояние между двумя узлами.
Нормированная центральность по близости: $C_c'(n_i) = \frac {g-1} {\sum_{i=1}^g d(n_i, n_j)} = (g-1)C_C(n_i)$
Централизация по близости: $C_C = \frac {\sum_{i=1}^g [C_C(n^*)-C_C(n_i)]} {(g-1)(g-2)/(2g-3)}$.
*Центральность по посредничеству - характеризует насколько важную роль данный узел играет на пути \"между\" парами узлов в сети, в том смысле, что пути между другими узлыми должны проходить через данный узел $n_i$ : $C_B(n_i) = \frac {\sum_{j<k} g_{jk}(n_i)} {g_{jk}} , n_i \ne j,k$ . $g_{ij}(n_i)$ - количество геодезичских расстояний (кратчайшие пути) между узлами $j$ и $k$, который включают узел $i$, $g_{jk}$ - геодезическое расстояние между $j$ и $k$.<br>
Нормированная центральность по посредничеству: $C_B'(n_i) = \frac {C_B(n_i)} {C_2^{g-1}}$ .<br>
Централизация по посредничеству: $C_B = \frac {2\sum_{i=1}^g [C_B(n^*)-C_B(n_i)]} {(g-1)^2(g-2)} = \frac {\sum_{i=1}^g [C_B(n^*)-C_B(n_i)]} {g-1}$ .
* Центральность по информации - обобщает все меры центральности, учитываются все комбинированные пути от одного актира к другому: $C_I(n_i)=\frac {1} {c_{ii}+(T-2R)/g}$ , $T=\sum_{i=1}^g c_{ii} = tr C$ , $R = \sum_{j=1}^g c_{ij}$ , при этом создается матрица $A$ размера $g\times g$ с элементами $\{a_{ij}\}$ : $a_{ij} = 1\ +$ сумма значений всех линий, инцидентных с $n_i$ . $a_{ij} = \left\{ \begin{array}{cl}1 & \text {если узлы $n_i$ и $n_j$ не инцидентны}\\ 1-x_{ij} & \text {если узлы $n_i$ и $n_j$ инцидентны} \end{array} \right.$ ,<br> $C=A^{-1}$ - матрица с элементами $\{c_{ij}\}$ . <br>
Нормированная централизация по информации: $C_I'=\frac {C_I(n_i)} {\sum_i C_I(n_i)}$ .<br>
Централизация по информации (можно вычислить как дисперсию индексов информации актора): $S_I^2 = \frac {1} {g} \sum_{i=1}^g (C_I'(n_i) - \overline {C_I})^2$ , $\overline {C_I} = \frac {1} {g} \sum_i C_I(n_i)$ . <br>
"<br>

**Престиж** - мера, характеризующая актор с наибольшим числом входящих связей. Обозн.: $P_A(n_i)\ , i=1,..,g$ .<br>
* Престиж по степени - входящая степень каждого узла: $P_D(n_i) = d_I(n_i) = x_{+i}\ , d_I$ - входящие (Input). <br>
Нормированный престиж по степени: $P_D' = \frac {X_{+i}} {g-1}$ .<br>
<br> **Область видимости актора $I_i$** - множество акторов прямо или косвенно связанных с указанным актором (связи входящие в $i$-ый актор).<br>
* Престиж по близости - определяет насколько все акторы близки к заданному по расстоянию до него: $P_P(n_i) = \frac {\#I_i/(g-1)} {\sum_{j: n_j \in I_i} d(n_j, n_i) / \#I_i}~$,$~d(n_j, n_i)$ - растояние пути между двумя узлами.<br>

**Пункт 7**

```r
network_size <- gorder(to_graph)
cat("Размер сети:", network_size)
```

Самостоятельный рассчет:

```r
density <- 2 * gsize(to_graph) / ((network_size) * (network_size - 1))
cat("Плотность по самостоятельному рассчету:", density)
```

Рассчет с помощью библиотечной функции:

```r
cat("Плотность библиотечной функции:", edge_density(to_graph))
```

На графике 6 компонент связности: 5 обособленных по одному элементу и 1 большая с несколькими элементами

```r
components_number <- count_components(to_graph)
cat("Количетсво компонент связности:", components_number)
```

Диаметр вычислен с помощью функции diameter (max геодезическое расстояние между двумя узлами):

```r
D <- diameter(to_graph, directed=FALSE)
cat("Диаметр сети:", D)
```

Коэффициент кластеризации был вычислен с помощью функции transitivity (функция ищет транзитивность, по которой определяется коэффициент кластеризации):

```r
claster_coef <- transitivity(to_graph)
cat("Коэффициент кластеризации:", claster_coef)
```

**Пункт 8**

Средняя степень узлов была вычислена по формуле $\overline{d}=\frac{1}{g}*\sum_{i=1}^{g}d(n_{i})$, g - число акторов в сети:

```r
degrees <- igraph::degree(to_graph)
avg_degree <- mean(degrees)
cat("Средняя степень узлов", avg_degree)
```

Как видно по формуле, средняя степень узлов была вычислена верно

Дисперсия была вычислена по формуле $S^2_{D}=\frac{1}{g}*\sum_{i=1}^{g}d(n_{i} - \overline{d})^2$:

```r
dispersion <- mean((degrees - avg_degree) ** 2)
cat("Дисперсия:", dispersion)
```

Как видно по формуле, дисперсия была вычислена верно

**Пункт 9**

```r
membership <- igraph::components(to_graph, mode="strong")$membership
counts <- igraph::components(to_graph, mode="strong")$csize
isolated <- membership[counts[membership] == 1]
without_isolated <- delete_vertices(to_graph, names(isolated))
vertexes <- gorder(without_isolated)

# центральность по степени
degree_centrality <- igraph::degree(without_isolated) / (vertexes - 1)

# центральность по близости
closeness_centrality <- igraph::closeness(without_isolated, normalized = TRUE)

#центральность по посредничеству
betweenness_centrality <- igraph::betweenness(without_isolated, directed=FALSE, normalized=TRUE)

# центральность по информации
information_centrality <- infocent(as.matrix(as_adjacency_matrix(without_isolated)), rescale=TRUE)

res <- data.frame(degree_centrality=degree_centrality,
                        betweenness_centrality=closeness_centrality,
                        closeness_centrality=betweenness_centrality,
                        inf_centrality=information_centrality)
res
```

**Пункт 10**

```r
centralizations <- data.frame(measures = c("degree", "closeness", "betweenness", "information"),
                              meanings = c((sum(max(degree_centrality) - degree_centrality)) / (vertexes - 2),
                              (sum(max(closeness_centrality) - closeness_centrality)) / ((vertexes - 2) * (vertexes - 1) / (2 * vertexes - 3)),
                              (sum(max(betweenness_centrality) - betweenness_centrality)) / (vertexes - 1),
                              mean(information_centrality)))
centralizations
```

Из данной таблицы видно, что сеть больше централизована по близости, меньше по информации

```r
degree_prestige <- igraph::degree(to_graph2, mode="in") / (network_size2 - 1)

dists <- distances(to_graph2, mode="out")
count_proximity_prestige <- function(x){
  x <- x[is.finite(x)]
  I <- length(x)
  sum_dists <- sum(x)
  return(I ** 2 / (sum_dists * (network_size2 - 1)))
}

res2 = data.frame(Pd=degree_prestige, Pp=apply(dists, FUN=count_proximity_prestige, MARGIN=2))
res2
```
