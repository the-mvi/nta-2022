## Основные теоретические положения
Актор - это дискретные индивидуальные, корпоративные или коллективные социальные единицы.  
Отношение - совокупность связей определенного вида между членами группы.  
Социальная сеть - объект, представляющий собой совокупность конечного набора (или наборов) акторов и их отношения (или отношений), определенных на них.  
Граф - это модель социальной сети с дихотомическим отношением, т.е. связь либо присутсвует, либо отсутствует между каждой парой акторов.  
Социоматрица - матрица X размерность g x g описывающая смежные узлы.  
Сеть - объект, представляющий собой совокупность конечного набора акторов и отношения определенное на них.  
Актор - дискретное индивидуальное корпоративное социальная единица.  
Размер сети - количество акторов в сети.  
Плотность сети - доля имеющихся связей по отношению к максимально возможному количеству связей в сети с таким же количеством акторов.  
Плотность для ориентированной сети: $$\Delta = \frac{L}{A_g^2}$$  
Плотность для неориентированной сети: $$\Delta = \frac{L}{C_g^2}$$  
Компонента - подгруппа в которой все акторы связаны друг с другом прямо или косвенно.  
Диаметр сети - наибольшая длина любого из кратчайших путей между всеми парами узлов.  
Коэффициент кластеризации - доля закрытых треугольников (триад, где наблюдаются все связи) по отношению к общему количеству треугольников.  
Степень узла - число ребер, инцидентных узлу $n_i$  
Средняя степень: $$\bar{d} = \frac{1}{g}\sum_{i=1}^gd(n_i)$$
Дисперсия степеней узла: $$S_d^2 = \frac{1}{g}\sum_{i=1}^g(d(n_i)-\bar{d})^2$$
Центральность - мера, характеризующая актора с большим количеством связей(прямых или косвенных).  
Центральность по степени: $$C_D(n_i) = d(n_i)$$  
Нормированная центральность по степени: $$\frac{d(n_i)}{g-1}$$  
Центральность по близости - более важным считается узел, наиболее близкий к остальным.
$$C_C(n_i)=(\sum_{i=1}^gd(n_i,n_j))^{-1}$$
Нормированная центральность по близости: $$C_C^{'}(n_i)=(g-1)C_C(n_i)$$
Центральность по посредничеству показывает, насколько важную роль узел играет на пути между другими узлами.
$$C_B(n_i)=\sum_{i<k}g_{ik}(n_i)/g_{jk}$$, где $g_{jk}$ расстояние между узлами j и k, а
$g_{jk}(n_i)$ - количество кратчайших путей между узлами j и k через i.  
Центральность по информации показывает, сколько информации содержится в путях, которые начинаются и заканчиваются у определенного актора.
$$ A = \{a_{ij}\}$$ - матрица размера g x g, где $$ a_{ii} = 1 + \sum_{j=1}^{g}{x_{ij}}$$
$$ a_{ij} = 1 \text{ , if } n_{i} \text{ and } n_{j} \text{ are not connected}$$
$$ a_{ij} = 1-x_{ij} \text{ , if } n_{i} \text{ and } n_{j} \text{ are connected}$$
$$ x_{ij} \text{- weight of edge between }n_{i} \text{ and } n_{j}$$
$$ C = A^{-1}$$
$$C_I(n_{i}) =  \frac{1}{c_{ii}+(T-2R)/g}$$, где $$T = trace(C)$$ $$R = \sum_{j=1}^{g}c_{ij}$$
$$C_I^{'}=\frac{C_I(n_i)}{\sum_{i}C_I(n_i)}$$
Централизация показывает, насколько равномерно распределены центральности акторов в сети.  
Централизация по степени:
$$C_D = \frac{\sum_{i=1}^g(C_D(n^{*})-C_D(n_i))}{(g-1)(g-2)}$$
Централизация по близости:
$$C_C = \frac{\sum_{i=1}^g(C_C^{'}(n^{*})-C_C^{'}(n_i))}{(g-1)(g-2)/(2g-3)}$$
Централизация по посредничеству:
$$C_B = \frac{\sum_{i=1}^g(C_B^{'}(n^{*})-C_B^{'}(n_i))}{g-1}$$  
Точка сочленения - узел, при удалении которого увеличивается количество компонент связности.

Мост - ребро, при удалении которого одна компонента связности разделяется на 2.

Престиж - мера, характеризующая фктора с большим количеством входящих связей.

Престиж по степени - входящая степень каждого узла. $$P_D(n_i) = d_I(n_i)=x_{+i}$$

Нормированный престиж по степени: $$P_D^{'}(n_i) =\frac{x_{+i}}{g-1}$$

Престиж по близости - мера, определяющая, насколько все акторы близки к заданному по расстоянию до него. $$ P_P(n_i) = \frac{\#I_i/(g-1)}{\sum_{j:n_j \in I_i}(d(n_j,n_i)/\#I_i)}$$

12. Выполнить пп. 2–8 для направленной сети из файла krackhardt_friendship.csv.

```r
data <- read.csv("krackhardt_friendship.csv", skip=2, header=FALSE, sep=",")
sociomatrix <- data.matrix(data)
sociomatrix

G <- graph.adjacency(sociomatrix, mode="directed")
edgeList <- get.edgelist(G)
head(edgeList, 5)

actorNames <- 1:dim(sociomatrix)[1]
rownames(sociomatrix) <- colnames(sociomatrix) <- actorNames
sociomatrix


net <- as.network(sociomatrix, directed=TRUE)
network.vertex.names(net) <- actorNames

plot(G, layout=layout.fruchterman.reingold(G), main="Алгоритм Фрюхтермана Рейнгольда")
plot(G, layout=layout_with_kk(G), main="Алгоритм Камада-Каваи")

summary(net, print.adj = FALSE)

cat("net size: ", network.size(net), "\n")
cat("net density: ", gden(net, mode="digraph"), "\n")
g <- gorder(G)
cat("manually calculated net density: ", network.edgecount(net) / (factorial(g)/factorial(g-2)), "\n")
cat("net components: ", components(net), "\n")
geoDists <- geodist(component.largest(net, result = "graph"))
cat("net diameter: ", max(geoDists$gdist), "\n")
cat("net transitivity: ", gtrans(net, mode = "digraph"), "\n")

degree_var <- function(net, meanDegree, mode){
  return (1 / network.size(net)*sum((degree(net, cmode=mode) - meanDegree)^2))
}
cat("mean in-degree of nodes: ", mean(degree(net, cmode="indegree")), "\n")
cat("mean in-degree variance of nodes: ", degree_var(net, mean(degree(net, cmode="indegree")), "indegree"), "\n")
cat("mean out-degree of nodes: ", mean(degree(net, cmode="outdegree")), "\n")
cat("mean out-degree variance of nodes: ", degree_var(net, mean(degree(net, cmode="indegree")), "outdegree"), "\n")
```

13. Заполнить таблицу для различных мер престижа.

```r
I <- shortest.paths(G, to=V(G), mode="out")
I[!is.finite(I)] <- 0

ICount <- rep(0, 21)

for (i in 1:nrow(I))
  for (j in 1:ncol(I))
    if (I[i, j] != 0)
      ICount[i] <- ICount[i] + 1

prestige <- data.frame(
  P_Dn = prestige(net, cmode="indegree") / (g-1), # нормированный престиж по степени
  P_Pn = (ICount / (g - 1)) / (rowSums(I) / ICount) # нормированный престиж по близости
)

prestige
```
