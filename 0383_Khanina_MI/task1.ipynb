## Основные теоретические положения
<br>**Акторы**  -  дискретные индивидуальные, корпоративные или коллективные социальные единицы.
<br>**Реляционная связь**  - свойство, описывающее взаимодействие между двумя акторами.
<br>**Подгруппа** - любое подмножество акторов и все связи между ними.
<br>**Группа** - совокупность всех акторов, связи между которыми должны быть изучены.
<br>**Отношение** - совокупность связей определённого вида между членами группы.
<br>**Социальная сеть** - объект, представляющий собой совокупность конечного набора акторов и отношений, определённых на них.

### Графовое:
Граф - модель социальной сети с ненаправлеными дихроматическими отношениями, т.е. связь либо присутствует, либо отсутствует между кадой парой акторов:

$\mathcal{G} = \mathcal{G}(\mathcal{N}, \mathcal{L})$
<br>
$\mathcal{N} = \{n_1, n_2, ..., n_g\}$ - набор узлов
<br>
$\mathcal{L} = \{l_1, l_2, ..., l_l\}$ - набор линий.
<br><br>
$l = <n_i, n_j>$ - направленная (ориентированная) связь.
<br>
$l = (n_i, n_j)$ - ненаправленная(неориентированная) связь.
<br><br>
**Подграф:** <br>
Граф $\mathcal{G_s} = \mathcal{G_s}(\mathcal{N_s}, \mathcal{L_s})$ **называется подграфом графа** $\mathcal{G}(\mathcal{N}, \mathcal{L})$, **если** $\mathcal{N_s} ⊆ \mathcal{N}$ и $\mathcal{L_s} ⊆ \mathcal{L}$
<br><br>
<br>**Маршрут** - последовательность узлов и линий, начинающихся и заканчивающихся узлами, в которой каждый узел ициндентен линиям, следующим и предшевствующим ему в последовательности.
<br>
<br>**Тропа** - это маршрут, в котором все линии различны, однако некоторые узлы могут быть включены более одного раза.
<br>**Путь** - это маршрут, все узлы и линии которого, различны.

<br><br>
### Социометрическое:
**Социоматрица** - матрица $\mathcal{X}$ размерности $\mathcal{g} \times  \mathcal{g}$, описывающая смежные узлы в сети:
<br>$\mathcal{X} = \{\mathcal{X_{ij}, \mathcal{r}}\}$, где $\mathcal{i}, \mathcal{j} = 1, ..., \mathcal{g}$
<br><br>**Матрица инцидентности** - матрица $\mathcal{I}$ размерности $\mathcal{g} \times \mathcal{L}$, описывающая какие линии каким узлам инцидентны.
<br><br>
**Матрица достижимости**: $\mathcal{X}^{\mathcal{[R]}} = \{\mathcal{X_{ij}}^{\mathcal{[R]}}\}$,
<br>где $\mathcal{X_{ij}}^{\mathcal{[R]}} = 1_{\mathcal{X_{ij}}^{\mathcal{[Σ]}} \geq 1}$
<br><br>
### Числовые характеристики сети:

<br>Размер
<br>Плотность
<br>Компонента
<br>Диаметр
<br>Коэффициент кластеризации

<br><br>
**Размер** - количество акторов сети.
<br><br>
**Плотность** - доля имеющихся связей по отношению к максимально возможному количеству связей:
<br>
Для ор. сетей:<br><br>
$Δ = \frac{L}{A_{g}^{2}}$, <br>где $L$ - количество связей.
<br><br>
Дла неор. сетей:<br><br>
$Δ = \frac{L}{\begin{pmatrix}
g\\
2
\end{pmatrix}}$ 
<br><br>
**Компонента** - подгруппа, в которой все акторы связаны друг с другом прямо или косвенно.
<br><br>
**Диаметр полной сети** - наибольшая длина любого из кратчайших путей между всеми парами узлов.
<br><br>
**Степень узла** - число прямых, инцидентных узлу n (неор. граф).
<br><br>
Средняя степень узла:<br>
$\mathcal{\overline{d}} = \frac{1}{g} ⋅ \sum_{i=1}^{g}{d(n_i)}$
<br><br>Дисперсия степеней узлов:<br>
$S_D^2 = \frac{1}{g} \cdot \sum_{i=1}^{g} {(d(n_i) - \overline{d})^2}$
<br><br>
Если d узла = 0, то узел называется "изолированным".
<br><br>
**Коэффициент кластеризации** $∈ [0,1]$ - доля закрытых треугольников (триад, где наблюдаются все свзяи) по отношению к общему количеству открытых и закрытых треугольников.
<br><br>

### Алгоритмы укладки сетей. Силовые алгоритмы:

## Меры важности


**Центральность** - (centrality) мера, характеризующая актор с большим количеством связей (прямых или косвенных). [Для ненаправленных сетей]
<br>
Обозначение: $C_A(n_i)$, где $i = 1, ..., g$
<br><br>
**Престиж** - (presiege) мера, характеризующая актор с большим количеством входящих связей. [Для направленных сетей]
<br>
Обозначение $P_A(n_i)$, где $i = 1, ..., g$
<br><br><br>
**Централизация группы** 
<br><br>
**Индекс центральности** (centralization index):
<br>
$C_A = \frac {\sum_{i=1}^{g}{[C_A(n^*) - C_A(n_i)}]} {max \sum_{i=1}^{g}{[C_A(n^*) - C_A(n_i)}]}$
<br><br>

<br>**Центральность по степени** (degree centrality) - степерь каждого актора
<br>
$C_D(n_i) = d(n_i)$
<br>
<br>
<br>**Нормированная центральность по степени:**
<br>
$C_D^{'}(n_i) = \frac{d(n_i)}{g-1}$
<br><br>
<br>**Централизация по степени:**
<br>
$C_D = \frac {\sum_{i=1}^{g}{[C_D(n^*) - C_D(n_i)}]} {max \sum_{i=1}^{g}{[C_D(n^*) - C_D(n_i)}]} = \frac{\sum_{i=1}^{g}{[(C_D(n^*) - C_D(n_i)]}}{(g-1)(g-2)}$
<br><br>
<br>**Централизация по степеням:**
<br>
$S_D^2 = \frac{1}{g} \cdot \sum_{i=1}^{g}{(C_D(n_i) - \overline{C}_D^2}$
<br><br><br>
<br>**Центральность по близости** (closenss centrality) - более важными узлами сети считаются узлы наиболее близкие к остальным узлам сети:
<br>
$C_C(n_i) = (\sum_{i=1}^{g}{d(n_i,n_j)})^{-1}$ ,где $d(n_i, n_j)$ - расстояние между двумя узлами<br><br>
<br>**Нормированная центральность по близости:**<br>
$C_C^{'}(n_i) = \frac{g-1}{\sum_{i=1}^{g}{d(n_i, d_j)}} = (g-1)⋅C_C(n_i)$<br><br>
<br>**Централизация по близости:**<br>
$C_C = \frac{\sum_{i=1}^{g}{[(C_C(n^*) - C_C^{'}(n_i)]}}{(g-1)(g-2)/(2g-3)}$<br><br>
<br> Можно использовать дисперсию:<br>
$S_D^2 = \frac{1}{g}⋅\sum_{i=1}^{g}{(C_C^{'}(n_i) - \overline{C}_C)}$<br><br>
<br>**Центральность по посредничеству** (betweeness centrality) - характеризует насколько важную роль данный узел играет на пути "между" парами других узлов сети, в том смысле, что пути между другими узлами должны проходить через данный узел:<br><br>
$C_B(n_i) = \sum_{j<k}{g_{j,k}(n_i)/g_{jk}}$, где $g_{j,k}(n_i)$ - количество геодезических расстояний между узлами j и k, которые включают узел i<br>
$g_{jk}$ - геодезическое расстояние между узлами j и k <br><br>
<br>**Нормированная центральность по посредничеству:**<br>
$C_B^{'}(n_i) = \frac{C_B(n_i)}{\begin{pmatrix}
g-1\\
2
\end{pmatrix}}$<br><br>
<br>**Централизация по посредничеству:**<br>
$C_B = \frac{2⋅\sum_{i=1}^{g}{[C_B(n^*) - C_B(n_i)]}}{(g-1)^2(g-1)} = 
\frac{\sum_{i=1}^{g}{[C_B^{'}(n^*) - C_B(n_i)]}}{g-1}$<br><br>
<br>**Центральность по информации** - показывает сколько информации содержится в путях, которые начинаются (и заканчиваются) у определённого актора:<br>
$C_I(n_i) = \frac{1}{C_{ii} + (T - 2R)/g}$, где $T = \sum_{i=1}^{g}{C_{ii}} = trC_i$, $R = \sum_{i=1}^{g}{C_{ij}}$<br><br>
<br>**Нормированая центральность по информации:**<br>
$C_{I}^{'}(n_i) = \frac{C_I(n_i)}{\sum_{j}{C_I(n_j)}}$
Централизация по информации:<br>
$S_i^2 = \frac{1}{g} ⋅ \sum_{i=1}^{g}{(C_I(n_i) - \overline{C}_I)^2}$, где $\overline{C}_I = \sum_{i}{C_i(n_i)}$

<br><br><br>
### Престиж
<br>
**Престиж** (presiege) - мера, характеризующая актор с большим количеством входящих связей.
<br><br>


Престиж по степени (degree presitege) - входящая степень каждого узла.<br>
$P_D(n_i) = d_i(n_i) = x_{+i}$<br><br>
Нормированный престиж по степени:<br>
$P_D^{'}(n_i) = \frac{x_{+i}}{g-1}$<br><br><br>


Престиж по близости:<br><br>
Область влияния актора (influence domain of actor) - $I_i$ - множество акторов, прямо и косвенно связанных с указаным актором.<br><br>
Престиж по близости (proximity priestige) - определяет на сколько все акторы близки к заданному по расстоянию до него<br><br>
Престиж по близости определяется как:<br>
$P_P(n_i) = \frac{\#I_i / (g-1)} {\sum_{j⋅n ∈ I_i}{d(n_i, n_j)/\#I_i}}$, где $d(n_j, n_i)$ - расстояние пути между двумя узлами<br><br><br>

**Престиж по близости для групп:**<br><br>
Используется дисперсия:<br>
$S_P^2 = \frac{1}{g} ⋅ \sum{(P_P(n_i) - \overline{P}_P)^2}$, где $\overline{P}_P = \frac{1}{g} ⋅ \sum_{i=1}^{g}{P_P(n_i)}$

<br><br>

<br>

**Точки сочленения (cutpoint)** - узел, при удалении которого увеличивается количество компонент связности.

**Мост (bridge)** - ребро, при удалении которого одна компонента связности разделяется на 2.

**Подгруппа (subgroup)** - подмножество узлов: $\mathcal{N_S} ⊆ \mathcal{N}$, графа $\mathcal{G}(\mathcal{N}, \mathcal{L})$

**Сплочённые подгруппы (social subgroup)** - множество акторов, которые соединены между собой посредствам многочисленных, сильных и прямых связей.

**Клика (clique)** - максимальный полный подграф из 3х и более узлов

**$n$-клика ($n$-clique)** - максимальный подграф, в котором наибольшее геодезическое расстояние между любыми двумя узлами не превышает $n$.

**$n$-кланы ($n$-clans)** - n-клики, в которых геодезические расстояния $d(i,j)$ между всеми узлами в подграфе не превышает $n$ для путей внутри подграфа.

**$n$-клубы ($n$-clubs)** - максимальный подграф диаметра $\leq n$.

**$k$-ядро ($k$-core)** - максимальный подграф, в котором каждая вершина связана с минимум $k$ другими вершинами этого же подграфа.<br><br>
Свойства $k$-ядер:
  <br> Вложенные
  <br> Не перекрываются
  <br> Лекго определить
