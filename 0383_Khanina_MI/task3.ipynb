## –¶–µ–ª—å —Ä–∞–±–æ—Ç—ã
–û–≤–ª–∞–¥–µ—Ç—å –Ω–∞–≤—ã–∫–∞–º–∏ —Ä–∞–±–æ—Ç—ã —Å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º–∏ –ø–∞–∫–µ—Ç–∞–º–∏ –¥–ª—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ —Å–æ—Ü–∏–∞–ª—å–Ω—ã—Ö —Å–µ—Ç–µ–π.

## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è


**–î–∏–∞–¥–∞** - —ç–ª–µ–º–µ–Ω—Ç —Å–µ—Ç–∏, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–±—â–∏–π —Å–æ–±–æ–π –ø–∞—Ä—É –∞–∫—Ç–æ—Ä–æ–≤ –∏ (–≤–æ–∑–º–æ–∂–Ω—É—é) —Å–≤—è–∑—å –º–µ–∂–¥—É –Ω–∏–º–∏;<br>
 $D_{ij} = (X_{ij}, X_{ji}, i \neq j)$
<br><br>

**–¶–µ–Ω–∑—É—Å –¥–∏–∞–¥** - —Ç—Ä–æ–π–∫–∞ —á–∏—Å–µ–ª $(M,A,N)$, –≥–¥–µ:
1. $M = \sum_{i<j}{X_{ij}X_{ji}}$
2. $A = X_{++} - 2M$, –≥–¥–µ $X_{++}$ - —Å—É–º–º–∞ –≤—Å–µ—Ö —ç–ª-—Ç–æ–≤ –≤ —Å–æ—Ü–∏–æ–º–∞—Ç—Ä–∏—Ü–µ
3. $N = C_g^2 - A - M$

**–í–∏–¥—ã –¥–∏–∞–¥**:
<br>
1. –í–∑–∞–∏–º–Ω–∞—è –¥–∏–∞–¥–∞: $D_{ij} = (1,1)$
2. –ê—Å—Å–∏–º–µ—Ç—Ä–∏—á–Ω–∞—è –¥–∏–∞–¥–∞: $D_{ij} = (1,0), D_{ji} = (0,1)$
3. –ù—É–ª–µ–≤–∞—è –¥–∏–∞–¥–∞: $D_{ij} = (0,0)$
<br><br>

**–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ö–∞—Ü–∞ - –ü–∞—É—ç–ª–ª–∞** - —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞ –∏–∑–º–µ—Ä–µ–Ω–∏—è —Ç–µ–Ω–¥–µ–Ω—Ü–∏–∏ –∞–∫—Ç–æ—Ä–æ–≤ –≤ –≥—Ä—É–ø–ø–µ<br><br>
$- \inf < \rho_{KP} \leq 1$<br><br>
–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ –æ–∂–∏–¥–∞–Ω–∏–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥ –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞—è, —á—Ç–æ –≤—ã–±–æ—Ä –¥–µ–ª–∞–µ—Ç—Å—è –∞–∫—Ç–æ—Ä–∞–º–∏ —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º.
<br><br>
–§–∏—Å–∫–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≤—ã–±–æ—Ä:<br>
1. –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è, —á—Ç–æ –∫–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä –¥–µ–ª–∞–µ—Ç $d$ –≤—ã–±–æ—Ä–æ–≤ –∏–∑ $(g-1)$ –∞–∫—Ç–æ—Ä–æ–≤, –¥–æ—Å—Ç—É–ø–Ω—ã—Ö –¥–ª—è –≤—ã–±–æ—Ä–∞
2. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤–∑–∞–∏–º–Ω–æ–≥–æ –≤—ã–±–æ—Ä–∞ –º–µ–∂–¥—É –ª—é–±–æ–π –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–π –ø–∞—Ä–æ–π –∞–∫—Ç–æ—Ä–æ–≤ —Ä–∞–≤–Ω–∞ $d^2/(g-1)^2$
3. –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥:<br><br>
$E(M) = \frac{g(g-1)}{2}\cdot\frac{d^2}{(g-1)^2} = \frac{gd^2}{2(g-1)}$<br><br>
4. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –æ–±—â–∞—è –¥–∏–∞–¥–∞, –≤ –∫–æ—Ç–æ—Ä–æ–π —É—á–∞—Å—Ç–≤—É—é—Ç –∞–∫—Ç–æ—Ä—ã, —è–≤–ª—è–µ—Ç—Å—è –≤–∑–∞–∏–º–Ω–æ–π –¥–∏–∞–¥–æ–π:<br><br>
$P(i \rightarrow j \wedge j \rightarrow i) = P(j \rightarrow i | i \rightarrow j)$<br><br>
5. –£—Å–ª–æ–≤–Ω–≤–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å:<br><br>
$P(j \rightarrow i | i \rightarrow j) = P(j \rightarrow i) + \rho_{KP} P(j \nrightarrow i)$ <br><br>
6. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º:<br><br>
$P(i \rightarrow j \wedge j \rightarrow i) = \frac{g}{g-1}(\frac{d}{d-1} + \rho_{KP}\frac{g-1-d}{g-1})$<br><br>
7. –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥ —Å —É—á—ë—Ç–æ–º –¥–∞–Ω–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏:<br><br>
$ E(M | \rho) = \frac{gd^2}{2(g-1)}(1 - \rho_{KP} + \frac{gd}{2}\rho_{KP})$ <br><br>
8. –û—Ü–µ–Ω–∫–∞ $\rho_{KP}$:<br><br>
$\hat{\rho}_{KP} = \frac{2(g-1)M-gd^2}{gd(g-1-d)}$
<br><br>
**–°–≤–æ–±–æ–¥–Ω—ã–π –≤—ã–±–æ—Ä:**<br><br>
1. –ü—É—Å—Ç—å i-—ã–π –∞–∫—Ç–æ—Ä –≤—ã–±–∏—Ä–∞–µ—Ç $d_O(n_i) = x_{i+}$
2. –ü—É—Å—Ç—å $L = \sum_i{x_{i+}}, L_2 = \sum_i{x_{i+}^2}$ ($L_2$ = —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ —Å—É–º–º—ã —ç–ª-—Ç–æ–≤ —Å—Ç—Ä–æ–∫ —Å–æ—Ü–∏–æ–º–∞—Ç—Ä–∏—Ü—ã)
3. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –∞–∫—Ç–æ—Ä—ã –∏–º–µ—é—Ç –≤–∑–∞–∏–º–Ω—ã–µ –æ–Ω—Ç–Ω–æ—à—â–µ–Ω–∏–µ —Ä–∞–≤–Ω–∞ $x_{i+}x_{j+}/(g-1)^2$
4. –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥:<br><br>
$E(M) = \frac{L^2 - L_2}{2(g-1)^2}$<br><br>
5. –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥ —Å —É—á—ë—Ç–æ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏:<br><br>
$E(M | \rho{'}) = \frac{L^2 - L_2}{2(g-1)}(1 - \rho_{KP}^{'} + \frac{L}{2}\rho_{KP})$<br><br>
6. –û—Ü–µ–Ω–∫–∞ $\rho_{KP}$ –ø–æ –º–µ—Ç–æ–¥—É –º–æ–º–µ–Ω—Ç–æ–≤:<br><br>
$\hat{\rho}_{KP} = \frac{2(g-1)^2M - L^2 + L_2}{L(g-1)^2 - L^2 + L_2}$
<br><br><br>
**–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–æ–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è** - –î–∞–Ω–Ω—ã–π –∏–Ω–¥–µ–∫—Å –æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–º –∑–Ω–∞—á–µ–Ω–∏—è—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥;<br>
–ó–Ω–∞—á–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–∞: $\rho_b = \frac{M - M_{min}}{M_{max} - M_{min}}$<br><br>

**–í—ã—á–∏—Å–ª–µ–Ω–∏–µ $M_{min}, M_{max}$**

1. $M_{min} = \underset{0 <= t <= g}{max}f(t)$
2. $M_{max} = floor[ {1/2 \sum_{i=1}^g{x_{i+} - \underset{0 <= t <= g}{max}g(t)}}]$
3. –§—É–Ω–∫—Ü–∏—è $f(t)$: $f(t) = \sum_{i=1}^t{x_{i+} - t(g-1) - C_{t}^2}$
4. –§—É–Ω–∫—Ü–∏—è g(t): $g(t) = \sum_{i=1}^t{x_{i+} - t(t-1) - \sum_{i=t+1}^g{max(t, x_{i+})}}$<br><br>

**–û—Ü–µ–Ω–∫–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ P**

1. –ß–∏—Å–ª–æ –¥—É–≥ $L$ ~ $B(g(g-1), P_0)$ –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —á–∏—Å–ª–æ –¥—É–≥ –≤ –æ—Ä–≥—Ä–∞—Ñ–µ —Ä–∞–≤–Ω–æ l:
$P(L= l) = C_{g(g-1)}^l \cdot P_0(1 - P_0)^{g(g-1)-l}$
2. –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ: $E(L) = P_0\cdot g(g-1)$
3. –î–∏—Å–ø–µ—Ä—Å–∏—è: $D(L) = P_0(1-P_0)g(g-1)$ <br><br>

**–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –≥—Ä–∞—Ñ–∞—Ö**

1. $\mathcal{g}_d$ - –æ—Ä–≥—Ä–∞—Ñ —Å $g$ —É–∑–ª–∞–º–∏
2. $G_d(\mathcal{N})$ - –º–Ω–æ–∂–µ—Å—Ç–≤–æ –≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö –æ—Ä–≥—Ä–∞—Ñ–æ–≤ –±–µ–∑ –ø–µ—Ç–µ–ª—å —Å $g$ —É–∑–ª–∞–º–∏
3. –°–æ—Ü–∏–æ–º–∞—Ç—Ä–∏—Ü–∞ $\mathbb{X} = {X_{ij}}$ - —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞
4. $x = x_{ij}$ - –æ–¥–Ω–∞ –∏–∑ –≤–æ–∑–º–æ–∂–Ω—ã—Ö —Å–æ—Ü–∏–æ–º–∞—Ç—Ä–∏—Ü <br><br>

**–†–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ë–µ—Ä–Ω—É–ª–ª–∏**

1. #$G_d(\mathcal{N}) = 2^{g(g-1)}$
2. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏—Ç—å —Å–æ—Ü–∏–æ–º–∞—Ç—Ä–∏—Ü—É $x$: $P(X-x) = 2^{g(1-g)}$
3. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—è–≤–ª–µ–Ω–∏—è –¥—É–≥–∏ –º–µ–∂–¥—É –∞–∫—Ç–æ—Ä–∞–º–∏: 
$P(X_{ij} = 1) = 1/2$ if $i \neq j$,if $i = j => 0$
4. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—è–≤–ª–µ–Ω–∏—è –¥—É–≥–∏ –º–µ–∂–¥—É –∞–∫—Ç–æ—Ä–∞–º–∏:
$P(X_{ij} = 1) = P_{ij}$ if $i \neq j$, if $i = j => 0$
<br><br>

**–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –≥–∏–ø–æ—Ç–µ–∑—ã**

1. –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: $H_0 : L$ ~ $B(g(g-1), P_0)$
2.  –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ :
$z_l = \frac{l - E(L)}{\sqrt{D(L)}} = \frac{l - P_0\cdot g(g-1)}{\sqrt{P_0(1-P_0)g(g-1)}}$, –≥–¥–µ $l = \sum_{i=1}^{g}{\sum_{j=1}^{g}{x_{ij}}}$
3. –î–∞–Ω–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —è–≤–ª—è–µ—Ç—Å—è –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏.<br><br>

**–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –æ—Ü–µ–Ω–∫–∏ P**

1. –û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ P:
$\hat{\rho} = \frac{\sum_{i=1}^{g}{\sum_{j=1}^{g}{x_{ij}}}}{g(g-1)} = \frac{l}{g(g-1)}$ 
2. –î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª: $\hat{P}_{lower} \leq P \leq \hat{P}_{upper}$
3. $\hat{P}_{lower} = \hat{P} - z_{\alpha/2}\cdot \sqrt{\hat{P}(1-\hat{P})/g(g-1)}$
4. $\hat{P}_{upper} = \hat{P} + z_{\alpha/2}\cdot \sqrt{\hat{P}(1-\hat{P})/g(g-1)}$
5. –ó–¥–µ—Å—å $z_{\alpha/2}$: $\phi(z_{\alpha/2}) = 0.5 - \alpha/2$

2. –û—Ç–æ–±—Ä–∞–∑–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–π –æ—Ä–≥—Ä–∞—Ñ.

```r
plot(random_graph, main = '–°–ª—É—á–∞–π–Ω—ã–π –æ—Ä–≥—Ä–∞—Ñ –Ω–∞ –æ—Å–Ω–æ–≤–µ –º–æ–¥–µ–ª–∏ –≠—Ä–¥–µ—à–∞‚Äî–†–µ–Ω—å–∏', label=colnames(random_graph))
```

3. –í—ã—á–∏—Å–ª–∏—Ç—å ¬´–ø—è—Ç–∏—á–∏—Å–ª–æ–≤—É—é —Å–≤–æ–¥–∫—É¬ª –¥–ª—è –≥—Ä–∞—Ñ–∞ Gùëë.

```r
random_graph_net <- asNetwork(random_graph)

cat("–†–∞–∑–º–µ—Ä:\n", network.size(random_graph_net))

cat("\n–ü–ª–æ—Ç–Ω–æ—Å—Ç—å:\n", gden(random_graph_net))

cat("\n–ö–æ–º–ø–æ–Ω–µ–Ω—Ç—ã:\n", igraph::components(random_graph)$no)

components <- component.largest(random_graph_net,result="graph")
gd <- geodist(components)
cat("\n–î–∏–∞–º–µ—Ç—Ä:\n", max(gd$gdist))

cat("\n–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏:\n", gtrans(random_graph_net))
```

4. –í—ã—á–∏—Å–ª–∏—Ç—å —Ü–µ–Ω–∑—É—Å –¥–∏–∞–¥. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```r
cat("–¶–µ–Ω–∑—É—Å –¥–∏–∞–¥:\n")
census <- dyad_census(random_graph)
census
```

–í—ã–≤–æ–¥—ã: –ß–∏—Å–ª–æ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥/–¥–≤—É–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö –æ—Ç–Ω–æ—à–µ–Ω–∏–π = 1125;

–ß–∏—Å–ª–æ –∞—Å—Å–∏–º–µ—Ç—Ä–∏—á–Ω—ã—Ö –¥–∏–∞–¥/–æ–¥–Ω–æ–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö –æ—Ç–Ω–æ—à–µ–Ω–∏–π = 1629;

–ß–∏—Å–ª–æ –ø–∞—Ä –∞–∫—Ç–æ—Ä–æ–≤ (–º–µ–∂–¥—É –∫–æ—Ç–æ—Ä—ã–º–∏ –Ω–µ—Ç —Å–≤—è–∑–∏) = 567.


5. –í—ã—á–∏—Å–ª–∏—Ç—å –æ—Ü–µ–Ω–∫—É –∏–Ω–¥–µ–∫—Å–∞ –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ö–∞—Ü–∞‚Äî–ü–∞—É—ç–ª–∞ –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–æ–º –≤—ã–±–æ—Ä–µ –∞–∫—Ç–æ—Ä–æ–≤ ùëë –∏ –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```r
d <- 0.5*g
fix_choice <- (2*(g-1)*census$mut-g*d*d)/(g*d*(g-1-d))
cat("–§–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≤—ã–±–æ—Ä –∞–∫—Ç–æ—Ä–æ–≤:\n")
fix_choice

matrix <- as.matrix(as_adjacency_matrix(random_graph))
L <- sum(rowSums(matrix))
L2 <- sum(rowSums(matrix)^2)
free_choice <- (2*(g - 1)^2 * census$mut - L^2 + L2)/(L*(g - 1)^2 - L^2 + L2)
cat("–°–≤–æ–±–æ–¥–Ω—ã–π –≤—ã–±–æ—Ä –∞–∫—Ç–æ—Ä–æ–≤:\n")
free_choice
```

–í—ã–≤–æ–¥—ã: –ü—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –≤—ã–±–æ—Ä–µ –∞–∫—Ç–æ—Ä—ã –ø—Ä–µ–¥–ø–æ—á–∏—Ç–∞—é—Ç –æ–±—Ä–∞–∑–æ–≤—ã–≤–∞—Ç—å –≤–∑–∞–∏–º–Ω—ã–µ —Å–≤—è–∑–∏, –≤ —Ç–æ –≤—Ä–µ–º—è –∫–∞–∫ –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ –∞–∫—Ç–æ—Ä—ã —Å–∫–ª–æ–Ω–Ω—ã –æ–±—Ä–∞–∑–æ–≤—ã–≤–∞—Ç—å –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–µ —Å–≤—è–∑–∏ –∏–ª–∏ –Ω–µ –æ–±—Ä–∞–∑–æ–≤—ã–≤–∞—Ç—å –∏—Ö –≤–æ–≤—Å–µ.

6. –í—ã—á–∏—Å–ª–∏—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å–≤—è–∑—å –±—É–¥–µ—Ç –≤–∑–∞–∏–º–Ω–æ–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Å–ª—É—á–∞—è.

```r
fix_prob <- (d / (g-1)) * ((d/(g-1)) + fix_choice*(g - 1 - d) / (g - 1))
cat("–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –¥–ª—è —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –≤—ã–±–æ—Ä–∞ –∞–∫—Ç–æ—Ä–æ–≤:\n")
fix_prob

free_prob <- (rowSums(matrix) * colSums(matrix))/(g - 1)^2
free_prob_avg <- sum(free_prob)/g
cat("–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –¥–ª—è —Å–≤–æ–±–æ–¥–Ω–æ–≥–æ –≤—ã–±–æ—Ä–∞ –∞–∫—Ç–æ—Ä–æ–≤:\n")
free_prob_avg
```

7. –í—ã—á–∏—Å–ª–∏—Ç—å –∏–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```r
random_graph_matr <- as.matrix(random_graph_net)
v_1 <- c()
v_2 <- c()
for(t in 0:g){
    s <- 0
    if (t>0) {
        for(i in 1:t) {
            s <- s + sum(random_graph_net[i, ])
        }
    }
    f_t <- s - t*(g - 1) - choose(t, 2)
    v_1 <- append(v_1, f_t)
    s2 <- 0
    if (t + 1 <= g){
        for(i in (t+1):g) {
            s2 <- s2 + min(t, sum(random_graph_matr[i, ]))
        }
    }
    g_t <- s - t*(t-1) - s2
    v_2 <- append(v_2, g_t)
}
M_min <- max(v_1)
M_max <- floor(0.5*(sum(random_graph_matr) - max(v_2)))
P_B <- (census$mut - M_min)/(M_max - M_min)
cat("–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è:", P_B)
```

–í—ã–≤–æ–¥—ã: –ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ—Ü–µ–Ω–∏—Ç—å –≤–∑–∞–∏–º–Ω–æ—Å—Ç—å –±–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤—ã–±–æ—Ä–æ–≤, —Ç–∞–∫ –∫–∞–∫ –æ–Ω –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤–∑–∞–∏–º–Ω—ã—Ö —Å–≤—è–∑–µ–π, –ø—Ä–∏ –º–æ–∏—Ö –¥–∞–Ω–Ω—ã—Ö –æ–Ω —Ä–∞–≤–µ–Ω 0.580196.

8. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —á–∏—Å–ª–æ –¥—É–≥ ùêø –≤ –æ—Ä–≥—Ä–∞—Ñ–µ
Gùëë –∏–º–µ–µ—Ç –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ ùêø ‚àº Bin(ùëî(ùëî ‚àí 1), 1/2) –ø—Ä–∏
—É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ ùõº. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è.

```r
cat("alpha = ", 0.04)
l <- sum(matrix)
binom.test(l, g*(g-1), 0.5)
```

9. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —á–∏—Å–ª–æ –¥—É–≥ ùêø –≤ –æ—Ä–≥—Ä–∞—Ñ–µ Gùëë
–∏–º–µ–µ—Ç –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ ùêø ‚àº Bin(ùëî(ùëî‚àí1), ùëÉ0) –ø—Ä–∏ —É—Ä–æ–≤–Ω–µ
–∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ ùõº. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –ø—Ä–∏
–∫–æ—Ç–æ—Ä–æ–º –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è.

–í—ã–≤–æ–¥—ã: –¢–∞–∫–∂–µ –∫–∞–∫ –∏ –≤ –ø—Ä–æ—à–ª–æ–º –ø—É–Ω–∫—Ç–µ, p-value –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å ùõº –æ—á–µ–Ω—å –º–∞–ª–æ, —Ç–æ –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è. –ó–Ω–∞—á–∏—Ç p-value –∏ —è–≤–ª—è–µ—Ç—Å—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º —É—Ä–æ–≤–Ω–µ–º –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è: p-value < 2.2e-16.

10. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùëù —Å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–π –Ω–∞–¥—ë–∂–Ω–æ—Å—Ç—å—é (1 ‚àí ùõº). –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```r
est <- (sum(matrix))/(g*(g-1))
cat("–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùëù:")
est

laplace_func <- 0.5 - (alpha/2)
cat("–§—É–Ω–∫—Ü–∏—è –õ–∞–ø–ª–∞—Å–∞:")
laplace_func
#–¢–∞–±–ª–∏—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
z_a <- 2.06

lower <- est - z_a * sqrt((est * (1 - est))/(g * (g - 1)))
upper <- est + z_a * sqrt((est * (1 - est))/(g * (g - 1)))     
cat("–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª:")                 
lower
upper
```
