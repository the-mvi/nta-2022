## Основные теоретические положения
**Актор** - это дискретные индивидуальные, корпоративные или коллективные социальные единицы.

**Отношение** - совокупность связей определенного вида между членами группы.

**Социальная сеть** - объект, представляющий собой совокупность конечного набора (или наборов) акторов и их отношения (или отношений), определенных на них.

**Граф** - это модель социальной сети с дихотомическим отношением, т.е. связь либо присутсвует, либо отсутствует между каждой парой акторов.

**Социоматрица** - матрицы $X$ размерности $g*g$, описывающая смежные узлы в сети.

**Размер сети** - количество акторов сети.

**Плотность** - доля имеющихся связей по отношению к максимально возможному количеству связей.

Для ориентированных сетей:
$\Delta = \frac{L}{A_g^2}$

Для неориентированных сетей:
$\Delta = \frac{L}{\binom{g}{2}}$

$g$ - число акторов, $L$ - число имеющихся связей в сети

**Компонента сети** - подгруппа, в которой все акторы связаны друг с другом прямо или косвенно.

**Диаметр полной сети** - наибольшая длина из кратчайших путец между всеми парами узлов.

**Степень узла $d(n_i)$** - число прямых, инцидентных данному узлу. 

Средняя степень узлов: 
$$\overline{d} = \frac{1}{g}\sum_{i=1}^{g}d(n_i)$$

Дисперсия степеней узлов: 
$$S_D^2 = \frac{1}{g}\sum_{i=1}^{g}(d(n_i)-\overline{d}))^2$$

**Транзитивность** - доля закрытых треугольников (триад, где наблюдаются все 3 связи) по отношению к общему количеству открытых и закрытых треугольников.

**Центральность** - мера, характеризующая актор с большим количеством связей (прямых или косвенных).
$$C_{A}(n_{i}), i=1...g$$

**Престиж** - мера, характеризующая актор с большим количеством входящих связей.
$$P_{A}(n_{i}), i=1...g$$

**Индекс централизации:**
$$C_{A}=\frac{\sum_{i=1}^{g}[C_{A}(n^{*}) - C_{A}(n_{i})]}{\max \sum_{i=1}^{g}[C_{A}(n^{*}) - C_{A}(N_{i})]}$$

**Центральность по степени** - степень каждого актора.
$$C_{D}(n_{i}) = d(n_{i})$$

**Нормированная центральность по степени:**
$$C_{D}^{'}(n_{i}) = \frac{d(n_{i})}{g-1}$$

**Централизация по степени:**
$$C_{D} = \frac{\sum_{i=1}^{g}[C_{D}(n^{*}) - C_{D}(n_{i})]}{(g-1)(g-2)}$$

**Центральность по близости** - более важными узлами считаются узлы, наиболее близкие к остальным узлам сети.
$$C_{C}(n_{i}) = (\sum_{j=1}^{g}{d(n_{i}, n_{j}}))^{-1}$$

**Нормированная центральность по близости:**
$$C_{C}^{'}(n_{i}) = \frac{g-1}{\sum_{j=1}^{g}{d(n_{i}, n_{j})}}$$

**Централизация по близости:**
$$C_{C} = \frac{\sum_{i=1}^{g}[C_{C}^{'}(n^{*}) - C_{C}^{'}(n_{i})]}{(g-1)(g-2)/(2g-3)}$$

**Центральность по посредничеству** характеризует, насколько важную роль данный узел играет на пути "между" парами других узлов сети, в том смысле, что пути между другими узлами должны проходить через данный узел.
$$C_{B}(n_{i}) = \sum_{j<k}{g_{jk}(n_{i})/g_{jk}},$$ 
где $g_{jk}(n_{i})$ - количество геодезических расстояний между узлами $j$ и $k$, которые включают узел $i$;  $g_{jk}$ - кодезическое расстояние между узлами $j$ и $k$.

**Нормированная центральность по посредничеству:**
$$C_{B}'(n_{i}) = \frac{C_{B}(n_{i})}{\binom{g-1}{2}}$$

**Централизация по посредничеству:**
$$C_{B} = \frac{\sum_{i=1}^{g}[C_{B}^{'}(n^{*} - C_{B}^{'}(n_{i}]}{(g-1)}$$

**Центральность по информации:**
$$C_{I}(n_{i}) = \frac{1}{c_{ij}+(T-2R)/g}$$

**Нормированная центральность по информации:**
$$C_{i}^{'}(n_{i}) = \frac{C_{I}(n_{i})}{\sum_{i=1}^{g} C_{I}(n_{i})}$$

**Централизация по информации:**
$$S_{I}^2 = \frac{1}{g}\sum_{i=1}^{g}(C_{I}^{'}(n_{i})-\overline{C_{I}})^2, $$
где $\overline{C_{I}} = \sum_{i=1}^{g}C_{I}(n_{i})/g$

**Точка сочленения** - это узел, при удалении которого увеличивается число компонент связности.

**Мост** - это ребро, при удалении которого одна компонента разделяется на две.

**Престиж по степени** - входящая степень каждого узла.
$$P_{D}(n_{i}) = d_{i}(n_{i}) = x_{+i}$$

**Нормализованный престиж по степени:**
$$P_{D}^{'}(n_{i}) = \frac{x_{+i}}{g-1}$$

**Престиж по близости** определяет насколько все акторы близки к заданному по расстоянию до него.
$$P_{P}(n_{i}) = \frac{\#I_{i}/(g-1)}{\sum_{j:n_{j} \in I_{i}} d(n_{j}, n_{i})/\#I_{i}}$$
