## Основные теоретические положения

**Подграф**
Для графа $G(N, L)$ подграфом $G_{S}$ называется граф $G_{S}=G_{S}(N_S, L_S)$ cостоящий из подмножества узлов $N_S \subseteq N$ и подмножества ребер  $N_L \subseteq L$, взязывающие узлы в $G_S$. 

**Группа** - это совокупность всех акторов, связи между которыми должны быть изучена.

**Подгруппа** - подмножество узлов $N_S$.

**Клика** - максимальный полный подграф из 3 или более узлов.

**n-клика** - максимальный подграф, в котором наибольшее геодезическое расстояние между любыми 2 узлами не превышает n.
$$d(i, j) \leq n  ,\forall n_i, n_j  \in  N_S$$

**n-клан** - это n-клика, в которой геодезическое расстояние $d(i, j)$ между всеми узлами в подграфе не превышает $n$ для путей внутри подграфа.

**n-клуб** - это максимальный подграф диаметра $\leq  n$.

**k-ядро** - это максимальный подграф, в котором каждая вершина связана минимум с $k$ другими вершинами этого же подграфа.

**k-плекс** - это максимальный подграф, содержащий $g_S$ узлов, в котором каждый узел смержен не менее чем с $(g_S-k)$ узлами подграфа.
$$d_S(i) \geq  g_S-k  ,\forall n_i  \in  N_S$$

**Модулярность** - характеристика структуры сети, в частности степень кластеризации узлов, когда внутри кластера наблюдается много узлов(высокая лотность), а между класстерами мало узлов(низкая плотность).

**Сеть аффилированности** - сеть, в которой участники аффилированы друг с другом на основе сотрудничества в рамках какой-то группы или совместного участия в определенном событии.

**Матрица инцидентности** - матрица $A={a_{ij}}$ размерности $g\times h$ с элементами
$$a_{ij}=\begin{cases} 
1, &\text{i связан с j} \\
0, &\text{иначе}
\end{cases}$$

**Двудольный граф** - граф, узлы которого можно разделить на 2 подмножества, и все ребра будут соединять узлы разных подмножеств.
Представление биграфа в виде социоматрицы размера $(g+h)*(g+h)$:
$$\bigl(\begin{smallmatrix}
O_g&A \\ A^T&O_h
\end{smallmatrix} \bigr)$$

**Гиперграф** - обобщение понятия графа, состоит из множества объектов и множества подмножеств объектов, в котором каждый объект принадлежит хотя бы одному подмножеству, причем ни одно подмножество не является пустым.

$$H= H(A, B),$$
$A=[a_{1}, a_{2}, ..., a_{g}]$ - множество объектов

$B=[B_{1}, B_{2}, ..., B_{h}]$ - множество ребер

**Двойственный гиперграф** для гиперграфа H=H(A, B)$:

$$H^{*}=H(B, A),$$
$B=[b_{1}, b_{2}, ..., b_{h}]$ - множество объектов

$A=[A_{1}, A_{2}, ..., A_{g}]$ - множество ребер

**Матрица соучастия акторов**:
$$x_{ij}^{N}=\sum_{k=1}^{h}a_{ik}a_{jk}=XX^{T}$$

**Матрица перекрытия событий**:
$$x_{ij}^{M}=\sum_{i=1}^{g}a_{ik}a_{j1}=X^{T}X$$

**Степень участия** - количество событий, с которыми связан каждый актор.
$$a_{i+}=\sum_{j=1}a_{ij}=x_{ii}$$

**Среднее количество событий, с которыми акторы аффилированы**:
$$\overline a_{i+}=\frac{1}{g}\sum_{i=1}\sum_{j=1}a_{ij}=\frac{a++}{g}=\frac{1}{g}\sum_{i=1}x_{ii}^N$$

**Размер событий**:
$$a_{ij}=\sum_{i=1}a_{ij}=x_{jj}$$


**Средний размер событий** - среднее количество акторов в каждом событии:
$$\overline a_{+j}=\frac{1}{h}\sum_{j=1}\sum_{i=1}a_{ij}=\frac{a++}{h}=\frac{1}{h}\sum_{j=1}x_{jj}^M$$

**Плотность для отношения соучастия**:
$$\Delta_{{N}}=\frac{1}{g(g-1)}\sum_{i=1}\sum_{j=1, i\neq j}x_{ij}^{N}$$

**Плотность для отношения перекрытия**:
$$\Delta_{{M}}=\frac{1}{h(h-1)}\sum_{k=1}\sum_{l=1, l\neq k}x_{kl}^{M}$$

**Диаметр аффилированной сети** - это длина самого длинного пути между любой парой акторов/событий.

**Клика уровня c в взвешенном графе** - максимальный полный подграф из 3х или более узлов, все из которых являются смежными на уровне $c$, т.е. между всеми парами узлов есть ребра со значениями большими или равными $с$.

**Клика уровня c для отношения соучастия** - это подграф, в котором все пары акторов совместно участвуют не менее, чем в $с$ событиях.полный подграф из 3х или более узлов, все из которых являются смежными на уровне $c$.

**Клика уровня c для отношения перекрытия** - это подграф, в котором все пары событий имеют не менее $c$ общих членов.

**Отношение шансов**

$x_{kl}^M $- число акторов, которые аффилированы с событиями $k$ и $l$

$x_{\overline{kl}}^M $- число акторов, которые не принадлежат ни одному событию$

$x_{k\overline{l}}^M$- число акторов, которые аффилированы с событием $k$, но не с событием $l$

$x_{\overline{k}l}^M $- число акторов, которые аффилированы с событием $l$, но не с событием $k$

Справедливо следующее:
$$x_{kl}^M+x_{\overline{kl}}^M+x_{k\overline{l}}^M+x_{\overline{k}l}^M=g$$

**Отношение шансов** - мера перекрытия между собятиями, которая не зависит от размера событий.

**Логарифм отношения шансов**:
$$ln\theta_{kl} = ln(\frac{x_{kl}^M x_{\overline{kl}}^M}{x_{k\overline{l}}^M x_{\overline{k}l}^M})$$

```r
#n-клики
n_clique <- function(graph, n) {
    n_graph <- connect(graph, n)
    return (max_cliques(n_graph, min=3))
}
# n_clique(graph_from_matrix, 4)

ncliques <- function(g, n){
    E(g)$weight <- 1 
    res <- list() 
    for (i in 1:n){ 
        cliq <- n_clique(g, i)
        print(paste(i, "-клики : "))
        print(cliq)
      }
}

ncliques(graph_from_matrix, 4)

```

```r
#n-кланы
n_clan <- function(graph, n) {
    new_graph <- connect(graph, n) 
    ncliques <- max_cliques(new_graph) 
    for (i in 1:length(ncliques)) {
        vertexes <- list()
        for (j in 1:length(ncliques[[i]])) {
          vertexes <- append(vertexes, ncliques[[i]][j]$name)
        }
        nodes <- V(graph)[name %in% vertexes]
        sub_g <- induced_subgraph(graph, nodes)
        if ((diameter(sub_g) <= n) && (length(vertexes) >= 3)) {
            print(ncliques[i])
        }
    }
}
n_clans <- function(graph, n){
    for (i in 1:n){ 
        n_clan(graph, i)
      }
}
n_clans(graph_from_matrix, 2)
```

Кроме того, была определена модулярность сети и найдено 4 подгруппы, в которых учатсники более всего связаны между собой с помощью алгоритмов нахождения сообществ InfoMAP, Louvain и Optimal. 

Также была иследована направленная сеть, сео, представляющая собой членство людей в клубах. Данная сеть была представлена в виде биграфа, гиперграфа и двойственного гиперграфа. Кроме этого, были построены сеть соучастия и сеть перекрытия, которые показади, что в среднем каждый участник состоит в 3 клубах. В каждом клубе примерно состоит 6-7 участников. А также были определы наиболее часто встечающиеся наборы клубом и вычислены основные метрики.
