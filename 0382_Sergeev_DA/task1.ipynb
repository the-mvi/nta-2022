## Основные теоретические положения
**Акторы(actors)** - это дискретные индивидуальные, корпоративные или коллективные социальные единицы.  
**Социоматрица (*sociomatrix*)** - матрица X размерности g x g, описывающая смежные узлы в сети:  
X = {$x_{ij;r}$} i,j = 1, …, g  

Числовые характеристики сети:
   
**Социальная сеть (social network)** - объект, представляющий собой совокупность конечного набора (или наборов) акторов и отношения (или отношений), определенных на них.
**Размер сети (network size)** - количество акторов сети.  
**Плотность (density)** - доля имеющихся связей по отношению к максимально возможному количеству связей  
$\varDelta = \frac{L}{A_g^2}$ (ориентированная сеть)  
$\varDelta = \frac{L}{(_2^g)}$ (неориентированная сеть)  
**Компонента (component)** - подгруппа, в которой все акторы связаны друг с другом прямо или косвенно  
**Диаметр полной сети (diametr)** - наибольшая длина любого из кратчайших путей между всеми парами узлов.  
**Тразитивность, коэф. кластеризации (transitivity)** - доля закрытых треугольников (триад где наблюдаются все три связи) по отношению к общему количеству открытых и закрытых теругольников (триад, где наблюдаются либо две, либо все три связи)

Алгоритмы укладки сетей:  

**Алгоритм Фрюхтермана-Рейнгольда** - Идея данного алгоритма заключается в том, чтобы рассматривать узлы как атомные частицы или небесные тела, между которыми действуют силы притяжения и отталкивания. Вводятся следующие силы: $F_a(d) = \frac {d^2} {k}$ - сила притяжения, $F_r(d) = \frac {k^2} {d}$ - сила отталкивания, где $d$ - расстояние между узлами, $k=const*\sqrt {\frac {area}{|V|}}$. В начале работы алгоритма узлы случайным образом располагаются в пространстве. Затем на каждой итерации силы притяжения и отталкивания смещаются, причем на это смещение также влияет параметр температуры $t$, значение которого начинается с $0.1*area$ и линейно уменьшается до 0.   
**Алгоритм Камада-Каваи** - Идея данного алгоритма состоит в том, что узлы соединяются пружинами, и задача сводится к тому, чтобы минимальная энергия пружинной системы соответствовала минимальной разнице между геометрическим расстоянием и графовым расстоянием между узлами (т.е. найти минимум энергии системы). Вводим сила пружины как $k_{ij}=\frac {const} {d^2_{ij}}$, где $d_{ij}$ - кратчайшее растояние пути между вершинами. Фиксируется некоторое число $ϵ$. На каждой итерации выбирается вершина, которая имеет максимальное значение $Δ_m=\sqrt {(\frac {δE} {δx_m})^2+(\frac {δE} {δy_m})^2}$, где $E$ - энергия пружинной системы, $(x_m,y_m)$ - координаты вершины. Далее нужно решить систему дифференциальных уравнений, благодря которой будет найдено смещение. Это будет продолжаться до тех пор, пока $Δ_m > ϵ$.

**Актор считается важным (prominence)**, если связи актора делают его видимым для других участников сети.  
**Центральность**(centrality) - мера, характеризующая актор с большим количеством связей(прямых или косвенных).  
    Обозначение: $C_A(n_i),\:i=1,...,g$  

**Центральность по степени**(degree centrality) - степень каждого актора: $С_D(n_i)=d(n_i)$   
**Нормированная центральность по степени**: $C^`_D(n_i)=\dfrac{d(n_i)}{g-1}$  
**Нормированная центральность по степени**: $C_D = \dfrac{\sum_{i=1}^g[C_D(n^*)-C_D(n_i)]}{max\sum_{i=1}^g[C_D(n^*)-C_D(n_i)]} = \dfrac{\sum_{i=1}^g[C_D(n^*)-C_D(n_i)]}{{(g-1)(g-2)}}$

**Центральность по близости**(closeness centrality) - более важными узлами считаются узлы, наиболее близкие к остальным узлам сети: $
С_с(n_i) = (\sum_{j=1}^g d(n_i, n_j))^{-1}$  
**Нормированная центральность по близости**: $C^`_C(n_i) = \dfrac{g-1}{\sum_{j=1}^gd(n_i, n_j)} = (g-1)C_c(n_i)$  
**Централизация по близости**: $C_С = \dfrac{\sum_{i=1}^g[C^`_C(n^*)-C_C^`(n_i)]}{(g-1)(g-2)/(2g-3)}$  

**Центральность по посредничеству**(*betweenness centrality*) характеризует, насколько важную роль данный узел играет на пути “между” парами других узлов сети, в том смысле, что пути между другими узлами должны проходить через данный узел: $$C_B(n_i) = \sum_{j<k}( g_{jk}(n_i)/g_{jk})$$ где:  
$g_{jk}(n_i)$ - количество геодезических расстояний между узлами j и k, которые включают узел i  
$g_{jk}$ - геодезическое расстояние между узлами j и k    
**Нормированная центральность по посредничеству**: $C^`_B(n_i) = \dfrac{C_b(n_i)}{\binom{g-1}{2}}$  
**Централизация по посредничеству**: $C_B=\frac{\sum_{i=1}^{g}(C_B'(n^\star)-C_B'(n_i))}{(g-1)}$

**Центральность по информации**(*information centrality*) показывает сколько информации содержится в путях, которые начинаются(заканчиваются) у определенного актора: $C_{I}(n_i) = (c_{ii} + (T - 2R)/k)^{-1}$, где $T=∑_{i=1}^{k}c_{ii}$, $R=∑_{j=1}^{k}c_{ij}$, $C=A^{-1}$, $A$ - матрица ${\{a_{ij}\}}$ размером $k\times k$ со значениями $a_{ii} = 1 + deg(n_i)$, $a_{ij} = 1 - x_{ij}$ ($x_{ij}$ - элемент матрицы смежности) если $i$ и $j$ - инцидентны, иначе - 1.<br>
**Нормированная центральность по информации**: $C_{I}^{'}(n_i)=\frac{C_{I}(n_i)}{\sum_{i=1}^{k}C_{I}(n_i)}$.<br>
**Централизация по информации**: $S_I^2=\frac {1} {k}∑_{i=1}^k[C_I^{'}(n_i) - \overline {C_I}]$, где $\overline {C_I} = ∑C_I(n_i)$.<br><br>

**Престиж (prestige)** - мера, характеризующая актор с большим количеством входящих связей.  
**Престиж по степени**: $P_D(n_i)=d_I(n_i)$, где $d_I(n_i)$ - входящая степень вершины.<br>
**Нормированный престиж по степени**: $P_D^{'}(n_i)=d_I(n_i)/(k-1)$.<br>

**Область влияния актора(influence domain of actor)** $I_i$ - множество акторов, прямо и косвенно связанных с указанным актором  
**Престиж по близости**:
$P_P(n_i)=\frac{\# I_i/g-1}{\sum_{j:n_j \ni l_i}d(n_j, n_i)/\# I_i}$  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$I_i$ - область влияния $i$-го актора

```r
library(igraph)
vertToDelete <- which(degree(net1)==0)
net1 <- delete.vertices(net1,vertToDelete)
actorsCount <- gorder(net1)

Cd <- centr_degree(net1)$res/(actorsCount-1)
Cc <- centr_clo(net1)$res
Cb <- 2*centr_betw(net1)$res/((actorsCount-1)*(actorsCount-2))

mWithoutEmpty <- m[which(rowSums(m)>0),which(colSums(m)>0)]
network2 <- network(mWithoutEmpty,directed = FALSE,matrix.type = "adjacency")
Ci <- infocent(network2,gmode = "graph")
Ci <- Ci/sum(Ci)
df <- data.frame(Cd,Cc,Cb,Ci)
rownames(df) <- rownames(mWithoutEmpty)
print(df)
```
