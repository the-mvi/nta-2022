## Основные теоретические положения

***клика*** - максимальный полный подграф из трёх или более узлов.

***n-клика*** - максимальный подграф, в котором наибольшее геодезическое расстояние между любыми двумя узлами не превышает n.

***n-клан*** -  n-клика, в которой геодезическое расстояние между всеми узлами в подграфе не превышает n для путей внутри подграфа.

***n-клуб*** - максимальный подграф диаметра ≤ n .

***k-плекс*** - максимальный подграф, содержащий $g_s$ узлов, в котором каждый узел смежен с не менее чем с $(g_s - k)$ узлами подграфа.

***k-ядро*** - максимальный подграф, в котором каждая вершина связана минимум с k другими вершинами этого же подграфа.
    
***Модулярность*** - характеристика структуры сети, в частности степень кластеризации узлов, когда внутри кластера наблюдается много узлов, а между кластерами мало узлов. показатель модулярности определяется как разница между долей связей внутри группы и ожидаемой доли связей, если бы связи были размещены случайно.

***Сеть аффилированности*** - сеть, в которой участники аффилированы друг с другом на основе сотрудничества в рамках какой-то группы или совместного участия в определенном событии. Сеть состоит из двух измерений.


*Представление сетей аффилированности:*

* Матрица инцидентности $A_{g\times h}$
* Двудольный граф
* Прелставление биграфа в виде социоматрицы размера $(g+h)\times (g+h)$ и вида $(^{0_g A}_{A^T 0_h})$
* Гиперграф ($\mathcal{H}=\mathcal{H}(\mathcal{A}, \mathcal{B})$, где $\mathcal{A}$ - множество объектов, $\mathcal{B}$ - множество событий)
* Двойственный гиперграф

**Одномодальные проекции:**

* Матрица соучастия акторов: 

  $X^{\cal{N}} = \{x_{ij}^{\cal{N}} \}$

  $x_{ij}^{\cal{N}} = \sum_{k = 1}^{h} a_{ik}a_{jk}$

* Матрица перекрытия событий

  $X^{\cal{M}} = \{x_{kl}^{\cal{M}} \}$

  $x_{kl}^{\cal{M}} = \sum_{i = 1}^{g} a_{ik}a_{jl}$


***Сепень участия*** - количество событий, с которым связан каждый актор.

$a_{i+} = \sum_{j = 1}^{h} a_{ij} = x_{ii}^{\cal{N}}$

***Средняя степень участия***: $\bar{a_{i+}} = \frac1g \sum_{i=1}^g x_{ii}^\mathcal{N}$

***Размер события***: $a_{+j} = \sum_{i = 1}^{g} a_{ij} = x_{ii}^{\cal{M}}$

***Средний размер события***- среднее количество акторов в каждом событии или средняя степень узлов, прндставляющих события в двудольном графе.
 $\bar{a_{+j}} = \frac1h \sum_{j=1}^h x_{jj}^\mathcal{M}$

***Плотность:***

* Отношения соучастия: $\Delta_{(\mathcal{N})} = \frac1{g(g-1)} \sum_{i=1}^g\sum_{j=1; j\ne i}^g x_{ij}^\mathcal{N}$ - $\in[0; h]$
* Отношения перекрытия: $\Delta_{(\mathcal{M})} = \frac1{h(h-1)} \sum_{k=1}^h\sum_{l=1; l\ne k}^h x_{kl}^\mathcal{M}$ - $\in[0; g]$

***Связная сеть аффилированности*** - аффилированная сеть, в которой все пары узлов достижимы.

***Диаметр аффилированной сети*** - длина самого длинного пути между любой парой актор и/или событий.

***Клика уровня c*** в взвешенном графе - - максимальный полный подграф из трех или более узлов, все из которых являются смежными на уровне c, то есть между всеми парами узлов есть ребра со значениями, большими или равными c.

***Клика уровня c*** для отношений соучастия - подграф, где все пары акторов совметсно участвуют не менее чем в c событиях.

***Клика уровня c*** для отношений перекрытия - подграф, где все пары событий имеет не менее c общих членов.

***Отношение шансов*** - мера перекрытия между событиями, которая не зависит от размера событий: $\Theta_{kl}=\frac{x_{kl}^\mathcal{M} x_{\bar{kl}}^\mathcal{M}}{x_{k\bar{l}}^\mathcal{M}x_{\bar{k}l}^\mathcal{M}}$

**Свойства сетей аффилированности:**
* Количество перекрывающихся связей между событиями частично зависит от количества событий, к которым принадлежат акторы.

* Количество связей соучастия между акторами частично зависит от размера событий.

* Актор, принадлежащий $a_{i+}$ событиям, создает $C_{a_{i+}}^2$ парных связей между событиями.

* Событие с участиеми $a_{+j}$ участников создает $C_{a_{+о}}^2$ связей между парами акторов.

* Плотность для дихотомического отношения - доля присутствующих связей.

* Плоность для ценностных отношений - среднее значение связей.

**Меры**. 

* Мера сплочённости по по Боке-Хусейну - отношение средней силы связи внутри подгруппы и средней силы связи между подгруппой и остальной группой: $M_{BH} = \frac{\frac{1}{g_s(g_s-1)}\sum_{i \in \mathcal{N}_s}\sum_{j \in \mathcal{N}_s}x_{ij}}{\frac{1}{g_s(g-g_s)}\sum_{i \in \mathcal{N}_s}\sum_{j \notin \mathcal{N}_s}x_{ij}}$
* Мера Альбы - гипергеометрическая вероятность того, что наблюлаемое количество рёбер в подграфе равно q: $P(L_s=q)=(_q^L)\frac{(_{(_2^{g_s})-q}^{(_2^g)-L})}{(_{(_2^{g_s})}^{(_2^g)})}$

Указать какие подгруппы плучились. Сделать вывод

```r
nclan <- function(g,n,ncliques){
    E(g)$weight <- 1 #just in case g has weights - does not modify original graph
    n.cand <- ncliques[[n]] #n-clique candidates to be an n-clan
    n.clan <- list() #initializes a list to store the n-clans
    n.clan.i <- 1 #initializes a list pointer
    for (n.cand.i in 1:length(n.cand)){ #loop over all of the candidates
        g.n.cand <- induced_subgraph(g,n.cand[[n.cand.i]]) #get the subgraph
        if (diameter(g.n.cand)<=n){ #check diameter of the subgraph
          n.clan[[n.clan.i]] <- n.cand[[n.cand.i]] #add n-clan to the list
          n.clan.i <- n.clan.i+1 #increment list pointer
        }
      }
    return(n.clan) #return the entire list
}
nclans <- function(g){
    g <- as.undirected(g)
    ncliques <- kCliques(ugraph(igraph.to.graphNEL(g))) #get cliques
    result <- c()
    for (i in names(cl)){
        result[[i]] <- nclan(g, i, ncliques)
    }
    return(result)
}

kplexes <- function(g) {
    result <- c()
    for (i in 2:(length(V(g))-6)) {
        for (j in 1:(i-1)) {
            plex <- kplex(g, j, i)
            if(length(plex) > 0) {
                result[[paste(toString(i), toString(j), "plex", sep="_")]] <- plex
            }
        }
    }
    return(result)
}

kplex <- function(g,k,m){
    g.sym <- as.undirected(g)
    #получаем подграф, степени вершин которых >= m-k
    k.plex <- list()
    g.sym.degmk <- induced_subgraph(g.sym,igraph::degree(g.sym)>=(m-k))
    if(length(V(g.sym.degmk)) < m) {
        return(k.plex)
    } 
    k.cand <- combn(V(g.sym.degmk)$name,m) #all candidate combinations with m members
    k.plex <- list() 
    k.plex.i <- 1 
    for (k.cand.i in 1:dim(k.cand)[2]){ 
        g.k.cand <- induced_subgraph(g.sym.degmk,k.cand[,k.cand.i]) 
        if (min(igraph::degree(g.k.cand))>=(m-k)){ #if minimum degree of sugraph is > m=k, k-plex!
#             plot(g.k.cand, layout=layout_with_kk, label=colnames(data))
            k.plex[[k.plex.i]] <- k.cand[,k.cand.i] 
            k.plex.i <- k.plex.i+1 
        }
    }
    return(k.plex) #return the entire list
}

kkernel <- function(g, k) {
    g.sym <- as.undirected(g)
    k.kernel <- list()
    g.sym.degk <- induced_subgraph(g.sym,igraph::degree(g.sym)>=k)
    if(length(V(g.sym.degk)) < k) {
        return(k.kernel)
    } 
    flag <- 0
    for (j in length(V(g.sym.degk)):1){
        if(flag) {
            break
        }
        k.cand <- combn(V(g.sym.degk)$name,j) 
        k.kernel <- list() 
        k.kernel.i <- 1 
        for (k.cand.i in 1:dim(k.cand)[2]){ 
            g.k.cand <- induced_subgraph(g.sym.degk,k.cand[,k.cand.i]) 
            if (min(igraph::degree(g.k.cand))>=k){ #if minimum degree of sugraph is > k, k-kernel!
                k.kernel[[k.kernel.i]] <- k.cand[,k.cand.i] 
                k.kernel.i <- k.kernel.i+1 
                flag <- 1
            }
        }
    }
    return(k.kernel) #return the entire list
}

kkernels <- function(g) {
    result <- c()
    for (k in 1:length(V(g))) {
        kernel <- kkernel(g, k)
        if(length(kernel) > 0) {
            result[[paste(toString(k), "kernel", sep="_")]] <- kernel
        }
    }
    return(result)
}

matr <- matr[!(apply(matr, 1, function(y) all(y == 0))),!(apply(matr, 1, function(y) all(y == 0)))]
without_alone_graph <- graph.adjacency(matr,  mode="undirected")

#n-клики
cat("n-cliques")
cl <- kCliques(igraph.to.graphNEL(graph))
cl

#n-кланы
cat("n-cliques => n-clans")
ncl <- nclans(graph)
ncl

#k-плексы
cat("i_j_plex => i - мощность множества вершин подграфа, j - k")
kpl <- kplexes(without_alone_graph)
kpl
         
#k-ядра
cat("k-kernels")
kke <- kkernels(without_alone_graph)
kke
```

Суть алгоритмов:

***Spinglass*** - Алгоритм поиска сообществ на графиках с помощью модели вращающегося стекла и имитированного отжига.

***Walktrap*** - Алгоритм поиска плотно связанных подграфов (сообществ), в графе с помощью случайных блужданий. Суть в том, что короткие случайные блуждания, как правило, остаются в одном и том же сообществе.

***Label propagation*** - Алгоритм для определения структуры сообщества в сетях. In работает, помечая вершины уникальными метками, а затем обновляя метки большинством голосов в окрестности вершины.

```r
gn <- induced_subgraph(graph, subcomponent(graph, 1))
spinglass <- cluster_spinglass(gn)
cat("Модулярность: ", modularity(spinglass))
plot(spinglass, graph, layout=layout_with_kk(graph), main="spinglass")

walktrap <- cluster_walktrap(graph)
cat("Модулярность: ", modularity(walktrap))
plot(walktrap, graph, layout=layout_with_kk(graph), main="walktrap")

label_prop <- cluster_label_prop(graph)
cat("Модулярность: ", modularity(label_prop))
plot(label_prop, graph, layout=layout_with_kk(graph), main="walktrap")
```

**№13** Для каждых пар событий определить отношение шансов. Сделать
выводы.


```r
count_0 <- 0
count_Inf <- 0
for(k in 1:(h-1)){
    for(l in (k+1):h){
        a <-  ceo_matr[, c(k, l)]
        res <- sum(colSums(t(a) == c(1, 1)) == ncol(a))*
        sum(colSums(t(a) == c(0, 0)) == ncol(a))/
        sum(colSums(t(a) == c(0, 1)) == ncol(a))/
        sum(colSums(t(a) == c(1, 0)) == ncol(a))   
        cat('(M', k, ', M', l, ') = ', res, '\n', sep="")
        count_0 <- count_0 + (res!=0)
        count_Inf <- count_Inf + (res==Inf)
    }
}
cat('Количество ненулевых значений:', count_0, '\n')
cat('Количество ненулевых значений (%):', count_0/(h*(h-1)/2)*100, '\n')
cat('Количество бесконечных значений:', count_Inf, '\n')
cat('Количество бесконечных значений (%):', count_Inf/(h*(h-1)/2)*100, '\n')
```

*Вывод*:  Cуществуют такие пары событий, где актор аффилирован лишь с одним из событий. Также существуют пары событий, где актор либо на аффелирован с обоими событиями, либо аффелирован хотя бы с одним из событий.
