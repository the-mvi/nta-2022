## Основные теоретические положения

**Актор** - дискретная индивидуальная или коллективная социальная единица.

**Реляционная связь** - свойство, описывабшее взаимодействие между двумя акторами.

**Группа** - совокупность всех акторов, связи между которыми должны быть изучены.

**Отношение** - совокупность связей определенного вида между членами группы.

**Социальная сеть** - объект, представляющий собой совокупность конечного набора (или наборов) акторов и отношения (отношений), определенных на них.

**Способы математического представления социальных сетей:**

1.   В виде графа.<br> Граф - это модель социальной сети $G=G(N,L)$, где $N$ - узлы, $L$ - связи. 
2.   В виде социоматрицы.<br> Социоматрица - матрица смежности $X=\{x_{ij}\}$ размеров $k\times k$, где $i,j=1,2,....,k$; $x_{ij}$ принимает значение $1$, если вершины $i$ и $j$ смежны, иначе - $0$.

**Числовые характеристики сети:**

1.   Размер - это количество акторов сети.
2.   Плотность - доля имеющихся связей по отношению к максимально возможному число связей.<br> Формулы расчета: $Δ=|L|/A^2_k$ - если связи ориентированные, $Δ=|L|/C^2_k$ - если связи неориентированные, где $L$ - множество связей, $k$ - число узлов сети.
3.   Число компонент. Компонента - это подгруппа акторов, в которой все акторы связаны друг с другом прямо или косвенно.
4.   Диаметр - это наибольшая длинна любого из кратчайших путей между всеми парами узлов сети.
5.   Степень узла - число инцидентных узлу $n_i$ связей.<br>Если связи ориентированные, то различают степень для входящих связей и степень для исходящих связей.<br>Средняя степень узлов вычисляется как $\overline d = \frac {1} {n}∑^k_{i=1}d(n_i)$.<br>Дисперсия степени узлов вычисляется как $S^2_D=\frac {1} {n}∑^k_{i=1}(d(n_i)-\overline d)^2$.
6.   Коэффициент кластеризации - доля закрытых треугольников (т.е. триад, где присутствую все 3 связи) по отношении к общему количеству закрытых или открытых треугольников (т.е. триад с 2 или 3 связями).

**Силовые алгоритмы укладки сетей:**

1.   Алгоритм Фрюхтермана-Рефнгольда.<br>Идея данного алгоритма заключается в том, чтобы рассматривать узлы как атомные частицы или небесные тела, между которыми действуют силы притяжения и отталкивания. Вводятся следующие силы: $F_a(d) = \frac {d^2} {k}$ - сила притяжения, $F_r(d) = \frac {k^2} {d}$ - сила отталкивания, где $d$ - расстояние между узлами, $k=const*\sqrt {\frac {area}{|V|}}$. В начале работы алгоритма узлы случайным образом располагаются в пространстве. Затем на каждой итерации силы притяжения и отталкивания смещаются, причем на это смещение также влияет параметр температуры $t$, значение которого начинается с $0.1*area$ и линейно уменьшается до 0.
2.   Алгоритм Камада-Каваи. <br>Идея данного алгоритма состоит в том, что узлы соединяются пружинами, и задача сводится к тому, чтобы минимальная энергия пружинной системы соответствовала минимальной разнице между геометрическим расстоянием и графовым расстоянием между узлами (т.е. найти минимум энергии системы). Вводим сила пружины как $k_{ij}=\frac {const} {d^2_{ij}}$, где $d_{ij}$ - кратчайшее растояние пути между вершинами. Фиксируется некоторое число $ϵ$. На каждой итерации выбирается вершина, которая имеет максимальное значение $Δ_m=\sqrt {(\frac {δE} {δx_m})^2+(\frac {δE} {δy_m})^2}$, где $E$ - энергия пружинной системы, $(x_m,y_m)$ - координаты вершины. Далее нужно решить систему дифференциальных уравнений, благодря которой будет найдено смещение. Это будет продолжаться до тех пор, пока $Δ_m > ϵ$.

**Актор** считается **важным**, если его связи делают его видимым для других участников сети.

**Меры важности:**

**Центральность** - мера, характеризующая актора с большим количеством прямых или косвенных связей.

**Централизация** - мера, характеризующая группу акторов.

Центральность по степени вычисляется как $C_{D}(n_i)=deg(n_i)$.<br>
Нормированная центральность по степени вычислается как $C_{D}^{'}(n_i)=\frac {deg(n_i)} {k-1}$.<br>
Централизация по степени вычисляется как $C_{D}=\frac {∑_{i=1}^k[C_D(n^*) - C_d(n_i)]} {max ∑_{i=1}^k[C_D(n^*) - C_D(n_i)]}=\frac {∑_{i=1}^k[C_D(n^*) - C_d(n_i)]} {(k-1)(k-2)}$.<br><br>


Центральность по близости вычисляется как $C_{C}(n_i)=(∑_{i=1}^kd(n_i,n_j))^{-1}$.<br>
Нормированная центральность по близости вычислается как $C_{C}^{'}(n_i)=(k-1)C_{C}(n_i)$.<br>
Централизация по близости вычисляется как $C_{C}=\frac {∑_{i=1}^k[C_C^{'}(n^*) - C_C^{'}(n_i)]} {(k-1)(k-2)/(2k-3)}$.<br><br>


Центральность по посредничеству вычисляется как $C_{B}(n_i)=∑_{j<k}g_{jk}(n_i)/g_{jk}$, $g_{jk}(n_i)$ - количество геодезических расстояний между $j$ и $k$, включающие $i$ $(j,k!=i)$, $g_{jk}$ - геодезическое расстояние между $j$ и $k$.<br>
Нормированная центральность по посредничеству вычислается как $C_{B}^{'}(n_i)=C_{B}(n_i)/C_{g-2}^k$.<br>
Централизация по посредничеству вычисляется как $C_{B}=\frac {2∑_{i=1}^k[C_B(n^*) - C_B(n_i)]} {(k-1)^2(k-2)}=\frac {∑_{i=1}^k[C_B^{'}(n^*) - C_B^{'}(n_i)]} {(k-1)}$.<br><br>


Центральность по информации вычисляется как $C_{I}(n_i) = (c_{ii} + (T - 2R)/k)^{-1}$, где $T=∑_{i=1}^{k}c_{ii}$, $R=∑_{j=1}^{k}c_{ij}$, $C=A^{-1}$, $A$ - матрица ${\{a_{ij}\}}$ размером $k\times k$ со значениями $a_{ii} = 1 + deg(n_i)$, $a_{ij} = 1 - x_{ij}$ ($x_{ij}$ - элемент матрицы смежности) если $i$ и $j$ - инцидентны, иначе - 1.<br>
Нормированная центральность по информации вычислается как $C_{I}^{'}(n_i)=\frac{C_{I}(n_i)}{\sum_{i=1}^{k}C_{I}(n_i)}$.<br>
Централизация по информации вычисляется как $S_I^2=\frac {1} {k}∑_{i=1}^k[C_I^{'}(n_i) - \overline {C_I}]$, где $\overline {C_I} = ∑C_I(n_i)$.<br><br>


**Престиж** - мера, характеризующая актома с наибольшим числом входных связей.

Престиж по степени вычисляется как $P_D(n_i)=d_I(n_i)$, где $d_I(n_i)$ - входящая степень вершины.<br>
Нормированный престиж по степени вычисляется как $P_D^{'}(n_i)=d_I(n_i)/(k-1)$.<br>
Нормированный престиж по степени вычисляется как $P_D(n_i)^{'}=P_D(n_i)/(k-1)$.<br>

**Область влияния актора $I_i$** - это множество акторов, прямо или косвенно связанных с указанным актором.

Престиж по близости вычисляется как $P_P(n_i)=\frac {|I_i|/(k-1)} {∑_{j: n_j∈I_i}d(n_j,n_i)/|I_i|}$, где $d(n_j,n_i)$ - расстояние от $n_j$ до $n_i$.

9. Заполнение таблицы для различных мер центральности (для изолированных вершин меры центральности вычислены не будут):

| $$n_i$$ | $$C'_D(n_i)$$ | $$C'_C(n_i)$$ | $$C'_B(n_i)$$ |  $$C'_I(n_i)$$ |
|:-----:|:-----:|---|---|---|
|       |       |   |   |   |



```r
try(detach(package:statnet), silent = TRUE)
suppressMessages(library(igraph))

dt <- data[which(rowSums(data) > 0), which(colSums(data) > 0)]
mg_without_lonely <- graph.adjacency(dt,  mode="undirected")
size <- ncol(dt)

#центральности и нормированная централизация по степеням
Cd <- centr_degree(mg_without_lonely, normalized = TRUE)
#центральности и нормированная централизация по близости
Cc <- centr_clo(mg_without_lonely,normalized = TRUE)
#центральности и нормированная централизация по посреднечеству
Cb <- centr_betw(mg_without_lonely, directed = FALSE,normalized = TRUE)

try(detach(package:igraph), silent = TRUE)
suppressMessages(library(statnet))
#центральности и нормированная централизация по информации
mn_without_vert <- network(dt, matrix.type = "adjacency", directed = FALSE)
Ci <- infocent(mn_without_vert, gmode = "graph")

centrality <- data.frame(n=colnames(dt), Cdn=Cd$res/(size-1), Ccn=Cc$res, Cbn=Cb$res*2/((size-1)*(size-2)), Cin=Ci/sum(Ci))
centrality
```

Выводы:<br>
Согласно всем мерам центральности - самой важной семьей во Флоренции является Medici (т.е. имеют наибольшее число прямых или косвенных связей с остальными семьями)

```r
centralization <- data.frame(Cd=Cd$centralization, Cc=Cc$centralization, Cb=Cb$centralization, Ci=(1/size)*sum((centrality$Cin-(1/size)*sum(Ci))^2))
rownames(centralization) = "value"
centralization
```

11.  Выявление точек сочленения и мостов сети с помощью соответствующих функций. Построиение графика сети, на котором выделены точки сочленения и мосты.

```r
try(detach(package:statnet), silent = TRUE)
suppressMessages(library(igraph))

#выделение точек сочленения
ap <- articulation_points(merital_graph)
cat("Сочленения:")
ap
#выделение мостов
bridge <- bridges(merital_graph)
cat("Мосты:")
bridge
```

```r
Pd <- prestige(friendship_network, gmode="digraph", cmode="indegree")

#находим геодезические расстояния
gd <- geodist(friendship_network)
#убираем строки вершин, из которых нет исходящих связей
#тогда сумма значений столбцов - будет суммой геодезических 
#расстояний от акторов, входящих в область влияния i-го актора,
#до самого i-го актора
gd <- gd$gdist[which(rowSums(gd$gdist) != Inf), ]
I <- nrow(gd)

#посчитали по формулке
Pp <- (I^2)/((s-1)*colSums(gd))

prest <- data.frame(n=network.vertex.names(friendship_network), Pd=Pd/(s-1), Pp=Pp)
prest
```
