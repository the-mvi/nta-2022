```r
nclan <- function(g,n,ncliques){
    E(g)$weight <- 1 #just in case g has weights - does not modify original graph
    n.cand <- ncliques[[n]] #n-clique candidates to be an n-clan
    n.clan <- list() #initializes a list to store the n-clans
    n.clan.i <- 1 #initializes a list pointer
    for (n.cand.i in 1:length(n.cand)){ #loop over all of the candidates
        g.n.cand <- induced_subgraph(g,n.cand[[n.cand.i]]) #get the subgraph
        if (diameter(g.n.cand)<=n){ #check diameter of the subgraph
          n.clan[[n.clan.i]] <- n.cand[[n.cand.i]] #add n-clan to the list
          n.clan.i <- n.clan.i+1 #increment list pointer
        }
      }
    return(n.clan) #return the entire list
}
nclans <- function(g){
    g <- as.undirected(g)
    ncliques <- kCliques(ugraph(igraph.to.graphNEL(g))) #get cliques
    result <- c()
    for (i in names(cl)){
        result[[i]] <- nclan(g, i, ncliques)
    }
    return(result)
}

kplexes <- function(g) {
    result <- c()
    for (i in 2:(length(V(g)))) {
        for (j in 1:(i-1)) {
            plex <- kplex(g, j, i)
            if(length(plex) > 0) {
                result[[paste(toString(i), toString(j), "plex", sep="_")]] <- plex
            }
        }
    }
    return(result)
}

kplex <- function(g,k,m){
    g.sym <- as.undirected(g)
    #–ø–æ–ª—É—á–∞–µ–º –ø–æ–¥–≥—Ä–∞—Ñ, —Å—Ç–µ–ø–µ–Ω–∏ –≤–µ—Ä—à–∏–Ω –∫–æ—Ç–æ—Ä—ã—Ö >= m-k
    k.plex <- list()
    g.sym.degmk <- induced_subgraph(g.sym,igraph::degree(g.sym)>=(m-k))
    if(length(V(g.sym.degmk)) < m) {
        return(k.plex)
    } 
    k.cand <- combn(V(g.sym.degmk)$name,m) #all candidate combinations with m members
    k.plex <- list() 
    k.plex.i <- 1 
    for (k.cand.i in 1:dim(k.cand)[2]){ 
        g.k.cand <- induced_subgraph(g.sym.degmk,k.cand[,k.cand.i]) 
        if (min(igraph::degree(g.k.cand))>=(m-k)){ #if minimum degree of sugraph is > m=k, k-plex!
            k.plex[[k.plex.i]] <- k.cand[,k.cand.i] 
            k.plex.i <- k.plex.i+1 
        }
    }
    return(k.plex) #return the entire list
}

kkernel <- function(g, k) {
    g.sym <- as.undirected(g)
    k.kernel <- list()
    g.sym.degk <- induced_subgraph(g.sym,igraph::degree(g.sym)>=k)
    if(length(V(g.sym.degk)) < k) {
        return(k.kernel)
    } 
    flag <- 0
    for (j in length(V(g.sym.degk)):1){
        if(flag) {
            break
        }
        k.cand <- combn(V(g.sym.degk)$name,j) 
        k.kernel <- list() 
        k.kernel.i <- 1 
        for (k.cand.i in 1:dim(k.cand)[2]){ 
            g.k.cand <- induced_subgraph(g.sym.degk,k.cand[,k.cand.i]) 
            if (min(igraph::degree(g.k.cand))>=k){ #if minimum degree of sugraph is > k, k-kernel!
                k.kernel[[k.kernel.i]] <- k.cand[,k.cand.i] 
                k.kernel.i <- k.kernel.i+1 
                flag <- 1
            }
        }
    }
    return(k.kernel) #return the entire list
}

kkernels <- function(g) {
    result <- c()
    for (k in 1:length(V(g))) {
        kernel <- kkernel(g, k)
        if(length(kernel) > 0) {
            result[[paste(toString(k), "kernel", sep="_")]] <- kernel
        }
    }
    return(result)
}

#n-–∫–ª–∏–∫–∏
cat("n-cliques")
cl <- kCliques(igraph.to.graphNEL(business_graph))
cl

#n-–∫–ª–∞–Ω—ã
cat("n-cliques => n-clans")
ncl <- nclans(business_graph)
ncl

#k-–ø–ª–µ–∫—Å—ã
cat("i_j_plex => i - –º–æ—â–Ω–æ—Å—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –≤–µ—Ä—à–∏–Ω –ø–æ–¥–≥—Ä–∞—Ñ–∞, j - k")
kpl <- kplexes(without_alone_graph)
kpl
         
#k-—è–¥—Ä–∞
cat("k-kernels")
kke <- kkernels(without_alone_graph)
kke
```

8. –í—ã—á–∏—Å–ª–∏—Ç—å —Å—Ç–µ–ø–µ–Ω—å —É—á–∞—Å—Ç–∏—è –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∞–∫—Ç–æ—Ä–∞, –∞ —Ç–∞–∫–∂–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π, —Å –∫–æ—Ç–æ—Ä—ã–º–∏ –∞–∫—Ç–æ—Ä –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```r
degreeOfPart<-compl[row(compl)==col(compl)]
for(i in 1:length(degreeOfPart)){
    cat("–ê–∫—Ç–æ—Ä", i, "—É—á–∞–≤—Å—Ç–≤—É–µ—Ç –≤ ", degreeOfPart[i], "—Å–æ–±—ã—Ç–∏—è—Ö\n")
}

meanDegree<-sum(degreeOfPart)/length(degreeOfPart)
cat("–°—Ä–µ–¥–Ω—è—è —Å—Ç–µ–ø–µ–Ω—å —É—á–∞—Å—Ç–∏—è: ", meanDegree)
```

9. –í—ã—á–∏—Å–ª–∏—Ç—å —Ä–∞–∑–º–µ—Ä –∏ —Å—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–º–µ—Ä –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Å–æ–±—ã—Ç–∏—è. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```r
size<-event[row(event)==col(event)]
for(i in 1:length(size)){
    cat("–°–æ–±—ã—Ç–∏–µ", i, "–∏–º–µ–µ—Ç —Ä–∞–∑–º–µ—Ä", size[i], "\n")
}

meanSize<-sum(size)/length(size)
cat("–°—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–º–µ—Ä —Å–æ–±—ã—Ç–∏—è: ", meanSize)
```

10. –í—ã—á–∏—Å–ª–∏—Ç—å –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –¥–ª—è —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è —Å–æ—É—á–∞—Å—Ç–∏—è –∏ –æ—Ç–Ω–æ—à–µ–Ω–∏—è –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è. –ü—Ä–æ–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä—É–π—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.

```r
densityPart <- (sum(compl)-sum(diag(compl)))/(nrow(compl)*(nrow(compl)-1))
cat("–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –¥–ª—è —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è —Å–æ—É—á–∞—Å—Ç–∏—è: ", densityPart, '\n')
densityEvent <- (sum(event)-sum(diag(event)))/(nrow(event)*(nrow(event)-1))
cat("–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –¥–ª—è —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è: ",densityEvent, '\n')
```

```r
cat("–î–∏–∞–º–µ—Ç—Ä —Å–µ—Ç–∏ –∞—Ñ—Ñ–µ–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏: ", diameter(ceo_graph))
```
