## Основные теоретические положения

**Группа** - совокупность всех акторов, связи между которыми должны быть изучены.

**Подгруппа** - любое подмножество акторов и все связи между ними.

**Сплочённая подгруппа** - множество акторов, которые соеденены между собой посредством многочисленных сильных и прямых связей.

**Клика** - Максимальный полный подграф из трёх и более элементов.

**n-клика** - максимальный подграф, в котором наибольшее геодезическое расстояние между любыми двумя узлами не превышает n. (Путь может проходить по узлам вне подграфа).

**n-клан** - n-клика, в которой геодезическое расстояние между любыми двумя узлами не превышает n для путей внутри подграфа. (Путь не может проходить по узлам вне подграфа).

**n-клуб** - максимальный подграф диаметра $d\le n$.

**k-ядро** - максимальный подграф, в котором каждая вершина связана минимум с k другими вершинами этого же подграфа.

**k-плекс** - максимальный подграф содержащий $g_s$ узлов, в котором каждый узел смежен не менее чем с ($g_s-k$) узлами подграфа.

**LS-множество** - множество узлов $\mathcal{N}_s$ подграфа $\mathcal{G}_s$, если любое собственное подмножество $\mathcal{Q} \subset \mathcal{N}_s$ имеет больше линий к узлам $\mathcal{N}_s \backslash Q$, чем к $\mathcal{N} \backslash \mathcal{N}_s$.

**$\lambda$-множество** - множество узлов $\mathcal{N}_s$ подграфа $\mathcal{G}_s$, если любая пара узлов в $\mathcal{N}_s$ имеют большее линейное соединение, чем любая пара узлов, состоящая из одного узла из $\lambda$-набора и второго узла извне.

**Меры сплочённости подгрупп:**

* мера сплочённости по по Боке-Хусейну - отношение средней силы связи внутри подгруппы и средней силы связи между подгруппой и остальной группой: $M_{BH} = \frac{\frac{1}{g_s(g_s-1)}\sum_{i \in \mathcal{N}_s}\sum_{j \in \mathcal{N}_s}x_{ij}}{\frac{1}{g_s(g-g_s)}\sum_{i \in \mathcal{N}_s}\sum_{j \notin \mathcal{N}_s}x_{ij}}$
* мера Альбы - гипергеометрическая вероятность того, что наблюлаемое количество рёбер в подграфе равно q: $P(L_s=q)=(_q^L)\frac{(_{(_2^{g_s})-q}^{(_2^g)-L})}{(_{(_2^{g_s})}^{(_2^g)})}$

**Виды связности подгрупп:**

* слабо n-связные - соединены полупутём не больше n (полупуть - неориентированный путь).
* односторонне n-связные - $\exists$ путь из i в j **или** из j в i не больше n.
* сильно n-связные - $\exists$ путь из i в j **и** из j в i не больше n; дуги могут различаться.
* рекурсивно n-связные - $\exists$ путь из i в j **и** из j в i не больше n; пути должны включать одни узлы и дуги.

**Модулярность** - характеристика структуры сети, в частности степень кластеризации узлов, когда внутри кластера наблюдается много связей (высокая плотность), а между кластерами мало связей (низка плотность) - разность между долей связей внутри группы и ожидаемой долей связей, если бы связи были размещены случайно ($\in [-\frac12;1]$).

**Сеть аффилированности** - сеть, в которой участники аффилированы друг с другом на основе сотрудничества в рамках какой-то группы или совместного участия в определённом событии. Сеть состоит из двух измерений: множества акторов $\mathcal{N}=\{n_1...n_g\}$ и множества событий $\mathcal{M}=\{m_1...m_h\}$

**Представление сетей аффилированности:**

* Матрица инцидентности $A_{g\times h}$
* Двудольный граф
* Прелставление биграфа в виде социоматрицы размера $(g+h)\times (g+h)$ и вида $(^{0_g A}_{A^T 0_h})$
* Гиперграф ($\mathcal{H}=\mathcal{H}(\mathcal{A}, \mathcal{B})$, где $\mathcal{A}$ - множество объектов, $\mathcal{B}$ - множество событий)
* Двойственный гиперграф

**Одномодальные проекции**

* матрица соучастия акторов: 

  $X^{\mathcal{N}} = \{X_{ij}^{\mathcal{N}}\}: X_{ij}^{\mathcal{N}} = \sum_{k=1}^ha_{ik}a_{jk}$; 
  
  $X^{\mathcal{N}} = XX^T$

* матрица перекрытия событий

  $X^{\mathcal{M}} = \{X_{kl}^{\mathcal{M}}\}: X_{kl}^{\mathcal{M}} = \sum_{i=1}^ga_{ik}a_{il}$; 

  $X^{\mathcal{M}} = X^TX$

**Сепень участия** - количество событий, с которыми связан каждый актор: $a_{i+} = \sum_{j=1}^h a_{ij} = x_{ii}^\mathcal{N}$

**Средняя степень участия** - $\bar{a_{i+}} = \frac1g \sum_{i=1}^g x_{ii}^\mathcal{N}$

**Размер события** - $a_{+j} = \sum_{i=1}^g a_{ij} = x_{jj}^\mathcal{M}$

**Средний размер события** = $\bar{a_{+j}} = \frac1h \sum_{j=1}^h x_{jj}^\mathcal{M}$

**Свойства одномодальных сетей:**

* Количество перекрывающих связей между событиями частично зависит от количества событий, к которым принадлежат акторы.
* Количество связей соучастия между акторами частично зависит от размера событий.
* Актор, принадлежащий $a_{i+}$ событиям, создаёт $(_2^{a_{i+}})$ парных связей между событиями.
* Событие с участием $a_{+j}$ участников создаёт $(_2^{a_{+j}})$ связей между парами акторов.

**Плотность:**

* для отношения соучастия: $\Delta_{(\mathcal{N})} = \frac1{g(g-1)} \sum_{i=1}^g\sum_{j=1; j\ne i}^g x_{ij}^\mathcal{N}$ - $\in[0; h]$
* для отношения перекрытия: $\Delta_{(\mathcal{M})} = \frac1{h(h-1)} \sum_{k=1}^h\sum_{l=1; l\ne k}^h x_{kl}^\mathcal{M}$ - $\in[0; g]$

**Связная сеть аффилированности** - аффилированная сеть, в которой все пары узлов достижимы.

**Диаметр аффилированной сети** - длина самого длинного пути между любой парой актор и/или событий.

**Клика уровня c** в взвешенном графе - максимальный полный подграф из ьрёх и более узлов, все из которых являются смежными на уровне c (медлу всеми парами узлов есть рёбра со значениями $\ge c$).

**Клика уровня c** для отношений соучастия - подграф, где все пары акторов совметсно участвуют не менее чем в c событиях.

**Клика уровня c** для отношений перекрытия - подграф, где все пары событий имеет не менее c общих членов.

**Отношение шансов** - мера перекрытия между событиями, которая не зависит от размера событий: $\Theta_{kl}=\frac{x_{kl}^\mathcal{M} x_{\bar{kl}}^\mathcal{M}}{x_{k\bar{l}}^\mathcal{M}x_{\bar{k}l}^\mathcal{M}}$

№1. Из файла `florentine_marital.csv` была извлечена социоматрица (в виде таблицы) с помощью функции `read.csv`. Полученная таблица была преобразована в социоматрицу и выведена на экран. Сеть была отображена с помощью алгоритма укладки Фрюхтермана–Рейнгольда. Для этого сначала по социоматрице был создан объект ненаправленного графа, а затем при отрисовке был задан соответсвующий алгоритм укладки.

```r
library(igraph)

data <- read.csv('./florentine_marital.csv', comment.char='#')
rownames(data) <- colnames(data)
sociomatrix <- as.matrix(data)
sociomatrix

graph <- graph.adjacency(sociomatrix, mode="undirected")
plot(graph, layout=layout_with_fr(graph), main="Aлгоритм укладки Фрюхтермана–Рейнгольда")
```

№2. Сеть была исследована на предмет подгрупп с помощью n-клик, n-кланов и k-ядер.

**Вывод**: Вся сеть (исключая изолированный элемент) представляет из себя 5-клику, из неё можно выделить подмножества 4- и 3-клик, причём в каждой полученной n-клике будет достаточно много акторов (примерно по десять акторов). Это показывает, что сеть семейств имеет тесные связи друг с другом, но всё равно может быть разделена на отдельные большие группы, где это связь ещё более сильна. Ситуация с n-кланами похожая, но, так как само определение n-кланов более жёсткое (пути не должны проходить вне подгрупп), то количество клик и участников в них получается ниже. Вся сеть представляет из себя 0-ядро (1-ядро, если убрать изолированный элемент), но есть подгруппа из 10-ти элементов, которая представляет из себя 2-ядро, то есть в этой группе семей каждая имеет связи минимум с другими двумя семьями из этой подгруппы.

```r
library(RBGL)

# n-клики
res <- kCliques(igraph.to.graphNEL(graph))
res

# n-кланы
ncliques <- res
res <- c()
for (n in names(ncliques)){
    cand <- ncliques[[n]]
    clan <- list()
    i <- 1
    for (cand.i in 1:length(cand)){
        g.n.cand <- induced_subgraph(graph,cand[[cand.i]])
        if (diameter(g.n.cand) <= n){ 
          clan[[i]] <- cand[[cand.i]]
          i <- i+1
        }
      }
    res[[n]] <- clan
}
names(res) <- paste(1:length(names(ncliques)), 'clan', sep='-')
res

# k-ядра
info <- coreness(graph, mode= "all")
res <- lapply(0:max(as.integer(info)), function(x) names(info)[which(as.integer(info) >= x)])
names(res) <- paste(1:length(res)-1, 'core', sep='-')
res
```

№3. С помощью библиотеки igraph были исследованы алгоритмы обноружения сообществ InfoMAP, Louvain и Optimal. 

Алгоритм Infomap пытается минимизировать функцию стоимости. Разделение основано на потоке, вызванном шаблоном соединений в данной сети.

Алгоритм Louvain основан на мере модульности и иерархическом подходе. Изначально каждая вершина назначается сообществу самостоятельно. На каждом этапе вершины жадно переназначаются сообществам: каждая вершина перемещается в сообщество, с которым она достигает наибольшего вклада в модульность. Когда ни одна вершина не может быть переназначена, каждое сообщество считается вершиной само по себе, и процесс начинается снова с объединенных сообществ. Процесс останавливается, когда остается только одна вершина или когда модульность больше не может быть увеличена за шаг.

Алгоритм Optimal вычисляет оптимальную структуру сообщества для графа с точки зрения максимальной оценки модулярности. Расчет выполняется путем преобразования задачи максимизации модулярности в задачу целочисленного программирования и последующего вызова библиотеки GLPK для ее решения.

**Результаты**

| Алгоритм            | Количество сообществ | Модулярность |
|---------------------|----------------------|--------------|
| InfoMAP             | 5                    | 0.4          |
| Louvain             | 4                    | 0.4         |
| Optimal             | 4                    | 0.4          |

**Вывод**: Не смотря на различие в методах нахождения сообществ, алгоритмы показывают схожие результаты (4-5 сообществ с показателем модулярности в 0.4). Небольшие различия в результатах также видны в наполнениях сообществ: есть группы акторов, которые в каждом методе оказываются в одном сообществе (например, "Acciaiuoli", "Medici", "Ridolfi" и "Tornabuoni"), а есть акторы, которые по результатам оказываются то в одном, то в другом сообществе (например, "Pazzi" и "Salviati").



```r
infomap <- cluster_infomap(graph)
louvain <- cluster_louvain(graph)
optimal <- cluster_optimal(graph)

plot(infomap, graph, layout=layout_with_fr(graph), main="Алгоритм InfoMAP")
plot(louvain, graph, layout=layout_with_fr(graph), main="Алгоритм Louvain")
plot(optimal, graph, layout=layout_with_fr(graph), main="Алгоритм Optimal")
```

№4. Из файла `сeo.csv` была извлечена матрица инцидентности. Столбцы были названы по номеру события, а ряды - по номеру актора.

```r
data <- read.csv('./ceo.csv', comment.char='#', header=FALSE)
incidence <- as.matrix(data)
g <- nrow(incidence)
h <- ncol(incidence)
rownames(incidence) <- paste('N', 1:g, sep='');
colnames(incidence) <- paste('M', 1:h, sep='');
incidence
```

№5. Сеть была представлена в виде матрицы биграфа размера $(g+h)\times (g+h)$ и вида $(^{0_g A}_{A^T 0_h})$. Столбцы и ряды были названы следующим образом: сначала идут номера акторов, затем номера событий. Затем полученная матрица была отображена в виде биграфа. Для корректного отображения вершины были разделены на 2 группы с разными координатами по оси X и цветами.

```r
res <- rbind(
    cbind(
        matrix(0,nrow = g,ncol = g), 
        incidence
    ), 
    cbind(
        t(incidence), 
        matrix(0,nrow = h, ncol = h)
    )
)
colnames(res) <- rownames(res) <- c(paste('N', 1:g, sep=''), paste('M', 1:h, sep=''));

bigraph <- graph.adjacency(res, mode = "undirected")
V(bigraph)$x <- c(rep(0, g), rep(1,h))
V(bigraph)$y <- c(1:g, (1:h)*1.6)
V(bigraph)$color<-c(rep('red', g), rep('yellow',h))
plot(bigraph)
```

№7. По матрице инцидентности сети аффилированности были построены ценностные матрицы соучастия акторов ($X^{\mathcal{N}} = XX^T$) и перекрытия событий ($X^{\mathcal{M}} = X^TX$). Полученные одномодальные сети были отображены в виде взвешенных графов: для этого при создании графа был поставлен флаг *weighted*, указывающий на его взвешенность, а при отображений были заданы веса рёбер по значению в ценностной матрице.

**Вывод**: Каждый директор посещает хотя бы один общий клуб с большинством других директоров (на это указывает количество ненулевых элементов в матрице).
```r
X <- incidence
X_N <- X %*% t(X)
X_N
graphX_N <- graph.adjacency(X_N, mode="undirected", weighted=TRUE)
plot(graphX_N, edge.label=E(graphX_N)$weight, layout=layout_with_kk(graphX_N), main='Сеть соучастия акторов')

X_M <- t(X) %*% X
X_M
graphX_M <- graph.adjacency(X_M, mode="undirected", weighted=TRUE)
plot(graphX_M, edge.label=E(graphX_M)$weight, layout=layout_with_kk(graphX_M), main='Сеть перекрытия событий')
```

№13. Для каждых пар событий было определено отношение шансов. Для подсчета отношения шансов для одной пары достаточно рассмотреть два соответсующих столбца из матрицы инцидентности и подсчитать количество пар (0, 0), (0, 1), (1, 0) и (1, 1) и подставтиь полученные значения в формулу.

**Вывод**: Около 62% пар клубов (65 пар из 105) имеют ненулевую меру, то есть имеют хотя бы одного общего директора ($X_{kl}^{\mathcal{M}} \ne 0$) и между собой разделяют всех директоров ($X_{\bar{kl}}^{\mathcal{M}} \ne 0$). Остальные 40 пар либо не имеют общих директоров ($X_{kl}^{\mathcal{M}} = 0$), либо есть директора, которые не посещают ни один из этих двух клубов ($X_{\bar{kl}}^{\mathcal{M}} = 0$). Около 7% пар (7 пар) имеют бесконечную меру, то есть список директоров одного из клубов не перекрывает списка директоров другого клуба ($X_{\bar{k}l}^{\mathcal{M}} \ne 0$ или $X_{k\bar{l}}^{\mathcal{M}} \ne 0$).

```r
countZero <- 0
countInf <- 0
for(k in 1:(h-1)){
    for(l in (k+1):h){
        a <- X[, c(k, l)]
        res <- sum(colSums(t(a) == c(1, 1)) == ncol(a))*
        sum(colSums(t(a) == c(0, 0)) == ncol(a))/
        sum(colSums(t(a) == c(0, 1)) == ncol(a))/
        sum(colSums(t(a) == c(1, 0)) == ncol(a))   
        cat('Theta(M', k, ', M', l, ') = ', res, '\n', sep="")
        countZero <- countZero + (res!=0)
        countInf <- countInf + (res==Inf)
    }
}
cat('Количество ненулевых значений:', countZero, '\n')
cat('Количество ненулевых значений (%):', countZero/(h*(h-1)/2)*100, '\n')
cat('Количество бесконечных значений:', countInf, '\n')
cat('Количество бесконечных значений (%):', countInf/(h*(h-1)/2)*100, '\n')
```
