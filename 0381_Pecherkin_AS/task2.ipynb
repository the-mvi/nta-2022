## Основные теоретические положения
Подграфом $G_S$ называется граф $G_S=G_S(\mu_S,L_S)$ состоящий из подмножеств узлов $\mu_S \le \mu$ и подмножеств линий(ребер) $l_s \le l$, связывающие узлы в $G_S$

N-клика - максимальный подграф, в котором наибольшее геодезическое расстояние между любыми двумя узлами не превышает n
$d_{i, j}\le n$

N - клан - n-клика в которой геодезические расстояния не превышают n для путей внутри подграфа.
N-клуб максимальный подграф диаметра $/le n$

K-ядро это максимальный подграф в котором каждая вершина связана минимум с k другими вершинами этого же подграфа

K-плекс - это максимальный подграф содержащий $g_s$ узлов, в котором каждый узел смещен не менее, чем с $(g_s - k)$ узлами подграфа.
$d_s(i) \ge g_s - k$
где $d_s(i)$ - степень узла $n_i$ в подграфе $G_s$

Модулярность - характеристика структуры сети в частности степень кластеризации узлов, когда внутри кластеров сосредоточено много узлов, а между кластерами - мало узлов
Показатель модулярность определяется как разница между ценой связей внутри группы и ожидаемой ценой связей, если бы связи были размещены случайно.

Сеть аффилированности - сеть, в которой участники аффилированы друг с другом на основе сотрудничества в рамках какой-то группы или совместного участия в определенном событии. Сеть состоит из двух измерений.

Матрица инцидентности - Матрица $A = \{a_{ij}\}$ размерности $g \times h$

$a_{ij}=
\begin{cases} 
1, если актор i связан с актором j\\ 
0, иначе
\end{cases}$

Двудольный граф - граф, вершины которого можно разделить на 2 подмножества, а все ребра соединяют узлы из разных подмножеств.

Гиперграф - обобщение понятия графа, состоит из множества объектов и множества подмножеств объектов, котором каждый объект принадлежит хотя бы одному подмножества хотя бы одному подмножеству , причем ни одно подмножество не является пустым.
$H=H(A,B)$,
где $A={a_1, a_2, ...,a_g}$
$B={b_1, b_2, ..., b_3}$

Двойственный гиперграф для гиперграфа $H=H(A, B)$

$H^*=H(B, A)$,
где $A={a_1, a_2, ...,a_g}$
$B={b_1, b_2, ..., b_3}$
Матрица соучастия акторов

$X^N=\{X_{ij}^N\}$

$X_{ij}^N=\sum_{k=1}^b a_{ik}a_{jk}$

$X^N = XX^T$

Матрица перекрытия событий

$X^M=\{x_{kl}^M\}$

$X_{kl}^M=\sum_{i=1}^g a_{ik}a_{ji}$

$X^M = X^TX$

Степень участия - это количество событий, с которыми связан каждый актор

$a_{i+}=\sum_{j=1}^n a_ij = X_{ii}^N$

Среднее количество событий, с которыми акторы аффилированы

$\bar{a_{i+}}=\frac1g\sum_{i=1}^g\sum_{j=1}^ha_{ii}=\frac{a_{++}}{g}=\frac1g\sum_{i=1}^g x_{ii}^N$

Размер событий

$a_{+j}=\sum_{i=1}^g a_ij = x_{jj}^M$

Средний размер события - среднее количество акторов в каждом событии или средняя степень узлов

Плотность для отношения соучастия

$\Delta(N)= \frac{1}{g(g-1)}\sum_{i=1}^g\sum_{j=1, i \neq j}^gx_{ij}^N$

Плотность для отношения перекрытия

$\Delta(M)= \frac{1}{h(h-1)}\sum_{i=1}^h\sum_{j=1, i \neq j}^hx_{kl}^M$

Диаметр сети аффилированности - это длина самого длинного пути между любой парой акторов и/или событий
Клика уровня c в взвешенном графе - максимальный полный подграф из трёх или более узлов, все из которых являются смежными на уровне c, т. е. между всеми парами узлов есть ребра с значением $\ge c$

Клика уровня c для отношения соучастия - подграф в котором все пары акторов совместно участвуют не менее чем в c событиях.

Клика уровня c для отношения перекрытия - подграф в котором все пары событий имеют не менее c общих членов.

Отношение шансов
$X_{kl}^M$- число акторов, которые аффилированы с событиями k и l.

$X_{\bar{kl}}^M$ - число акторов, которые не принадлежат ни одному событию.

$X_{k\bar{l}}^M$ - число акторов, которые аффилированы с событием k, но не c событием l

$X_{\bar{k}l}^M$ - число акторов, которые аффилированы с событием l, но не с событием k.

Мера перекрытия между событиями, которая не зависит от размера сети, называется отношением шансов

$\theta=\frac{x_{kl}/x_{\bar{k}l}}{x_{l\bar{l}}/x_{\bar{k}l}}=\frac{x_{kl}x_{\bar{k}l}}{x_{l\bar{l}}x_{\bar{k}l}}$

```r
nclique <- function(net,n){
  net <- connect(net, order=n)
  return(max_cliques(net,min=3))
}

clans<- function(net,n) {
    cliques<-nclique(net,n)
    clansarray<-c()
    for(cl in cliques) {
        sub<-subgraph(net,names(cl))
        if(diameter(sub) <= n) {
           cat('(',names(cl),")\n")
        }    
    }
    return (clansarray)
}

clubs<- function(net,n) {
    cliques<-nclique(net,n)
    clansarray<-c()
    for(cl in cliques) {
        sub<-subgraph(net,names(cl))
        if(diameter(sub) <= n) {
           cat('(',names(cl),")\n")
        }    
    }
    return (clansarray)
}

core <- function(net,n) {
    print(V(subgraph(net,degree(net)>=n)))
}

plex <- function(net,n) {
    print(V(subgraph(net,degree(net)>=n)))
}

print("1-cliques")
lapply(nclique(network,1), function(x) names(x))
print("2-cliques")
lapply(nclique(network,2), function(x) names(x))
print("3-cliques")
lapply(nclique(network,3), function(x) names(x))
print("4-cliques")
lapply(nclique(network,4), function(x) names(x))
print("5-cliques")
lapply(nclique(network,5), function(x) names(x))
print("6-cliques")
lapply(nclique(network,6), function(x) names(x))
 
print("1-clans")
lapply(clans(network,1), function(x) names(x))
print("2-clans")
lapply(clans(network,2), function(x) names(x))
print("3-clans")
lapply(clans(network,3), function(x) names(x))
print("4-clans")
lapply(clans(network,4), function(x) names(x))
print("5-clans")
lapply(clans(network,5), function(x) names(x))
print("6-clans")
lapply(clans(network,6), function(x) names(x))    
       
print("1-clubs")
lapply(clubs(network,1), function(x) names(x))
print("2-clubs")
lapply(clubs(network,2), function(x) names(x))
print("3-clubs")
lapply(clubs(network,3), function(x) names(x))
print("4-clubs")
lapply(clubs(network,4), function(x) names(x))
print("5-clubs")
lapply(clubs(network,5), function(x) names(x))
print("6-clubs")
lapply(clubs(network,6), function(x) names(x))    
  
print("0-core")
core(network,0)
print("1-core")
core(network,1)  
print("2-core")
core(network,2)  
print("3-core")
core(network,3)  
print("4-core")
core(network,4)  
print("5-core")
core(network,5)  
print("6-core")
core(network,6)  
       
       
print("0-plex")
plex(network,0)
print("1-plex")
plex(network,1)  
print("2-plex")
plex(network,2)  
print("3-plex")
plex(network,3)  
print("4-plex")
plex(network,4)  
print("5-plex")
plex(network,5)  
print("6-plex")
plex(network,6)         
     
    
```

Label propagation. Этот алгоритм использует концепцию соседства узла и распространение информации в сети. Изначально каждый узел помечен уникальным значением. Затем происходит итеративный процесс, в котором каждый узел получает метку, которая является наиболее распространенной в его окрестностях. Этот процесс продолжается до тех пор, пока метки не перестануют изменяться. Полученные сообщества определяются последней меткой.

Spinglass. Этот алгоритм является методом оптимизации, основанным на аналогии между статистической механикой сложной сети и физической моделью Spinglass.

Этот алгоритм использует иерархическое объединение. Здесь расстояние между двумя узлами определяется с точки зрения процесса случайного блуждания. Основная идея заключается в том, что если два узла находятся в одном сообществе, вероятность попасть в третий узел k, расположенный в том же сообществе, через случайное блуждание не должно сильно отличаться для i и j. Расстояние строится по суммируя эти различия по всем узлам с поправкой на степень.

13. Для каждых пар событий определить отношение шансов. Сделать выводы.

```r
kl<-matrix(0,15,15)
nkl<-matrix(0,15,15)
knl<-matrix(0,15,15)
nknl<-matrix(0,15,15)
for(i in 1:15) {
    for(j in 1:15) {
        kl[i,j] <- sum((data[,i]==1) & (data[,j]==1))
        nkl[i,j] <- sum((data[,i]!=1) & (data[,j]==1))
        knl[i,j] <- sum((data[,i]==1) & (data[,j]!=1))
        nknl[i,j] <- sum((data[,i]!=1) & (data[,j]!=1))
    }
 }
res<-kl*nknl/(nkl*knl)
for(i in 1:15) {
    for(j in (i+1):15) {
        cat('k=',i,'l=',j,'theta=',res[i,j],'\n')
    }
 }
```

Нулевые значения говорят о том, что события исключают друг друга.
У некоторых пар событий значение равно Inf, что говорит о том, что если актор попадает в одно событие, он точно попадет и во второе.
