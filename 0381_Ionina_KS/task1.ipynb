## Основные теоретические положения

**Акторы** - дискретные индивидуальные, корпоративные или коллективные социальные единицы.

**Реляционная связь** - свойство, описывающее взаимодействие между двумя акторами.

**Подгруппа** - любое подмножество акторов и все связи между ними.

**Группа** - совокупность всех акторов, связи между которыми должны быть изучены.

**Социальная сеть** - объект, который представляет собой совокупность конечного набора (наборов) аквторов и отношения (отношений), определенных на них.

**Математическое представление сетей.**

**1. Графовое.**

Граф - модель социальной сети с ненаправленным дихотомическим отношением. (т.е. связь илбо присутствует либо отсутствует).

$$Z=(N,L)$$

$$N=[n_1, n_2, n_3,..., n_n]-\;набор  \;  узлов.$$ 
$$L=[l_1, l_2, l_3,..., l_n]-\;набор  \;  линий.$$ 

Представление связей:
$$l=<n_i, n_j>- \;направленная  \; связь.$$ 
$$l=(n_i, n_j)- \;ненаправленная  \; связь.$$ 


**2. Социометрическое.**

Социоматрица - матрица X размерности g*g, описывающая смежные узлы сети.

Матрица инцидентности -
матрица размерности g*L, описывающая какие линии каким узлам инциденты.


**Числовые характеристики сети.**


*Размер сети* - количество акторов сети.


*Плотность* - доля имеющихся связей по отношению к максимально возможному количеству связей.

Для ориентированных сетей:
$$\Delta=\frac{L}{A\binom{2}{g}}$$

Для неориентированных сетей:
$$\Delta=\frac{L}{\binom{g}{2}}$$

где L- имеющиеся связи в сети.
*Компонента сети* - подгруппа, в которой все акторы связаны друг с другом прямо или косвенно.

*Диаметр полной сети* - наибольшая длина любого из кратчайших путей между всеми парами узлов.

*Степень узла* - число прямых, инцидентных данному узлу. Для ориентированных делится на входящую стпень и исходящую степень.

Средняя степень узлов:
$$\overline{d}=\frac{1}{g}\sum_{i=1}^{g} d(n_i)$$
Дисперсия степеней узлов:
$$S_{D}^{2}=\frac{1}{g}\sum_{i=1}^{g} (d(n_i) - \overline{d})^2$$

**Мост** — ребро, удаление которого увеличивает число компонент связности.

**Точка сочленения** - вершина графа, при удалении которой количество компонент связности возрастает.

**Центральность** - мера, характеризующая актор с большим количеством связей (прямых/косвенных)
$$C_{A}(n_{i}), i=1,...g$$

**Престиж** - мера, характеризующая актор с большим количеством входящих связей
$$P_{A}(n_{i}), i=1...g$$

**Центральность по степени** - степень каждого актора.
$$C_{D}(n_{i})=d(n_{i})$$

**Нормированная центральность по степени**:
$$C'_{D}(n_{i})=\frac{d(n_{i})}{g-1}$$

**Индекс централизации**:
$$C_{A}=\frac{\sum_{i=1}^{g}[C_{A}(n^{*} - C_{A}(n_{i}]}{max \sum_{i=1}^{g}[C_{A}(n^{*} - C_{A}(N_{i}]}$$

**Централизация по степени**:
$$C_{D}=\frac{\sum_{i=1}^{g}[C_{D}(n^{*} - C_{D}(n_{i}]}{(g-1)(g-2)}$$

**Центральность по близости** - более важными узлами считаются узлы, наиболее близкие к остальным узлам, если:
$$C_{C}(n_{i})=(\sum_{i=1}^{g}{d(n_{i}, n_{j}})^{-1},$$


**Нормированная центральность по близости**:
$$C_{C}'(n_{i})=\frac{g-1}{\sum_{i=1}^{g}{d(n_{i}, n_{j})}},$$

**Централизация по близости**:
$$C_{C}=\frac{\sum_{i=1}^{g}[C_{C}'(n^{*}) - C_{C}'(n_{i})]}{(g-1)(g-2)/(2g-3)}$$

**Центральность по посредничеству** характеризует, насколько важную роль данный узел играет на пути между парами других узлов сети, в том смысле, что пути между другими узлами должны проходить через данный узел.
$$C_{B}(n_{i})=\sum_{j<k}{g_{jk}(n_{i})/g_{jk}}$$

**Нормированная центральность по посредничеству**:
$$C_{B}'(n_{i})=\frac{C_{B}(n_{i})}{\binom{g-1}{2}},$$

**Централизация по посредничеству**:
$$C_{B}=\frac{\sum_{i=1}^{g}[C_{B}'(n^{*} - C_{B}'(n_{i}]}{(g-1)}$$

**Центральность по информации** показывает, сколько информации содержится в путях, которые начинаются и заканчиваются у определенного актора.

**Централизация по информации**:
$$S_{I}^2=\frac{1}{g}\sum_{i=1}{g}(C_{I}^{`}(n_{i})-\overline{C_{I}})^2$$

8.Вычислить среднюю степень узлов и дисперсию. Сделать выводы.

```r
D <- function(nd_network, degr) { #дисперсия
    return (1/network.size(nd_network)*(sum((degree(nd_network, gmode="graph") - degr)^2)))
}

degr <- mean(degree(nd_network, gmode="graph")) #среднее значение степени для ненаправленной сети
print('Средняя степень узлов')
print(degr)
print('Дисперсия степеней узлов')
print(D(nd_network, degr))
```

В результате анализа направленной сети  krackhardt_friendship было выявлено, что среди 21 актора,19 находятся в одной компоненте сильной связности, т.е. имеют прямые или косвенные связи друг с другом, а еще 2 актора "изолированы" (т.е. к ним связь есть, а от них - нет). Также у акторов анализируемой сети, в среднем, количество исходящих связей превышает количество входящих.
