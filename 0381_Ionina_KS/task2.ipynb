**–î–≤—É–¥–æ–ª—å–Ω—ã–π –≥—Ä–∞—Ñ** $-$  –≥—Ä–∞—Ñ, —É–∑–ª—ã –∫–æ—Ç–æ—Ä–æ–≥–æ –º–æ–∂–Ω–æ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –Ω–∞ –¥–≤–∞ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–∞, –∞ –≤—Å–µ  —Ä–µ–±—Ä–∞ —Å–æ–µ–¥–∏–Ω—è—é—Ç —É–∑–ª—ã, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–∏–µ —Ä–∞–∑–Ω—ã–º –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–∞–º.

**–ì–∏–ø–µ—Ä–≥—Ä–∞—Ñ** $-$  –æ–±–æ–±—â–µ–Ω–∏–µ –ø–æ–Ω—è—Ç–∏—è –≥—Ä–∞—Ñ–∞, –≥–∏–ø–µ—Ä–≥—Ä–∞—Ñ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –æ–±—ä–µ–∫—Ç–æ–≤ –∏ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤ –æ–±—ä–µ–∫—Ç–æ–≤, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–∞–∂–¥—ã–π –æ–±—ä–µ–∫—Ç –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ–º—É –ø–æ–¥–Ω–æ–∂–µ—Å—Ç–≤—É, –ø—Ä–∏—á–µ–º –Ω–∏ –æ–¥–Ω–æ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–æ –Ω–µ —è–≤–ª—è–µ—Ç—Å –ø—É—Å—Ç—ã–º.

**–ú–∞—Ç—Ä–∏—Ü–∞ —Å–æ—É—á–∞—Å—Ç–∏—è –∞–∫—Ç–æ—Ä–æ–≤**:  

$X^{N}={\{x_{ij}^{N}}\}$:

$x_{ij}^{N}=\sum_{k=1}^{h}a_{ik}a_{jk}$

$x^{N}=XX^{T}$


**–ú–∞—Ç—Ä–∏—Ü–∞ –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è —Å–æ–±—ã—Ç–∏–π**:  

$X^{M}={\{x_{kl}^{M}}\}$:

$x_{kl}^{M}=\sum_{i=1}^{g}a_{ik}a_{il}$

$x^{M}=X^{T}X$

**–°—Ç–µ–ø–µ–Ω—å —Å–æ—É—á–∞—Å—Ç–∏—è** $-$ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π, —Å –∫–æ—Ç–æ—Ä—ã–º–∏ —Å–≤—è–∑–∞–Ω –∫–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä.

$a_{i+}=\sum_{j=1}^{h}a_{ij}x_{ii}^{N}$

**–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π, —Å –∫–æ—Ç–æ—Ä—ã–º–∏ –∞–∫—Ç–æ—Ä—ã –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω—ã**

$\Large \bar{a}_{i+}=\frac{1}{g}\sum_{i=1}^{g}\sum_{j=1}^{h}a_{ij}=\frac{a_{++}}{g}=\frac{1}{g}\sum_{i=1}^{g}x_{ii}^{N}$


**–†–∞–∑–º–µ—Ä —Å–æ–±—ã—Ç–∏–π**

$\Large a_{ij}=\sum_{i=1}^{g}a_{ij}=x_{jj}^{M}$


**–°—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–º–µ—Ä —Å–æ–±—ã—Ç–∏–π**  $-$ —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–æ—Ä–æ–≤ –≤ –∫–∞–∂–¥–æ–º —Å–æ–±—ã—Ç–∏–∏ –∏–ª–∏ —Å—Ä–µ–¥–Ω—è—è —Å—Ç–µ–ø–µ–Ω—å —É–∑–ª–æ–≤, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â—Ç—Ö —Å–æ–±—ã—Ç–∏—è –≤ –¥–≤—É–¥–æ–ª—å–Ω–æ–º –≥—Ä–∞—Ñ–µ.


*–°–≤–æ–π—Å—Ç–≤–∞ –æ–¥–Ω–æ–º–æ–¥–∞–ª—å–Ω—ã—Ö —Å–µ—Ç–µ–π*

*   –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–≤—è–∑–µ–π —Å–æ—É—á–∞—Å—Ç–∏—è –º–µ–∂–¥—É –∞–∫—Ç–æ—Ä–∞–º–∏ —á–∞—Å—Ç–∏—á–Ω–æ –∑–∞–≤–∏—Å—è—Ç –æ—Ç —Ä–∞–∑–º–µ—Ä–∞ —Å–æ–±—ã—Ç–∏–π. 
*   –ê–∫—Ç–æ—Ä, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—â–∏–π $a_{i+}$ —Å–æ–±—ã—Ç–∏—è–º, —Å–æ–∑–¥–∞–µ—Ç $\Large \binom{a_{i+}}{2}$ –ø–∞—Ä–Ω—ã—Ö —Å–≤—è–∑–µ–π –º–µ–∂–¥—É —Å–æ–±—ã—Ç–∏—è–º–∏.

```r
infinity <- 0
zero <- 0
for(k in 1:14){
    for(l in (k+1):15){
        y <- matr[, c(k, l)]
        res <- sum(colSums(t(y) == c(1, 1)) == ncol(y))*
        sum(colSums(t(y) == c(0, 0)) == ncol(y))/
        sum(colSums(t(y) == c(0, 1)) == ncol(y))/
        sum(colSums(t(y) == c(1, 0)) == ncol(y))   
        cat('Theta(M', k, ', M', l, ') = ', res, '\n', sep="")
        zero <- zero + (res!=0)
        infinity <- infinity + (res==Inf)
    }
}
cat('–ù–µ–Ω—É–ª–µ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:', infinity, '\n')
cat('–ù–µ–Ω—É–ª–µ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (%):',infinity/(15*14/2)*100, '\n')
cat('–ë–µ—Å–∫–æ–Ω–µ—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è:',infinity, '\n')
cat('–ë–µ—Å–∫–æ–Ω–µ—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è (%):', infinity/(15*14/2)*100, '\n')
```
