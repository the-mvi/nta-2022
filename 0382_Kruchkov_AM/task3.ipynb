## Основные теоретические положения

- Индекс взаимности Каца-Пауэлла - характеристи измеренения тенденции акторов в группу:

$$
- \inf < \rho _{KP} ≤ 1
$$

Индекс взаимности основан на ожидании количества взаимных диад, предпологая, что выбор делается акторами случайным образом

- Предполагается, что каждый актор делает d выборов из (g-1) акторов, доступных для выбора
- Вероятность взаимного выбора между любой произвольной парой акторов равна $d^2/(g-1)^2$
- Математическое ожидание взаимных диад $EM = g(g-1)/2  d^2/(g-1)^2 = gd^2 /2(g-1)$
- Вероятность того, что общая диада, в которой участвуют акторы является взаимной диадой
    
    
- Условная вероятность:
    
    $P(j \rightarrow i | i \rightarrow j) = P (j \rightarrow i ) +   \rho _{KP} P (j \not \rightarrow i )$
    
    Таким образом 
    
    $P(j \rightarrow i ^ i \rightarrow j) = d/g-1 (d/g-1 + \rho _{KP} (g-1-d)/(g-1))$
    
- Математическое ожидание взаимных диад с учётом данной вероятности:
    
    $E(M | \rho ) = gd^2 /2(g-1) (1- \rho _{KP}) + gd/2  \rho _{KP}$
    
- Оценка  $\rho _{KP}$   по методу моментов:
    
     $\hat \rho _{KP} = (2(g-1)M-gd^2)/(gd(g-1-d))$
    
- Пусть i-ый актор выбирает $d_O(n_i)=x_{i+}$ других акторов
- Пусть
- Вероятность того, что акторы имеют взаимные отношения равна $x_{i+} x{j+} / (g-1)^2$
- Математического ожидание взаимных диад:
    
    $EM = L^2 - L_2/2(g-1)^2$
    
- Математического ожидание взаимных диад с учётом вероятности:
    
    $E(M | \rho ‘) = L^2 - L_2/2(g-1) * (1-\rho_{KP}‘ + L/2 \rho_{KP}$
    
- Оценка $\rho_{KP}$ по методу моментов:
    
    $$
    \rho_{KP}' = 2(g-1)^2M - L^2 + L_2 / (L(g-1)^2 - L^2 + L_2)

- Цензус диад(dyad census) - тройка <M, A, N>, где:
    - $M = \sum_{i<j} X_{ij} X_{ji}$
    - $A = X_{++} - 2M$
    - $N = \binom{g}{2} - A - M$

## Индекс взаимности Бандиопадхьяйя (**Bandyopadhyay)**

- Данный индекс  основан на минимальном и максимальном значениях количества взаимных диад
- Значение индекса:

$$
\rho_B = \frac{M-M{min}}{M{max}-M{min}}
$$

- $M_min = max_{0≤t≤g} f(t)$
- $M_{max} = \lfloor 1/2(\sum_{i=1}^g x_{i+} -  max_{0≤t≤g} f(t)) \rfloor$
- Функция f(t)

$$
f(t) = \sum_{i=1}^t x_{i+} - t(g-1) - \binom{t}{2}$$

- Функция g(t):

$$
g(t) = \sum_{i=1}^t x_{i+}- t(t-1) - \sum_{i=t+1}^g min(t, x_{i+})$$

## Статистический анализ количества дуг

- Число дуг: $L ~ B(g(g-1), 1/2)$
- Вероятность того, что число дуг в орграфе равно I:

$$
\mathbb P (L = l) = \binom{g(g-1)}{l} 2^{g(g-1)}$$

- Математическое ожидание:

$$
\mathbb(L)=\binom{g}{2}
$$

- Дисперсия:

$$
 \mathbb D(D) = \frac{1}{2} \binom{g}{2}
$$

### Статистические гипотезы - равномерное распределение

- Нулевая гипотеза: $H_0:L \sim B(g(g-1), 1/2)$
- Статистика:

$$
z_l = \frac{I-\mathbb E(L)} {\sqrt{\mathbb D(L)}} = \frac{I-g(g-1)/2}{\sqrt{g(g-1)/4}},
$$

где $I = \sum_{i=1}^g \sum_{j=1}^g x_{ij}$

- Данная статистика является приблизительно стандартным нормальным распределением при заданном уровне значимости

### Статистические гипотезы - Бернулли

- Нулевая гипотеза:  $H_0: L \sim B(g(g-1), P_0)$
- Статистика: ,

$$
z_l = \frac{I-\mathbb E(L)} {\sqrt{\mathbb D(L)}} = \frac{I-P_0g(g-1)/2}{\sqrt{P_0(1-P_0)g(g-1)}}
$$

 где $I = \sum_{i=1}^g \sum_{j=1}^g x_{ij}$

- Данная статистика является приблизительно стандартным нормальным распределением при заданном уровне значимости

### Доверительный интервал для оценки P

- Оценка парметра P:

$$
\hat P = \frac{\sum_{i=1}^g \sum_{j=1}^g x_{ij}}{g(g-1)} = \frac{l}{g(g-1)}
$$

- Доверительный интервал:

$$
 \hat P_{lower} ≤ P ≤ \hat P_{upper}
$$

- $\hat P_{lower}  = \hat P - z_{\alpha / 2} \sqrt{\hat P ( 1- \hat P) /g(g - 1) }$
- $\hat P_{upper}  = \hat P + z_{\alpha / 2} \sqrt{\hat P ( 1- \hat P) /g(g - 1) }$
- Здесь $z_{\alpha / 2} : Ф(z_{\alpha / 2} ) = 0.5 - \alpha / 2$

### 1. 
С помощью библиотеки igraph смоделировать случайный орграф
G𝑑 на основе модели Эрдеша—Реньи: G𝑑 = G (𝑔, 𝑝), где 𝑔 – количество акторов, 𝑝 – вероятность появления дуги между акторами. В R
для модели случайного графа Эрдеша—Реньи существует функция
erdos.renyi.game(). Ознакомьтесь с данной моделью и опишите
её.

Аргументы
- n - количество вершин в графе.

- p.or.m - либо вероятность существования ребра между двумя произвольными вершинами (граф G(n,p)), либо количество ребер в графе (для графов G(n,m)).

- type - Тип создаваемого случайного графа: gnp (граф G(n,p)) или gnm (граф G(n,m)).

- directed - ориентированный ли граф, по умолчанию FALSE.

- loops - добавлять ли петли, по умолчанию FALSE.

В графах G(n,p) граф имеет «n» вершин, и для каждого ребра вероятность того, что оно присутствует в графе, равна «p». Эквивалентно, вероятность создания каждого графа, имеющего $n$  узлов и $M$  ребер, равна
$p^M (1-p) ^{\binom{n}{2} - M}$
<br>
<br>
В графах G(n,m) граф имеет «n» вершин и «m» ребер, а «m» ребер выбираются равномерно случайным образом из множества всех возможных ребер. Этот набор также включает ребра петли, если параметр loops имеет значение TRUE.

```r
graph <-erdos.renyi.game(g, p, 'gnp', TRUE, TRUE)
graph
```

### 7. 
Вычислить индекс взаимности Бандиопадхьяйя. Сделать выводы

```r
functionF_t <- function(t){
  return(sum(rowSums(m)[1:t])-t*(g-1)-choose(t,2))
}
functionG_t <- function(t){
  smin <- sum(pmin(rep(t,g-t),rowSums(m)[(t+1):g]))
  return(sum(rowSums(m)[1:t])-t*(t-1)-smin)
}
g_all <- c(2,g-1)
f_all <-c(2,g)
f_all[g] <- functionF_t(g)
for(t in 2:(g-1)){
g_all[t] <- functionG_t(t)
f_all[t] <-functionF_t(t)
}

bandyopadhyay <- (dyad_census$mut-max(f_all))/(floor(0.5*sum(rowSums(m)[1:g]-max(g_all)))-max(f_all))
bandyopadhyay
```

Вывод: индекс взаимности практически совпадает с p -> он найден правильно
