## Основные теоретические положения

- Количество перекрывающихся связей между событиями частично зависит от количества событий, к которым принадлежат акторы.
- Количество связей соучастия между акторами частично зависит от размера событий
- Актор, принадлежащий $a_{i+}$ событиями, создает $\binom{a_{i+}}{2}$ парных связей между событиями
- Событие с участием $a_{+j}$ участников создаёт $\binom{a_{+j}}{2}$ связей между парами акторов
- Плотность для дихотомического отношения - доля присутствующих связей.
- Плотность для ценностных отношений - среднее значение связей.
- Полезным способом изучения достижимости в аффилированной сети является рассмотрение двудольного графа, в котором как акторы, так и события представлены в виде узлов
- Поскольку акторы смежны с событиями (и наоборот), ни одна пара акторов не является смежной, равно как и никакая пара событий не является смежной
- Если акторы n_i и n_j связаны с событием m_k, то говорят, что существует путь между узлами  $n_i$ и $n_j$
- Lemma
    
    Breiger(1974): Любая сеть аффилированности, которая связана в графе соучастия акторов, обязательно связана в графе перекрытий между событиями (если ни одно событие не является пустым)

 Сеть аффилированности (affilation network) - сеть, в которой участники аффилированы друг с другом на основе сотрудничества в рамках какой-то группы или совместного участия в определенном событии.
    - Сеть состоит из двух измерений (two-modes): множество акторов
  
  $\mathscr N = \{n_1, n_2, ..., n_g\}$  и множества событий (events) $\mathscr M = \{m_1, m_2, ..., m_h\}$
        
- Матрица инцедентности (affilation network) - матрица А = {a_ij} размерности  g x h c элементами
    
$$
a_{ji} =  \begin{equation*}  \left\{
    \begin{aligned}
      & 1, \:если\: актор\:i\: связан\: с \: о\\
      & 0, иначе
    \end{aligned}
  \right.
\end{equation*}
$$
    
- Двудольный граф(bipartite graph) - это граф, узлы которого можно разделить на два подмножества, а все рёбра соединяют узлы, принадлежащие разным подмножествам.
- Представление биграфа в виде социоматрицы размера (g+h) x (g+h):

$$
\begin{pmatrix}
0_g & A \\
A^T & 0_h 
\end{pmatrix}
$$
    
- Гиперграф (hypergraph) - обобщение понятние графа; гиперграф состоит из множества объектов и множества и множества подмножеств объектов, в котором каждый объект принадлежит хотя бы одному подмножеству, причём ни одно подмножество не является пустым:
    
    $$
    \mathscr H = \mathscr{H(A, B),}
    $$
    
    где 
    
    $\mathscr A = \{a_1, a_2, ..., a_g\} -$  множество объектов
    
    $\mathscr B = \{B_1, B_2, …, B_h\}$
    
- Двойственный гиперграф (dual hypergraph) для гиперграф $\mathscr H = \mathscr {H(A, B)}$:
    
    $$
    \mathscr {H^* = H(B, A)},
    $$
    
    где
    
    - $\mathscr{B = \{b_1, b_2, ..., b_h\}}$
    - $\mathscr{A} = \{A_1, A_2, ..., A_h\}$
- Матрица соучастия акторов (actor co-membership matrix):
    
    $X^\mathscr N = \{x_{ij}^\mathscr N\}:$
    
    $$
    x_{ij}^\mathscr N = \sum_{k=1}^h a_{ik}a_{jk} X^\mathscr N = XX^T
    $$
    
- Матрица перекрытия событий (event overlap matrix): $X^\mathscr M = \{x_{kl}^\mathscr M\}:$

$$
x_{kl}^\mathscr M = \sum_{k=1}^h a_{ik}a_{il} X^\mathscr M = X^TX
$$

- Степень участия (rates of participation) - это количество событий, с  которым связан каждый актор:

$$
a_{i+} = \sum_{j=1}^h a_{ij} = x_{ii}^\mathscr N
$$

- **Среднее количество событий с которыми акторы аффилированы:**

$$
\bar a_{i+} = 1/g \sum_{i=1}^g \sum_{j=1}^h a_{ij} = a_{++} / g = 1/g \sum ^g _{i=1} x_{ii}^\mathscr N 
$$

- **Размер событий** (size of events):

$$
a_{+j} = \sum_{i=1}^g = x_{jj}^\mathscr M
$$

- **Средний размер событий** - среднее количество акторов в каждом событии или среднее степень узлов, представляющих события в двудольном графе

$$
\bar a_{+j} = 1/g \sum_{i=1}^g \sum_{j=1}^h a_{ij} = a_{++} / h = 1/h \sum ^h _{i=1} x_{jj}^\mathscr M 
$$

- Плотность для отношения соучастия $\triangle_{(\mathscr N)}$:
    
    $$
    \triangle_{(\mathscr N)} = 1/(g(g-1)) \sum_{i=1}^{g}\sum_{j=1 j≠i)}^{g} x_{ij}^{\mathscr N}
    $$
    
- Плотность для отношения перекрытия $\triangle_{(\mathscr M)}$:
    
    $$
    \triangle_{(\mathscr M)} = 1/(h(h-1)) \sum_{k=1}^{h}\sum_{i=1 i≠k)}^{k} x_{kl}^{\mathscr M}
    $$
    
- Связная сеть аффилированности (connected affiliation network) - сеть в которой все пары узлов (как акторов, так и событий) достижимы.
- Диаметр аффилированной сети - это длина самого длинного пути между любой парой акторов и/или событий


- Взаимная дида: $D_{ij} = (1, 1)$
- Асимметричная диада $D_{ij} = (1, 0), D_{ij} = (0, 1)$
- Нулевая диада $D_{ij} = (0, 0)$
- Цензус диад(dyad census) - тройка <M, A, N>, где:
    - $M = \sum_{i<j} X_{ij} X_{ji}$
    - $A = X_{++} - 2M$
    - $N = \binom{g}{2} - A - M$
- Индекс взаимности Каца-Пауэлла - характеристи измеренения тенденции акторов в группу:
    
  $$
  - \inf < \rho _{KP} ≤ 1
  $$
  
    Индекс взаимности основан на ожидании количества взаимных диад, предпологая, что выбор делается акторами случайным образом
    
    - Предполагается, что каждый актор делает d выборов из (g-1) акторов, доступных для выбора
    - Вероятность взаимного выбора между любой произвольной парой акторов равна $d^2/(g-1)^2$
    - Математическое ожидание взаимных диад $EM = g(g-1)/2  d^2/(g-1)^2 = gd^2 /2(g-1)$
    - Вероятность того, что общая диада, в которой участвуют акторы является взаимной диадой
        
        
- Условная вероятность:
        
    $P(j \rightarrow i | i \rightarrow j) = P (j \rightarrow i ) +   \rho _{KP} P (j \not \rightarrow i )$
    
    Таким образом 
    
    $P(j \rightarrow i ^ i \rightarrow j) = d/g-1 (d/g-1 + \rho _{KP} (g-1-d)/(g-1))$
    
  - Маметиматческое ожидание взаимных диад с учётом данной вероятности:
      
      $E(M | \rho ) = gd^2 /2(g-1) (1- \rho _{KP}) + gd/2  \rho _{KP}$
      
  - Оценка  \rho _{KP} по методу моментов:
      
       $\hat \rho _{KP} = (2(g-1)M-gd^2)/(gd(g-1-d))$
      
  - Пусть i-ый актор выбирает $d_O(n_i)=x_{i+}$ других акторов
  - Вероятность того, что акторы имеют взаимные отношения равна $x_{i+} x{j+} / (g-1)^2$
  - Математического ожидание взаимных диад:
      
      $EM = L^2 - L_2/2(g-1)^2$
      
  - Математического ожидание взаимных диад с учётом вероятности:
      
      $E(M | \rho ‘) = L^2 - L_2/2(g-1) * (1-\rho_{KP}‘ + L/2 \rho_{KP}$
      
  - Оценка $\rho_{KP}$ по методу моментов:
      
      $$
      \rho_{KP}' = 2(g-1)^2M - L^2 + L_2 / (L(g-1)^2 - L^2 + L_2) 
      $$
