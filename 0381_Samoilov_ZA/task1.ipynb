## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è
**–ê–∫—Ç–æ—Ä** - —ç—Ç–æ –¥–∏—Å–∫—Ä–µ—Ç–Ω—ã–µ –∏–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω—ã–µ, –∫–æ—Ä–ø–æ—Ä–∞—Ç–∏–≤–Ω—ã–µ –∏–ª–∏ –∫–æ–ª–ª–µ–∫—Ç–∏–≤–Ω—ã–µ —Å–æ—Ü–∏–∞–ª—å–Ω—ã–µ –µ–¥–∏–Ω–∏—Ü—ã.

**–û—Ç–Ω–æ—à–µ–Ω–∏–µ** - —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å —Å–≤—è–∑–µ–π –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–≥–æ –≤–∏–¥–∞ –º–µ–∂–¥—É —á–ª–µ–Ω–∞–º–∏ –≥—Ä—É–ø–ø—ã.

**–°–æ—Ü–∏–∞–ª—å–Ω–∞—è —Å–µ—Ç—å** - –æ–±—ä–µ–∫—Ç, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∏–π —Å–æ–±–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å –∫–æ–Ω–µ—á–Ω–æ–≥–æ –Ω–∞–±–æ—Ä–∞ (–∏–ª–∏ –Ω–∞–±–æ—Ä–æ–≤) –∞–∫—Ç–æ—Ä–æ–≤ –∏ –∏—Ö –æ—Ç–Ω–æ—à–µ–Ω–∏—è (–∏–ª–∏ –æ—Ç–Ω–æ—à–µ–Ω–∏–π), –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö –Ω–∞ –Ω–∏—Ö.

**–ì—Ä–∞—Ñ** - —ç—Ç–æ –º–æ–¥–µ–ª—å —Å–æ—Ü–∏–∞–ª—å–Ω–æ–π —Å–µ—Ç–∏ —Å –¥–∏—Ö–æ—Ç–æ–º–∏—á–µ—Å–∫–∏–º –æ—Ç–Ω–æ—à–µ–Ω–∏–µ–º, —Ç.–µ. —Å–≤—è–∑—å –ª–∏–±–æ –ø—Ä–∏—Å—É—Ç—Å–≤—É–µ—Ç, –ª–∏–±–æ –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç –º–µ–∂–¥—É –∫–∞–∂–¥–æ–π –ø–∞—Ä–æ–π –∞–∫—Ç–æ—Ä–æ–≤.

**–°–æ—Ü–∏–æ–º–∞—Ç—Ä–∏—Ü–∞** - –º–∞—Ç—Ä–∏—Ü—ã  ùêó  —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ **g √ó g** , –æ–ø–∏—Å—ã–≤–∞—é—â–∞—è —Å–º–µ–∂–Ω—ã–µ —É–∑–ª—ã –≤ —Å–µ—Ç–∏.

**–†–∞–∑–º–µ—Ä —Å–µ—Ç–∏** - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–æ—Ä–æ–≤ —Å–µ—Ç–∏.

**–ü–ª–æ—Ç–Ω–æ—Å—Ç—å** - –¥–æ–ª—è –∏–º–µ—é—â–∏—Ö—Å—è —Å–≤—è–∑–µ–π –ø–æ –æ—Ç–Ω–æ—à–µ–Ω–∏—é –∫ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å–≤—è–∑–µ–π.

–î–ª—è –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —Å–µ—Ç–µ–π: $$\Delta = \frac{L}{A_g^2}$$
 

–î–ª—è –Ω–µ–æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —Å–µ—Ç–µ–π: $$\Delta = \frac{L}{C_g^2}$$
 

ùê† - —á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤, ùêã - —á–∏—Å–ª–æ –∏–º–µ—é—â–∏—Ö—Å—è —Å–≤—è–∑–µ–π –≤ —Å–µ—Ç–∏.

**–ö–æ–º–ø–æ–Ω–µ–Ω—Ç–∞ —Å–µ—Ç–∏** - –ø–æ–¥–≥—Ä—É–ø–ø–∞ —Å–µ—Ç–∏, –≤ –∫–æ—Ç–æ—Ä–æ–π –≤—Å–µ –∞–∫—Ç–æ—Ä—ã —Å–≤—è–∑–∞–Ω—ã –¥—Ä—É–≥ —Å –¥—Ä—É–≥–æ–º –ø—Ä—è–º–æ –∏–ª–∏ –∫–æ—Å–≤–µ–Ω–Ω–æ.

**–î–∏–∞–º–µ—Ç—Ä —Å–µ—Ç–∏** - –Ω–∞–∏–±–æ–ª—å—à–∞—è –¥–ª–∏–Ω–∞ –ª—é–±–æ–≥–æ –∏–∑ –∫—Ä–∞—Ç—á–∞–π—à–∏—Ö –ø—É—Ç–µ–π –º–µ–∂–¥—É –≤—Å–µ–º–∏ –ø–∞—Ä–∞–º–∏ —É–∑–ª–æ–≤.

**–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏** - –¥–æ–ª—è –∑–∞–∫—Ä—ã—Ç—ã—Ö —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–æ–≤ (—Ç—Ä–∏–∞–¥, –≥–¥–µ –Ω–∞–±–ª—é–¥–∞—é—Ç—Å—è –≤—Å–µ —Å–≤—è–∑–∏) –ø–æ –æ—Ç–Ω–æ—à–µ–Ω–∏—é –∫ –æ–±—â–µ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–æ–≤.

**–°—Ç–µ–ø–µ–Ω—å —É–∑–ª–∞** - —á–∏—Å–ª–æ —Ä–µ–±–µ—Ä, –∏–Ω—Ü–∏–¥–µ–Ω—Ç–Ω—ã—Ö —É–∑–ª—É $n_i$

**–î–∏—Å–ø–µ—Ä—Å–∏—è —Å—Ç–µ–ø–µ–Ω–µ–π —É–∑–ª–∞:** $$S_d^2 = \frac{1}{g}\sum_{i=1}^g(d(n_i)-\bar{d})^2,$$

–≥–¥–µ d - c—Ä–µ–¥–Ω—è—è —Å—Ç–µ–ø–µ–Ω—å: $$\bar{d} = \frac{1}{g}\sum_{i=1}^gd(n_i)$$

**–¶–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å** - –º–µ—Ä–∞, —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É—é—â–∞—è –∞–∫—Ç–æ—Ä—ã —Å –±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —Å–≤—è–∑–µ–π (–ø—Ä—è–º—ã—Ö –∏–ª–∏ –∫–æ—Å–≤–µ–Ω–Ω—ã—Ö).

–¶–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ —Å—Ç–µ–ø–µ–Ω–∏:  $$C_D^{'}(n_i) =\frac{d(n_i)}{g-1}$$

–¶–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ –±–ª–∏–∑–æ—Å—Ç–∏: $$C_C^{'}(n_i)=(g-1)(\sum_{i=1}^gd(n_i,n_j))^{-1}$$

–¶–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ –ø–æ—Å—Ä–µ–¥–Ω–∏—á–µ—Å—Ç–≤—É:  $$C_B(n_i)=\sum_{i<k}g_{ik}(n_i)/g_{jk}$$ , –≥–¥–µ $g_{jk}$ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É —É–∑–ª–∞–º–∏ j –∏ k, –∞ $g_{jk}(n_i)$ - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫—Ä–∞—Ç—á–∞–π—à–∏—Ö –ø—É—Ç–µ–π –º–µ–∂–¥—É —É–∑–ª–∞–º–∏ j –∏ k —á–µ—Ä–µ–∑ i.

–¶–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏: $$C_I^{'}=\frac{C_I(n_i)}{\sum_{i}C_I(n_i)}$$

$$ A = \{a_{ij}\}$$
$$ a_{ii} = 1 + \sum_{j=1}^{g}{x_{ij}}$$
$$ a_{ij} = 1 \text{ , if } n_{i} \text{ and } n_{j} \text{ are not connected}$$
$$ a_{ij} = 1-x_{ij} \text{ , if } n_{i} \text{ and } n_{j} \text{ are connected}$$
$$ x_{ij} \text{- weight of edge between }n_{i} \text{ and } n_{j}$$
$$ C = A^{-1}$$
$$C_I(n_{i}) = \frac{1}{c_{ii}+(T-2R)/g}$$ $$T = trace(C)$$ $$R = \sum_{j=1}^{g}c_{ij}$$

**–¶–µ–Ω—Ç—Ä–∞–ª–∏–∑–∞—Ü–∏—è** - –º–µ—Ä–∞, —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É—é—â–∞—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω—É—é —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ—Å—Ç—å —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç–∏ –∞–∫—Ç–æ—Ä–æ–≤ –≤ —Å–µ—Ç–∏

–¶–µ–Ω—Ç—Ä–∞–ª–∏–∑–∞—Ü–∏—è –ø–æ —Å—Ç–µ–ø–µ–Ω–∏: $$C_D = \frac{\sum_{i=1}^g(C_D(n^{*})-C_D(n_i))}{(g-1)(g-2)}$$
 
 –¶–µ–Ω—Ç—Ä–∞–ª–∏–∑–∞—Ü–∏—è –ø–æ –±–ª–∏–∑–æ—Å—Ç–∏: $$C_C = \frac{\sum_{i=1}^g(C_C^{'}(n^{*})-C_C^{'}(n_i))}{(g-1)(g-2)/(2g-3)}$$
 
 –¶–µ–Ω—Ç—Ä–∞–ª–∏–∑–∞—Ü–∏—è –ø–æ –ø–æ—Å—Ä–µ–¥–Ω–∏—á–µ—Å—Ç–≤—É: $$C_B = \frac{\sum_{i=1}^g(C_B^{'}(n^{*})-C_B^{'}(n_i))}{g-1}$$
 
–¶–µ–Ω—Ç—Ä–∞–ª–∏–∑–∞—Ü–∏—è –ø–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏: $$C_I =\frac{1}{g}\sum_{i=1}^g(C_I^{'}(n^{*})-\bar{C_I}(n_i))^2$$

**–¢–æ—á–∫–∞ —Å–æ—á–ª–µ–Ω–µ–Ω–∏—è** - —É–∑–µ–ª, –ø—Ä–∏ —É–¥–∞–ª–µ–Ω–∏–∏ –∫–æ—Ç–æ—Ä–æ–≥–æ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç —Å–≤—è–∑–Ω–æ—Å—Ç–∏.

**–ú–æ—Å—Ç** - —Ä–µ–±—Ä–æ, –ø—Ä–∏ —É–¥–∞–ª–µ–Ω–∏–∏ –∫–æ—Ç–æ—Ä–æ–≥–æ –æ–¥–Ω–∞ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞ —Å–≤—è–∑–Ω–æ—Å—Ç–∏ —Ä–∞–∑–¥–µ–ª—è–µ—Ç—Å—è –Ω–∞ 2.

**–ü—Ä–µ—Å—Ç–∏–∂** - –º–µ—Ä–∞, —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É—é—â–∞—è —Ñ–∫—Ç–æ—Ä–∞ —Å –±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –≤—Ö–æ–¥—è—â–∏—Ö —Å–≤—è–∑–µ–π.

**–ü—Ä–µ—Å—Ç–∏–∂ –ø–æ —Å—Ç–µ–ø–µ–Ω–∏** - –≤—Ö–æ–¥—è—â–∞—è —Å—Ç–µ–ø–µ–Ω—å –∫–∞–∂–¥–æ–≥–æ —É–∑–ª–∞. $$P_D(n_i) = d_I(n_i)=x_{+i}$$

–ù–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –ø—Ä–µ—Å—Ç–∏–∂ –ø–æ —Å—Ç–µ–ø–µ–Ω–∏: $$P_D^{'}(n_i) =\frac{x_{+i}}{g-1}$$

**–ü—Ä–µ—Å—Ç–∏–∂ –ø–æ –±–ª–∏–∑–æ—Å—Ç–∏** - –º–µ—Ä–∞, –æ–ø—Ä–µ–¥–µ–ª—è—é—â–∞—è, –Ω–∞—Å–∫–æ–ª—å–∫–æ –≤—Å–µ –∞–∫—Ç–æ—Ä—ã –±–ª–∏–∑–∫–∏ –∫ –∑–∞–¥–∞–Ω–Ω–æ–º—É –ø–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—é –¥–æ –Ω–µ–≥–æ. $$ P_P(n_i) = \frac{\#I_i/(g-1)}{\sum_{j:n_j \in I_i}(d(n_j,n_i)/\#I_i)}$$

## –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏

–° –ø–æ–º–æ—â—å—é –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö –ø–∞–∫–µ—Ç–æ–≤ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫ –ø—Ä–æ–≤–µ—Å—Ç–∏ –∞–Ω–∞–ª–∏–∑ –∑–∞–¥–∞–Ω–Ω—ã—Ö —Å–æ—Ü–∏–∞–ª—å–Ω—ã—Ö —Å–µ—Ç–µ–π, –≤—ã—á–∏—Å–ª–∏—Ç—å –æ–ø–∏—Å–∞—Ç–µ–ª—å–Ω—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏. –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ–¥–µ—Ä–∂–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å.

```r
filename <- "/content/sample_data/florentine_business.csv"
florentine_data <- read.table(filename, sep=",")
florentine_matrix <- as.matrix(florentine_data)
headers <- florentine_matrix[1, ]
florentine_matrix <- florentine_matrix[-1, ]
class(florentine_matrix) <- "numeric"
florentine_matrix
```

2. –ü–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å —Å–æ—Ü–∏–æ–º–∞—Ç—Ä–∏—Ü—É –≤ —Å–ø–∏—Å–æ–∫ —Ä—ë–±–µ—Ä. –í—ã–≤–µ—Å—Ç–∏ –ø–µ—Ä–≤—ã–µ 5 —ç–ª–µ–º–µ–Ω—Ç–æ–≤ —Å–ø–∏—Å–∫–∞ —Ä—ë–±–µ—Ä.

```r
florentine_graph <- graph.adjacency(florentine_matrix)
edge_list <- get.edgelist(florentine_graph)
head(edge_list, 5)
```

```r
rownames(florentine_matrix) <- colnames(florentine_matrix) <- headers
florentine_matrix
```

4. –° –ø–æ–º–æ—â—å—é —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π –ø–æ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π —Å–æ—Ü–∏–æ–º–∞—Ç—Ä–∏—Ü–µ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –Ω–µ–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–Ω—É—é —Å–µ—Ç—å. –û–±—ä—è—Å–Ω–∏—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä—ã, –ø–µ—Ä–µ–¥–∞–≤–∞–µ–º—ã–µ –≤ —Ñ—É–Ω–∫—Ü–∏—é.

5. –û—Ç–æ–±—Ä–∞–∑–∏—Ç—å —Å–µ—Ç—å —Å –ø–æ–º–æ—â—å—é –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤ —É–∫–ª–∞–¥–∫–∏ –§—Ä—é—Ö—Ç–µ—Ä–º–∞–Ω–∞‚Äì–†–µ–π–Ω–≥–æ–ª—å–¥–∞ –∏ –ö–∞–º–∞–¥–∞‚Äì–ö–∞–≤–∞–∏. –û–±—ä—è—Å–Ω–∏—Ç—å, –≤ —á—ë–º —Ä–∞–∑–ª–∏—á–∏–µ –º–µ–∂–¥—É –∞–ª–≥–æ—Ä–∏—Ç–º–∞–º–∏ –∏ —É–∫–ª–∞–¥–∫–∞–º–∏.

```r
plot(network, 
     vertex.size=5, 
     vertex.color="red", 
     vertex.label.dist=1.5, 
     main="Fruchterman-Reingold Algorithm", 
     layout=layout.fruchterman.reingold
    )
plot(network, 
     vertex.size=5, 
     vertex.color="purple", 
     vertex.label.dist=1.5, 
     main="Kamada-Kawai Algorithm", 
     layout=layout_with_kk
    )
```

9. –ó–∞–ø–æ–ª–Ω–∏—Ç—å —Ç–∞–±–ª–∏—Ü—É –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –º–µ—Ä —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç–∏. *–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ: –º–µ—Ä—ã —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç–∏ –≤—ã—á–∏—Å–ª–∏—Ç—å –¥–ª—è –Ω–µ–∏–∑–æ–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –≤–µ—Ä—à–∏–Ω*

```r
updated_florentine_matrix <- florentine_matrix[!apply(florentine_matrix == 0, 1, all), !apply(florentine_matrix == 0, 1, all)]
G <- graph_from_adjacency_matrix(updated_florentine_matrix, mode = "undirected")

G_order <- gorder(G)
G_degree <- degree(G)

inv_matrix <- solve(laplacian_matrix(G) + 1)
T <- sum(diag(inv_matrix))
R <- colSums(inv_matrix) 
CI <- 1/(diag(inv_matrix) + (T - 2*R)/G_order)

CD_ni <- G_degree / (G_order - 1)
CC_ni <- closeness(G, normalized = TRUE)
CB_ni <- betweenness(G, normalized = TRUE)
CI_ni <- CI / sum(CI)

centrality <- data.frame(
    ni = 1:G_order, # actors
    CD_ni = CD_ni, # –¶–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ —Å—Ç–µ–ø–µ–Ω–∏ (–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω–∞—è)
    CC_ni = CC_ni, # –¶–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ –±–ª–∏–∑–æ—Å—Ç–∏ (–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω–∞—è)
    CB_ni = CB_ni, # –¶–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ –ø–æ—Å—Ä–µ–¥–Ω–µ—á–µ—Å—Ç–≤—É (–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω–∞—è)
    CI_ni = CI_ni # –¶–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç—å –ø–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ (–Ω–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–Ω–∞—è)
)
centrality
```

–ú–µ—Ä—ã —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç–∏ –ø–æ–∑–≤–æ–ª—è—é—Ç —Å—É–¥–∏—Ç—å –æ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–∏ –∞–∫—Ç–æ—Ä–æ–≤ –≤ —Å–µ—Ç–∏. –ß–µ–º –±–æ–ª—å—à–µ –º–µ—Ä—ã —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ—Å—Ç–∏ –¥–ª—è –∞–∫—Ç–æ—Ä–∞ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–π —Å–µ—Ç–∏, —Ç–µ–º –±–æ–ª—å—à–µ —Å–≤—è–∑–µ–π –∏–º–µ–µ—Ç –¥–∞–Ω–Ω—ã–π –∞–∫—Ç–æ—Ä —Å –¥—Ä—É–≥–∏–º–∏ –∞–∫—Ç–æ—Ä–∞–º–∏.

10. –í—ã—á–∏—Å–ª–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ —Ü–µ–Ω—Ç—Ä–∞–ª–∏–∑–∞—Ü–∏–∏ –¥–ª—è –∫–∞–∂–¥–æ–π –º–µ—Ä—ã. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```r
cat("Centralization by degree: ", centralize(CD_ni, (G_order - 1) * (G_order - 2)))
cat("\nCentralization by proximity: ", centralize(CC_ni, (G_order - 1) * (G_order - 2) / (2 * G_order - 3)))
cat("\nCentralization by mediation: ", centralize(CB_ni, (G_order - 1)))
cat("\nCentralization by information: ", var(CI_ni))
```

```r
x_plus_i <- degree(network, mode="in", normalized="true")
g_order <- gorder(network)

i <- 1
l <- list()
for(vertex in V(network)) {
    num_Ii <- length(subcomponent(network, vertex)) - 1
    d_nj_ni <- distances(network, v=vertex)
    l[[i]] <- (num_Ii/(g_order - 1))/(sum(d_nj_ni/num_Ii))
    i <- i + 1
}
res <- unlist(l)

data.frame(
    n_i=1:g_order,
    PD_ni=x_plus_i,
    PP_ni=res
)
```
