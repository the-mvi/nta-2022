## Основные теоретические положения
**Группа** - совокупность всех акторов, связи между которыми должны быть изучены.

**Подгруппа** - любое подмножество акторов и все связи между ними.

**Клика** - максимальный полный подграф из 3х и более узлов.

**n-клика** - максимальный подграф, в котором наибольшее геодезичесоке расстояние между любыми двумя узлами не превышает $n$.

$d(i,j)\leq n \: \forall n_{i}, n_{j} 	
\in N_{s}$

**n-кланы** - это *n-клика*, в которой геодезическое расстояние $d(i,j)$ между всеми узлами в подграфе не превышает $n$ для путей внутри подграфа.

**n-клубы** - максимальный подграф диаметра  $ \: \leq n$ .

**k-плекс** -  максимальный подграф, содержащий $ g_{s} $ узлов, в котором каждый узел смежен не менее, чем с $ (g_{s} - k) $ узлами подграфа:
$g_{s}(i) \geqslant g_{s} - k \:\: \forall n_{i}	
\in N_{s}$

**k-ядро** -  максимальный подграф, в которой каждая вершина связана минимум с  $k$ другими вершинами этого же подграфа.

**LS-множества.**
   $ \\ $ 
  Дан подграф $ I $ c множеством узлов $N_{s}$. Множество узлов $N_{s}$ будем называть **LS-множеством**, если любое собственное подмножество $ Q $ c $N_{s}$ имеет больше линий к узлам  $N_{s}|Q$, чем к $N|N_{s}$.

**$\lambda$-множества.** Рассматриваются пары узлов из множества $N_{s}$ подграфа $Y_{s}$. Множество узлов  $N_{s}$ называется $\lambda$-множеством, если любая пара узлов в $N_{s}$ имеет большое линейное соединение, чем любая другая пара узлов, состоящая из одного узла лямбда-набора и второго узла извне:

$N_{s} \subseteq N : \: \forall i,j,k \in N_{s} I \in N_{s}|N \Rightarrow \lambda(i,j)>\lambda(k,l) ,
 где \: \lambda(i,j)$ - минимальное количество линий , которое необходимо удалить из графа, чтобы не осталось путей между двумя узлами. 


**Модулярность** $-$ это характеристика структуры сети, в частности степень кластеризации узлов, когда внутри кластера наблюдается много узлов (высокая плотность), а между кластерми малов узлов (низкая плотность). 

**Сеть аффилированности** $-$ сеть, в которой участники аффилированы друг с другом на основе сотрудничества в рамках какой-то группы или совместного участия в определенном событии. Сеть состоит из 2 измерений: множества акторов $N={\{n_{1}, n_{2}, ..., n_{g}}\}$
 и множества событий $M={\{m_{1}, m_{2}, ..., m_{n}}\}$.



**Матрица инцедентности** $-$ матрица $ A \: A={\{a_{ij}}\}$ размерности $g*h $ c элементами $a_{ij}=1 -$ если актор $i$ связан с событием $ j$, $a_{ij}=0 -$ иначе.

**Двудольный граф** $-$  граф, узлы которого можно разделить на два подмножества, а все  ребра соединяют узлы, принадлежащие разным подмножествам.

**Гиперграф** $-$  обобщение понятия графа, гиперграф состоит из множества объектов и множества подмножеств объектов, в котором каждый объект принадлежит хотя бы одному подножеству, причем ни одно подмножество не являетс пустым.

**Матрица соучастия акторов**:  

$X^{N}={\{x_{ij}^{N}}\}$:

$x_{ij}^{N}=\sum_{k=1}^{h}a_{ik}a_{jk}$

$x^{N}=XX^{T}$


**Матрица перекрытия событий**:  

$X^{M}={\{x_{kl}^{M}}\}$:

$x_{kl}^{M}=\sum_{i=1}^{g}a_{ik}a_{il}$

$x^{M}=X^{T}X$

**Степень соучастия:** $-$ количество событий, с которыми связан каждый актор.

$a_{i+}=\sum_{j=1}^{h}a_{ij}x_{ii}^{N}$

**Размер событий:**

$\Large a_{ij}=\sum_{i=1}^{g}a_{ij}=x_{jj}^{M}$

**Средний размер событий**  $-$ среднее количество акторов в каждом событии или средняя степень узлов, представляющтх события в двудольном графе.

**Среднее количество событий, с которыми акторы аффилированы**

$\Large \bar{a}_{i+}=\frac{1}{g}\sum_{i=1}^{g}\sum_{j=1}^{h}a_{ij}=\frac{a_{++}}{g}=\frac{1}{g}\sum_{i=1}^{g}x_{ii}^{N}$

```r
#n-кланы
n_clan <- function(graph, n) {
    new_graph <- connect(graph, n) 
    ncliques <- max_cliques(new_graph) 
    for (i in 1:length(ncliques)) {
        vertexes <- list()
        for (j in 1:length(ncliques[[i]])) {
          vertexes <- append(vertexes, ncliques[[i]][j]$name)
        }
        nodes <- V(graph)[name %in% vertexes]
        sub_g <- induced_subgraph(graph, nodes)
        if ((diameter(sub_g) <= n) && (length(vertexes) >= 3)) {
            print(ncliques[i])
        }
    }
}
n_clans <- function(graph, n){
    for (i in 1:n){ 
        n_clan(graph, i)
      }
}
n_clans(g, 2)
```

```r
#n-клубы
nclique <- function(graph,n){
  graph <- connect(graph, order=n)
  return(max_cliques(graph,min=3))
}

nclubs<- function(graph,n) {
    cliques<-nclique(graph,n)
    res<-c()
    for(cl in cliques) {
        sub<-subgraph(graph,names(cl))
        if(diameter(sub) <= n) {
           cat('(',names(cl),")\n")
        }    
    }
    return (graph)
}

print("1")
print(nclubs(g, 1))
print("2")
print(nclubs(g, 2))
print("3")
print(nclubs(g, 3))
print("4")
print(nclubs(g, 4))
print("5")
print(nclubs(g, 5))
print("6")
print(nclubs(g, 6))
```

```r
#k-ядра
kkernel <- function(g, k) {
    g.sym <- as.undirected(g)
    k.kernel <- list()
    g.sym.degk <- induced_subgraph(g.sym,igraph::degree(g.sym)>=k)
    if(length(V(g.sym.degk)) < k) {
        return(k.kernel)
    } 
    flag <- 0
    for (j in length(V(g.sym.degk)):1){
        if(flag) {
            break
        }
        k.cand <- combn(V(g.sym.degk)$name,j) 
        k.kernel <- list() 
        k.kernel.i <- 1 
        for (k.cand.i in 1:dim(k.cand)[2]){ 
            g.k.cand <- induced_subgraph(g.sym.degk,k.cand[,k.cand.i]) 
            if (min(igraph::degree(g.k.cand))>=k){
                k.kernel[[k.kernel.i]] <- k.cand[,k.cand.i] 
                k.kernel.i <- k.kernel.i+1 
                flag <- 1
            }
        }
    }
    return(k.kernel)
}

kkernels <- function(g) {
    result <- c()
    for (k in 1:length(V(g))) {
        kernel <- kkernel(g, k)
        if(length(kernel) > 0) {
            result[[paste(toString(k), "ядро", sep=" ")]] <- kernel
        }
    }
    return(result)
}

print(kkernels(g))
```

```r
#k-плексы
kplex <- function(g,k,m){
    g.sym <- as.undirected(g)
    
    k.plex <- list()
    g.sym.degmk <- induced_subgraph(g.sym,igraph::degree(g.sym)>=(m-k))
    if(length(V(g.sym.degmk)) < m) {
        return(k.plex)
    } 
    k.cand <- combn(V(g.sym.degmk)$name,m) 
    k.plex <- list() 
    k.plex.i <- 1 
    for (k.cand.i in 1:dim(k.cand)[2]){ 
        g.k.cand <- induced_subgraph(g.sym.degmk,k.cand[,k.cand.i]) 
        if (min(igraph::degree(g.k.cand))>=(m-k)){
            k.plex[[k.plex.i]] <- k.cand[,k.cand.i] 
            k.plex.i <- k.plex.i+1 
        }
    }
    return(k.plex)
}

kplexes <- function(g) {
    result <- c()
    for (i in 2:(length(V(g))-6)) {
        for (j in 1:(i-1)) {
            plex <- kplex(g, j, i)
            if(length(plex) > 0) {
                result[[paste(toString(i), toString(j), "plex", sep="_")]] <- plex
            }
        }
    }
    return(result)
}

print(kplexes(g))
```

###13. Для каждых пар событий определить отношение шансов. 


```r
zero <- 0
inf <- 0
for(k in 1:14){
    for(l in (k+1):15){
        a <- matr[, c(k, l)]
        R <- sum(colSums(t(a) == c(1, 1)) == ncol(a))*
        sum(colSums(t(a) == c(0, 0)) == ncol(a))/
        sum(colSums(t(a) == c(0, 1)) == ncol(a))/
        sum(colSums(t(a) == c(1, 0)) == ncol(a))   
        cat('(', k, ', ', l, ') -> ', R, '\n', sep="")
        zero <- zero + (R!=0)
        inf <- inf + (R==Inf)
    }
}
cat('Ненулевых значений:', zero, '\n')
cat('Ненулевых значений:', zero/(15*14/2)*100, '%\n')
cat('Бесконечных значений:', inf, '\n')
cat('Бесконечных значений:', inf/(15*14)/2*100, '%\n')
```
