## Основные теоретические положения

<b>Подгруппа</b> - подмножествоо узлов $N_s$<br>

<b>Клика</b> - максимальный полный подграф из трех или более узлов.

<b>n-клика</b> - максимальный подграф, в котором наибольшее геодезическое расстояние между любыми двумя узлами не превышает n: $d(i, j)<=n; n_i, n_j $ принадлежат $N_s$

<b>n-кланы</b> - n-клика, в которой d(i, j) между всеми узлами в подграфе не превышает n для всех путей внутри графа.

<b>n-коубы</b> - максимально полный полграф диаметра <= n

<b>k-ядро</b> - максимальный подграф, в котором каждая вершина связана минимум с k другими вершинами этого же подграфа.

<b>k-плекс</b> - это максимальный подграф, содержащий $g_s$ узлов, в котором каждый узел смежен не мене чем с $g_s - k$ узлами подграфа.

<b>Модулярность</b> - характеристика структуры сети, в частности степени кластеризации узлов, когда внутри кластера наблюдается много узлов (высокая плотность), а между кластерами мало узлов (низкая плотность)

<b>Сеть аффилированности</b> - сеть, в которой участники аффилированны друг с другом на основе сотрудничества в рамках какой-то группы или совместного участия в определнном событии. Сеть состоит из двух измерений: множества акторов $N = {n_1,...,n_g}$ и множества событий $M = {m_1, ... , m_h}$

<b>Матрица инцидентности</b> - $A = {a_{ij}}$ размерности gxh с элементами $a_{ij} \begin{cases} {1, d(i,j)!=0} \\{0} \end{cases}$

<b>Двудольный граф</b> - это граф, узлы которого можно разделить на два подмножества, а все ребра соединяют узлы, принадлежащие разным подмножествам.

<b>Матрица соучастия акторов</b>: $X^N = {x_{ij}^N}$: $x_{ij}^N = \sum_{k = 1}^h{a_{ik}a_{jk}} = XX^T$

<b>Матрица перекрытия событий</b>: $X^M = {x_{ij}^M}$: $x_{kl}^M = \sum_{l = 1}^g{a_{il}a_{jl}} = X^TX$

<b>Степень участия</b> - это количество событий, с которыми связан каждый актор: $a_{i+} = X_{ii}^N$

<b>Средняя степень участия</b> - количество событий, с которыми актор аффилирован: $\bar{a_{i+}} = \frac{1}{g}\sum_{i = 1}^gx_{ii}^N$

<b>Размер события</b>: $a_{+j} = x_{jj}^M$

<b>Средний размер событий</b> - количество акторов в каждом событии: $\bar{a_{+j}} = \frac{1}{h}\sum_{j = 1}^h{x_{jj}^M}$

<b>Плотность для отношения соучастия</b>: $\delta_{N} = \frac{1}{g(g-1)}\sum_{i = 1}^{g}\sum_{j = 1}^{g}x_{ij}^N$

<b>Плотность для отношения перекрытия</b>: $\delta_{M} = \frac{1}{h(h-1)}\sum_{k = 1}^{h}\sum_{l = 1}^{h}x_{kl}^M$

<b>Диаметр сети аффилированности</b> - это длина самого длинного пути между любой парой акторов и/или событий

<b>Клика уровня с для отношения соучастия</b> - подграф, в котором все пары акторов участвую не менее чем в с событиях

<b>Клика уровня с для отношения перекрытия</b> - подграф, в котором все пары событий имеют не менее с общих членов

<b>$x_{kl}^M$</b> - число акторов, которые аффилированы с событиями k и l

<b>$x_{\bar{kl}}^M$</b> - число акторов, которые не принадлежат ни одному событию

<b>$x_{k\bar{l}}^M$</b> - число акторов, которы аффилированы с событием k но не с l

<b>$x_{\bar{k}l}^M$</b> - число акторов, аффилированных с l, но не с k

Мерой критерия между событиями, которая не зависит от размера событий, называетют <b>отношение шансов</b>: $\theta_{kl} = \frac{\frac{x_{kl}^M}{x_{\bar{k}l}^M}}{\frac{x_{k\bar{l}}^M}{x_{\bar{kl}}}} = \frac{\frac{x_{kl}^M}{x_{\bar{kl}}^M}}{\frac{x_{\bar{k}l}^M}{x_{k\bar{l}}^M}}$

n-клики:

```r
get_cliques <- function(graph,order){
  graph <- connect(graph, order=order)
  return(max_cliques(graph,min=3))
}

task_2_1 <- function(net){
  graph <- graph.adjacency(as.matrix(net), mode="undirected")
  order <- 1
  len <- 2
  res <-list()
  while(len > 1){
    cliques <- get_cliques(graph,order)
    cliques<-lapply(cliques, function(x) names(x))
    names(cliques) <- rep(paste(order,"clique",sep= "-"),times = length(cliques))
    res <- c(res,cliques)
    order <- order+1
    len <- length(cliques)
  }
  return (res)
}

task_2_1(g)
```

```r
is_clan <-function(graph, clique, name){
  subgraph <- subgraph(graph,clique)
  order <- unlist(strsplit(name,"-"))[1]
  diameter <- diameter(subgraph,directed = FALSE)
  if(diameter > order){
    return(FALSE)
  }
  return (TRUE)
}


task_2_2 <-function(net){
  graph <- graph.adjacency(as.matrix(net), mode="undirected")
  cliques <- task_two_1(net)
  mask <- 1:length(cliques)
  mask <-sapply(mask, function(x) is_clan(graph,cliques[[x]],names(cliques)[x]))
  names(cliques)<-sapply(names(cliques), function(x) paste(unlist(strsplit(x,"-"))[1],"-clan",sep = ""))
  return(cliques[mask])
}

task_2_2(g)
```

Суть алгоритмов:

**InfoMAP**:
В алгоритме применяет подход, основанный на коде Хаффмана. Центральный объект алгоритма - случайный ходок, исследующий сеть с вероятностью согласно заданной матрицей переходов для цепи Маркова. Путь ходока кодируется с помощью кода Хаффмана. Затем эту длину пути хотим минимизоровать путем иерархического кодирования. При попадании в сообщество - записываем его уникальный код, затем пишутся коды вершин внутри сообщества, а когда ходок выходит из сообщества - записывается уникальный код выхода. Таким образом, получаем минимизированный путь ходока.

**Louvain**:
Основной принцип алгоритма Лувена является стратегией слияния, стандартом которой является коэффициент модульности. Инициализируем каждую вершину как отдельное сообщество. Затем рассматриваются слияния узлов в одно сообщество. Рассчитывается модульност. Затем выбирается соединение вершин так, что коэффициент модульности для этого объединения наибольший. Если вершина не может быть переназначена, то объединяются уже полученные на прошлом шаге сообщества. Если за шаг модульность увеличена быть не может, то алгоритм завершает работу.

**Optimal**:
Идея алгоритма в максимальной оценки модулярности. Реализуется засчет перехода в задачу, подходящую для библиотеки GLPK, благодаря чему и получаем решение.

5. Отобразить сеть в виде биграфа.

```r
create_graph <- function(A){
  hight_zero <-matrix(0,nrow = nrow(A),ncol = nrow(A))
  low_zero <-matrix(0,nrow = ncol(A), ncol = ncol(A))
  res_hight <-cbind(hight_zero, A)
  res_low <-cbind(t(A),low_zero)
  res <- rbind(res_hight,res_low)
  event_names <- sapply(1:ncol(A), function (x) paste(x,"event",sep=" "))
  actors_names <-sapply(1:nrow(A),function(x) paste(x, "actor",sep =" "))
  all_names <-c(actors_names,event_names)
  mask <-sapply(1:(nrow(A)+ncol(A)), function(x) if (x <= nrow(A)) TRUE else FALSE)
  rownames(res) <-all_names
  colnames(res) <-all_names
  graph <-graph.adjacency(res,mode = "undirected")
  V(graph)$type <-mask
  return (graph)
  
}

task_5 <-function(A){
  graph <- create_graph(A)
  colors <-c("red","blue")
  my_layout <-layout_as_bipartite(graph,V(graph)$type,hgap = 0.1, vgap = 0.5,maxiter = 100)
  my_layout[,1] <- my_layout[,1] - 2.7
  plot.igraph(graph, layout= my_layout,rescale = T, vertex.color = colors[as.numeric(V(graph)$type)+1])
}

task_five(mat)
```

10. Вычислить плотность для соотношения соучастия и отношения перекрытия. Проинтерпретируйте полученные значения.

```r
task_10 <-function(A){
  participation_matrix <- A %*% t(A)
  overlap_matrix <- t(A) %*% A
  g <- nrow(participation_matrix)
  h <-nrow(overlap_matrix)
  participation_density <- (sum(participation_matrix)-sum(diag(participation_matrix)))/(g*(g-1))
  overlap_density <- (sum(overlap_matrix)-sum(diag(overlap_matrix)))/(h*(h-1))
  return (list(participation_density = participation_density, overlap_density = overlap_density))
}

print(task_10(mat))
```

**Выводы:** В обоих случаях плотность достаточно низкая

12. Определить клики уровня 𝑐 = {4,3} для отношения соучастия и клики уровня 𝑐 = {6, 5, 4, 3, 2} для отношения перекрытия событий. Сделать выводы.


```r
get_network_cliques <- function (matrix, level){
  graph <-graph.adjacency(matrix,weighted = TRUE,mode = "undirected")
  mask <- E(graph)$weight < level
  to_delete <- E(graph)[mask]
  graph <- delete_edges(graph,to_delete)
  return (cliques(graph,min = 3))
}

get_all_level_cliquies <- function (matrix,levels){
  res <- list()
  for(i in levels){
    new_list <- get_network_cliques(matrix,i)
    names(new_list) <- rep(paste(i,"clique",sep =" level "), length(new_list))
    res <- c(res,new_list)
  }
  return (res)
  
}

task_12_1 <- function(A){
  levels <- c(3,4)
  participation_matrix <- A %*% t(A)
  return(get_all_level_cliquies (participation_matrix,levels))
}

print(task_12_1(mat))
```

```r
task_12_2 <-function(A){
  levels <-c(2,3,4,5,6)
  overlap_matrix <- t(A) %*% A
  return(get_all_level_cliquies(overlap_matrix,levels))
}

print(task_12_2(mat))
```

**Выводы:** для отношений сооучастия и перекрытия были выведены клики соответсвующих уровней.

13. Для каждых пар событий определить отношение шансов. Сделать выводы.

```r
task_13 <- function(A){
  x_k_l <- t(A) %*% A
  x_not_kl <- (1-t(A))%*%(1-A)
  x_not_k_l <- (1-t(A))%*%A
  x_k_not_l <- t(A)%*%(1-A)
  
  res <- c()
 for(i in 2:nrow(x_k_l)){
   for(j in 1:(i-1)){
     name <- paste(i,"and",j,"events",sep=" ")
     chance <- (x_k_l[i,j]*x_not_kl[i,j])/(x_not_k_l[i,j]*x_k_not_l[i,j])
     res[name] <-chance
   }
 }
 return(res)
}

print(task_13(mat))
```

**Выводы:** есть большое количестве пар событий, которые исключают попадние актора в оба элемента этой пары; также в матрице есть 2 значения inf, которые говорят о том, что если актор попадет в одно событие из пары, то он гарантированно попадет во второе событие.
