# –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∞—è —Ä–∞–±–æ—Ç–∞ ‚Ññ3: –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ —Å–æ—Ü–∏–∞–ª—å–Ω—ã—Ö —Å–µ—Ç–µ–π.
–í—ã–ø–æ–ª–Ω–∏–ª —Å—Ç—É–¥–µ–Ω—Ç –≥—Ä. 0381 –ö–æ—Ç–æ–≤ –î–º–∏—Ç—Ä–∏–π. –í–∞—Ä–∏–∞–Ω—Ç ‚Ññ9

## –¶–µ–ª—å —Ä–∞–±–æ—Ç—ã
–û–≤–ª–∞–¥–µ—Ç—å –Ω–∞–≤—ã–∫–∞–º–∏ —Ä–∞–±–æ—Ç—ã —Å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º–∏ –ø–∞–∫–µ—Ç–∞–º–∏ –¥–ª—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ —Å–æ—Ü–∏–∞–ª—å–Ω—ã—Ö —Å–µ—Ç–µ–π.

## –û—Å–Ω–æ–≤–Ω—ã–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –ø–æ–ª–æ–∂–µ–Ω–∏—è
<b>–î–∏–∞–¥–∞</b> - –ø–∞—Ä–∞ –∞–∫—Ç–æ—Ä–æ–≤ –∏ –≤–æ–∑–º–æ–∂–Ω–∞—è —Å–≤—è–∑—å –º–µ–∂–¥—É –Ω–∏–º–∏
$D_{ij}=(x_{ij}, x_{ji}), i \neq j$

<b>–í–∑–∞–∏–º–Ω–∞—è –¥–∏–∞–¥–∞</b> $D_{ij}=(1,1)$

<b>–ê—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞—è –¥–∏–∞–¥–∞</b> $D_{ij}=(1,0), D_{ij}=(0,1)$

<b>–ù—É–ª–µ–≤–∞—è –¥–∏–∞–¥–∞</b> $D_{ij}=(0,0)$

<b>–¶–µ–Ω–∑—É—Å –¥–∏–∞–¥</b> - —Ç—Ä–æ–π–∫–∞ —á–∏—Å–µ–ª (M, A, N),
–≥–¥–µ
$M=\sum_{in < j} x_{ij}x_{ji}$

$A=X_{++} - 2M$

$N=C_{g}^2-A-M$

$M=\frac12tr(XX)$

$A=tr(XX^T)-tr(X^TX)$

$N=C_g^2 - tr(XX^T) + \frac12tr(XX)$

<b>–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ö–∞—Ü–∞-–ü–∞—É—ç–ª–ª–∞</b> - —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞ –∏–∑–º–µ—Ä–µ–Ω–∏—è —Ç–µ–Ω–¥–µ–Ω—Ü–∏–∏ –∞–∫—Ç–æ—Ä–æ–≤

$-\infty < p_{kp} \le 1$

–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ –æ–∂–∏–¥–∞–Ω–∏–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥, –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞—è, —á—Ç–æ –≤—ã–±–æ—Ä –¥–µ–ª–∞–µ—Ç—Å—è –∞–∫—Ç–æ—Ä–∞–º–∏ —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º.

<b>–§–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≤—ã–±–æ—Ä</b>

–ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è, —á—Ç–æ –∫–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä –¥–µ–ª–∞–µ—Ç d –≤—ã–±–æ—Ä–æ–≤ –∏–∑ $(g - 1)$ –∞–∫—Ç–æ—Ä–æ–≤, –¥–æ—Å—Ç—É–ø–Ω—ã—Ö –¥–ª—è –≤—ã–±–æ—Ä–∞. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤–∑–∞–∏–º–Ω–æ–≥–æ –≤—ã–±–æ—Ä–∞ –º–µ–∂–¥—É –ª—é–±–æ–π –ø–∞—Ä–æ–π –∞–∫—Ç–æ—Ä–æ–≤ —Ä–∞–≤–Ω–∞ $d^2/(g-1)^2$

<b>–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥</b>

$\mathbb{E}=\frac{g(g-1)}{2}=\frac{d^2}{(g-1)^2}=\frac{gd^2}{2(g-1)}$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –æ–±—â–∞—è –¥–∏–∞–¥–∞, –≤ –∫–æ—Ç–æ—Ä–æ–π —É—á–∞—Å—Ç–≤—É—é—Ç –∞–∫—Ç–æ—Ä—ã —è–≤–ª—è–µ—Ç—Å—è –≤–∑–∞–∏–º–Ω–æ–π –¥–∏–∞–¥–æ–π

$P(i \rightarrow j \wedge j \rightarrow i)=P(i \rightarrow j) * P(j \rightarrow i | i \rightarrow j )$

–£—Å–ª–æ–≤–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å 

$P(j \rightarrow i | i \rightarrow j )=P(j \rightarrow i)+p_{kp}P(j \not\rightarrow i)$

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º 

$P(i \rightarrow j \wedge j \rightarrow i)=\frac{d}{g-1}(\frac{d}{g-1}+p_{kp}\frac{g-1-d}{g-1})$

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥ —Å —É—á—ë—Ç–æ–º –¥–∞–Ω–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏

$\mathbb{E}(M|p) = \frac{gd^2}{2(g-1)}(1-p_{kp}+\frac{gd}{2}p_{kp}$

–û—Ü–µ–Ω–∫–∞ $\hat{p_{kp}}=\frac{2(g-1)M-gd^2}{gd(g-1-d)}$

<b>–°–≤–æ–±–æ–¥–Ω—ã–π –≤—ã–±–æ—Ä</b> 

–ü—É—Å—Ç—å i-—ã–π –∞–∫—Ç–æ—Ä –≤—ã–±–∏—Ä–∞–µ—Ç

$d_o(n_i)=x_{i+}$

–ü—É—Å—Ç—å $L=\sum{x_{i+}} L=\sum{x_{i+}^2}$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –∞–∫—Ç–æ—Ä—ã –∏–º–µ—é—Ç –≤–∑–∞–∏–º–Ω–æ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ, —Ä–∞–≤–Ω–∞ 
$X_i+x_j/(g-1)^2$

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥
$\mathbb{E}=\frac{L^2-L_2}{2(g-1)^2}$

–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥ —Å —É—á—ë—Ç–æ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
$\mathbb{E}(M|p^{'})=\frac{L^2-L_2}{2(g-1)}(1-p_{kp}^{'}+\frac{L}{2}p_{kp})$

–û—Ü–µ–Ω–∫–∞ –ø–æ –º–µ—Ç–æ–¥—É –º–æ–º–µ–Ω—Ç–æ–≤

$\hat{p}_{kp}^{'}=\frac{2(g-1)^2M-L^2+L_2}{L(g-1)^2-L^2+L_2}$

–ò–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è - –æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–º –∑–Ω–∞—á–µ–Ω–∏—è—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥.
–ó–Ω–∞—á–µ–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–∞:
$\rho_\beta=\frac{M-M_{min}}{M_{max}-M_{min}}$

$M_{min}=max_{o\le t \le g}f(t)$

$M_max=\left\lfloor\frac12(\sum_{i=1}^gx_{i+}-max_{0\le t \le g} g(t))\right\rfloor$

$f(t)=\sum_{i=1}^tx_{i+}-t(g-1)-C_t^2$

$g(t) = \sum_{i=1}^tx_{i+}-t(t-1)-\sum_{i=t+1}min(t,x_{i+})$

**–°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞**

$H_0: L \sim B(g(g-1),\frac12)$

–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞:
$z_i=\frac{I-\mathbb{E}(L)}{\sqrt{D(L)}}=\frac{I-g(g-1)/2}{\sqrt{g(g-1)}/4}$,

–≥–¥–µ $I=\sum_{i=1}^g\sum_{j=1}^gx_ij$
–î–∞–Ω–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —è–≤–ª—è–µ—Ç—Å—è –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ —Å—Ç–Ω–¥–∞—Ä—Ç–Ω—ã–º –Ω–æ–º–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º —É—Ä–æ–≤–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏.
–û—Ü–µ–Ω–∫–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ p
–ß–∏—Å–ª–æ –¥—É–≥ $L \sim B(g(g-1)P_0)$ –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —á–∏—Å–ª–æ –¥—É–≥ –≤ –æ—Ä–≥—Ä–∞—Ñ–µ —Ä–∞–≤–Ω–æ I, $P(L=I)=C_{g(g-1)}p_0^{'}(1-P_0)^{g(g-1)-L}$

**–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ**

$\mathbb{E}(L)=P_0g(g-1)$

**–î–∏—Å–ø–µ—Ä—Å–∏—è**

$\mathbb{D}=P_0(1-P_0)g(g-1)$

–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ $H_0: L \sim B(g(g-1)P_0)$ 

$z_i=\frac{I-\mathbb{E}(L)}{\sqrt{D(L)}}=\frac{I-P_0g(g-1)}{\sqrt{P_0(1-P_0)g(g-1)}}$

–î–∞–Ω–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —è–≤–ª—è–µ—Ç—Å—è –ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏.

**–û—Ü–µ–Ω–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞**
$\hat{p}=\frac{\sum_{i=1}^g\sum_{j=1}^gx_{ij}}{g(g-1)}=\frac{I}{g(g-1)}$

**–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª** 

$\hat{p}_{lower} \le p \le \hat{p}_{upper}$

$\hat{p}_{lower}=\hat{p}-z_{\alpha/2}\sqrt{\hat{p}(1-\hat{p}/g(g-1)}$

$\hat{p}_{upper}=\hat{p}+z_{\alpha/2}\sqrt{\hat{p}(1-\hat{p}/g(g-1)}$

–ó–¥–µ—Å—å $z_{\alpha/2} : \Phi(z_{\alpha/2})=0.5-\alpha/2$

## –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏
–° –ø–æ–º–æ—â—å—é –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏—Ö –ø–∞–∫–µ—Ç–æ–≤ –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫ –ø—Ä–æ–≤–µ—Å—Ç–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ —Å–ª—É—á–∞–π–Ω–æ —Å–º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∞. –ü–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ–¥–µ—Ä–∂–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å.

## –í—ã–ø–æ–ª–Ω–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã
–ö–∞–∂–¥—ã–π –ø—É–Ω–∫—Ç —Ä–∞–±–æ—Ç—ã (*–ø–æ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏*) –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —Å –ø–æ–º–æ—â—å—é –∫–æ–¥–∞.
–†–µ–∫–æ–º–µ–Ω–¥–æ–≤–∞–Ω–Ω—ã–µ —è–∑—ã–∫–∏ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è ‚Äì R –∏–ª–∏ Python.

```r
#–ø–æ–¥–∫–ª—é—á–µ–Ω–∏–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –ø–∞–∫–µ—Ç–æ–≤

library(igraph)
library(statnet)

```

####1. –° –ø–æ–º–æ—â—å—é –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ igraph —Å–º–æ–¥–µ–ª–∏—Ä–æ–≤–∞—Ç—å —Å–ª—É—á–∞–π–Ω—ã–π –æ—Ä–≥—Ä–∞—Ñ Gùëë –Ω–∞ –æ—Å–Ω–æ–≤–µ –º–æ–¥–µ–ª–∏ –≠—Ä–¥–µ—à–∞‚Äî–†–µ–Ω—å–∏: Gùëë = G (ùëî, ùëù), –≥–¥–µ ùëî ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–æ—Ä–æ–≤, ùëù ‚Äì –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—è–≤–ª–µ–Ω–∏—è –¥—É–≥–∏ –º–µ–∂–¥—É –∞–∫—Ç–æ—Ä–∞–º–∏. –í R –¥–ª—è –º–æ–¥–µ–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∞ –≠—Ä–¥–µ—à–∞‚Äî–†–µ–Ω—å–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ñ—É–Ω–∫—Ü–∏—è erdos.renyi.game(). –û–∑–Ω–∞–∫–æ–º—å—Ç–µ—Å—å —Å –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª—å—é –∏ –æ–ø–∏—à–∏—Ç–µ –µ—ë. –°–æ–≥–ª–∞—Å–Ω–æ –≤–∞—Ä–∏–∞–Ω—Ç—É 9: g = 71, p = 0.71

```r
g <- 95
p <- 0.63
random_graph <- erdos.renyi.game(g, p, "gnp", directed = T)
net <- network(as_adjacency_matrix(random_graph), directed = T)
```

–ò—Å–ø–æ–ª—å–∑—É–µ–º –≤–∞—Ä–∏–∞–Ω—Ç —Å G(g, p). –í –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –º–µ–∂–¥—É –≤–µ—Ä—à–∏–Ω–∞–º–∏ –±—É–¥—É—Ç —Å–ª—É—á–∞–π–Ω–æ —Ä–∞—Å—Å—Ç–∞–≤–ª—è—Ç—å—Å—è —Ä—ë–±—Ä–∞ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é $p$. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ —Å $g$ —É–∑–ª–∞–º–∏ –∏ $L$ —Ä—ë–±—Ä–∞–º–∏ —Ä–∞–≤–Ω–∞:

$p^L (1-p)^{(^2_n) - L}$.


####2. –û—Ç–æ–±—Ä–∞–∑–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–π –æ—Ä–≥—Ä–∞—Ñ.

```r
plot(random_graph, layout = layout_with_fr, vertex.label = V(random_graph)$name, edge.arrow.size=0.1)
```

####3. –í—ã—á–∏—Å–ª–∏—Ç—å ¬´–ø—è—Ç–∏—á–∏—Å–ª–æ–≤—É—é —Å–≤–æ–¥–∫—É¬ª –¥–ª—è –≥—Ä–∞—Ñ–∞ Gùëë.

```r
cat("size: ", network.size(net))
cat("\ndensity: ", gden(net))
cat("\n#components: ", components(net))
cat("\ndiameter: ", diameter(random_graph))
cat("\ngtrans: ", gtrans(net, mode = "graph"))
```

####4. –í—ã—á–∏—Å–ª–∏—Ç—å —Ü–µ–Ω–∑—É—Å –¥–∏–∞–¥. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```r
census <- dyad_census(random_graph)
census
```

–í—ã–≤–æ–¥—ã:

–í–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥ 1762

–û–¥–Ω–æ–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö –æ—Ç–Ω–æ—à–µ–Ω–∏–π 2098

–ù–µ—Å–≤—è–∑–Ω—ã—Ö –∞–∫—Ç–æ—Ä–æ–≤ 605

####5. –í—ã—á–∏—Å–ª–∏—Ç—å –æ—Ü–µ–Ω–∫—É –∏–Ω–¥–µ–∫—Å–∞ –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ö–∞—Ü–∞‚Äî–ü–∞—É—ç–ª–∞ –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–æ–º –≤—ã–±–æ—Ä–µ –∞–∫—Ç–æ—Ä–æ–≤ ùëë –∏ –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

–í–æ–∑—å–º—ë–º d = 40.

–û—Ü–µ–Ω–∫–∞ –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –≤—ã–±–æ—Ä–µ –ø–æ –º–µ—Ç–æ–¥—É –º–æ–º–µ–Ω—Ç–æ–≤:

```r
d <- 40
KPEstimationFixed = (2*(g-1) * census$mut - g*d*d) / (g*d*(g-1-d))
cat("Estimation with fixed d:", KPEstimationFixed)
```

–û—Ü–µ–Ω–∫–∞ –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ:

```r
adj_matrix <- as_adjacency_matrix(random_graph)
matr <- as.matrix(adj_matrix)
L_list <- rowSums(matr)
L <- sum(L_list)
L2_list <- rep(0, nrow(matr))
for(row in 1:nrow(matr))
  L2_list[row] <- L_list[row]*L_list[row]

L2 <- sum(L2_list)
KPEstimationFree <- (2 * (g - 1) * (g - 1) * census$mut - L * L + L2) / (L * (g-1)* (g-1) - L*L + L2)
cat("Estimation free:", KPEstimationFree)
```

–í—ã–≤–æ–¥—ã: –û—Ü–µ–Ω–∫–∞ –ø—Ä–∏ —Å–≤–æ–±–æ–¥–Ω–æ–º –≤—ã–±–æ—Ä–µ –ø–æ–ª—É—á–∏–ª–∞—Å—å -0.007297036, –∞ –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –∏ —Ä–∞–≤–Ω–æ–º 50 –ø—Ä–∏–º–µ—Ä–Ω–æ 0.46, —Ä–∞–∑–Ω–∏—Ü–∞ –Ω–∞ –ø–æ—Ä—è–¥–æ–∫.

####6. –í—ã—á–∏—Å–ª–∏—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å–≤—è–∑—å –±—É–¥–µ—Ç –≤–∑–∞–∏–º–Ω–æ–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Å–ª—É—á–∞—è.

–§–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–π

```r
p_fixed <- (d / (g-1)) * ((d/(g-1)) + KPEstimationFixed*(g - 1 - d) / (g - 1))
p_fixed
```

–°–≤–æ–±–æ–¥–Ω—ã–π

```r
p_free = 0
for(row in 1:(nrow(matr) - 1))
  for(row2 in (row + 1):nrow(matr))
    p_free <- p_free + (L_list[row] * L_list[row2])/((g-1)*(g-1))
p_free = p_free * 2 / (g*(g-1))
p_free
```

–ù–µ—Å–º–æ—Ç—Ä—è –Ω–∞ –±–æ–ª—å—à—É—é —Ä–∞–∑–Ω–∏—Ü—É –≤ –æ—Ü–µ–Ω–∫–∞—Ö, –∏—Ç–æ–≥–æ–≤—ã–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª—É—á–∏–ª–∏—Å—å –ø–æ—á—Ç–∏ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º–∏ ~40%.

####7. –í—ã—á–∏—Å–ª–∏—Ç—å –∏–Ω–¥–µ–∫—Å –≤–∑–∞–∏–º–Ω–æ—Å—Ç–∏ –ë–∞–Ω–¥–∏–æ–ø–∞–¥—Ö—å—è–π—è. –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã.

```r
f <- function(t) {
  return (sum(L_list[1:t]) - t * (g - 1) - (factorial(t)/(factorial(2) * factorial(t-2))))
}

g_func <- function(t) {
  res_sum <- 0
  for(i in (t + 1):g)
    res_sum <- res_sum + min(t, L_list[i])
  return (sum(L_list[1:t]) - t* (t-1) - res_sum)
}

f_res = rep(2, g)
for(t in 2:g)
  f_res[t] = f(t)
  
g_res = rep(2, g - 1)
for(t in 2:(g - 1))
  g_res[t] = g_func(t)
  
Mmin <- max(f_res)
Mmax <- floor(0.5 * (L - max(g_res)))

index_B <- (census$mut - Mmin) / (Mmax - Mmin)

index_B
```

–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∑–∞–∏–º–Ω—ã—Ö –¥–∏–∞–¥ –ø—Ä–∏–º–µ—Ä–Ω–æ 2/3 –æ—Ç –∏—Ö –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞.

####8. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —á–∏—Å–ª–æ –¥—É–≥ ùêø –≤ –æ—Ä–≥—Ä–∞—Ñ–µ Gùëë –∏–º–µ–µ—Ç –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ ùêø ‚àº Bin(ùëî(ùëî ‚àí 1), 1/2) –ø—Ä–∏ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ ùõº. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è. ùõº = 0.03

```r
alpa <- 0.03
binom.test(L, g*(g-1), 1-alpa)
```

p-value < ùõº => –æ—Ç–∫–ª–æ–Ω—è–µ–º –≥–∏–ø–æ—Ç–µ–∑—É. –ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –º–µ–Ω—å—à–µ p-value, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –Ω–µ –±–æ–ª—å—à–µ, —á–µ–º 2.2–µ-16

####9. –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —á–∏—Å–ª–æ –¥—É–≥ ùêø –≤ –æ—Ä–≥—Ä–∞—Ñ–µ Gùëë –∏–º–µ–µ—Ç –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ ùêø ‚àº Bin(ùëî(ùëî‚àí1), ùëÉ0) –ø—Ä–∏ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ ùõº. –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è. 

$P_0$ = 0.55, ùõº = 0.03

```r
p0 = 0.55
binom.test(L, g*(g-1), p0, conf.level = 1-alpa)
```

p-value < ùõº => –æ—Ç–≤–µ—Ä–≥–∞–µ–º –≥–∏–ø–æ—Ç–µ–∑—É. –ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –º–µ–Ω—å—à–µ p-value, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –Ω–µ –±–æ–ª—å—à–µ, —á–µ–º 2.2–µ-16

####10.  –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùëù —Å –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–π –Ω–∞–¥—ë–∂–Ω–æ—Å—Ç—å—é (1 ‚àí ùõº). –°–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥—ã

```r
X <- as_adjacency_matrix(random_graph, sparse=FALSE)
p_cap <- sum(X)/(g*(g-1))
z_alpha_over_two <- 0.0376082877
p_lower <- p_cap - z_alpha_over_two * sqrt(p_cap * (1 - p_cap)/(g*(g-1)))
p_upper <- p_cap + z_alpha_over_two * sqrt(p_cap * (1 - p_cap)/(g*(g-1)))
cat('—Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é 99% –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç ', p_lower, ' –¥–æ ', p_upper, ' –Ω–∞–∫—Ä—ã–≤–∞–µ—Ç –∏—Å—Ç–∏–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ p')
```
