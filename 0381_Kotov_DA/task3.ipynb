## Основные теоретические положения
<b>Диада</b> - пара акторов и возможная связь между ними
$D_{ij}=(x_{ij}, x_{ji}), i \neq j$

<b>Взаимная диада</b> $D_{ij}=(1,1)$

<b>Асимметричная диада</b> $D_{ij}=(1,0), D_{ij}=(0,1)$

<b>Нулевая диада</b> $D_{ij}=(0,0)$

<b>Цензус диад</b> - тройка чисел (M, A, N),
где
$M=\sum_{in < j} x_{ij}x_{ji}$

$A=X_{++} - 2M$

$N=C_{g}^2-A-M$

$M=\frac12tr(XX)$

$A=tr(XX^T)-tr(X^TX)$

$N=C_g^2 - tr(XX^T) + \frac12tr(XX)$

<b>Индекс взаимности Каца-Пауэлла</b> - характеристика измерения тенденции акторов

$-\infty < p_{kp} \le 1$

Индекс взаимности основан на ожидании количества взаимных диад, предполагая, что выбор делается акторами случайным образом.

<b>Фиксированный выбор</b>

Предполагается, что каждый актор делает d выборов из $(g - 1)$ акторов, доступных для выбора. Вероятность взаимного выбора между любой парой акторов равна $d^2/(g-1)^2$

<b>Математическое ожидание взаимных диад</b>

$\mathbb{E}=\frac{g(g-1)}{2}=\frac{d^2}{(g-1)^2}=\frac{gd^2}{2(g-1)}$

Вероятность того, что общая диада, в которой участвуют акторы является взаимной диадой

$P(i \rightarrow j \wedge j \rightarrow i)=P(i \rightarrow j) * P(j \rightarrow i | i \rightarrow j )$

Условная вероятность 

$P(j \rightarrow i | i \rightarrow j )=P(j \rightarrow i)+p_{kp}P(j \not\rightarrow i)$

Таким образом 

$P(i \rightarrow j \wedge j \rightarrow i)=\frac{d}{g-1}(\frac{d}{g-1}+p_{kp}\frac{g-1-d}{g-1})$

Математическое ожидание взаимных диад с учётом данной вероятности

$\mathbb{E}(M|p) = \frac{gd^2}{2(g-1)}(1-p_{kp}+\frac{gd}{2}p_{kp}$

Оценка $\hat{p_{kp}}=\frac{2(g-1)M-gd^2}{gd(g-1-d)}$

<b>Свободный выбор</b> 

Пусть i-ый актор выбирает

$d_o(n_i)=x_{i+}$

Пусть $L=\sum{x_{i+}} L=\sum{x_{i+}^2}$

Вероятность, что акторы имеют взаимное отношение, равна 
$X_i+x_j/(g-1)^2$

Математическое ожидание взаимных диад
$\mathbb{E}=\frac{L^2-L_2}{2(g-1)^2}$

Математическое ожидание взаимных диад с учётом вероятности
$\mathbb{E}(M|p^{'})=\frac{L^2-L_2}{2(g-1)}(1-p_{kp}^{'}+\frac{L}{2}p_{kp})$

Оценка по методу моментов

$\hat{p}_{kp}^{'}=\frac{2(g-1)^2M-L^2+L_2}{L(g-1)^2-L^2+L_2}$

Индекс взаимности Бандиопадхьяйя - основан на минимальном и максимальном значениях количества взаимных диад.
Значение индекса:
$\rho_\beta=\frac{M-M_{min}}{M_{max}-M_{min}}$

$M_{min}=max_{o\le t \le g}f(t)$

$M_max=\left\lfloor\frac12(\sum_{i=1}^gx_{i+}-max_{0\le t \le g} g(t))\right\rfloor$

$f(t)=\sum_{i=1}^tx_{i+}-t(g-1)-C_t^2$

$g(t) = \sum_{i=1}^tx_{i+}-t(t-1)-\sum_{i=t+1}min(t,x_{i+})$

**Статистическая гипотеза**

$H_0: L \sim B(g(g-1),\frac12)$

Статистика:
$z_i=\frac{I-\mathbb{E}(L)}{\sqrt{D(L)}}=\frac{I-g(g-1)/2}{\sqrt{g(g-1)}/4}$,

где $I=\sum_{i=1}^g\sum_{j=1}^gx_ij$
Данная статистика является приблизительно стндартным номальным распределением при заданном урове значимости.
Оценка вероятности p
Число дуг $L \sim B(g(g-1)P_0)$ Вероятность того, что число дуг в орграфе равно I, $P(L=I)=C_{g(g-1)}p_0^{'}(1-P_0)^{g(g-1)-L}$

**Математическое ожидание**

$\mathbb{E}(L)=P_0g(g-1)$

**Дисперсия**

$\mathbb{D}=P_0(1-P_0)g(g-1)$

Нулевая гипотеза $H_0: L \sim B(g(g-1)P_0)$ 

$z_i=\frac{I-\mathbb{E}(L)}{\sqrt{D(L)}}=\frac{I-P_0g(g-1)}{\sqrt{P_0(1-P_0)g(g-1)}}$

Данная статистика является приблизительно стандартным нормальным распределением при заданном уровне значимости.

**Оценка параметра**
$\hat{p}=\frac{\sum_{i=1}^g\sum_{j=1}^gx_{ij}}{g(g-1)}=\frac{I}{g(g-1)}$

**Доверительный интервал** 

$\hat{p}_{lower} \le p \le \hat{p}_{upper}$

$\hat{p}_{lower}=\hat{p}-z_{\alpha/2}\sqrt{\hat{p}(1-\hat{p}/g(g-1)}$

$\hat{p}_{upper}=\hat{p}+z_{\alpha/2}\sqrt{\hat{p}(1-\hat{p}/g(g-1)}$

Здесь $z_{\alpha/2} : \Phi(z_{\alpha/2})=0.5-\alpha/2$

####1. С помощью библиотеки igraph смоделировать случайный орграф G𝑑 на основе модели Эрдеша—Реньи: G𝑑 = G (𝑔, 𝑝), где 𝑔 – количество акторов, 𝑝 – вероятность появления дуги между акторами. В R для модели случайного графа Эрдеша—Реньи существует функция erdos.renyi.game(). Ознакомьтесь с данной моделью и опишите её. Согласно варианту 9: g = 71, p = 0.71

```r
g <- 95
p <- 0.63
random_graph <- erdos.renyi.game(g, p, "gnp", directed = T)
net <- network(as_adjacency_matrix(random_graph), directed = T)
```

Используем вариант с G(g, p). В данном случае между вершинами будут случайно расставляться рёбра с вероятностью $p$. Вероятность построить граф с $g$ узлами и $L$ рёбрами равна:

$p^L (1-p)^{(^2_n) - L}$.

####5. Вычислить оценку индекса взаимности Каца—Пауэла при фиксированом выборе акторов 𝑑 и при свободном выборе. Сделать выводы.

Возьмём d = 40.

Оценка при фиксированном выборе по методу моментов:

```r
d <- 40
KPEstimationFixed = (2*(g-1) * census$mut - g*d*d) / (g*d*(g-1-d))
cat("Estimation with fixed d:", KPEstimationFixed)
```

Оценка при свободном выборе:

```r
adj_matrix <- as_adjacency_matrix(random_graph)
matr <- as.matrix(adj_matrix)
L_list <- rowSums(matr)
L <- sum(L_list)
L2_list <- rep(0, nrow(matr))
for(row in 1:nrow(matr))
  L2_list[row] <- L_list[row]*L_list[row]

L2 <- sum(L2_list)
KPEstimationFree <- (2 * (g - 1) * (g - 1) * census$mut - L * L + L2) / (L * (g-1)* (g-1) - L*L + L2)
cat("Estimation free:", KPEstimationFree)
```

Выводы: Оценка при свободном выборе получилась -0.007297036, а при фиксированном и равном 50 примерно 0.46, разница на порядок.

####6. Вычислить вероятность того, что связь будет взаимной для каждого случая.

Фиксированный

```r
p_fixed <- (d / (g-1)) * ((d/(g-1)) + KPEstimationFixed*(g - 1 - d) / (g - 1))
p_fixed
```

Свободный

```r
p_free = 0
for(row in 1:(nrow(matr) - 1))
  for(row2 in (row + 1):nrow(matr))
    p_free <- p_free + (L_list[row] * L_list[row2])/((g-1)*(g-1))
p_free = p_free * 2 / (g*(g-1))
p_free
```

Несмотря на большую разницу в оценках, итоговые вероятность получились почти одинаковыми ~40%.

####7. Вычислить индекс взаимности Бандиопадхьяйя. Сделать выводы.

```r
f <- function(t) {
  return (sum(L_list[1:t]) - t * (g - 1) - (factorial(t)/(factorial(2) * factorial(t-2))))
}

g_func <- function(t) {
  res_sum <- 0
  for(i in (t + 1):g)
    res_sum <- res_sum + min(t, L_list[i])
  return (sum(L_list[1:t]) - t* (t-1) - res_sum)
}

f_res = rep(2, g)
for(t in 2:g)
  f_res[t] = f(t)
  
g_res = rep(2, g - 1)
for(t in 2:(g - 1))
  g_res[t] = g_func(t)
  
Mmin <- max(f_res)
Mmax <- floor(0.5 * (L - max(g_res)))

index_B <- (census$mut - Mmin) / (Mmax - Mmin)

index_B
```

Количество взаимных диад примерно 2/3 от их максимально возможного количества.

####9. Проверить нулевую гипотезу о том, что число дуг 𝐿 в орграфе G𝑑 имеет биномиальное распределение 𝐿 ∼ Bin(𝑔(𝑔−1), 𝑃0) при уровне значимости 𝛼. Определить максимальный уровень значимости при котором нулевая гипотеза не отвергается.

####10.  Построить доверительный интервал для оценки параметра 𝑝 с доверительной надёжностью (1 − 𝛼). Сделать выводы

```r
X <- as_adjacency_matrix(random_graph, sparse=FALSE)
p_cap <- sum(X)/(g*(g-1))
z_alpha_over_two <- 0.0376082877
p_lower <- p_cap - z_alpha_over_two * sqrt(p_cap * (1 - p_cap)/(g*(g-1)))
p_upper <- p_cap + z_alpha_over_two * sqrt(p_cap * (1 - p_cap)/(g*(g-1)))
cat('с вероятностью 99% интервал от ', p_lower, ' до ', p_upper, ' накрывает истинное значение параметра p')
```
