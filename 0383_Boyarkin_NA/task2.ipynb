```r
find.nclan <- function(g, n.cliques){
  for (n in 1:5){
    for (i in 1:length(n.cliques[[n]])) { 
      if (diameter(induced_subgraph(g, n.cliques[[n]][[i]])) <= n){
        print(n)
        print(n.cliques[[n]][[i]])
      }
    }
  }
}

find.nclan(net_florentine_business, n.cliques)
```

Ğ’Ñ‹Ğ²Ğ¾Ğ´: Ğ¸Ğ· Ğ´Ğ°Ğ½Ğ½Ğ¾Ğ¹ Ñ‚Ğ°Ğ±Ğ»Ğ¸Ñ†Ñ‹ Ğ¼Ğ¾Ğ¶Ğ½Ğ¾ Ğ·Ğ°Ğ¼ĞµÑ‚Ğ¸Ñ‚ÑŒ, Ñ‡Ñ‚Ğ¾ 1-ÑĞ´Ñ€Ğ¾ Ğ¿Ğ¾Ğ´Ğ³Ñ€Ğ°Ñ„ ÑĞ¾Ğ´ĞµÑ€Ğ¶Ğ¸Ñ‚ Ğ²ĞµÑ€ÑˆĞ¸Ğ½Ñ‹: Barbadori, Bischeri, Castellani, Ginori, Guadagni, Lamberteschi, Medici, Pazzi, Peruzzi, Salviati, Tornabuoni; 2-ÑĞ´Ñ€Ğ¾: Barbadori, Bischeri, Castellani, Ginori, Guadagni, Lamberteschi, Medici, Peruzzi.

```r
find_k_plexes <- function(g, k) {
  k_plexes <- list()
  for (v in V(g)$name) {
    neighbors <- neighbors(g, v)$name
    if (length(neighbors) >= k) {
      is_k_plex <- TRUE
      for (u in neighbors) {
        if (length(intersect(neighbors, neighbors(g, u))) < length(neighbors) - k) {
          is_k_plex <- FALSE
          break
        }
      }
      if (is_k_plex) {
        k_plexes[[length(k_plexes) + 1]] <- c(neighbors, v)
      }
    }
  }
  return(k_plexes)
}


for(i in c(1, 2, 3, 4, 5)){
  cat('k = ')
  cat(i, '\n')
  print(find_k_plexes(net_florentine_business, i))
}
```

### 13. Ğ”Ğ»Ñ ĞºĞ°Ğ¶Ğ´Ñ‹Ñ… Ğ¿Ğ°Ñ€ ÑĞ¾Ğ±Ñ‹Ñ‚Ğ¸Ğ¹ Ğ¾Ğ¿Ñ€ĞµĞ´ĞµĞ»Ğ¸Ñ‚ÑŒ Ğ¾Ñ‚Ğ½Ğ¾ÑˆĞµĞ½Ğ¸Ğµ ÑˆĞ°Ğ½ÑĞ¾Ğ². Ğ¡Ğ´ĞµĞ»Ğ°Ñ‚ÑŒ Ğ²Ñ‹Ğ²Ğ¾Ğ´Ñ‹.

```r
transpose_matrix_ceo <- t(incidence_matrix_ceo)
nkl <- (1 - transpose_matrix_ceo) %*% (1 - incidence_matrix_ceo)
nk_l <- (1 - transpose_matrix_ceo) %*% incidence_matrix_ceo
k_nl <- transpose_matrix_ceo %*% (1 - incidence_matrix_ceo)
count.zero <- 0
count.inf <- 0

for(i in 2:nrow(event_matrix)){
   for(j in 1:(i - 1)){
     cat('i = ', i)
     cat(' ')
     cat('j = ', j)
     cat('\nres = ')
     res <- (event_matrix[i,j] * nkl[i,j]) / (nk_l[i,j] * k_nl[i,j])
     cat(res)
     cat('\n')
     if (res == 0){
       count.zero <- count.zero + 1
     }
     if (res == Inf){
       count.inf <- count.inf + 1
     }
   }
}
print('ĞšĞ¾Ğ»-Ğ²Ğ¾ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ¸Ğ¹ Ñ Ğ±ĞµÑĞºĞ¾Ğ½ĞµÑ‡Ğ½Ğ¾ÑÑ‚ÑŒÑ: ')
print(count.inf)
print('ĞšĞ¾Ğ»-Ğ²Ğ¾ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ¸Ğ¹ Ñ 0: ')
print(count.zero)
```
