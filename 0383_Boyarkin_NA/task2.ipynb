***k-—è–¥—Ä–æ*** - —ç—Ç–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–∞–∂–¥–∞—è –≤–µ—Ä—à–∏–Ω–∞ —Å–≤—è–∑–∞–Ω–∞ –º–∏–Ω–∏–º—É–º —Å $k$ –¥—Ä—É–≥–∏–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏ —ç—Ç–æ–≥–æ –∂–µ –ø–æ–¥–≥—Ä–∞—Ñ–∞

***LS-–º–Ω–æ–∂–µ—Å—Ç–≤–æ***: –î–∞–Ω –ø–æ–¥–≥—Ä–∞—Ñ $Y$ —Å –º–Ω–æ–∂–µ—Å—Ç–≤–æ–º —É–∑–ª–æ–≤ $N_s$. –ú–Ω–æ–∂–µ—Å—Ç–≤–æ —É–∑–ª–æ–≤ $N_s$ –±—É–¥–µ—Ç –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è $LS$-–º–Ω–æ–∂–µ—Å—Ç–≤–æ–º, –µ—Å–ª–∏ –ª—é–±–æ–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–µ –ø–æ–¥–º–Ω–æ–∂–µ—Å—Ç–≤–æ $Q$ —Å $N_s$ –∏–º–µ–µ—Ç –±–æ–ª—å—à–µ –ª–∏–Ω–∏–π –∫ —É–∑–ª–∞–º $N_s|Q$, —á–µ–º –∫ $N|N_s$ 

***$\lambda$-–º–Ω–æ–∂–µ—Å—Ç–≤–æ***: –†–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—é—Ç—Å—è –ø–∞—Ä—ã —É–∑–ª–æ–≤ –∏–∑ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ $N_s$ –ø–æ–¥–≥—Ä–∞—Ñ–∞ $Y_s$. –ú–Ω–æ–∂–µ—Å—Ç–≤–æ —É–∑–ª–æ–≤ $N_s$ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è $\lambda$-–º–Ω–æ–∂–µ—Å—Ç–≤–æ–º, –µ—Å–ª–∏ –ª—é–±–∞—è –ø–∞—Ä–∞ —É–∑–ª–æ–≤ –≤ $N_s$ –∏–º–µ–µ—Ç –±–æ–ª—å—à–µ–µ –ª–∏–Ω–µ–π–Ω–æ–µ —Å–æ–µ–¥–∏–Ω–µ–Ω–∏–µ, —á–µ–º –ª—é–±–∞—è –¥—Ä—É–≥–∞—è –ø–∞—Ä–∞ —É–∑–ª–æ–≤, —Å–æ—Å—Ç–æ—è—â–∞—è –∏–∑ –æ–¥–Ω–æ–≥–æ —É–∑–ª–∞ –∏–∑ –ª—è–º–±–¥–∞-–Ω–∞–±–æ—Ä–∞ –∏ –≤–æ—Ç–æ—Ä–æ–≥–æ —É–∑–ª–∞ –∏–∑–≤–Ω–µ:
$$N_s \subseteq N: ‚àÄ i,j,k \in N_s, l \in N|N_s \Rightarrow \lambda(i, j) > \lambda (k, l)$$
–≥–¥–µ $\lambda(i, j)$ - –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∏–Ω–∏–π, –∫–æ—Ç–æ—Ä–æ–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —É–¥–∞–ª–∏—Ç—å –∏–∑ –≥—Ä–∞—Ñ–∞, —á—Ç–æ–±—ã –Ω–µ –æ—Å—Ç–∞–ª–æ—Å—å –ø—É—Ç–µ–π –º–µ–∂–¥—É –¥–≤—É–º—è —É–∑–ª–∞–º–∏

***–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–¥–≥—Ä–∞—Ñ***: –ø–æ–¥–≥—Ä–∞—Ñ —è–≤–ª—è–µ—Ç—Å—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –ø–æ –æ—Ç–Ω–æ—à–µ–Ω–∏—é –∫ —Å–≤–æ–π—Å—Ç–≤—É, –∑–∞–¥–∞–Ω–Ω–æ–º—É –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ, –µ—Å–ª–∏ —ç—Ç–æ —Å–≤–æ–π—Å—Ç–≤–æ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –¥–ª—è —ç—Ç–æ–≥–æ –ø–æ–¥–≥—Ä–∞—Ñ–∞, –Ω–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–µ—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å—Å—è, –µ—Å–ª–∏ –∫ –ø–æ–¥–≥—Ä–∞—Ñ—É –¥–æ–±–∞–≤–ª–µ–Ω –∫–∞–∫–æ–π-–ª–∏–±–æ —É–∑–µ–ª –∏–ª–∏ —É–∑–ª—ã

2) –û–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ —Å–≤—è–∑–∞–Ω–Ω–∞—è $n$-–∫–ª–∏–∫–∞ - –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –≤—Å–µ —É–∑–ª—ã –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ $n$-—Å–≤—è–∑–∞–Ω—ã –∏ –Ω–µ—Ç –¥–æ–ø–æ–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —É–∑–ª–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ —Ç–∞–∫–¥–µ –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ $n$-—Å–≤—è–∑–Ω—ã —Å–æ –≤—Å–µ–º–∏ —É–∑–ª–∞–º–∏ –ø–æ–¥–≥—Ä–∞—Ñ–∞

3) –°–∏–ª—å–Ω–æ —Å–≤—è–∑–Ω–∞—è $n$-–∫–ª–∏–∫–∞ - –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –≤—Å–µ —É–∑–ª—ã —Å–∏–ª—å–Ω–æ $n$-—Å–≤—è–∑–Ω—ã, –∏ –Ω–µ—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —É–∑–ª–æ–≤, —Ç–∞–∫–∂–µ —Å–∏–ª—å–Ω–æ $n$-—Å–≤—è–∑–Ω—ã—Ö —Å–æ –≤—Å–µ–º–∏ —É–∑–ª–∞–º–∏ –ø–æ–¥–≥—Ä–∞—Ñ–∞.

4) –†–µ–∫—É—Ä—Å–∏–≤–Ω–∞—è $n$-–∫–ª–∏–∫–∞ - –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –≤—Å–µ —É–∑–ª—ã —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ $n$-—Å–≤—è–∑–Ω—ã, –∏ –Ω–µ—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —É–∑–ª–æ–≤, —Ç–∞–∫–∂–µ —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ $n$-—Å–≤—è–∑–Ω—ã—Ö —Å–æ –≤—Å–µ–º–∏ —É–∑–ª–∞–º–∏ –ø–æ–¥–≥—Ä–∞—Ñ–∞

***–ú–æ–¥—É–ª—è—Ä–Ω–æ—Å—Ç—å*** - —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã —Å–µ—Ç–∏, –≤ —á–∞—Å—Ç–Ω–æ—Å—Ç–∏ —Å—Ç–µ–ø–µ–Ω—å –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ —É–∑–ª–æ–≤, –∫–æ–≥–¥–∞ –≤–Ω—É—Ç—Ä–∏ –∫–ª–∞—Å—Ç–µ—Ä–∞ –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è –º–Ω–æ–≥–æ —É–∑–ª–æ–≤ (–≤—ã—Å–æ–∫–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å), –∞ –º–µ–∂–¥—É –∫–ª–∞—Å—Ç–µ—Ä–º–∏ –º–∞–ª–æ–≤ —É–∑–ª–æ–≤ (–Ω–∏–∑–∫–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å). –û–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∫–∞–∫ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –¥–æ–ª–µ–π —Å–≤—è–∑–µ–π –≤–Ω—É—Ç—Ä–∏ –≥—Ä—É–ø–ø—ã –∏ –æ–∂–∏–¥–∞–µ–º–æ–π –¥–æ–ª–µ–π —Å–≤—è–∑–µ–π, –µ—Å–ª–∏ –±—ã —Å–≤—è–∑–∏ –±—ã–ª–∏ —Ä–∞–∑–º–µ—â–µ–Ω—ã —Å–ª—É—á–∞–π–Ω–æ.

***–°–µ—Ç—å –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏*** - —Å–µ—Ç—å, –≤ –∫–æ—Ç–æ—Ä–æ–π —É—á–∞—Å—Ç–Ω–∏–∫–∏ –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω—ã –¥—Ä—É–≥ —Å –¥—Ä—É–≥–æ–º –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å–æ—Ç—Ä—É–¥–Ω–∏—á–µ—Å—Ç–≤–∞ –≤ —Ä–∞–º–∫–∞—Ö –∫–∞–∫–æ–π-—Ç–æ –≥—Ä—É–ø–ø—ã –∏–ª–∏ —Å–æ–≤–º–µ—Å—Ç–Ω–æ–≥–æ —É—á–∞—Å—Ç–∏—è –≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º —Å–æ–±—ã—Ç–∏–∏. –°–µ—Ç—å —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ 2 –∏–∑–º–µ—Ä–µ–Ω–∏–π: –º–Ω–æ–∂–µ—Å—Ç–≤–∞ –∞–∫—Ç–æ—Ä–æ–≤ $N = \{n_1, n_2, ..., n_g\}$ –∏ –º–Ω–æ–∂–µ—Å—Ç–≤–∞ —Å–æ–±—ã—Ç–∏–π $M = \{m_1, m_2, ... , m_n\}$

***–ú–∞—Ç—Ä–∏—Ü–∞ —Å–æ—É—á–∞—Å—Ç–∏—è –∞–∫—Ç–æ—Ä–æ–≤***:
$$X^N = \{x_{ij}^N\}$$

$$x_{ij}^N = \sum_{k=1}^ha_{ik}a_{jk}$$

$$X^N = AA^T$$

***–ú–∞—Ç—Ä–∏—Ü–∞ –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è —Å–æ–±—ã—Ç–∏–π***:
$$X^M = \{x_{kl}^M\}$$

$$x_{kl}^M = \sum_{i=1}^ga_{ik}a_{il}$$

$$X^M = A^TA$$

***–°—Ç–µ–ø–µ–Ω—å —É—á–∞—Å—Ç–∏—è*** - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π, —Å –∫–æ—Ç–æ—Ä—ã–º–∏ —Å–≤—è–∑–∞–Ω –∫–∞–∂–¥—ã–π –∞–∫—Ç–æ—Ä:
$$a_{i+} = \sum_{j=1}^ha_{ij} = x_{ii}^N$$

***–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π, —Å –∫–æ—Ç–æ—Ä—ã–º –∞–∫—Ç–æ—Ä—ã –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω—ã***:

$$\overline{a_{i+}} = \frac{1}{g}\sum_{i=1}^g\sum_{j=1}^ha_{ij} = \frac{a_{++}}{g} = \frac{1}{g}\sum_{i=1}^gx_{ii}^N$$

***–†–∞–∑–º–µ—Ä —Å–æ–±—ã—Ç–∏–π***:
$$a_{+j} = \sum_{i=1}^ga_{ij} = x_{jj}^M$$

***–°—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–º–µ—Ä —Å–æ–±—ã—Ç–∏–π*** - —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–æ—Ä–æ–≤ –≤ –∫–∞–∂–¥–æ–º —Å–æ–±—ã—Ç–∏–∏ –∏–ª–∏ —Å—Ä–µ–¥–Ω—è—è —Å—Ç–µ–ø–µ–Ω—å —É–∑–ª–æ–≤, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—â–∏—Ö —Å–æ–±—ã—Ç–∏—è –≤ –¥–≤—É–¥–æ–ª—å–Ω–æ–º –≥—Ä–∞—Ñ–µ:
$$\overline{a_{+j}} = \frac{1}{n}\sum_{i=1}^g\sum_{j=1}^ha_{ij} = \frac{a_{++}}{h} = \frac{1}{h}\sum_{j=1}^hx_{jj}^M$$

***–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏—è —Å–æ—É—á–∞—Å—Ç—è $Œî_{(N)}$***:
$$Œî_{(N)} = \frac{1}{g(g-1)}\sum_{i=1}^g\sum_{j=1}^gx_{ij}^N, j \neq i$$

***–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏—è –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è $Œî_{(M)}$***:
$$Œî_{(M)} = \frac{1}{h(h-1)}\sum_{k=1}^h\sum_{l=1}^hx_{kl}^M, l \neq k$$

***–î–∏–∞–º–µ—Ç—Ä –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —Å–µ—Ç–∏*** - –¥–ª–∏–Ω–∞ —Å–∞–º–æ–≥–æ –¥–ª–∏–Ω–Ω–æ–≥–æ –ø—É—Ç–∏ –º–µ–∂–¥—É –ª—é–±–æ–π –ø–∞—Ä–æ–π –∞–∫—Ç–æ—Ä–æ–≤ –∏/–∏–ª–∏ —Å–æ–±—ã—Ç–∏–π

***–ö–ª–∏–∫–∞ —É—Ä–æ–≤–Ω—è $c$ –≤ –≤–∑–≤–µ—à–µ–Ω–Ω–æ–º –≥—Ä–∞—Ñ–µ*** - –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –ø–æ–ª–Ω—ã–π –ø–æ–¥–≥—Ä–∞—Ñ –∏–∑ 3 –∏–ª–∏ –±–æ–ª–µ–µ —É–∑–ª–æ–≤, –≤—Å–µ –∏–∑ –∫–æ—Ç–æ—Ä—ã–∑ —è–≤–ª—è—é—Ç—Å—è —Å–º–µ–∂–Ω—ã–º–∏ –Ω–∞ —É—Ä–æ–≤–Ω–µ $c$, —Ç.–µ. –º–µ–∂–¥—É –≤—Å–µ–º–∏ –ø–∞—Ä–∞–º–∏ —É–∑–ª–æ–≤ –µ—Å—Ç—å —Ä–µ–±—Ä–∞ —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –±–æ–ª—å—à–∏–º–∏ –∏–ª–∏ —Ä–∞–≤–Ω—ã–º–∏ $c$

***–ö–ª–∏–∫–∞ —É—Ä–æ–≤–Ω—è $c$ –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏—è —Å–æ—É—á–∞—Å—Ç–∏—è*** - –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –≤—Å–µ –ø–∞—Ä—ã –∞–∫—Ç–æ—Ä–æ–≤ —Å–æ–≤–º–µ—Å—Ç–Ω–æ —É—á–∞–≤—Å—Ç–≤—É—é—Ç –Ω–µ –º–µ–Ω–µ–µ, —á–µ–º –≤ $c$ —Å–æ–±—ã—Ç–∏—è—Ö

***–ö–ª–∏–∫–∞ —É—Ä–æ–≤–Ω—è $c$ –¥–ª—è –æ—Ç–Ω–æ—à–µ–Ω–∏—è –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è*** - –ø–æ–¥–≥—Ä–∞—Ñ, –≤ –∫–æ—Ç–æ—Ä–æ–º –≤—Å–µ –ø–∞—Ä—ã —Å–æ–±—ã—Ç–∏–π –∏–º–µ—é—Ç –Ω–µ –º–µ–Ω–µ–µ $c$ –æ–±—â–∏—Ö —á–ª–µ–Ω–æ–≤

$x_{kl}^M$ - —á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω—ã —Å —Å–æ–±—ã—Ç–∏—è–º–∏ $k$ –∏ $l$

$x_{\overline{kl}}^M$ - —á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—Ç –Ω–∏ –æ–¥–Ω–æ–º—É —Å–æ–±—ã—Ç–∏—é 

$x_{k\overline{l}}^M$ - —á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã —Å —Å–æ–±—ã—Ç–∏–µ–º $k$, –Ω–æ –Ω–µ —Å —Å–æ–±—ã—Ç–∏–µ–º $l$

$x_{\overline{k}l}^M$ - —á–∏—Å–ª–æ –∞–∫—Ç–æ—Ä–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –∞—Ñ—Ñ–∏–ª–∏—Ä–æ–≤–∞–Ω—ã —Å —Å–æ–±—ã—Ç–∏–µ–º $l$, –Ω–æ –Ω–µ —Å —Å–æ–±—ã—Ç–∏–µ–º $k$

–°–ø—Ä–∞–≤–µ–¥–ª–∏–≤–æ —Å–ª–µ–¥—É—â–µ–µ:
$$x_{kl}^M + x_{k\overline{l}}^M + x_{k\overline{l}}^M + x_{\overline{k}l}^M$$

***–û—Ç–Ω–æ—à–µ–Ω–∏–µ —à–∞–Ω—Å–æ–≤*** - –º–µ—Ä–∞ –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏—è –º–µ–∂–¥—É —Å–æ–±—ã—Ç–∏—è–º–∏, –∫–æ—Ç–æ—Ä–∞—è –Ω–µ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —Ä–∞–∑–º–µ—Ä–∞ —Å–æ–±—ã—Ç–∏–π:
  $$\theta_{kl} = \frac{x_{kl}^M/x_{\overline{k}l}^M}{x_{k\overline{l}}^M/x_{\overline{kl}}^M} = \frac{x_{kl}^Mx_{\overline{kl}}^M}{x_{k\overline{l}}^Mx_{\overline{k}l}^M}$$

  ***–õ–æ–≥–∞—Ä–∏—Ñ–º –æ—Ç–Ω–æ—à–µ–Ω–∏—è —à–∞–Ω—Å–æ–≤***:
  $$ln\theta_{kl} = ln(\frac{x_{kl}^Mx_{\overline{kl}}^M}{x_{k\overline{l}}^Mx_{\overline{k}l}^M})$$

## –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏
